## Introduction

This project provides a rigorous Python implementation of the architectural pattern outlined by Karpathy (2026). The core of this repository is the iPython Notebook `llm_knowledge_bases_karpathy_draft.ipynb`, which contains a comprehensive suite of 20 orchestrated tasks to replicate and formalize the workflow.

The pipeline addresses a critical vulnerability in modern LLM usage: the reliance on ephemeral chat interactions that fail to accumulate durable knowledge. By treating the LLM not as a stateless oracle, but as a **knowledge organization engine** operating over a version-controllable filesystem substrate, the system enables cumulative knowledge accretion.

The codebase operationalizes this paradigm by:
-   **Ingesting** raw documents (PDFs, web clippings, repositories) into an immutable, append-only `raw/` directory.
-   **Compiling** these sources incrementally into a structured Markdown wiki, extracting concepts, and maintaining bidirectional links.
-   **Retrieving** information deterministically using auto-maintained indices and a field-aware BM25 search engine, bypassing the need for complex vector RAG at the ~400K word scale.
-   **Synthesizing** grounded answers that strictly enforce citation invariants, rendering them into durable artifacts (reports, slides, figures) that are filed back into the wiki.
-   **Auditing** the knowledge base via LLM linting to detect contradictions and impute missing data with strict provenance.

## Theoretical Background

The implemented methods formalize the workflow into a rigorous systems architecture.

**1. System Architecture:**
The system $\mathcal{S}$ is formalized as an 8-tuple defining the state and the operators acting upon that state:
$$ \mathcal{S} = \langle \mathcal{D}_{\text{raw}}, \mathcal{K}, \Phi_{\text{compile}}, \Phi_{\text{qa}}, \Phi_{\text{render}}, \Phi_{\text{lint}}, \mathcal{T}, \mathcal{L} \rangle $$

**2. Knowledge Graph Representation:**
The compiled wiki $\mathcal{K}$ is not a flat directory, but a directed labeled multigraph embedded in the filesystem:
$$ \mathcal{K} = (V, E, \lambda_V, \lambda_E) $$
Where nodes $V$ are Markdown files, edges $E$ are wikilinks, and $\lambda_V$ maps nodes to content and metadata (e.g., provenance, timestamps).

**3. Deterministic Local Retrieval:**
To support tool-assisted Q&A without vector embeddings, the system implements a field-aware BM25 scoring function:
$$ \text{score}(v, q) = \sum_{i=1}^m \text{idf}(t_i) \cdot \frac{\text{tf}_w(t_i, v) \cdot (k_1 + 1)}{\text{tf}_w(t_i, v) + k_1 \cdot \left(1 - b + b \cdot \frac{|v|}{\text{avgdl}}\right)} $$
Where $\text{tf}_w$ applies configured boosts to titles, headings, body text, and tags.

**4. The Compounding Cycle:**
The system executes a recurring transformation cycle, ensuring that generated artifacts $\mathcal{A}^{(t)}$ are reintegrated into the knowledge base:
$$ \mathcal{D}_{\text{raw}}^{(t)} \xrightarrow{\Phi_{\text{compile}}} \mathcal{K}^{(t)} \xrightarrow{\Phi_{\text{qa}}} \mathcal{A}^{(t)} \xrightarrow{\text{file}} \mathcal{K}^{(t+1)} \xrightarrow{\Phi_{\text{lint}}} \mathcal{K}_{\text{corrected}}^{(t+1)} $$

## Features

The provided iPython Notebook (`llm_knowledge_bases_karpathy_draft.ipynb`) implements the full research pipeline, including:

-   **Cryptographic Determinism:** Utilizes full-file chunked SHA-256 hashing and canonical JSON serialization to guarantee cross-platform byte-for-byte determinism for run manifests and delta detection.
-   **AST-Aware Markdown Parsing:** Integrates `markdown-it-py` to parse Abstract Syntax Trees, ensuring that wikilinks and section headers are extracted securely, ignoring false positives inside fenced code blocks.
-   **Secure Docker Sandboxing:** Neutralizes the severe security risks of executing LLM-generated Python code (for matplotlib figures) by isolating execution within an ephemeral, network-disabled Docker container.
-   **Computational Linguistics (NLP):** Replaces brittle regex heuristics with `spaCy` for Named Entity Recognition (NER) to detect unsupported factual claims, and `nltk` for formal Sentence Boundary Disambiguation (SBD) during summary extraction.
-   **$O(1)$ Incremental Graph Updates:** Updates the in-memory `networkx` multigraph incrementally during artifact filing, avoiding computationally naive $O(|V| + |E|)$ full filesystem rebuilds.
-   **Configuration-Driven Design:** All study parameters, prompt templates, decoding settings, and directory schemas are managed in an external `config.yaml` file, ensuring strict methodological reproducibility.

## Methodology Implemented

The core analytical steps directly implement the workflow from the source note:

1.  **Initialization & Governance (Tasks 1-3, 7-9):** Validates the `config.yaml` via deep recursive schema traversal. Materializes the filesystem substrate (`raw/`, `wiki/`, `logs/`) and enforces strict symlink-aware write policies. Registers hashed prompt templates.
2.  **Ingestion & Normalization (Tasks 4-6):** Acquires heterogeneous sources, validates them via binary magic-byte inspection, and normalizes them to UTF-8 derivatives. Computes cryptographic hashes to detect incremental deltas ($\Delta \mathcal{D}$).
3.  **Incremental Compilation (Task 10):** Assembles semantic chunks and invokes the LLM to compile changed raw sources into structured Markdown source summaries and concept pages, updating the graph topology.
4.  **Indexing & Retrieval (Tasks 11-12):** Generates deterministic Maps of Content (MOCs) and builds the field-aware BM25 inverted index to support routing.
5.  **Grounded Q&A & Rendering (Tasks 13-14):** Routes queries using the search tool, synthesizes answers with strict citation enforcement, and renders durable artifacts (Markdown reports, Marp slides, matplotlib figures).
6.  **Filing & Linting (Tasks 15-16):** Files generated artifacts back into the wiki graph. Invokes the LLM to audit the wiki for contradictions and missing data, extracting minimal JSON-formatted patches.
7.  **Optional Fine-Tuning & Evaluation (Tasks 17-20):** Generates synthetic supervision data, splits it using exact integer arithmetic to prevent cluster leakage, executes LoRA fine-tuning with CPU-offloaded merging, and evaluates the system's Citation Correctness Rate (CCR).

## Core Components (Notebook Structure)

The notebook is structured as a logical pipeline with modular orchestrator functions for each of the 20 major tasks. All functions are self-contained, fully documented with strict type hints and comprehensive docstrings, and designed for professional-grade execution.

## Key Callable: `run_knowledge_base_system`

The project is designed around a single, top-level user-facing interface function:

-   **`run_knowledge_base_system`:** This apex orchestrator function runs the entire automated research pipeline from end-to-end. A single call to this function reproduces the entire workflow, managing static initialization, the dynamic operational cycle (ingest $\rightarrow$ compile $\rightarrow$ Q&A $\rightarrow$ file $\rightarrow$ lint), optional ML branches, and the final cryptographic reproducibility audit.

## Usage

The notebook provides a complete, step-by-step guide. The primary workflow is to execute the final cell, which demonstrates how to load the configuration, generate synthetic acquisition requests, and use the top-level orchestrator to execute the pipeline:

```python
import os
import yaml
import pathlib
from typing import Dict, Any

# 1. Load the master configuration from the YAML file.
# (Assumes config.yaml is in the working directory)
def load_study_configuration(filepath: str = "config.yaml") -> Dict[str, Any]:
    if not isinstance(filepath, str):
        raise TypeError(f"filepath must be a string, got {type(filepath)}.")
    try:
        with open(filepath, "r", encoding="utf-8") as file:
            config = yaml.safe_load(file)
        print(f"Successfully loaded configuration from {filepath}")
        return config
    except FileNotFoundError as e:
        print(f"CRITICAL ERROR: {filepath} not found in the working directory.")
        raise e

config = load_study_configuration("config.yaml")

# 2. Generate synthetic inputs (Example using generator provided in the notebook)
# In production, these would be driven by a UI or external event queue.
acq_reqs, user_queries, formats, eval_qs = generate_synthetic_pipeline_inputs()

# 3. Execute the entire knowledge base system.
if __name__ == "__main__":
    repo_root = pathlib.Path("./kb_workspace").resolve()
    
    if config:
        print("\nInitiating LLM Knowledge Base System...")
        
        system_artifacts = run_knowledge_base_system(
            raw_config=config,
            repo_root=repo_root,
            acquisition_requests=acq_reqs,
            queries=user_queries,
            requested_formats=formats,
            eval_questions=eval_qs
        )
        
        # 4. Access results
        print("\n" + "="*80)
        print("SYSTEM EXECUTION COMPLETE")
        print("="*80)
        
        pipeline_manifest = system_artifacts.get("pipeline_manifest")
        if pipeline_manifest:
            run_summary = pipeline_manifest.run_summary
            print(f"\n[Run Summary] ID: {run_summary['run_id']}")
            print(f"  - Wiki Files Written: {run_summary['wiki_files_written']}")
            print(f"  - Lint Issues Found: {run_summary['lint_issues_found_count']}")
            
        eval_bundle = system_artifacts.get("evaluation_bundle")
        if eval_bundle:
            baseline_report = eval_bundle.baseline_report
            print("\n[Baseline Evaluation KPIs]")
            print(f"  - Citation Coverage Rate (CCR): {baseline_report.ccr_coverage:.2%}")
            print(f"  - Citation Correctness Rate: {baseline_report.ccr_correct:.2%}")
            print(f"  - Total Hallucinations Detected: {baseline_report.total_hallucinations}")
```

## Output Structure

The pipeline returns a master dictionary containing two primary artifacts, serialized to disk under `logs/`:
-   **`EndToEndRunManifest`**: Contains the `run_summary` (counts of files written, lint issues, warnings), the `cycle_state` (cryptographic hashes of the graph before and after filing), and the optional `fine_tuning_manifest`.
-   **`FinalEvaluationBundle`**: Contains the `reproducibility_bundle` (deep order-invariant comparisons of repeated runs), the `baseline_report` (KPIs like Citation Correctness Rate), and the `acceptance_decision` for the fine-tuned model.

The filesystem substrate (`kb_workspace/`) will contain:
-   **`raw/`**: The immutable ingested sources and `manifest.yaml`.
-   **`wiki/`**: The compiled Markdown graph, including `sources/`, `concepts/`, `index/`, and `derived/` (reports, slides, figures).

## Customization

The pipeline is highly customizable via the `config.yaml` file. Users can modify study parameters such as:
-   **LLM Decoding Settings:** Adjust `temperature`, `top_p`, and `max_output_tokens` independently for compilation, Q&A, and linting tasks.
-   **Retrieval Parameters:** Modify the BM25 `field_boosts` (e.g., weighting tags higher than body text) or adjust the `snippet_chars` length.
-   **Linting Policies:** Define custom `what_counts_as_contradiction` rules or alter the `web_imputation_policy` to restrict authoritative domains.
-   **Fine-Tuning Hyperparameters:** Alter the LoRA `rank_r`, `learning_rate`, or `train_val_test_split_policy` fractions.

# "Paper"

Title: "*LLM Knowledge Bases*"

Authors: Andrej Karpathy

Social Media Submission Date: 2 April 2026

Link: https://x.com/karpathy/status/2039805659525644595

Full Post:

*LLM Knowledge Bases*

Something I'm finding very useful recently: using LLMs to build personal knowledge bases for various topics of research interest. In this way, a large fraction of my recent token throughput is going less into manipulating code, and more into manipulating knowledge (stored as markdown and images). The latest LLMs are quite good at it. So:

Data ingest:
I index source documents (articles, papers, repos, datasets, images, etc.) into a raw/ directory, then I use an LLM to incrementally "compile" a wiki, which is just a collection of .md files in a directory structure. The wiki includes summaries of all the data in raw/, backlinks, and then it categorizes data into concepts, writes articles for them, and links them all. To convert web articles into .md files I like to use the Obsidian Web Clipper extension, and then I also use a hotkey to download all the related images to local so that my LLM can easily reference them.

IDE:
I use Obsidian as the IDE "frontend" where I can view the raw data, the the compiled wiki, and the derived visualizations. Important to note that the LLM writes and maintains all of the data of the wiki, I rarely touch it directly. I've played with a few Obsidian plugins to render and view data in other ways (e.g. Marp for slides).

Q&A:
Where things get interesting is that once your wiki is big enough (e.g. mine on some recent research is ~100 articles and ~400K words), you can ask your LLM agent all kinds of complex questions against the wiki, and it will go off, research the answers, etc. I thought I had to reach for fancy RAG, but the LLM has been pretty good about auto-maintaining index files and brief summaries of all the documents and it reads all the important related data fairly easily at this ~small scale.

Output:
Instead of getting answers in text/terminal, I like to have it render markdown files for me, or slide shows (Marp format), or matplotlib images, all of which I then view again in Obsidian. You can imagine many other visual output formats depending on the query. Often, I end up "filing" the outputs back into the wiki to enhance it for further queries. So my own explorations and queries always "add up" in the knowledge base.

Linting:
I've run some LLM "health checks" over the wiki to e.g. find inconsistent data, impute missing data (with web searchers), find interesting connections for new article candidates, etc., to incrementally clean up the wiki and enhance its overall data integrity. The LLMs are quite good at suggesting further questions to ask and look into.

Extra tools:
I find myself developing additional tools to process the data, e.g. I vibe coded a small and naive search engine over the wiki, which I both use directly (in a web ui), but more often I want to hand it off to an LLM via CLI as a tool for larger queries.

Further explorations:
As the repo grows, the natural desire is to also think about synthetic data generation + finetuning to have your LLM "know" the data in its weights instead of just context windows.

TLDR: raw data from a given number of sources is collected, then compiled by an LLM into a .md wiki, then operated on by various CLIs by the LLM to do Q&A and to incrementally enhance the wiki, and all of it viewable in Obsidian. You rarely ever write or edit the wiki manually, it's the domain of the LLM. I think there is room here for an incredible new product instead of a hacky collection of scripts.


# Summary

## Central thesis / motivation
Karpathy reports that he is increasingly using modern LLMs not primarily for code generation, but as **knowledge manipulation engines** to construct and maintain **personal research knowledge bases**. The knowledge base is represented as a **local, file-based wiki** (Markdown + images), enabling iterative accumulation of structured understanding rather than ephemeral chat outputs.

## Data ingest: building a raw corpus
He begins with a curated but heterogeneous research corpus—e.g.:

- articles and web pages  
- academic papers  
- repositories / codebases  
- datasets  
- images and figures  

These items are stored in a `raw/` directory. Web pages are converted into Markdown using **Obsidian Web Clipper**, and related images are downloaded locally so the LLM can reference them.

**Implicit design rationale (inference):** localizing all artifacts (text + images) reduces dependency on link rot, paywalls, and network access, and improves reproducibility of subsequent summarization and Q&A.


## “Compilation”: LLM-driven transformation into a wiki
He describes an incremental “compile” step: an LLM takes the raw corpus and produces a **structured wiki**, implemented as a directory tree of `.md` files. The wiki includes:

- summaries of source documents  
- backlinks (cross-references between notes)  
- categorization into concepts/topics  
- concept articles that synthesize multiple sources  
- linking structure across concepts and sources  

Crucially: he states the **LLM writes and maintains the wiki**, and he “rarely touches it directly.”

**Implicit requirements (inference):**
- Some stable conventions for filenames, headings, and linking so the agent can maintain consistency over time.
- Lightweight metadata (e.g., source URL, ingest date, tags) to support indexing and traceability.
- An incremental update protocol (e.g., “only recompile affected nodes”) to avoid rewriting the entire wiki each time.

## Obsidian as the “IDE frontend”
Obsidian functions as the primary interface to:

- browse `raw/` documents  
- view and navigate the compiled wiki  
- render derived artifacts (slides, images, etc.)  

He experiments with plugins such as **Marp** for slides.

**Interpretation:** Obsidian is used as a *human-facing IDE* for inspection and presentation, while the LLM acts as the *primary author/maintainer* of the knowledge base.

## Q&A over the wiki without “fancy RAG” (at small scale)
Once the wiki becomes sufficiently large (he gives an example: ~100 articles, ~400K words), he can ask complex questions. The LLM “goes off” to research answers by consulting relevant wiki materials.

He remarks that he expected to require sophisticated retrieval-augmented generation (RAG), but in this regime the LLM performs well by:

- maintaining index files  
- keeping brief summaries of documents  
- reading key related artifacts as needed  

**Key boundary condition he implies:** this works because the knowledge base is still at a “small scale” where hierarchical summaries and selective reading remain tractable.

## Outputs are artifacts, not just chat responses
Rather than returning answers only in a terminal/chat transcript, he prefers that the agent produce **first-class, viewable artifacts**:

- Markdown reports  
- Marp slide decks  
- matplotlib-generated images  

These are then opened in Obsidian.

**Core workflow consequence:** answers become *durable knowledge objects* that can be re-used, linked, and improved.

## “Filing” outputs back into the knowledge base
He frequently “files” generated outputs back into the wiki, so exploration is **cumulative**: each query and synthesis potentially expands and strengthens the knowledge base for future queries.

**Econometric analogy (inference):** this resembles building a growing panel dataset—each new query adds observations/features, improving future inference capacity, provided drift and inconsistencies are controlled.

## Linting / health checks for integrity
He runs LLM-based “health checks” to:

- detect inconsistent data  
- impute missing data (sometimes with web search tools)  
- identify interesting cross-links and candidate new articles  
- generally improve integrity and coverage  

He notes that LLMs can propose additional questions to pursue, i.e., they can act as **active research assistants** suggesting “next experiments.”

**Implicit issue (inference):** any imputation or web-augmented filling requires provenance tracking; otherwise the wiki risks accumulating plausible but unverified statements.

## Extra tools: bespoke search and CLI tools
He develops additional tooling, including a “small and naive” search engine over the wiki, used both directly (web UI) and as a tool invoked by an LLM via CLI for larger queries.

**Interpretation:** this is an agent-tool ecosystem: the LLM is not operating alone; it is orchestrating utilities that expose the wiki as a queryable substrate.

## Future direction: synthetic data + fine-tuning
As the repo grows, he anticipates interest in:

- generating synthetic instruction/data from the wiki  
- fine-tuning models so knowledge migrates from “context” into “weights”  

This is framed as a natural escalation path when context-window-based consultation becomes limiting.

**Caution (inference):** fine-tuning trades easy editability for model-internalization; correcting errors becomes harder unless you maintain strong provenance and retraining pipelines.

## Concluding claim: product opportunity
He characterizes his current setup as “a hacky collection of scripts” and suggests there is room for a compelling product that formalizes:

- ingest → compilation → indexing → Q&A → artifact generation → integrity checks

**Methodology:**
1.  **File I/O:** Open `config.yaml` in read mode using strict UTF-8 encoding to prevent byte corruption.
2.  **Parsing:** Use `yaml.safe_load` to convert the YAML structure into a nested Python dictionary, neutralizing arbitrary code execution vulnerabilities.
3.  **Validation:** Catch `FileNotFoundError` and `yaml.YAMLError` to ensure the pipeline fails gracefully if the contract is missing or malformed.

    handler = logging.StreamHandler()
    handler.setFormatter(logging.Formatter("%(asctime)s | %(levelname)-8s | %(message)s"))
    logger.addHandler(handler)
    logger.setLevel(logging.INFO)

def load_study_configuration(filepath: str = "config.yaml") -> Dict[str, Any]:
    """
    Loads the study configuration parameters from a YAML file into a Python dictionary.

    Purpose:
        To ingest the deterministic hyperparameters, directory schemas, and
        evaluation metrics defined in the external configuration file. This ensures
        reproducibility by separating code from configuration.

    Inputs:
        filepath (str): The relative or absolute path to the YAML configuration file.
                        Default is "config.yaml".

    Processes:
        1.  File Access: Attempts to open the specified file in read mode with UTF-8 encoding.
        2.  Parsing: Uses PyYAML's safe_load to parse the YAML structure securely.
        3.  Validation: Catches file existence errors and parsing errors.

    Outputs:
        Dict[str, Any]: A nested dictionary containing the study configuration.
                        Returns an empty dictionary if the file is not found.

    Raises:
        TypeError: If filepath is not a string.
        yaml.YAMLError: If the file contains invalid YAML syntax.
    """
    # Validate input type
    if not isinstance(filepath, str):
        raise TypeError(f"filepath must be a string, got {type(filepath).__name__}.")

    try:
        # Open the file stream with strict encoding
        with open(filepath, "r", encoding="utf-8") as file:
            # Parse the YAML content safely
            loaded_config = yaml.safe_load(file)

        # Log success to console
        logger.info(f"Successfully loaded configuration from {filepath}")

        return loaded_config or {}

    except FileNotFoundError:
        logger.error(f"Warning: {filepath} not found. Please ensure the file exists.")
        return {}
    except yaml.YAMLError as e:
        # Handle parsing errors explicitly
        logger.error(f"Error parsing YAML file {filepath}: {e}")
        raise

# Load the configuration into the 'config' dictionary
# Note: Ensure 'config.yaml' is in your working directory.
config = load_study_configuration()

### **Summary of the Execution Flow**

1.  **Static Initialization:** The pipeline validates the `config` dictionary (deep schema validation), verifies the host environment, materializes the `raw/` and `wiki/` directories, and registers the hashed prompt templates.
2.  **Data Ingestion & Normalization:** The synthetic `acquisition_requests` are downloaded/cloned into `raw/`, validated via magic-byte inspection, and normalized into UTF-8 derivatives.
3.  **Incremental Compilation:** The system computes cryptographic hashes, detects deltas, and invokes the LLM to compile the derivatives into a structured Markdown wiki graph.
4.  **Index Refresh & Search:** The system generates deterministic MOCs and builds a BM25 inverted index over the compiled wiki.
5.  **Grounded Q&A & Rendering:** The system routes the synthetic `queries`, synthesizes answers with strict citation enforcement, and renders the requested `markdown` and `marp` artifacts.
6.  **Filing & Linting:** The artifacts are filed back into the wiki graph ($O(1)$ update), and the LLM lints the wiki for contradictions and missing data.
7.  **Evaluation:** The system evaluates the baseline (and optionally fine-tuned) model against the synthetic `eval_questions`, computing the Citation Correctness Rate and outputting the final audit report.

````python
# Task 0: GPT 5.2 Chat Utilities

# ===========================================================================================
# Task 0: GPT 5.2 Chat Utilities
# ===========================================================================================

# Define the exact literal type for supported GPT-5.2 reasoning-effort values.
ReasoningEffort = Literal["none", "low", "medium", "high", "xhigh"]

# Define the allowed image-suffix to MIME-type mapping used for multimodal
# input validation and data-URI construction.
_SUFFIX_TO_MIME: Dict[str, str] = {
    ".jpg": "image/jpeg",
    ".jpeg": "image/jpeg",
    ".png": "image/png",
    ".gif": "image/gif",
    ".webp": "image/webp",
}

# Define the helper that constructs a text-only Responses API input payload.
def build_text_input(
    system_prompt: str,
    user_text: str,
) -> List[Dict[str, str]]:
    """
    Construct a minimal two-message structured input list for the Responses API.

    Purpose:
        Build a text-only message payload using the GPT-5.x Responses API
        message structure, with:
        1. a developer message carrying system-level constraints, and
        2. a user message carrying the task or question.

    Inputs:
        system_prompt:
            Developer-level instruction string that encodes behavioral
            constraints, grounding rules, and output contracts.
        user_text:
            User-facing query or instruction string.

    Processes:
        1. Validate that both inputs are strings.
        2. Create a two-element list of message dictionaries.
        3. Assign the first message the ``developer`` role.
        4. Assign the second message the ``user`` role.

    Outputs:
        List[Dict[str, str]]:
            Structured Responses API input payload containing exactly two
            message objects.

    Raises:
        TypeError:
            Raised if either input is not a string.

    Notes:
        For GPT-5.x reasoning models in the Responses API, ``developer`` is the
        correct high-trust role for system-level instructions. This callable is
        a payload-construction primitive and does not itself issue a model call.
    """
    # Validate that the system prompt is a string.
    if not isinstance(system_prompt, str):
        # Raise a type error because Responses API message construction requires
        # textual string content.
        raise TypeError(
            "build_text_input expected 'system_prompt' to be a string."
        )

    # Validate that the user text is a string.
    if not isinstance(user_text, str):
        # Raise a type error because Responses API message construction requires
        # textual string content.
        raise TypeError(
            "build_text_input expected 'user_text' to be a string."
        )

    # Return the minimal two-message structured input list for the Responses API.
    return [
        # Insert the developer-role message carrying system-level constraints.
        {"role": "developer", "content": system_prompt},
        # Insert the user-role message carrying the task-specific content.
        {"role": "user", "content": user_text},
    ]


# Define the helper that constructs a multimodal Responses API input payload
# containing both text and a local image.
def build_image_input(
    system_prompt: str,
    user_text: str,
    image_path: pathlib.Path,
) -> List[Dict[str, Any]]:
    """
    Construct a multimodal Responses API input list with an embedded
    base64-encoded local image.

    Purpose:
        Build a developer-plus-user payload where the user message contains:
        1. a text instruction part, and
        2. an image part encoded as a local base64 data URI.

    Inputs:
        system_prompt:
            Developer-level instruction string encoding system constraints.
        user_text:
            User-facing textual instruction about the image.
        image_path:
            Local filesystem path to the image file to embed.

    Processes:
        1. Validate input types.
        2. Resolve the image path to a pathlib.Path if needed.
        3. Read the raw image bytes.
        4. Base64-encode the bytes.
        5. Infer the MIME type from the file suffix.
        6. Raise if the image format is unsupported.
        7. Construct a data URI of the form:
           ``data:<mime_type>;base64,<base64_string>``.
        8. Build the structured developer/user message list required by the
           Responses API.

    Outputs:
        List[Dict[str, Any]]:
            Structured multimodal Responses API input payload.

    Raises:
        TypeError:
            Raised if the textual inputs are not strings or the image path is
            not path-like.
        FileNotFoundError:
            Propagated if the image file does not exist.
        ValueError:
            Raised if the image suffix is unsupported.

    Notes:
        The Responses API accepts images either as public URLs or as base64
        data URIs. This callable implements the local-file base64 route.
    """
    # Validate that the system prompt is a string.
    if not isinstance(system_prompt, str):
        # Raise a type error because Responses API message construction requires
        # textual string content.
        raise TypeError(
            "build_image_input expected 'system_prompt' to be a string."
        )

    # Validate that the user text is a string.
    if not isinstance(user_text, str):
        # Raise a type error because Responses API message construction requires
        # textual string content.
        raise TypeError(
            "build_image_input expected 'user_text' to be a string."
        )

    # Validate that the image path is path-like.
    if not isinstance(image_path, pathlib.Path):
        # Raise a type error because binary file access requires a pathlib.Path.
        raise TypeError(
            "build_image_input expected 'image_path' to be a pathlib.Path "
            "instance."
        )

    # Read the raw bytes of the local image file.
    raw_bytes: bytes = image_path.read_bytes()

    # Encode the image bytes into a base64 UTF-8 string.
    b64_image: str = base64.standard_b64encode(raw_bytes).decode("utf-8")

    # Define the local suffix-to-MIME mapping exactly as in the original logic.
    suffix_to_mime: Dict[str, str] = {
        ".jpg": "image/jpeg",
        ".jpeg": "image/jpeg",
        ".png": "image/png",
        ".gif": "image/gif",
        ".webp": "image/webp",
    }

    # Infer the MIME type from the lowercase file suffix.
    mime_type: str = suffix_to_mime.get(image_path.suffix.lower(), "")

    # Reject unsupported image formats immediately.
    if not mime_type:
        # Raise a value error that names both the offending suffix and the
        # supported set.
        raise ValueError(
            f"Unsupported image format: {image_path.suffix}. "
            f"Supported: {list(suffix_to_mime.keys())}"
        )

    # Construct the inline base64 data URI accepted by the Responses API.
    data_uri: str = f"data:{mime_type};base64,{b64_image}"

    # Return the structured developer-plus-user multimodal input payload.
    return [
        # Add the developer-role message carrying system-level instructions.
        {"role": "developer", "content": system_prompt},
        # Add the user-role message carrying both text and image content parts.
        {
            "role": "user",
            "content": [
                # Insert the textual instruction content part.
                {"type": "input_text", "text": user_text},
                # Insert the inline image content part encoded as a data URI.
                {"type": "input_image", "image_url": data_uri},
            ],
        },
    ]


# Define the primitive text-extraction wrapper around the Responses API for
# text-only or already-constructed multimodal message arrays.
def call_gpt52(
    client: OpenAI,
    input_messages: List[Dict[str, Any]],
    reasoning_effort: str = "high",
    max_output_tokens: int = 8_000,
) -> str:
    """
    Issue a single Responses API call to ``gpt-5.2`` and return the text output.

    Purpose:
        Provide a thin, production-oriented wrapper over
        ``client.responses.create`` for the GPT-5.2 reasoning model.

    Inputs:
        client:
            Instantiated OpenAI SDK client.
        input_messages:
            Structured Responses API message array created by helper builders.
        reasoning_effort:
            GPT-5.2 reasoning depth control. Valid semantic values are
            ``none``, ``low``, ``medium``, ``high``, and ``xhigh``.
        max_output_tokens:
            Hard ceiling on generated output tokens.

    Processes:
        1. Validate input types.
        2. Issue a Responses API call using:
           - model = ``gpt-5.2``
           - input = structured message array
           - reasoning = ``{"effort": reasoning_effort}``
           - max_output_tokens = requested cap
        3. Extract the model's textual output via ``response.output_text``.
        4. Return the extracted text.

    Outputs:
        str:
            The textual model response aggregated by the Responses API.

    Raises:
        TypeError:
            Raised if inputs are structurally invalid.
        openai.APIError:
            Propagated from the SDK for authentication, rate-limit, or API
            failures.

    Notes:
        This callable intentionally omits ``temperature``, ``top_p``,
        ``frequency_penalty``, and ``presence_penalty`` because GPT-5.2 is a
        reasoning model and those controls are not the correct levers here.
    """
    # Validate that the client is an OpenAI client instance.
    if not isinstance(client, OpenAI):
        # Raise a type error because model invocation requires a valid SDK
        # client object.
        raise TypeError(
            "call_gpt52 expected 'client' to be an OpenAI instance."
        )

    # Validate that the input_messages payload is a list.
    if not isinstance(input_messages, list):
        # Raise a type error because the Responses API input payload must be
        # list-structured here.
        raise TypeError(
            "call_gpt52 expected 'input_messages' to be a list of message "
            "dictionaries."
        )

    # Validate that the reasoning effort is a string.
    if not isinstance(reasoning_effort, str):
        # Raise a type error because reasoning settings must be textual.
        raise TypeError(
            "call_gpt52 expected 'reasoning_effort' to be a string."
        )

    # Validate that the output-token cap is an integer.
    if not isinstance(max_output_tokens, int):
        # Raise a type error because token budgeting requires an integer.
        raise TypeError(
            "call_gpt52 expected 'max_output_tokens' to be an integer."
        )

    # Issue the Responses API call using the verified GPT-5.2 model string.
    raw_response: OpenAIResponse = client.responses.create(
        # Specify the existing GPT-5.2 Thinking model string.
        model="gpt-5.2",
        # Pass the structured message array.
        input=input_messages,
        # Express reasoning depth via the reasoning.effort parameter.
        reasoning={"effort": reasoning_effort},
        # Enforce the output-token ceiling.
        max_output_tokens=max_output_tokens,
    )

    # Extract the textual output using the canonical Responses API accessor.
    output_text: str = raw_response.output_text

    # Return the extracted output text.
    return output_text


# Define the primitive Responses API wrapper with the web_search tool enabled.
def call_gpt52_with_web_search(
    client: OpenAI,
    input_messages: List[Dict[str, Any]],
    reasoning_effort: str = "high",
    max_output_tokens: int = 8_000,
    allowed_domains: Optional[List[str]] = None,
) -> Tuple[str, List[Dict[str, str]]]:
    """
    Issue a Responses API call to ``gpt-5.2`` with the ``web_search`` tool
    enabled and return both the response text and source citations.

    Purpose:
        Provide a production-oriented wrapper over ``client.responses.create``
        that injects the ``web_search`` tool into the call, optionally
        constrained to authoritative domains.

    Inputs:
        client:
            Instantiated OpenAI SDK client.
        input_messages:
            Structured Responses API message array.
        reasoning_effort:
            GPT-5.2 reasoning depth control.
        max_output_tokens:
            Hard ceiling on generated output tokens.
        allowed_domains:
            Optional list of allowed search domains. Domains must omit the
            ``http://`` or ``https://`` prefix.

    Processes:
        1. Validate input types.
        2. Construct the web search tool configuration:
           ``{"type": "web_search"}``.
        3. Optionally inject domain filters if ``allowed_domains`` is provided.
        4. Issue the Responses API call with the tool in ``tools=[...]``.
        5. Extract the response text.
        6. Iterate over all output items and content parts.
        7. Extract any ``url_citation`` annotations into provenance records.
        8. Return the text plus the citations list.

    Outputs:
        Tuple[str, List[Dict[str, str]]]:
            1. Textual model response.
            2. Ordered list of citation dictionaries, each with ``title`` and
               ``url`` keys.

    Raises:
        TypeError:
            Raised if inputs are structurally invalid.
        openai.APIError:
            Propagated from the SDK for authentication, rate-limit, or API
            failures.

    Notes:
        This callable implements the correct architecture for search-augmented
        GPT-5.2 calls: web search is a tool, not a separate model.
    """
    # Validate that the client is an OpenAI client instance.
    if not isinstance(client, OpenAI):
        # Raise a type error because model invocation requires a valid SDK
        # client object.
        raise TypeError(
            "call_gpt52_with_web_search expected 'client' to be an OpenAI "
            "instance."
        )

    # Validate that the input_messages payload is a list.
    if not isinstance(input_messages, list):
        # Raise a type error because the Responses API input payload must be
        # list-structured here.
        raise TypeError(
            "call_gpt52_with_web_search expected 'input_messages' to be a list "
            "of message dictionaries."
        )

    # Validate that the reasoning effort is a string.
    if not isinstance(reasoning_effort, str):
        # Raise a type error because reasoning settings must be textual.
        raise TypeError(
            "call_gpt52_with_web_search expected 'reasoning_effort' to be a "
            "string."
        )

    # Validate that the output-token cap is an integer.
    if not isinstance(max_output_tokens, int):
        # Raise a type error because token budgeting requires an integer.
        raise TypeError(
            "call_gpt52_with_web_search expected 'max_output_tokens' to be an "
            "integer."
        )

    # Validate allowed_domains when provided.
    if allowed_domains is not None and not isinstance(allowed_domains, list):
        # Raise a type error because domain filtering requires a list of domain
        # strings.
        raise TypeError(
            "call_gpt52_with_web_search expected 'allowed_domains' to be a "
            "list of strings or None."
        )

    # Construct the stable web_search tool configuration.
    web_search_tool: Dict[str, Any] = {"type": "web_search"}

    # Inject domain filters when allowed_domains is provided.
    if allowed_domains:
        # Add the allowed_domains filter to enforce provenance constraints.
        web_search_tool["filters"] = {
            "allowed_domains": allowed_domains
        }

    # Issue the Responses API call with the web search tool enabled.
    raw_response: OpenAIResponse = client.responses.create(
        # Specify the existing GPT-5.2 Thinking model string.
        model="gpt-5.2",
        # Pass the structured message array.
        input=input_messages,
        # Inject the web search tool into the call.
        tools=[web_search_tool],
        # Express reasoning depth via the reasoning.effort parameter.
        reasoning={"effort": reasoning_effort},
        # Enforce the output-token ceiling.
        max_output_tokens=max_output_tokens,
    )

    # Extract the primary textual output using the canonical accessor.
    output_text: str = raw_response.output_text

    # Initialize the ordered citation list.
    citations: List[Dict[str, str]] = []

    # Iterate over all output items produced by the response.
    for output_item in raw_response.output:
        # Only message-like output items carry content parts.
        if hasattr(output_item, "content"):
            # Iterate over each content part in the output message.
            for content_part in output_item.content:
                # Only content parts with annotations can carry URL citations.
                if hasattr(content_part, "annotations"):
                    # Iterate over each annotation attached to the content part.
                    for annotation in content_part.annotations:
                        # Capture only URL citation annotations as provenance
                        # records.
                        if annotation.type == "url_citation":
                            citations.append(
                                {
                                    "title": annotation.title,
                                    "url": annotation.url,
                                }
                            )

    # Return both the textual response and the extracted citations.
    return output_text, citations


# Define the fused production-grade GPT-5.2 query callable.
def query_gpt52(
    *,
    prompt: str,
    system_prompt: str = (
        "You are a rigorous research assistant. "
        "Do not fabricate facts. "
        "Cite sources explicitly. "
        "If you do not know, say so."
    ),
    image_path: Optional[pathlib.Path] = None,
    enable_web_search: bool = False,
    allowed_search_domains: Optional[List[str]] = None,
    reasoning_effort: ReasoningEffort = "high",
    max_output_tokens: int = 8_000,
    api_key: Optional[str] = None,
    log_run_manifest: bool = True,
) -> Dict[str, object]:
    """
    Issue a single, governed query to OpenAI's ``gpt-5.2`` reasoning model via
    the Responses API, with optional image attachment and optional web search.

    Purpose:
        Provide the primary production-grade interface for a continuously
        compounding, locally grounded research and decision memory system. This
        callable encapsulates:

        1. Correct model ID usage:
           - ``gpt-5.2``, not ``gpt-5.2-high``.
        2. Correct control levers for a reasoning model:
           - ``reasoning={"effort": ...}``
           - ``max_output_tokens``
           - prompt construction
        3. Optional multimodal local-image attachment via base64 data URI.
        4. Optional web search via the ``web_search`` tool.
        5. Provenance capture, including:
           - citations,
           - image SHA-256 hash,
           - model identifier,
           - reasoning effort,
           - token usage,
           - finish reason.

    Inputs:
        prompt:
            User-facing natural language query or instruction.
        system_prompt:
            Developer-role instruction encoding behavioral constraints,
            anti-fabrication requirements, and output contracts.
        image_path:
            Optional local image path for multimodal analysis.
        enable_web_search:
            Whether to inject the ``web_search`` tool.
        allowed_search_domains:
            Optional authoritative domains allowed for web search.
        reasoning_effort:
            Reasoning-depth control for GPT-5.2. Supported values are:
            ``none``, ``low``, ``medium``, ``high``, and ``xhigh``.
        max_output_tokens:
            Hard ceiling on generated output tokens.
        api_key:
            Optional explicit API key. If omitted, the callable falls back to
            the ``OPENAI_API_KEY`` environment variable.
        log_run_manifest:
            Whether to emit the structured run manifest at DEBUG level.

    Processes:
        1. Validate input types.
        2. Resolve the API key from the argument or environment.
        3. Instantiate the OpenAI client.
        4. Construct the developer message.
        5. Construct the user content parts.
        6. Optionally attach a local image:
           - read bytes,
           - compute SHA-256,
           - infer MIME type,
           - base64-encode,
           - build a data URI,
           - append an ``input_image`` content part.
        7. Assemble the final Responses API input array.
        8. Optionally construct the web search tool configuration.
        9. Issue the Responses API call using:
           - ``model="gpt-5.2"``
           - ``reasoning={"effort": reasoning_effort}``
           - ``max_output_tokens=max_output_tokens``
           - optional ``tools=[{"type": "web_search", ...}]``
        10. Extract:
            - textual output,
            - citations from output annotations,
            - token usage,
            - finish reason.
        11. Assemble and optionally log the run manifest.
        12. Return the run manifest.

    Outputs:
        Dict[str, object]:
            Structured run manifest containing:
            - ``response_text``:
              Textual model output.
            - ``citations``:
              Ordered list of provenance citation dictionaries.
            - ``model``:
              Model string used.
            - ``reasoning_effort``:
              Reasoning depth used.
            - ``usage``:
              Token usage dictionary.
            - ``image_sha256``:
              SHA-256 hash of the attached image, or None.
            - ``finish_reason``:
              API-reported terminal response status.

    Raises:
        TypeError:
            Raised if input types are invalid.
        EnvironmentError:
            Raised if no API key can be resolved.
        FileNotFoundError:
            Raised if an image path is provided but the file does not exist.
        ValueError:
            Raised if an unsupported image format is supplied.
        openai.APIError:
            Propagated from the SDK for authentication, rate-limit, or API
            failures.

    Notes:
        This callable intentionally omits:
        - ``temperature``
        - ``top_p``
        - ``frequency_penalty``
        - ``presence_penalty``
        - ``seed``

        because GPT-5.2 is a reasoning model and those are not the correct
        production controls here.

        This callable uses the Responses API rather than Chat Completions
        because that is the recommended surface for GPT-5.x reasoning models.
    """
    # Validate that the user-facing prompt is a string.
    if not isinstance(prompt, str):
        # Raise a type error because prompt construction requires a string.
        raise TypeError(
            "query_gpt52 expected 'prompt' to be a string."
        )

    # Validate that the developer/system prompt is a string.
    if not isinstance(system_prompt, str):
        # Raise a type error because developer message construction requires a
        # string.
        raise TypeError(
            "query_gpt52 expected 'system_prompt' to be a string."
        )

    # Validate that image_path is either None or a pathlib.Path.
    if image_path is not None and not isinstance(image_path, pathlib.Path):
        # Raise a type error because image handling requires pathlib.Path.
        raise TypeError(
            "query_gpt52 expected 'image_path' to be a pathlib.Path or None."
        )

    # Validate that enable_web_search is a boolean.
    if not isinstance(enable_web_search, bool):
        # Raise a type error because tool selection requires a boolean flag.
        raise TypeError(
            "query_gpt52 expected 'enable_web_search' to be a boolean."
        )

    # Validate that allowed_search_domains is either None or a list.
    if (
        allowed_search_domains is not None
        and not isinstance(allowed_search_domains, list)
    ):
        # Raise a type error because domain filtering requires a list of domain
        # strings.
        raise TypeError(
            "query_gpt52 expected 'allowed_search_domains' to be a list of "
            "strings or None."
        )

    # Validate that reasoning_effort is a string.
    if not isinstance(reasoning_effort, str):
        # Raise a type error because the reasoning-effort parameter is textual.
        raise TypeError(
            "query_gpt52 expected 'reasoning_effort' to be a string."
        )

    # Validate that max_output_tokens is an integer.
    if not isinstance(max_output_tokens, int):
        # Raise a type error because token budgeting requires an integer.
        raise TypeError(
            "query_gpt52 expected 'max_output_tokens' to be an integer."
        )

    # Validate that api_key is either None or a string.
    if api_key is not None and not isinstance(api_key, str):
        # Raise a type error because credentials must be text.
        raise TypeError(
            "query_gpt52 expected 'api_key' to be a string or None."
        )

    # Validate that log_run_manifest is a boolean.
    if not isinstance(log_run_manifest, bool):
        # Raise a type error because logging control requires a boolean.
        raise TypeError(
            "query_gpt52 expected 'log_run_manifest' to be a boolean."
        )

    # Resolve the API key from the explicit argument or the environment.
    resolved_key: str = api_key or os.environ.get("OPENAI_API_KEY", "")

    # Raise immediately if no API key is available.
    if not resolved_key:
        # Raise an environment error because authenticated API access is
        # mandatory.
        raise EnvironmentError(
            "No OpenAI API key found. "
            "Set the OPENAI_API_KEY environment variable or pass `api_key` "
            "explicitly."
        )

    # Instantiate the OpenAI client with the resolved API key.
    client: OpenAI = OpenAI(api_key=resolved_key)

    # Build the developer-role message carrying system-level constraints.
    developer_message: Dict[str, str] = {
        "role": "developer",
        "content": system_prompt,
    }

    # Initialize the user content parts with the textual prompt.
    user_content_parts: List[Dict[str, str]] = [
        {"type": "input_text", "text": prompt},
    ]

    # Initialize the image provenance hash to None.
    image_sha256: Optional[str] = None

    # Optionally attach a local image.
    if image_path is not None:
        # Resolve the image path to an absolute canonical path.
        image_path = pathlib.Path(image_path).resolve()

        # Reject missing or non-file image paths immediately.
        if not image_path.is_file():
            # Raise a file-not-found error because multimodal attachment
            # requires a real file.
            raise FileNotFoundError(
                f"Image file not found: {image_path}"
            )

        # Resolve the image MIME type from the allowed suffix mapping.
        mime_type: str = _SUFFIX_TO_MIME.get(image_path.suffix.lower(), "")

        # Reject unsupported image suffixes immediately.
        if not mime_type:
            # Raise a value error because only supported image formats can be
            # embedded as Responses API image parts.
            raise ValueError(
                f"Unsupported image format: '{image_path.suffix}'. "
                f"Supported: {list(_SUFFIX_TO_MIME.keys())}"
            )

        # Read the raw image bytes from disk.
        raw_bytes: bytes = image_path.read_bytes()

        # Compute the SHA-256 hash of the raw image bytes for provenance
        # tracking.
        image_sha256 = hashlib.sha256(raw_bytes).hexdigest()

        # Encode the image bytes as a base64 UTF-8 string.
        b64_str: str = base64.standard_b64encode(raw_bytes).decode("utf-8")

        # Construct the inline data URI required by the Responses API.
        data_uri: str = f"data:{mime_type};base64,{b64_str}"

        # Append the image content part to the user message content array.
        user_content_parts.append(
            {"type": "input_image", "image_url": data_uri}
        )

    # Assemble the full input message array.
    input_messages: List[Dict[str, Any]] = [
        # Add the developer-role constraint message.
        developer_message,
        # Add the user-role message, using a plain string when text-only and a
        # content-part list when multimodal.
        {
            "role": "user",
            "content": (
                user_content_parts[0]["text"]
                if len(user_content_parts) == 1
                else user_content_parts
            ),
        },
    ]

    # Initialize the tool array as empty.
    tools: List[Dict[str, Any]] = []

    # Optionally enable the web search tool.
    if enable_web_search:
        # Construct the stable web_search tool specification.
        web_search_tool: Dict[str, Any] = {"type": "web_search"}

        # Optionally inject domain filters when provided.
        if allowed_search_domains:
            web_search_tool["filters"] = {
                "allowed_domains": allowed_search_domains
            }

        # Append the tool configuration to the tools list.
        tools.append(web_search_tool)

    # Emit a DEBUG-level diagnostic before issuing the API call.
    logger.debug(
        "Issuing gpt-5.2 Responses API call | model=gpt-5.2 | "
        "reasoning_effort=%s | max_output_tokens=%d | "
        "web_search=%s | image_attached=%s",
        reasoning_effort,
        max_output_tokens,
        enable_web_search,
        image_path is not None,
    )

    # Execute the Responses API call.
    raw_response: OpenAIResponse = client.responses.create(
        # Specify the verified GPT-5.2 Thinking model string.
        model="gpt-5.2",
        # Pass the structured input messages.
        input=input_messages,
        # Express reasoning depth through reasoning.effort.
        reasoning={"effort": reasoning_effort},
        # Enforce the output-token ceiling.
        max_output_tokens=max_output_tokens,
        # Inject tools only when the tools list is non-empty.
        **({"tools": tools} if tools else {}),
    )

    # Extract the canonical textual output from the response.
    response_text: str = raw_response.output_text

    # Initialize the ordered citation list.
    citations: List[Dict[str, str]] = []

    # Iterate over all output items emitted by the response.
    for output_item in raw_response.output:
        # Only output items with content parts can carry annotations.
        if hasattr(output_item, "content"):
            # Iterate over each content part in the output item.
            for content_part in output_item.content:
                # Only content parts with annotations can carry citations.
                if hasattr(content_part, "annotations"):
                    # Iterate over each annotation.
                    for annotation in content_part.annotations:
                        # Capture only URL citation annotations.
                        if annotation.type == "url_citation":
                            citations.append(
                                {
                                    "title": annotation.title,
                                    "url": annotation.url,
                                }
                            )

    # Initialize the usage dictionary.
    usage: Dict[str, int] = {}

    # Extract token-usage statistics when the response provides them.
    if raw_response.usage is not None:
        usage = {
            "input_tokens": raw_response.usage.input_tokens,
            "output_tokens": raw_response.usage.output_tokens,
            "total_tokens": raw_response.usage.total_tokens,
        }

    # Resolve the finish reason from the response status field.
    finish_reason: str = getattr(raw_response, "status", "unknown")

    # Assemble the structured run manifest for audit and reproducibility.
    run_manifest: Dict[str, object] = {
        "response_text": response_text,
        "citations": citations,
        "model": "gpt-5.2",
        "reasoning_effort": reasoning_effort,
        "usage": usage,
        "image_sha256": image_sha256,
        "finish_reason": finish_reason,
    }

    # Emit the run manifest to the DEBUG log when requested.
    if log_run_manifest:
        logger.debug("Run manifest: %s", run_manifest)

    # Return the fully structured run manifest.
    return run_manifest
````

````python
# Task 1: Validate the appropriateness, completeness, and quality of the study configuration

# ===========================================================================================
# Task 1: Validate the appropriateness, completeness, and quality of the study configuration
# ===========================================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 1: Validate the structural integrity of the configuration dictionary
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass to hold the structural validation report.
@dataclass(frozen=True)
class StructuralValidationReport:
    """
    Immutable report describing whether the supplied study configuration
    satisfies the required deep structural contract.

    Purpose:
        Record the outcome of the recursive structural integrity check over the
        configuration dictionary, ensuring no nested malformations exist and
        that all required sections appear at the correct dot-paths.

    Inputs:
        This dataclass is populated by ``validate_structural_integrity``.

    Processes:
        No computation is performed here. It is a passive immutable report.

    Outputs:
        missing_required_sections:
            Ordered list of missing required dot-paths.
        null_but_not_permitted_sections:
            Ordered list of present dot-paths that are null, empty, or
            type-invalid where that is not permitted.
        unexpected_sections:
            Ordered list of unexpected dot-paths not recognized by the formal
            schema.
        structural_status:
            Boolean pass/fail indicator of overall deep structural validity.
    """

    # Store the ordered list of missing required dot-paths.
    missing_required_sections: List[str]

    # Store the ordered list of invalid null, empty, or type-mismatched paths.
    null_but_not_permitted_sections: List[str]

    # Store the ordered list of unexpected dot-paths.
    unexpected_sections: List[str]

    # Store the overall structural pass/fail result.
    structural_status: bool


# Define the function to validate structural integrity.
def validate_structural_integrity(
    config: Dict[str, Any],
) -> Tuple[StructuralValidationReport, Dict[str, Any]]:
    """
    Recursively validate the structural integrity of the study configuration.

    Purpose:
        Enforce the deep configuration schema required by the study's execution
        contract. The function traverses the configuration recursively,
        verifying key existence, nested type correctness, and nullability
        constraints at all depths.

    Inputs:
        config:
            Raw study configuration dictionary supplied by the caller.

    Processes:
        1. Validate that ``config`` is a dictionary.
        2. Define the formal recursive schema mapping expected keys to expected
           nested structures or types.
        3. Define the set of dot-paths explicitly allowed to be empty.
        4. Perform a recursive depth-first traversal over the configuration and
           the schema in parallel.
        5. Record:
           - missing required paths,
           - null/empty/type-invalid paths,
           - unexpected paths.
        6. Compute the overall structural status.
        7. Raise a ``ValueError`` if structural validity fails.
        8. Canonicalize the configuration by recursively sorting dictionary
           keys.
        9. Return the immutable structural report and the canonicalized deep
           copy.

    Outputs:
        Tuple[StructuralValidationReport, Dict[str, Any]]:
            1. Immutable structural validation report.
            2. Canonicalized deep copy of the configuration.

    Raises:
        TypeError:
            Raised if ``config`` is not a dictionary.
        ValueError:
            Raised if the configuration is structurally malformed.

    Notes:
        This callable enforces the structural portion of the run-contract gate
        used before any downstream pipeline execution can occur.
    """
    # Validate that the supplied configuration is a dictionary.
    if not isinstance(config, dict):
        # Raise a type error because recursive schema validation requires a
        # mapping-like root object.
        raise TypeError(
            "validate_structural_integrity expected 'config' to be a "
            "dictionary."
        )

    # Define the formal recursive schema mapping expected keys to expected
    # nested types.
    formal_schema: Dict[str, Any] = {
        "study_metadata": {
            "source_url": str,
            "date": str,
            "author": str,
            "title": str,
            "document_type": str,
            "explicit_equations_in_source": int,
            "explicit_neural_network_architectures_named_in_source": list,
            "explicit_prompt_templates_in_source": list,
        },
        "workflow_overview": {
            "tldr_pipeline": {
                "raw_collection": bool,
                "llm_compiles_wiki": bool,
                "llm_operates_with_tools": bool,
                "viewable_in_obsidian": bool,
            },
            "human_edit_policy": {
                "human_rarely_edits_wiki_directly": bool,
                "llm_maintains_wiki": bool,
            },
        },
        "filesystem_conventions": {
            "raw_directory": str,
            "wiki_directory": str,
            "assets_directory": str,
            "logs_directory": str,
            "eval_directory": str,
            "compiled_wiki_representation": {
                "format": str,
                "file_extension": str,
                "organization": str,
                "suggested_subdirs": dict,
            },
            "image_localization_requirement": {
                "download_related_images_locally": bool,
                "allowed_image_extensions": list,
                "linking_policy": dict,
            },
            "link_conventions": dict,
            "immutability_rules": dict,
        },
        "raw_input_schemas": dict,
        "input_source_types_indexed_into_raw": dict,
        "frontend_and_tools": dict,
        "scale_regime_explicit_example": dict,
        "llm_provider_and_auth": dict,
        "llm_decoding_settings": dict,
        "prompt_templates": dict,
        "scripted_calls_framework_details": dict,
        "retrieval_and_search_parameters": dict,
        "linting_thresholds_and_policies": dict,
        "outputs_and_visualizations": dict,
        "network_architecture_parameters": dict,
        "fine_tuning_exploration": dict,
        "methodological_notes": dict,
    }

    # Define the set of dot-paths explicitly allowed to be empty or null-like.
    allowed_empty: Set[str] = {
        "study_metadata.explicit_neural_network_architectures_named_in_source",
        "study_metadata.explicit_prompt_templates_in_source",
        "network_architecture_parameters",
    }

    # Initialize the missing-path accumulator.
    missing_sections: List[str] = []

    # Initialize the invalid-null-or-type-mismatch accumulator.
    null_sections: List[str] = []

    # Initialize the unexpected-path accumulator.
    unexpected_sections: List[str] = []

    # Define the recursive traversal and validation helper.
    def traverse_and_validate(
        current_config: Any,
        current_schema: Any,
        current_path: str
    ) -> None:
        """
        Recursively traverse the configuration and validate it against the
        formal schema.

        Purpose:
            Walk the nested configuration tree and record structural violations
            at exact dot-path granularity.

        Inputs:
            current_config:
                Current configuration subtree or scalar value.
            current_schema:
                Current schema subtree or expected scalar type.
            current_path:
                Dot-path representing the current traversal location.

        Processes:
            1. If the schema node is a dictionary:
               - require the config node to be a dictionary,
               - check for missing expected keys,
               - recurse into present children,
               - record unexpected extra keys.
            2. If the schema node is a Python type:
               - detect null/empty violations unless explicitly allowed,
               - detect runtime type mismatches.

        Outputs:
            None. Violations are accumulated into the outer-scope lists.

        Raises:
            No exceptions are raised directly here; errors are accumulated for
            the outer callable to adjudicate.
        """
        # Branch on whether the schema node expects a dictionary subtree.
        if isinstance(current_schema, dict):
            # If the configuration node is not a dictionary, record the path as
            # invalid and stop descending.
            if not isinstance(current_config, dict):
                # Record the invalid subtree path.
                null_sections.append(current_path)

                # Stop recursive descent on this branch.
                return

            # Iterate over all schema-expected child keys.
            for expected_key, expected_type in current_schema.items():
                # Construct the child's dot-path.
                child_path = (
                    f"{current_path}.{expected_key}"
                    if current_path else expected_key
                )

                # Record the path as missing if the expected key is absent.
                if expected_key not in current_config:
                    missing_sections.append(child_path)
                else:
                    # Recurse into the present child node.
                    traverse_and_validate(
                        current_config[expected_key],
                        expected_type,
                        child_path
                    )

            # Record unexpected keys that appear in the configuration but not in
            # the schema.
            for actual_key in current_config.keys():
                # Check whether the current config key is unexpected.
                if actual_key not in current_schema:
                    # Construct the unexpected child's dot-path.
                    child_path = (
                        f"{current_path}.{actual_key}"
                        if current_path else actual_key
                    )

                    # Record the unexpected path.
                    unexpected_sections.append(child_path)

        # Branch on whether the schema node expects a scalar Python type.
        elif isinstance(current_schema, type):
            # Determine whether the configuration value is null or empty.
            is_empty = (
                current_config is None
                or (
                    isinstance(current_config, (dict, list, str))
                    and len(current_config) == 0
                )
            )

            # Record null/empty violations unless the path is explicitly allowed.
            if is_empty and current_path not in allowed_empty:
                null_sections.append(current_path)

            # Record explicit type mismatches when the value is non-null and of
            # the wrong runtime type.
            elif (
                current_config is not None
                and not isinstance(current_config, current_schema)
            ):
                null_sections.append(
                    f"{current_path} (Type mismatch: expected "
                    f"{current_schema.__name__})"
                )

    # Start recursive traversal from the configuration root.
    traverse_and_validate(config, formal_schema, "")

    # Compute the overall structural validity flag.
    status: bool = (
        len(missing_sections) == 0
        and len(null_sections) == 0
    )

    # Construct the immutable structural validation report.
    report = StructuralValidationReport(
        missing_required_sections=sorted(missing_sections),
        null_but_not_permitted_sections=sorted(null_sections),
        unexpected_sections=sorted(unexpected_sections),
        structural_status=status,
    )

    # Halt immediately if deep structural validation fails.
    if not status:
        # Raise a value error that explicitly reports all missing and invalid
        # paths.
        raise ValueError(
            "Deep structural validation failed.\n"
            f"Missing: {report.missing_required_sections}\n"
            f"Null/Invalid: {report.null_but_not_permitted_sections}"
        )

    # Define the recursive canonicalization helper used to stabilize ordering.
    def canonicalize_dict(d: Any) -> Any:
        """
        Recursively sort dictionary keys to ensure deterministic
        representation.

        Purpose:
            Produce a canonical nested representation suitable for hashing and
            exact reproducibility comparisons.

        Inputs:
            d:
                Arbitrary nested configuration object.

        Processes:
            1. If the object is a dictionary:
               - sort keys lexicographically,
               - recursively canonicalize values.
            2. If the object is a list:
               - recursively canonicalize all elements.
            3. Otherwise:
               - return the value unchanged.

        Outputs:
            Any:
                Canonicalized nested object.
        """
        # Canonicalize dictionaries by sorting keys and recursively processing
        # values.
        if isinstance(d, dict):
            return {k: canonicalize_dict(d[k]) for k in sorted(d.keys())}

        # Canonicalize lists by recursively processing each element.
        elif isinstance(d, list):
            return [canonicalize_dict(i) for i in d]

        # Return primitive values unchanged.
        else:
            return d

    # Create a deep copy of the input configuration to avoid mutating caller
    # state.
    canonical_config: Dict[str, Any] = canonicalize_dict(copy.deepcopy(config))

    # Return the report and the canonicalized configuration.
    return report, canonical_config


# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 2: Validate the semantic integrity of each parameter block
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable effective runtime settings container.
@dataclass(frozen=True)
class EffectiveRuntimeSettings:
    """
    Immutable container for effective task-level runtime settings.

    Purpose:
        Store the translated per-task runtime settings that will be used during
        execution after semantic validation has completed.

    Inputs:
        This dataclass is populated upstream by the semantic validation layer.

    Processes:
        No computation is performed here. It is a passive immutable settings
        record.

    Outputs:
        task_settings:
            Mapping from task name to effective runtime parameter dictionaries.
    """

    # Store the per-task effective runtime settings.
    task_settings: Dict[str, Dict[str, Any]]


# Define the immutable semantic validation report.
@dataclass(frozen=True)
class SemanticValidationReport:
    """
    Immutable report describing semantic consistency.

    Purpose:
        Store the semantic pass/fail results required before a configuration can
        be frozen into a run contract.

    Inputs:
        This dataclass is populated upstream by semantic validation.

    Processes:
        No computation is performed here. It is a passive immutable report.

    Outputs:
        directory_policy_valid:
            Whether filesystem directory policies are coherent.
        model_mappings_complete:
            Whether all required tasks have model mappings.
        retrieval_conventions_valid:
            Whether retrieval conventions are sufficiently specified.
        split_ratios_valid:
            Whether train/validation/test fractions sum to 1.0.
        warnings:
            Ordered list of non-fatal semantic warnings.
    """

    # Store directory policy validity.
    directory_policy_valid: bool

    # Store model-mapping completeness.
    model_mappings_complete: bool

    # Store retrieval convention validity.
    retrieval_conventions_valid: bool

    # Store split-ratio validity.
    split_ratios_valid: bool

    # Store non-fatal semantic warnings.
    warnings: List[str]


# Define the function to validate semantic integrity.
def validate_semantic_integrity(
    config: Dict[str, Any],
) -> Tuple[SemanticValidationReport, EffectiveRuntimeSettings]:
    """
    Validate semantic integrity of the canonicalized study configuration and
    derive effective task-level runtime settings.

    Purpose:
        This function enforces cross-field consistency constraints that are
        necessary for reproducible execution of the study pipeline.

    Inputs:
        config:
            Canonicalized study configuration dictionary, typically produced by
            ``validate_structural_integrity``.

    Processes:
        1. Validate directory immutability and allowed/forbidden write-root
           coherence.
        2. Validate that all required tasks have model/runtime mappings.
        3. Validate that retrieval conventions are explicit.
        4. Validate that fine-tuning train/validation/test split fractions sum
           to 1.0.
        5. Derive effective per-task runtime settings.
        6. Return both the semantic validation report and the effective runtime
           settings.

    Outputs:
        A tuple of:
        1. SemanticValidationReport:
           Immutable summary of semantic validation findings.
        2. EffectiveRuntimeSettings:
           Immutable effective runtime settings for downstream model calls.

    Raises:
        TypeError:
            Raised when the supplied ``config`` is not a dictionary.
        ValueError:
            Raised when any critical semantic constraint is violated.

    Notes:
        The original callable's logic is preserved. This version adds explicit
        input validation, clearer reporting, and line-by-line documentation.

        Important:
            The model translation branch is intentionally preserved exactly as
            written in the original callable, including the check against the
            model identifier ``"gpt-5.2"``.
    """
    # Validate that the supplied configuration is a dictionary because semantic
    # checks require structured key-based access.
    if not isinstance(config, dict):
        # Raise a type error immediately because semantic validation cannot
        # proceed on non-dictionary inputs.
        raise TypeError(
            "validate_semantic_integrity expected 'config' to be a dictionary, "
            f"but received {type(config).__name__}."
        )

    # Initialize the list that will accumulate non-fatal semantic warnings.
    warnings: List[str] = []

    # --- Validate Directory Policies ---

    # Extract the filesystem conventions block, defaulting to an empty
    # dictionary if it is absent.
    fs_conv = config.get("filesystem_conventions", {})

    # Validate that the filesystem conventions block is dictionary-like before
    # reading nested fields.
    if not isinstance(fs_conv, dict):
        # Raise a value error because nested semantic access assumes a mapping.
        raise ValueError(
            "Semantic validation failed: 'filesystem_conventions' must be a "
            "dictionary."
        )

    # Extract the immutability rules nested under filesystem conventions.
    immutability = fs_conv.get("immutability_rules", {})

    # Validate that immutability rules are dictionary-like.
    if not isinstance(immutability, dict):
        # Raise a value error because the immutability policy must be a mapping.
        raise ValueError(
            "Semantic validation failed: 'immutability_rules' must be a "
            "dictionary."
        )

    # Extract the scripted-calls framework block that contains write-validation
    # rules.
    scripted_calls = config.get("scripted_calls_framework_details", {})

    # Validate that scripted call details are dictionary-like.
    if not isinstance(scripted_calls, dict):
        # Raise a value error because downstream nested access requires a
        # dictionary.
        raise ValueError(
            "Semantic validation failed: "
            "'scripted_calls_framework_details' must be a dictionary."
        )

    # Extract the write-validation rules from the scripted-calls block.
    write_rules = scripted_calls.get("validation_rules_before_writing_files", {})

    # Validate that write rules are dictionary-like.
    if not isinstance(write_rules, dict):
        # Raise a value error because write-validation rules must be structured.
        raise ValueError(
            "Semantic validation failed: "
            "'validation_rules_before_writing_files' must be a dictionary."
        )

    # Read the append-only policy for raw/ from the immutability rules.
    raw_append_only: bool = immutability.get("raw_is_append_only", False)

    # Check whether 'raw/' is explicitly listed among forbidden write roots.
    raw_forbidden: bool = "raw/" in write_rules.get("forbidden_write_roots", [])

    # Extract the allowed write roots and convert them to a set for set-based
    # inclusion checking.
    allowed_roots: set = set(write_rules.get("allowed_write_roots", []))

    # Define the expected allowed write roots from the study configuration
    # contract.
    expected_allowed: set = {"wiki/", "assets/", "logs/"}

    # Validate that the observed allowed roots are a subset of the expected
    # allowed roots.
    roots_valid: bool = allowed_roots.issubset(expected_allowed)

    # Compute the overall directory policy validity:
    # valid iff raw/ is append-only, raw/ is forbidden as a write root, and
    # allowed write roots are properly constrained.
    dir_policy_valid: bool = (
        raw_append_only and raw_forbidden and roots_valid
    )

    # Halt if the directory policy is semantically inconsistent because the
    # study requires immutable raw sources and constrained writes.
    if not dir_policy_valid:
        # Raise a value error to stop execution before any downstream file logic
        # can rely on an invalid policy.
        raise ValueError(
            "Semantic validation failed: Directory policies are inconsistent."
        )

    # --- Validate Model-to-Task Mappings ---

    # Extract the task-level decoding settings from the configuration.
    decoding_settings = (
        config.get("llm_decoding_settings", {}).get("task_defaults", {})
    )

    # Validate that task_defaults is dictionary-like.
    if not isinstance(decoding_settings, dict):
        # Raise a value error because per-task runtime derivation requires a
        # mapping from task name to settings.
        raise ValueError(
            "Semantic validation failed: "
            "'llm_decoding_settings.task_defaults' must be a dictionary."
        )

    # Define the complete list of required task names that must have model
    # mappings for the pipeline to be executable end-to-end.
    required_tasks: List[str] = [
        "wiki_compile_incremental",
        "wiki_index_refresh",
        "wiki_qa_synthesis",
        "artifact_render_markdown_report",
        "artifact_render_marp_slides",
        "artifact_render_matplotlib_plan_or_code",
        "wiki_lint_health_check",
        "synthetic_data_generation",
    ]

    # Check that every required task is present in the decoding settings mapping.
    mappings_complete: bool = all(
        task in decoding_settings for task in required_tasks
    )

    # Halt if any required task lacks runtime settings because downstream
    # orchestrators cannot invoke the pipeline completely.
    if not mappings_complete:
        # Raise a value error describing the missing semantic completeness.
        raise ValueError(
            "Semantic validation failed: Incomplete model-to-task mappings."
        )

    # --- Validate Retrieval Conventions ---

    # Extract the search-engine design block from retrieval configuration.
    search_design = (
        config.get("retrieval_and_search_parameters", {}).get(
            "search_engine_design",
            {},
        )
    )

    # Validate that the search-engine design block is dictionary-like.
    if not isinstance(search_design, dict):
        # Raise a value error because retrieval parameter extraction requires a
        # mapping.
        raise ValueError(
            "Semantic validation failed: 'search_engine_design' must be a "
            "dictionary."
        )

    # Extract the tokenization sub-block from the search-engine design.
    tokenization = search_design.get("tokenization", {})

    # Validate that tokenization configuration is dictionary-like.
    if not isinstance(tokenization, dict):
        # Raise a value error because tokenization settings must be a mapping.
        raise ValueError(
            "Semantic validation failed: 'tokenization' must be a dictionary."
        )

    # Validate that the retrieval conventions are fully explicit and match the
    # expected reproducibility-critical defaults.
    retrieval_valid: bool = (
        tokenization.get("lowercase") is True
        and tokenization.get("strip_punctuation") is True
        and tokenization.get("split_on_whitespace") is True
        and tokenization.get("stopwords") == "english_basic"
        and tokenization.get("stemming") is False
        and "field_boosts" in search_design
        and "snippet_policy" in search_design
        and "top_k_default" in search_design
        and "max_results_returned_to_llm" in search_design
    )

    # Halt if retrieval conventions are underspecified because ranking and
    # context selection would otherwise drift silently across runs.
    if not retrieval_valid:
        # Raise a value error indicating the retrieval specification is invalid.
        raise ValueError(
            "Semantic validation failed: Invalid retrieval conventions."
        )

    # --- Validate Fine-Tuning Split Ratios ---

    # Extract the train/validation/test split policy block from the fine-tuning
    # configuration.
    split_policy = (
        config.get("fine_tuning_exploration", {}).get(
            "train_val_test_split_policy",
            {},
        )
    )

    # Validate that the split policy is dictionary-like.
    if not isinstance(split_policy, dict):
        # Raise a value error because split-ratio access assumes a mapping.
        raise ValueError(
            "Semantic validation failed: "
            "'train_val_test_split_policy' must be a dictionary."
        )

    # Read the training split fraction, defaulting to 0.0 if absent.
    train_frac: float = split_policy.get("train_fraction", 0.0)

    # Read the validation split fraction, defaulting to 0.0 if absent.
    val_frac: float = split_policy.get("val_fraction", 0.0)

    # Read the test split fraction, defaulting to 0.0 if absent.
    test_frac: float = split_policy.get("test_fraction", 0.0)

    # Validate that all split fractions are numeric because arithmetic and
    # isclose checks require numeric operands.
    if not all(isinstance(x, (int, float)) for x in [train_frac, val_frac, test_frac]):
        # Raise a value error because split fractions must be numeric.
        raise ValueError(
            "Semantic validation failed: Train/validation/test fractions must "
            "all be numeric."
        )

    # Compute the total fraction:
    # total_frac = train_frac + val_frac + test_frac.
    total_frac: float = train_frac + val_frac + test_frac

    # Validate the normalization equation from the reproduction specification:
    # train_fraction + val_fraction + test_fraction = 1.0.
    splits_valid: bool = math.isclose(total_frac, 1.0, rel_tol=1e-9)

    # Halt if the split ratios do not sum to 1.0 because the data partition is
    # then semantically invalid.
    if not splits_valid:
        # Raise a value error containing the observed total for diagnosis.
        raise ValueError(
            "Semantic validation failed: "
            f"Split ratios sum to {total_frac}, expected 1.0."
        )

    # --- Derive Effective Runtime Settings ---

    # Initialize the dictionary that will accumulate effective per-task runtime
    # settings for downstream invocation.
    effective_settings_dict: Dict[str, Dict[str, Any]] = {}

    # Iterate over every task-specific decoding block in the configuration.
    for task_name, settings in decoding_settings.items():
        # Validate that the current task's settings payload is dictionary-like.
        if not isinstance(settings, dict):
            # Raise a value error because each task requires a mapping of runtime
            # parameters.
            raise ValueError(
                "Semantic validation failed: Task settings for "
                f"'{task_name}' must be a dictionary."
            )

        # Extract the declared model identifier for the current task.
        model_id: str = settings.get("model_id", "")

        # Preserve the original logic exactly: if the model is literally
        # 'gpt-5.2', translate the parameters into an effective reasoning-model
        # control block.
        if model_id == "gpt-5.2":
            # Record a warning that stochastic parameters are being ignored in
            # favor of a fixed reasoning-control abstraction.
            warnings.append(
                f"Task '{task_name}': Translating gpt-5.2 parameters. "
                "Ignoring temperature, top_p, penalties."
            )

            # Construct the effective task settings for the reasoning model.
            effective_settings_dict[task_name] = {
                # Preserve the translated model identifier.
                "model_id": "gpt-5.2",

                # Preserve the original callable's default mapping to high
                # reasoning effort.
                "reasoning_effort": "high",

                # Preserve the original maximum output token budget if present,
                # or default to 8000 otherwise.
                "max_output_tokens": settings.get("max_output_tokens", 8000),
            }

        # If the current model is not the exact reasoning-model identifier used
        # in the translation branch, preserve its settings unchanged.
        else:
            # Deep-copy the settings so downstream code cannot mutate the source
            # configuration by reference.
            effective_settings_dict[task_name] = copy.deepcopy(settings)

    # Instantiate the immutable effective runtime settings record.
    effective_runtime = EffectiveRuntimeSettings(
        task_settings=effective_settings_dict
    )

    # Instantiate the immutable semantic validation report.
    report = SemanticValidationReport(
        directory_policy_valid=dir_policy_valid,
        model_mappings_complete=mappings_complete,
        retrieval_conventions_valid=retrieval_valid,
        split_ratios_valid=splits_valid,
        warnings=warnings,
    )

    # Return both the semantic validation report and the effective runtime
    # settings for downstream orchestration.
    return report, effective_runtime


# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 3: Validate appropriateness and completeness and freeze the run contract
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable frozen run contract.
@dataclass(frozen=True)
class FrozenRunContract:
    """
    Immutable run contract produced after structural and semantic validation.

    Purpose:
        Represent the validated and frozen configuration state, anchored by a
        cryptographically deterministic SHA-256 hash.

    Inputs:
        This dataclass is populated by ``freeze_run_contract``.

    Processes:
        No computation is performed here. It is a passive immutable contract
        record.

    Outputs:
        original_config:
            Canonicalized original configuration dictionary.
        resolved_defaults:
            Reproducibility-critical defaults injected during freezing.
        effective_runtime_settings:
            Effective task-level runtime settings.
        structural_report:
            Structural validation report.
        semantic_report:
            Semantic validation report.
        epistemic_disclaimers:
            Ordered list of mandatory disclaimers about irreducible uncertainty.
        contract_hash:
            Deterministic SHA-256 hash of the canonical serialized contract.
    """

    # Store the canonicalized original configuration.
    original_config: Dict[str, Any]

    # Store injected reproducibility-critical defaults.
    resolved_defaults: Dict[str, Any]

    # Store the effective runtime settings.
    effective_runtime_settings: EffectiveRuntimeSettings

    # Store the structural validation report.
    structural_report: StructuralValidationReport

    # Store the semantic validation report.
    semantic_report: SemanticValidationReport

    # Store the mandatory epistemic disclaimers.
    epistemic_disclaimers: List[str]

    # Store the deterministic SHA-256 hash of the contract.
    contract_hash: str


# Define the run-contract freezing callable.
def freeze_run_contract(
    config: Dict[str, Any],
    structural_report: StructuralValidationReport,
    semantic_report: SemanticValidationReport,
    effective_runtime: EffectiveRuntimeSettings,
) -> FrozenRunContract:
    """
    Freeze the validated study configuration into an immutable run contract
    using canonical JSON serialization.

    Purpose:
        Convert a validated configuration into a final immutable contract by:
        1. injecting reproducibility-critical defaults,
        2. normalizing floating-point values,
        3. serializing to canonical JSON,
        4. hashing the canonical UTF-8 byte sequence.

    Inputs:
        config:
            Canonicalized configuration dictionary.
        structural_report:
            Structural validation report.
        semantic_report:
            Semantic validation report.
        effective_runtime:
            Effective per-task runtime settings.

    Processes:
        1. Validate input types.
        2. Enforce that structural validation passed.
        3. Enforce that semantic validation passed.
        4. Inject BM25 retrieval defaults:
           \[
           k_1 = 1.5,\qquad b = 0.75
           \]
        5. Define epistemic disclaimers.
        6. Recursively normalize floating-point values to fixed-precision
           strings.
        7. Serialize to canonical JSON with sorted keys and whitespace-free
           separators.
        8. Encode the canonical JSON as UTF-8 bytes.
        9. Compute the SHA-256 hash of the canonical byte stream.
        10. Return the immutable FrozenRunContract.

    Outputs:
        FrozenRunContract:
            Final immutable run contract.

    Raises:
        TypeError:
            Raised if inputs are invalid.
        ValueError:
            Raised if structural or semantic validation failed.
    """
    # Validate that config is a dictionary.
    if not isinstance(config, dict):
        # Raise a type error because contract freezing requires a configuration
        # dictionary.
        raise TypeError(
            "freeze_run_contract expected 'config' to be a dictionary."
        )

    # Validate that structural_report is a StructuralValidationReport.
    if not isinstance(structural_report, StructuralValidationReport):
        # Raise a type error because the structural report is required in typed
        # form.
        raise TypeError(
            "freeze_run_contract expected 'structural_report' to be a "
            "StructuralValidationReport."
        )

    # Validate that semantic_report is a SemanticValidationReport.
    if not isinstance(semantic_report, SemanticValidationReport):
        # Raise a type error because the semantic report is required in typed
        # form.
        raise TypeError(
            "freeze_run_contract expected 'semantic_report' to be a "
            "SemanticValidationReport."
        )

    # Validate that effective_runtime is an EffectiveRuntimeSettings.
    if not isinstance(effective_runtime, EffectiveRuntimeSettings):
        # Raise a type error because effective runtime settings must be typed.
        raise TypeError(
            "freeze_run_contract expected 'effective_runtime' to be an "
            "EffectiveRuntimeSettings."
        )

    # Enforce that structural validation succeeded.
    if not structural_report.structural_status:
        # Raise a value error because freezing a structurally invalid contract
        # is prohibited.
        raise ValueError(
            "Cannot freeze run contract: structural validation failed."
        )

    # Enforce that semantic validation succeeded on all required dimensions.
    if not (
        semantic_report.directory_policy_valid
        and semantic_report.model_mappings_complete
        and semantic_report.retrieval_conventions_valid
        and semantic_report.split_ratios_valid
    ):
        # Raise a value error because freezing a semantically invalid contract
        # is prohibited.
        raise ValueError(
            "Cannot freeze run contract: semantic validation failed."
        )

    # Inject the BM25 defaults required for reproducible retrieval scoring.
    resolved_defaults: Dict[str, Any] = {
        "retrieval_bm25_constants": {
            "k_1": 1.5,
            "b": 0.75,
        }
    }

    # Define the mandatory epistemic disclaimers.
    epistemic_disclaimers: List[str] = [
        "Proprietary models are treated as black-box API endpoints.",
        "Internal neural architectures are not fabricated.",
        "Exact bitwise reproducibility across provider-side model revisions is not guaranteed.",
    ]

    # Define the recursive float normalizer used to eliminate cross-platform
    # float formatting drift.
    def normalize_floats(obj: Any) -> Any:
        """
        Recursively convert floats to fixed-precision strings.

        Purpose:
            Stabilize floating-point serialization so contract hashing remains
            identical across environments.

        Inputs:
            obj:
                Arbitrary nested object potentially containing floats.

        Processes:
            1. If obj is a float:
               - convert to a 6-decimal-place string.
            2. If obj is a dictionary:
               - recursively normalize all values.
            3. If obj is a list:
               - recursively normalize all elements.
            4. Otherwise:
               - return the object unchanged.

        Outputs:
            Any:
                Float-normalized nested object.
        """
        # Convert floats to fixed-precision strings.
        if isinstance(obj, float):
            return "{:.6f}".format(obj)

        # Recurse into dictionaries.
        elif isinstance(obj, dict):
            return {k: normalize_floats(v) for k, v in obj.items()}

        # Recurse into lists.
        elif isinstance(obj, list):
            return [normalize_floats(i) for i in obj]

        # Return non-float primitives unchanged.
        else:
            return obj

    # Normalize floats across the configuration.
    normalized_config = normalize_floats(config)

    # Serialize to canonical JSON with sorted keys and no whitespace separators.
    canonical_json_str: str = json.dumps(
        normalized_config,
        sort_keys=True,
        separators=(",", ":")
    )

    # Encode the canonical JSON string as UTF-8 bytes.
    canonical_bytes: bytes = canonical_json_str.encode("utf-8")

    # Compute the SHA-256 contract hash.
    contract_hash: str = hashlib.sha256(canonical_bytes).hexdigest()

    # Construct and return the immutable frozen run contract.
    return FrozenRunContract(
        original_config=config,
        resolved_defaults=resolved_defaults,
        effective_runtime_settings=effective_runtime,
        structural_report=structural_report,
        semantic_report=semantic_report,
        epistemic_disclaimers=epistemic_disclaimers,
        contract_hash=contract_hash,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the orchestrator function for Task 1.
def task_1_orchestrator(raw_config: Dict[str, Any]) -> FrozenRunContract:
    """
    Orchestrate Task 1 end-to-end by validating and freezing the study
    configuration.

    Purpose:
        This function is the hard-gating entry point for Task 1. It coordinates
        structural validation, semantic validation, and final run-contract
        freezing in the required order.

    Inputs:
        raw_config:
            The raw, caller-supplied study configuration dictionary.

    Processes:
        1. Perform structural validation and canonicalization.
        2. Perform semantic validation and derive effective runtime settings.
        3. Freeze the validated configuration into an immutable run contract.

        The orchestration preserves the study's control-flow requirement that no
        downstream work may proceed unless the configuration first passes both
        structural and semantic validation.

    Outputs:
        FrozenRunContract:
            Immutable validated run contract for all downstream stages.

    Raises:
        TypeError:
            Raised when ``raw_config`` is not a dictionary.
        ValueError:
            Propagated from structural or semantic validation if the
            configuration is invalid.

    Notes:
        This callable preserves the original orchestration logic exactly while
        adding explicit input validation and documentation.
    """
    # Validate that the caller supplied a dictionary because the orchestrator
    # delegates to validators that require mapping-style access.
    if not isinstance(raw_config, dict):
        # Raise a type error immediately to prevent invalid delegation.
        raise TypeError(
            "task_1_orchestrator expected 'raw_config' to be a dictionary, but "
            f"received {type(raw_config).__name__}."
        )

    # Step 1: Perform structural validation and obtain the canonicalized
    # configuration. This step enforces the top-level schema contract and raises
    # immediately if required sections are missing or invalidly empty.
    structural_report, canonical_config = validate_structural_integrity(
        raw_config
    )

    # Step 2: Perform semantic validation and derive effective runtime settings.
    # This step enforces coherence of directory policies, task mappings,
    # retrieval conventions, and fine-tuning split ratios.
    semantic_report, effective_runtime = validate_semantic_integrity(
        canonical_config
    )

    # Step 3: Freeze the validated configuration into the immutable run
    # contract. This step injects BM25 defaults, adds epistemic disclaimers, and
    # computes the SHA-256 contract hash.
    frozen_contract: FrozenRunContract = freeze_run_contract(
        config=canonical_config,
        structural_report=structural_report,
        semantic_report=semantic_report,
        effective_runtime=effective_runtime,
    )

    # Return the final frozen run contract for downstream pipeline stages.
    return frozen_contract
````

````python
# Task 2 — Validate the execution environment and fix reproducibility controls

# ==============================================================================
# Task 2: Validate the execution environment and fix reproducibility controls
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 1: Validate the Python execution environment and toolchain
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass to hold the environment validation results.
class EnvironmentManifest:
    """
    Immutable record describing the validated Python execution environment and
    exact system binary versions.

    Purpose:
        Store the environment facts required to determine reproducibility,
        including exact package versions, exact parsed binary versions, and
        filesystem access status.

    Inputs:
        This dataclass is populated by ``validate_execution_environment``.

    Processes:
        No computation is performed here. It is a passive immutable environment
        record.

    Outputs:
        python_version:
            Exact Python interpreter version string.
        python_packages:
            Mapping of required Python package names to installed versions.
        system_binaries:
            Mapping of required system tools to parsed semantic versions or
            explicit unavailability markers.
        filesystem_access_verified:
            Whether the current working directory supports read/write access.
    """

    # Store the exact Python interpreter version.
    python_version: str

    # Store installed Python package versions.
    python_packages: Dict[str, str]

    # Store parsed system binary versions or availability markers.
    system_binaries: Dict[str, str]

    # Store whether filesystem read/write access is available.
    filesystem_access_verified: bool


# Define the environment-validation callable.
def validate_execution_environment(
    contract: FrozenRunContract,
) -> EnvironmentManifest:
    """
    Validate the Python runtime, required packages, and exact system binary
    versions.

    Purpose:
        Establish whether the execution environment satisfies the minimum
        reproducibility and tooling requirements of the study pipeline.

    Inputs:
        contract:
            Immutable run contract. It is validated for type consistency even
            though the present callable does not inspect deep configuration
            fields.

    Processes:
        1. Validate that the contract is a FrozenRunContract.
        2. Read the exact Python version string.
        3. Enforce a minimum Python version of 3.10.
        4. Resolve installed versions of mandatory Python packages using
           importlib.metadata.
        5. Probe required system binaries via subprocess:
           - git --version
           - pdftotext -v
        6. Parse semantic versions from stdout/stderr using a regex.
        7. Mark unavailable or unparsable binaries explicitly.
        8. Verify read/write access in the current working directory.
        9. Return the immutable EnvironmentManifest.

    Outputs:
        EnvironmentManifest:
            Immutable summary of the validated execution environment.

    Raises:
        TypeError:
            Raised if the contract is invalid.
        EnvironmentError:
            Raised if Python is too old, a required package is missing, or
            filesystem permissions are insufficient.
    """
    # Validate that the contract is a FrozenRunContract.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because environment validation requires the frozen
        # contract type.
        raise TypeError(
            "validate_execution_environment expected 'contract' to be a "
            "FrozenRunContract."
        )

    # Build the exact Python version string.
    py_version: str = (
        f"{sys.version_info.major}."
        f"{sys.version_info.minor}."
        f"{sys.version_info.micro}"
    )

    # Enforce the minimum Python version.
    if sys.version_info < (3, 10):
        # Raise an environment error if the interpreter is too old.
        raise EnvironmentError(
            f"Python 3.10+ is required. Current version: {py_version}"
        )

    # Initialize the package version mapping.
    resolved_packages: Dict[str, str] = {}

    # Define the mandatory Python packages.
    required_packages: List[str] = ["openai", "matplotlib", "pyyaml"]

    # Probe each required package for its installed version.
    for pkg in required_packages:
        try:
            # Resolve the installed version from package metadata.
            version: str = importlib.metadata.version(pkg)

            # Record the resolved version.
            resolved_packages[pkg] = version
        except importlib.metadata.PackageNotFoundError as exc:
            # Raise a fatal environment error if a mandatory package is missing.
            raise EnvironmentError(
                f"Mandatory Python package '{pkg}' is not installed."
            ) from exc

    # Initialize the system binary version mapping.
    resolved_binaries: Dict[str, str] = {}

    # Define the required system tools and their version commands.
    system_tools: Dict[str, List[str]] = {
        "git": ["git", "--version"],
        "pdftotext": ["pdftotext", "-v"]
    }

    # Compile the semantic version regex.
    semver_pattern = re.compile(r"(\d+\.\d+(?:\.\d+)?)")

    # Probe each system tool for its exact version.
    for tool_name, cmd in system_tools.items():
        try:
            # Execute the tool with its version flag under a strict timeout.
            result = subprocess.run(
                cmd,
                capture_output=True,
                text=True,
                timeout=5,
                check=False
            )

            # Concatenate stdout and stderr because some tools emit version
            # strings to stderr.
            output = result.stdout + result.stderr

            # Search for a semantic version string.
            match = semver_pattern.search(output)

            # Record the parsed version when available.
            if match:
                resolved_binaries[tool_name] = match.group(1)
            else:
                # Record an explicit parse failure state.
                resolved_binaries[tool_name] = "UNAVAILABLE (Parse Failed)"
        except (subprocess.TimeoutExpired, FileNotFoundError, OSError):
            # Record an explicit unavailable state on execution failure.
            resolved_binaries[tool_name] = "UNAVAILABLE"

    # Verify current working directory read/write permissions.
    fs_access: bool = os.access(".", os.R_OK | os.W_OK)

    # Reject environments lacking filesystem permissions.
    if not fs_access:
        # Raise an environment error because the pipeline requires read/write
        # access.
        raise EnvironmentError(
            "Insufficient filesystem permissions in the current working "
            "directory."
        )

    # Construct and return the immutable environment manifest.
    return EnvironmentManifest(
        python_version=py_version,
        python_packages=resolved_packages,
        system_binaries=resolved_binaries,
        filesystem_access_verified=fs_access,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 2: Validate credentials and model access
# -------------------------------------------------------------------------------------------------------------------------------

# Define an enumeration to represent the strict access states of a model.
class AccessStatus(Enum):
    """
    Enumeration of strict model access states.

    Purpose:
        Provide a constrained vocabulary for describing whether a required model
        is available, unavailable, or not required.

    Inputs:
        This enumeration is used by ModelAccessManifest.

    Processes:
        No computation is performed here.

    Outputs:
        AVAILABLE:
            The model is explicitly available.
        UNAVAILABLE:
            The model is not accessible.
        NOT_REQUIRED:
            The model is not needed because the corresponding branch is
            disabled.
    """

    # Represent that the model is available.
    AVAILABLE = "AVAILABLE"

    # Represent that the model is unavailable.
    UNAVAILABLE = "UNAVAILABLE"

    # Represent that the model is not required.
    NOT_REQUIRED = "NOT_REQUIRED"


# Define a frozen dataclass to hold the model access validation results.
@dataclass(frozen=True)
class ModelAccessManifest:
    """
    Immutable record describing validated credential presence and model access.

    Purpose:
        Store the outcome of API-key validation, base-URL validation, and
        model-availability verification.

    Inputs:
        This dataclass is populated by ``validate_credentials_and_models``.

    Processes:
        No computation is performed here. It is a passive immutable access
        record.

    Outputs:
        api_key_present:
            Whether the OPENAI_API_KEY environment variable is present and
            non-empty.
        api_base_url_verified:
            Whether the configured API base URL exactly matches the expected
            OpenAI v1 endpoint.
        model_statuses:
            Mapping from required model IDs to AccessStatus values.
    """

    # Store API-key presence.
    api_key_present: bool

    # Store API base-URL correctness.
    api_base_url_verified: bool

    # Store per-model access statuses.
    model_statuses: Dict[str, AccessStatus]


# Define the function to validate credentials and model access.
def validate_credentials_and_models(
    contract: FrozenRunContract,
) -> ModelAccessManifest:
    """
    Validate API credentials and definitively verify model access via a live
    models-endpoint query.

    Purpose:
        Prevent the pipeline from executing when credentials are absent, the API
        base URL is incorrect, or required model IDs are not actually available
        to the supplied API key.

    Inputs:
        contract:
            Immutable frozen run contract containing the provider configuration
            and required model identifiers.

    Processes:
        1. Validate that the contract is a FrozenRunContract.
        2. Read OPENAI_API_KEY from the environment.
        3. Enforce that the key is present and non-empty.
        4. Extract and validate the configured API base URL.
        5. Query the provider's /v1/models endpoint using urllib.
        6. Parse the JSON response to extract available model IDs.
        7. Determine access for:
           - the primary reasoning model,
           - the fine-tuning base model, if enabled.
        8. Return the immutable ModelAccessManifest.

    Outputs:
        ModelAccessManifest:
            Immutable summary of credentials and model access.

    Raises:
        TypeError:
            Raised if the contract is invalid.
        EnvironmentError:
            Raised if the API key is missing or the base URL is wrong.
    """
    # Validate that the contract is a FrozenRunContract.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because credential validation requires the frozen
        # contract.
        raise TypeError(
            "validate_credentials_and_models expected 'contract' to be a "
            "FrozenRunContract."
        )

    # Read the API key from the environment without logging it.
    api_key = os.environ.get("OPENAI_API_KEY", "")

    # Determine whether the API key is present and non-empty.
    api_key_present: bool = len(api_key) > 0

    # Reject missing or empty API keys.
    if not api_key_present:
        raise EnvironmentError(
            "OPENAI_API_KEY environment variable is missing or empty."
        )

    # Extract the configured API base URL from the contract.
    configured_base_url: str = (
        contract.original_config
        .get("llm_provider_and_auth", {})
        .get("api_base_url", "")
    )

    # Define the expected OpenAI v1 base URL.
    expected_base_url: str = "https://api.openai.com/v1"

    # Validate exact base-URL equality.
    api_base_url_verified: bool = (configured_base_url == expected_base_url)

    # Reject mismatched API base URLs.
    if not api_base_url_verified:
        raise EnvironmentError(
            f"API Base URL mismatch. Expected: {expected_base_url}, "
            f"Got: {configured_base_url}"
        )

    # Initialize the set of available model IDs.
    available_models: Set[str] = set()

    # Construct the authenticated GET request to the models endpoint.
    req = urllib.request.Request(f"{expected_base_url}/models")

    # Attach the Authorization bearer token header.
    req.add_header("Authorization", f"Bearer {api_key}")

    try:
        # Issue the HTTP request with a strict timeout.
        with urllib.request.urlopen(req, timeout=10) as response:
            # Decode the response body as UTF-8.
            response_body = response.read().decode("utf-8")

            # Parse the JSON payload.
            models_data = json.loads(response_body)

            # Extract the set of available model IDs from the "data" array.
            if "data" in models_data:
                available_models = {
                    m.get("id")
                    for m in models_data["data"]
                    if "id" in m
                }
    except (urllib.error.URLError, json.JSONDecodeError) as e:
        # Preserve the original behavior by logging a warning and defaulting to
        # an empty available-model set.
        print(
            f"WARNING: Failed to fetch available models from provider: {e}"
        )

    # Initialize the model-status mapping.
    model_statuses: Dict[str, AccessStatus] = {}

    # Extract the configured model-ID mapping.
    model_ids_config: Dict[str, str] = (
        contract.original_config
        .get("llm_provider_and_auth", {})
        .get("model_ids", {})
    )

    # Extract the configured primary model ID.
    primary_model: str = model_ids_config.get("question_answering", "gpt-5.2")

    # Extract the configured fine-tuning model ID.
    ft_model: str = model_ids_config.get(
        "fine_tuning_base",
        "openai/gpt-oss-20b"
    )

    # Determine whether the fine-tuning branch is enabled.
    ft_enabled: bool = (
        contract.original_config
        .get("fine_tuning_exploration", {})
        .get("enabled", False)
    )

    # Resolve the primary model access status.
    if primary_model in available_models:
        model_statuses[primary_model] = AccessStatus.AVAILABLE
    else:
        model_statuses[primary_model] = AccessStatus.UNAVAILABLE

    # Resolve the fine-tuning model access status.
    if ft_enabled:
        if ft_model in available_models:
            model_statuses[ft_model] = AccessStatus.AVAILABLE
        else:
            model_statuses[ft_model] = AccessStatus.UNAVAILABLE
    else:
        model_statuses[ft_model] = AccessStatus.NOT_REQUIRED

    # Construct and return the immutable model-access manifest.
    return ModelAccessManifest(
        api_key_present=api_key_present,
        api_base_url_verified=api_base_url_verified,
        model_statuses=model_statuses,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 3: Fix deterministic controls and document irreducible nondeterminism
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass to hold the reproducibility controls.
@dataclass(frozen=True)
class ReproducibilityControls:
    """
    Immutable record describing deterministic controls, run-level safety caps,
    prompt-template hashes, and the explicit nondeterminism disclaimer.

    Purpose:
        This dataclass stores the reproducibility controls fixed during Task 2 so
        that downstream orchestration can execute under a stable, inspectable
        contract.

    Inputs:
        This object is produced by ``fix_reproducibility_controls`` from the
        frozen run contract.

    Processes:
        The associated callable:
        1. fixes the global pseudo-random seed,
        2. reads run-level safety caps from the configuration,
        3. computes SHA-256 hashes of all prompt templates,
        4. records the irreducible hosted-model nondeterminism disclaimer.

    Outputs:
        global_seed:
            Fixed integer random seed.
        max_total_tokens:
            Run-level total token budget.
        max_api_calls:
            Run-level API call budget.
        prompt_hashes:
            Mapping from prompt-template names to SHA-256 hashes.
        nondeterminism_disclaimer:
            Mandatory explanatory note on hosted-model reproducibility limits.
    """

    # Store the fixed global pseudo-random seed.
    global_seed: int

    # Store the maximum total tokens allowed in a single run.
    max_total_tokens: int

    # Store the maximum number of API calls allowed in a single run.
    max_api_calls: int

    # Store SHA-256 hashes of all configured prompt templates to detect prompt
    # drift.
    prompt_hashes: Dict[str, str]

    # Store the explicit disclaimer documenting irreducible nondeterminism.
    nondeterminism_disclaimer: str


# Define the function to fix reproducibility controls.
def fix_reproducibility_controls(
    contract: FrozenRunContract,
) -> ReproducibilityControls:
    """
    Fix deterministic controls, compute prompt-template hashes, and establish
    run-level safety caps for the study pipeline.

    Purpose:
        This function enforces the reproducibility controls required before any
        model calls are made. It fixes the global pseudo-random seed, reads
        token/call caps from the frozen contract, hashes prompt templates to
        detect silent drift, and records the irreducible nondeterminism
        disclaimer for proprietary hosted models.

    Inputs:
        contract:
            The immutable run contract produced by Task 1. The function reads:
            - ``llm_decoding_settings.global_safety_limits``
            - ``prompt_templates``

    Processes:
        1. Validate that the supplied object is a ``FrozenRunContract``.
        2. Fix the global seed at 42 by seeding Python's ``random`` module.
        3. Read safety limits from the configuration with documented defaults.
        4. Compute SHA-256 hashes of all configured prompt templates.
        5. Construct the explicit nondeterminism disclaimer.

    Outputs:
        ReproducibilityControls:
            Immutable record containing deterministic controls and safety caps.

    Raises:
        TypeError:
            Raised if ``contract`` is not a ``FrozenRunContract`` instance, or if
            the prompt-template block is not a dictionary.
        AttributeError:
            Raised if a prompt value does not support UTF-8 string encoding in
            the way required by the original logic.

    Notes:
        This callable preserves the original logic exactly.

        There is no direct equation from the article implemented here. However,
        this step operationalizes the article's reproducibility discipline by
        stabilizing seed-based behavior and hashing prompt text to detect drift.
    """
    # Validate that the supplied contract is structurally compatible with the
    # immutable Task 1 run contract interface.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because the function requires a frozen run contract.
        raise TypeError(
            "fix_reproducibility_controls expected 'contract' to be an "
            "instance of FrozenRunContract."
        )

    # Define the fixed global seed exactly as specified by the original callable.
    fixed_seed: int = 42

    # Apply the fixed seed to Python's built-in pseudo-random generator.
    random.seed(fixed_seed)

    # If additional random-number generators such as numpy or torch were in use,
    # they would be seeded here in a larger implementation. The current logic
    # intentionally leaves them untouched.

    # Extract the global safety limits block from the frozen configuration.
    safety_limits: Dict[str, int] = (
        contract.original_config
        .get("llm_decoding_settings", {})
        .get("global_safety_limits", {})
    )

    # Validate that the safety-limits block is dictionary-like.
    if not isinstance(safety_limits, dict):
        # Raise a type error because downstream access assumes a mapping.
        raise TypeError(
            "fix_reproducibility_controls expected "
            "'llm_decoding_settings.global_safety_limits' to be a dictionary."
        )

    # Resolve the maximum total token budget, preserving the original default of
    # 2,000,000 if the field is absent.
    max_tokens: int = safety_limits.get("max_total_tokens_per_run", 2000000)

    # Resolve the maximum API call budget, preserving the original default of
    # 200 if the field is absent.
    max_calls: int = safety_limits.get("max_calls_per_run", 200)

    # Initialize the dictionary that will map prompt names to SHA-256 hashes.
    prompt_hashes: Dict[str, str] = {}

    # Extract the configured prompt templates block from the frozen contract.
    prompts: Dict[str, str] = contract.original_config.get("prompt_templates", {})

    # Validate that the prompt-template block is dictionary-like.
    if not isinstance(prompts, dict):
        # Raise a type error because prompt hashing requires name-to-text
        # mapping semantics.
        raise TypeError(
            "fix_reproducibility_controls expected 'prompt_templates' to be a "
            "dictionary."
        )

    # Iterate over each configured prompt template to compute its SHA-256 hash.
    for prompt_name, prompt_text in prompts.items():
        # Validate that the prompt name is a string because it is used as a key
        # in the resulting hash manifest.
        if not isinstance(prompt_name, str):
            # Raise a type error because prompt names must be strings.
            raise TypeError(
                "fix_reproducibility_controls expected all prompt names to be "
                "strings."
            )

        # Validate that the prompt text is a string because UTF-8 encoding is
        # required by the original hashing logic.
        if not isinstance(prompt_text, str):
            # Raise a type error because prompt hashing requires string inputs.
            raise TypeError(
                "fix_reproducibility_controls expected all prompt templates to "
                "be strings."
            )

        # Encode the prompt text as UTF-8 bytes prior to hashing.
        prompt_bytes: bytes = prompt_text.encode("utf-8")

        # Compute the SHA-256 hexadecimal digest of the encoded prompt bytes.
        prompt_hash: str = hashlib.sha256(prompt_bytes).hexdigest()

        # Store the computed prompt hash under its prompt-template name.
        prompt_hashes[prompt_name] = prompt_hash

    # Define the mandatory disclaimer documenting the irreducible limits of
    # hosted-model reproducibility.
    disclaimer: str = (
        "Exact bitwise reproducibility is not guaranteed for proprietary hosted "
        "models like gpt-5.2. Reproducibility must be evaluated based on schema "
        "validity, citation correctness, write-scope discipline, and retrieval "
        "consistency, rather than raw string equivalence."
    )

    # Construct and return the immutable reproducibility-controls record.
    return ReproducibilityControls(
        global_seed=fixed_seed,
        max_total_tokens=max_tokens,
        max_api_calls=max_calls,
        prompt_hashes=prompt_hashes,
        nondeterminism_disclaimer=disclaimer,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass to hold the complete execution environment context.
@dataclass(frozen=True)
class ExecutionEnvironmentContext:
    """
    Immutable aggregate record encapsulating the fully validated execution
    environment for the study pipeline.

    Purpose:
        This dataclass is the final Task 2 output. It packages the validated
        local environment state, model-access state, and reproducibility controls
        into a single immutable object for downstream tasks.

    Inputs:
        This object is produced by ``task_2_orchestrator`` from the outputs of:
        - ``validate_execution_environment``
        - ``validate_credentials_and_models``
        - ``fix_reproducibility_controls``

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        environment_manifest:
            Validated Python/package/system-binary/filesystem state.
        model_access_manifest:
            Validated API credential and model-access state.
        reproducibility_controls:
            Deterministic controls, prompt hashes, and safety caps.
    """

    # Store the validated environment manifest generated in Step 1.
    environment_manifest: EnvironmentManifest

    # Store the validated credential and model-access manifest generated in Step
    # 2.
    model_access_manifest: ModelAccessManifest

    # Store the reproducibility controls generated in Step 3.
    reproducibility_controls: ReproducibilityControls


# Define the orchestrator function for Task 2.
def task_2_orchestrator(
    contract: FrozenRunContract,
) -> ExecutionEnvironmentContext:
    """
    Orchestrate Task 2 end-to-end by validating the local execution environment,
    validating credentials/model access, and fixing reproducibility controls.

    Purpose:
        This function is the hard-gating entry point for Task 2. It composes the
        three Task 2 callables in the required order and returns a single
        immutable execution context for downstream tasks.

    Inputs:
        contract:
            The immutable run contract produced by Task 1.

    Processes:
        1. Validate the Python execution environment and local toolchain.
        2. Validate API credentials and infer model access states.
        3. Fix deterministic controls and compute prompt-template hashes.
        4. Aggregate the three resulting manifests into one immutable execution
           context.

    Outputs:
        ExecutionEnvironmentContext:
            Immutable aggregated context used by downstream tasks.

    Raises:
        TypeError:
            Raised if ``contract`` is not a ``FrozenRunContract`` instance.
        EnvironmentError:
            Propagated from environment and credential validation steps when a
            hard prerequisite is missing.
        ValueError:
            Propagated indirectly if nested validation logic raises it in future
            compatible extensions.

    Notes:
        This callable preserves the original orchestration logic exactly. It adds
        only explicit type validation and detailed documentation.
    """
    # Validate that the supplied contract is structurally compatible with the
    # immutable Task 1 run contract interface.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error immediately because all downstream Task 2 callables
        # require the frozen contract.
        raise TypeError(
            "task_2_orchestrator expected 'contract' to be an instance of "
            "FrozenRunContract."
        )

    # Step 1: Validate the local Python environment, required packages, required
    # system binaries, and filesystem permissions.
    env_manifest: EnvironmentManifest = validate_execution_environment(contract)

    # Step 2: Validate API credential presence, verify the API base URL, and
    # infer model access states.
    model_manifest: ModelAccessManifest = validate_credentials_and_models(
        contract
    )

    # Step 3: Fix deterministic controls, compute prompt hashes, and establish
    # run-level safety caps.
    repro_controls: ReproducibilityControls = fix_reproducibility_controls(
        contract
    )

    # Aggregate the three validated outputs into a single immutable execution
    # environment context.
    execution_context: ExecutionEnvironmentContext = ExecutionEnvironmentContext(
        environment_manifest=env_manifest,
        model_access_manifest=model_manifest,
        reproducibility_controls=repro_controls,
    )

    # Return the complete execution environment context for downstream tasks.
    return execution_context
````

````python
# Task 3 — Materialize the filesystem substrate exactly as specified

# ==============================================================================
# Task 3: Materialize the filesystem substrate exactly as specified
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 1: Materialize the root persistent namespace
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass to hold the state of a single directory.
@dataclass(frozen=True)
class DirectoryState:
    """
    Immutable record describing the materialized state of one filesystem
    directory within the study substrate.

    Purpose:
        Represent whether a directory required by the study's filesystem
        substrate exists, whether it was created during the current run, and
        whether it is valid as a directory.

    Inputs:
        This dataclass is populated by ``initialize_root_namespace``.

    Processes:
        No computation is performed here. It is a passive immutable state record.

    Outputs:
        absolute_path:
            Resolved absolute path of the directory.
        semantic_role:
            Logical role of the directory in the study, such as
            ``raw_directory`` or ``wiki_directory``.
        created_new:
            Whether the directory was created during the current run.
        is_valid:
            Whether the path exists and is a directory.

    Notes:
        This dataclass supports the filesystem substrate specified in the study:
        \[
        \mathcal{F} = \{ \text{raw/}, \text{wiki/}, \text{assets/},
        \text{logs/}, \text{eval/} \}
        \]
    """

    # Store the fully resolved absolute path of the materialized directory.
    absolute_path: pathlib.Path

    # Store the semantic role of the directory in the study contract.
    semantic_role: str

    # Store whether the directory was created during this execution.
    created_new: bool

    # Store whether the path is structurally valid as an existing directory.
    is_valid: bool


# Define the function to initialize the root namespace.
def initialize_root_namespace(
    contract: FrozenRunContract,
) -> Dict[str, DirectoryState]:
    """
    Idempotently materialize the root persistent namespace required by the study.

    Purpose:
        Create and validate the top-level filesystem substrate exactly as
        specified by the study configuration. This corresponds to the root
        namespace:
        \[
        \mathcal{F} = \{ \text{raw/}, \text{wiki/}, \text{assets/},
        \text{logs/}, \text{eval/} \}
        \]

    Inputs:
        contract:
            The frozen run contract containing ``filesystem_conventions`` and
            the configured root directory names.

    Processes:
        1. Validate the contract type and configuration payload.
        2. Read configured root directory paths.
        3. Resolve each configured path to an absolute path.
        4. If the path already exists, verify it is a directory.
        5. If the path does not exist, create it idempotently.
        6. Record the materialized state of each root directory.

    Outputs:
        Dict[str, DirectoryState]:
            Mapping from semantic directory roles to immutable directory-state
            records.

    Raises:
        TypeError:
            Raised if ``contract`` is not a ``FrozenRunContract`` or if required
            configuration sections are not dictionaries.
        FileExistsError:
            Raised if a target directory path is occupied by a file.
        PermissionError:
            Raised if the process lacks permission to create a required
            directory.

    Notes:
        This callable preserves the original code path and directory-creation
        logic exactly for valid inputs.
    """
    # Validate that the supplied contract is structurally compatible with the
    # immutable run contract interface.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because Task 3 requires the frozen study contract.
        raise TypeError(
            "initialize_root_namespace expected 'contract' to be an instance "
            "of FrozenRunContract."
        )

    # Validate that the original configuration is dictionary-like before nested
    # access.
    if not isinstance(contract.original_config, dict):
        # Raise a type error because nested configuration access requires a
        # mapping.
        raise TypeError(
            "initialize_root_namespace expected 'contract.original_config' to "
            "be a dictionary."
        )

    # Extract the filesystem conventions block from the original configuration.
    fs_config: Dict[str, Any] = contract.original_config.get(
        "filesystem_conventions",
        {},
    )

    # Validate that the filesystem conventions block is dictionary-like.
    if not isinstance(fs_config, dict):
        # Raise a type error because directory lookup requires mapping access.
        raise TypeError(
            "initialize_root_namespace expected 'filesystem_conventions' to be "
            "a dictionary."
        )

    # Define the required semantic root roles exactly as required by the study's
    # filesystem substrate.
    root_roles: List[str] = [
        "raw_directory",
        "wiki_directory",
        "assets_directory",
        "logs_directory",
        "eval_directory",
    ]

    # Initialize the dictionary that will hold the materialized root states.
    root_states: Dict[str, DirectoryState] = {}

    # Iterate over each required root role and materialize its directory.
    for role in root_roles:
        # Extract the configured relative path string for the current role,
        # defaulting to the role-prefix directory if missing.
        rel_path_str: str = fs_config.get(role, f"{role.split('_')[0]}/")

        # Validate that the configured relative path is a string.
        if not isinstance(rel_path_str, str):
            # Raise a type error because path materialization requires a string
            # or path-like representation.
            raise TypeError(
                f"initialize_root_namespace expected '{role}' to resolve to a "
                "string path."
            )

        # Convert the relative path string into a resolved absolute pathlib.Path.
        target_path: pathlib.Path = pathlib.Path(rel_path_str).resolve()

        # Determine whether the target path already exists.
        path_exists: bool = target_path.exists()

        # Initialize the created_new flag, which will be updated if the
        # directory is created during this run.
        created_new: bool = False

        # Branch on whether the target path already exists.
        if path_exists:
            # If the target exists, validate that it is a directory and not a
            # file or other non-directory object.
            if not target_path.is_dir():
                # Raise a fatal error if a file occupies the required directory
                # path.
                raise FileExistsError(
                    f"Cannot create directory '{target_path}'; a file already "
                    "exists at this location."
                )

        else:
            try:
                # Idempotently create the directory, including parents, exactly
                # as in the original implementation.
                target_path.mkdir(parents=True, exist_ok=True)

                # Mark that this directory was newly created in the current run.
                created_new = True

            except PermissionError as e:
                # Raise a fatal permission error preserving the original
                # semantics and diagnostic detail.
                raise PermissionError(
                    f"Permission denied when creating directory "
                    f"'{target_path}': {e}"
                ) from e

        # Record the immutable materialized state of the current root directory.
        root_states[role] = DirectoryState(
            absolute_path=target_path,
            semantic_role=role,
            created_new=created_new,
            is_valid=target_path.is_dir(),
        )

    # Return the complete mapping of semantic root roles to directory states.
    return root_states


# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 2: Create the raw corpus layout and initialize the manifest
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass to hold the raw layout manifest.
@dataclass(frozen=True)
class RawLayoutManifest:
    """
    Immutable record describing the materialized raw corpus layout and the raw
    manifest state.

    Purpose:
        Represent the initialized ``raw/`` subdirectory structure and the
        ``manifest.yaml`` required for append-only raw corpus tracking.

    Inputs:
        This dataclass is populated by ``initialize_raw_corpus_layout``.

    Processes:
        No computation is performed here. It is a passive immutable state record.

    Outputs:
        subdirectories:
            Mapping from semantic raw source categories to absolute subdirectory
            paths.
        manifest_path:
            Absolute path to the raw manifest file.
        manifest_initialized_new:
            Whether the manifest file was created during the current run.

    Notes:
        This dataclass supports the study's raw corpus structure, including
        representative subdirectories for web articles, papers, repositories,
        datasets, and images.
    """

    # Store the mapping from semantic raw-subdirectory names to absolute paths.
    subdirectories: Dict[str, pathlib.Path]

    # Store the absolute path of the raw manifest file.
    manifest_path: pathlib.Path

    # Store whether the manifest file was initialized during this run.
    manifest_initialized_new: bool


# Define the function to initialize the raw corpus layout.
def initialize_raw_corpus_layout(
    contract: FrozenRunContract,
    root_states: Dict[str, DirectoryState],
) -> RawLayoutManifest:
    """
    Idempotently create the raw corpus subdirectory layout and initialize the raw
    manifest file if it does not already exist.

    Purpose:
        Materialize the ``raw/`` subtree required by the study and create an
        initial schema-valid append-only ``manifest.yaml`` if absent.

    Inputs:
        contract:
            The frozen run contract containing ``raw_input_schemas`` and related
            raw layout settings.
        root_states:
            Mapping of root semantic roles to materialized directory states from
            ``initialize_root_namespace``.

    Processes:
        1. Validate the contract and root-state inputs.
        2. Resolve the valid ``raw/`` root directory.
        3. Read the raw layout schema from the configuration.
        4. Create expected raw subdirectories under ``raw/``.
        5. Create ``manifest.yaml`` if it does not already exist.
        6. Preserve append-only semantics by not overwriting an existing
           manifest.

    Outputs:
        RawLayoutManifest:
            Immutable summary of the raw layout state and manifest state.

    Raises:
        TypeError:
            Raised if inputs have incompatible types or malformed nested schema
            blocks.
        RuntimeError:
            Raised if the ``raw_directory`` root state is absent or invalid.
        PermissionError:
            Propagated if the filesystem denies directory or file creation.
        OSError:
            Propagated for lower-level filesystem write failures.
        yaml.YAMLError:
            Could propagate if YAML serialization fails unexpectedly.

    Notes:
        The initial manifest written by this callable has the schema:
        ``{"version": 1, "created_at": <iso_utc>, "items": []}``.

        There is no standalone article equation here. This step implements the
        filesystem substrate required for later realization of:
        \[
        d = \langle \text{id}(d), \text{type}(d), \text{path}(d),
        \text{metadata}(d), \text{content}(d) \rangle
        \]
        and the manifest index over raw artifacts.
    """
    # Validate that the supplied contract is structurally compatible with the
    # immutable run contract interface.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because this callable requires the frozen contract.
        raise TypeError(
            "initialize_raw_corpus_layout expected 'contract' to be an "
            "instance of FrozenRunContract."
        )

    # Validate that the root_states input is dictionary-like.
    if not isinstance(root_states, dict):
        # Raise a type error because root-state lookup requires a mapping.
        raise TypeError(
            "initialize_raw_corpus_layout expected 'root_states' to be a "
            "dictionary."
        )

    # Retrieve the raw-directory state from the materialized root-state mapping.
    raw_root_state: Optional[DirectoryState] = root_states.get("raw_directory")

    # Validate that the raw root exists and is a valid directory state.
    if not raw_root_state or not raw_root_state.is_valid:
        # Halt because raw layout creation requires a valid raw root directory.
        raise RuntimeError(
            "Raw root directory is invalid or missing. Cannot build raw layout."
        )

    # Validate that the retrieved raw root state is of the expected dataclass
    # type.
    if not isinstance(raw_root_state, DirectoryState):
        # Raise a type error because downstream access assumes DirectoryState.
        raise TypeError(
            "initialize_raw_corpus_layout expected 'raw_directory' to map to a "
            "DirectoryState."
        )

    # Extract the absolute raw-root path from the validated directory state.
    raw_root_path: pathlib.Path = raw_root_state.absolute_path

    # Extract the raw-input schema block from the frozen configuration.
    raw_schema: Dict[str, Any] = (
        contract.original_config
        .get("raw_input_schemas", {})
        .get("raw_directory_layout_schema", {})
    )

    # Validate that the raw directory layout schema is dictionary-like.
    if not isinstance(raw_schema, dict):
        # Raise a type error because nested layout lookup requires a mapping.
        raise TypeError(
            "initialize_raw_corpus_layout expected "
            "'raw_input_schemas.raw_directory_layout_schema' to be a "
            "dictionary."
        )

    # Extract the expected raw subdirectory mapping from the schema.
    expected_subdirs: Dict[str, str] = raw_schema.get("expected_subdirs", {})

    # Validate that the expected subdirectory specification is dictionary-like.
    if not isinstance(expected_subdirs, dict):
        # Raise a type error because subdirectory materialization requires a
        # mapping from semantic keys to paths.
        raise TypeError(
            "initialize_raw_corpus_layout expected 'expected_subdirs' to be a "
            "dictionary."
        )

    # Initialize the mapping that will store materialized raw subdirectory paths.
    materialized_subdirs: Dict[str, pathlib.Path] = {}

    # Iterate over every expected raw subdirectory in the schema.
    for subdir_key, rel_path_str in expected_subdirs.items():
        # Validate that the configured relative path is a string.
        if not isinstance(rel_path_str, str):
            # Raise a type error because path derivation requires a string.
            raise TypeError(
                f"initialize_raw_corpus_layout expected subdirectory path for "
                f"'{subdir_key}' to be a string."
            )

        # Extract only the basename so that joining with raw_root_path does not
        # duplicate the 'raw/' prefix.
        subdir_name: str = pathlib.Path(rel_path_str).name

        # Construct the absolute target subdirectory path under the validated
        # raw root.
        target_subdir: pathlib.Path = (raw_root_path / subdir_name).resolve()

        # Idempotently create the raw subdirectory exactly as in the original
        # implementation.
        target_subdir.mkdir(parents=True, exist_ok=True)

        # Record the materialized subdirectory path.
        materialized_subdirs[subdir_key] = target_subdir

    # Extract the required-files block from the raw schema.
    required_files_block: Dict[str, Any] = raw_schema.get("required_files", {})

    # Validate that the required-files block is dictionary-like.
    if not isinstance(required_files_block, dict):
        # Raise a type error because manifest lookup requires a mapping.
        raise TypeError(
            "initialize_raw_corpus_layout expected 'required_files' to be a "
            "dictionary."
        )

    # Resolve the configured manifest path string, preserving the original
    # default.
    manifest_rel_path: str = required_files_block.get(
        "manifest",
        "raw/manifest.yaml",
    )

    # Validate that the manifest path is a string.
    if not isinstance(manifest_rel_path, str):
        # Raise a type error because manifest-path derivation requires a string.
        raise TypeError(
            "initialize_raw_corpus_layout expected manifest path to be a "
            "string."
        )

    # Extract the manifest filename component so that it is created directly
    # under the raw root.
    manifest_name: str = pathlib.Path(manifest_rel_path).name

    # Construct the absolute target path of the raw manifest file.
    manifest_target_path: pathlib.Path = (raw_root_path / manifest_name).resolve()

    # Initialize the flag that will indicate whether the manifest was created
    # during the current run.
    manifest_created_new: bool = False

    # Preserve append-only semantics by writing an initial manifest only if it
    # does not already exist.
    if not manifest_target_path.exists():
        # Generate the current UTC timestamp in ISO 8601 format with a trailing
        # 'Z' to indicate UTC.
        current_time_iso: str = datetime.datetime.utcnow().isoformat() + "Z"

        # Construct the initial empty raw manifest payload exactly as in the
        # original implementation.
        initial_manifest_data: Dict[str, Any] = {
            "version": 1,
            "created_at": current_time_iso,
            "items": [],
        }

        # Open the target manifest path for UTF-8 writing.
        with manifest_target_path.open("w", encoding="utf-8") as f:
            # Serialize the initial manifest payload to YAML with block style and
            # preserved key order.
            yaml.dump(
                initial_manifest_data,
                f,
                default_flow_style=False,
                sort_keys=False,
            )

        # Mark the manifest as newly created.
        manifest_created_new = True

    # Construct and return the immutable raw layout manifest.
    return RawLayoutManifest(
        subdirectories=materialized_subdirs,
        manifest_path=manifest_target_path,
        manifest_initialized_new=manifest_created_new,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 3: Create the compiled wiki layout and establish write governance
# -------------------------------------------------------------------------------------------------------------------------------

# Define a class to act as the centralized write governance service.
class WritePolicyRegistry:
    """
    Centralized service for enforcing filesystem write boundaries.

    Purpose:
        Maintain resolved allowed and forbidden root prefixes and validate
        whether a candidate write path is permissible under the study's
        immutability and write-scope rules.

    Inputs:
        allowed_roots:
            Set of absolute root paths under which writes are permitted.
        forbidden_roots:
            Set of absolute root paths under which writes are forbidden.

    Processes:
        1. Store the allowed and forbidden root sets.
        2. Validate candidate write paths by resolving them absolutely.
        3. Reject any path that falls under a forbidden root.
        4. Accept only paths that fall under at least one allowed root.

    Outputs:
        This class exposes:
        - ``allowed_roots``
        - ``forbidden_roots``
        - ``validate_write_path(...)``

    Notes:
        This class operationalizes the study's immutability policy, especially
        the requirement that writes must not target ``raw/`` while remaining
        permitted under configured writable roots such as ``wiki/``, ``assets/``,
        and ``logs/``.
    """

    # Define the constructor used to initialize the write-policy registry.
    def __init__(
        self,
        allowed_roots: Set[pathlib.Path],
        forbidden_roots: Set[pathlib.Path],
    ) -> None:
        """
        Initialize the write-policy registry.

        Purpose:
            Store the absolute allowed and forbidden root sets used to validate
            candidate write paths.

        Inputs:
            allowed_roots:
                Set of absolute writable root paths.
            forbidden_roots:
                Set of absolute non-writable root paths.

        Processes:
            1. Validate the types of the supplied root sets.
            2. Store them exactly as provided.

        Outputs:
            None. State is stored on the instance.

        Raises:
            TypeError:
                Raised if the supplied roots are not sets of ``pathlib.Path``
                objects.
        """
        # Validate that allowed_roots is a set because downstream policy checks
        # assume set semantics.
        if not isinstance(allowed_roots, set):
            # Raise a type error because the constructor contract requires a set.
            raise TypeError(
                "WritePolicyRegistry expected 'allowed_roots' to be a set of "
                "pathlib.Path objects."
            )

        # Validate that forbidden_roots is a set because downstream policy checks
        # assume set semantics.
        if not isinstance(forbidden_roots, set):
            # Raise a type error because the constructor contract requires a set.
            raise TypeError(
                "WritePolicyRegistry expected 'forbidden_roots' to be a set of "
                "pathlib.Path objects."
            )

        # Validate that every allowed root is a pathlib.Path instance.
        if not all(isinstance(p, pathlib.Path) for p in allowed_roots):
            # Raise a type error because path operations require pathlib.Path
            # instances.
            raise TypeError(
                "WritePolicyRegistry expected all allowed roots to be "
                "pathlib.Path instances."
            )

        # Validate that every forbidden root is a pathlib.Path instance.
        if not all(isinstance(p, pathlib.Path) for p in forbidden_roots):
            # Raise a type error because path operations require pathlib.Path
            # instances.
            raise TypeError(
                "WritePolicyRegistry expected all forbidden roots to be "
                "pathlib.Path instances."
            )

        # Store the set of absolute writable roots exactly as provided.
        self.allowed_roots: Set[pathlib.Path] = allowed_roots

        # Store the set of absolute forbidden roots exactly as provided.
        self.forbidden_roots: Set[pathlib.Path] = forbidden_roots

    # Define the write-path validator used downstream by file-writing stages.
    def validate_write_path(self, target_path: pathlib.Path) -> bool:
        """
        Validate whether a candidate path is permitted for writing.

        Purpose:
            Enforce the study's write-governance rule that a path must be under
            at least one allowed root and under no forbidden root.

        Inputs:
            target_path:
                Candidate file or directory path to validate.

        Processes:
            1. Resolve the candidate path absolutely to neutralize traversal
               tricks.
            2. Reject immediately if the resolved path is under any forbidden
               root.
            3. Accept only if the resolved path lies under at least one allowed
               root.

        Outputs:
            bool:
                ``True`` if the path is permitted for writing, else ``False``.

        Raises:
            TypeError:
                Raised if ``target_path`` is not a ``pathlib.Path`` instance.

        Notes:
            This method implements the path-governance analogue of the study's
            write-scope constraint: writes may target only governed writable
            roots and may not target immutable roots such as ``raw/``.
        """
        # Validate that the candidate target path is a pathlib.Path instance.
        if not isinstance(target_path, pathlib.Path):
            # Raise a type error because path resolution requires pathlib.Path.
            raise TypeError(
                "WritePolicyRegistry.validate_write_path expected "
                "'target_path' to be a pathlib.Path instance."
            )

        # Resolve the candidate path absolutely to neutralize traversal such as
        # '..' components before governance checks.
        resolved_target: pathlib.Path = target_path.resolve()

        # Iterate over forbidden roots to reject any target falling within one.
        for forbidden in self.forbidden_roots:
            # Check whether the resolved target is relative to the current
            # forbidden root.
            if resolved_target.is_relative_to(forbidden):
                # Reject immediately if the target falls under a forbidden root.
                return False

        # Initialize the allowed-path flag to False until an allowed root match
        # is found.
        is_allowed: bool = False

        # Iterate over allowed roots to determine whether the target lies within
        # at least one writable root.
        for allowed in self.allowed_roots:
            # Check whether the resolved target is relative to the current
            # allowed root.
            if resolved_target.is_relative_to(allowed):
                # Mark the target as allowed because it falls under a permitted
                # write root.
                is_allowed = True

                # Break early because one allowed match is sufficient.
                break

        # Return the final allow/deny decision.
        return is_allowed


# Define a frozen dataclass to hold the wiki layout manifest.
@dataclass(frozen=True)
class WritePolicyRegistry:
    """
    Centralized service for enforcing strict filesystem write boundaries.

    Purpose:
        Maintain resolved allowed and forbidden filesystem root prefixes and
        validate candidate write paths using strict path resolution to
        neutralize symlink traversal attacks.

    Inputs:
        allowed_roots:
            Set of writable root paths.
        forbidden_roots:
            Set of forbidden root paths.

    Processes:
        1. Strictly resolve existing allowed and forbidden roots.
        2. Resolve candidate write paths incrementally to discover their true
           physical location.
        3. Reject candidate paths under forbidden roots.
        4. Accept candidate paths only if they lie under allowed roots.

    Outputs:
        validate_write_path(target_path):
            Boolean policy decision for the candidate path.
    """

    # Define the write-policy constructor.
    def __init__(
        self,
        allowed_roots: Set[pathlib.Path],
        forbidden_roots: Set[pathlib.Path],
    ) -> None:
        """
        Initialize the strict write-policy registry.

        Purpose:
            Capture the canonical allowed and forbidden roots after strict
            resolution.

        Inputs:
            allowed_roots:
                Set of candidate writable roots.
            forbidden_roots:
                Set of candidate forbidden roots.

        Processes:
            1. Validate that both inputs are sets.
            2. Strictly resolve only existing allowed roots.
            3. Strictly resolve only existing forbidden roots.
            4. Store the resolved roots on the instance.

        Outputs:
            None. State is stored on the instance.

        Raises:
            TypeError:
                Raised if inputs are not sets of pathlib.Path objects.
        """
        # Validate that allowed_roots is a set and forbidden_roots is a set.
        if not isinstance(allowed_roots, set) or not isinstance(forbidden_roots, set):
            raise TypeError(
                "WritePolicyRegistry expected sets of pathlib.Path objects."
            )

        # Strictly resolve all existing allowed roots and discard non-existent
        # ones.
        self.allowed_roots: Set[pathlib.Path] = {
            p.resolve(strict=True)
            for p in allowed_roots
            if p.exists()
        }

        # Strictly resolve all existing forbidden roots and discard non-existent
        # ones.
        self.forbidden_roots: Set[pathlib.Path] = {
            p.resolve(strict=True)
            for p in forbidden_roots
            if p.exists()
        }

    # Define the write-path validation method.
    def validate_write_path(
        self,
        target_path: pathlib.Path
    ) -> bool:
        """
        Validate whether a candidate path is permitted for writing using strict
        symlink-aware resolution.

        Purpose:
            Enforce write governance securely by reconstructing the true
            physical path of the candidate write target before comparing it
            against allowed and forbidden roots.

        Inputs:
            target_path:
                Candidate file or directory path to validate.

        Processes:
            1. Validate that target_path is a pathlib.Path.
            2. Convert it to an absolute path.
            3. Walk upward until the deepest existing parent is found.
            4. Strictly resolve that parent to expand all symlinks.
            5. Reconstruct the full physical path by appending non-existent
               suffix components.
            6. Reject if the physical path lies under any forbidden root.
            7. Accept only if the physical path lies under some allowed root.

        Outputs:
            bool:
                True if the path is write-permitted, else False.

        Raises:
            TypeError:
                Raised if target_path is not a pathlib.Path.
        """
        # Validate that target_path is a pathlib.Path.
        if not isinstance(target_path, pathlib.Path):
            # Raise a type error because strict path resolution requires
            # pathlib.Path.
            raise TypeError(
                "validate_write_path expected 'target_path' to be a "
                "pathlib.Path."
            )

        # Convert the target path to an absolute path.
        current_path = target_path.absolute()

        # Initialize the list of non-existent path components.
        non_existent_parts = []

        # Walk upward until an existing ancestor is found or the root is
        # reached.
        while (
            not current_path.exists()
            and current_path.parent != current_path
        ):
            # Prepend the current non-existent path component.
            non_existent_parts.insert(0, current_path.name)

            # Move to the parent path.
            current_path = current_path.parent

        try:
            # Strictly resolve the deepest existing parent directory.
            resolved_parent = current_path.resolve(strict=True)
        except OSError:
            # Reject the path if strict resolution fails.
            return False

        # Reconstruct the true physical path by appending non-existent child
        # components.
        true_physical_path = resolved_parent
        for part in non_existent_parts:
            true_physical_path = true_physical_path / part

        # Reject immediately if the true physical path lies under a forbidden
        # root.
        for forbidden in self.forbidden_roots:
            if true_physical_path.is_relative_to(forbidden):
                return False

        # Accept if the true physical path lies under any allowed root.
        for allowed in self.allowed_roots:
            if true_physical_path.is_relative_to(allowed):
                return True

        # Reject by default when no allowed root matches.
        return False


# Define the function to initialize the wiki layout and write policy.
def initialize_wiki_layout_and_policy(
    contract: FrozenRunContract,
    root_states: Dict[str, DirectoryState],
) -> WikiLayoutManifest:
    """
    Idempotently create the compiled wiki layout and instantiate the centralized
    write-governance policy.

    Purpose:
        Materialize the wiki subdirectories specified by the study and construct
        the write-policy registry that enforces allowed and forbidden write
        roots.

    Inputs:
        contract:
            Frozen run contract containing the wiki layout schema, artifact-path
            proposal, and write-validation rules.
        root_states:
            Mapping of root semantic roles to materialized directory states from
            ``initialize_root_namespace``.

    Processes:
        1. Validate inputs and resolve the valid ``wiki/`` root.
        2. Read configured wiki subdirectories and create them idempotently.
        3. Read configured derived-artifact directories and create them
           idempotently.
        4. Read allowed and forbidden write roots from the configuration.
        5. Resolve those roots to absolute paths.
        6. Instantiate the centralized ``WritePolicyRegistry``.

    Outputs:
        WikiLayoutManifest:
            Immutable summary of wiki layout state and write-governance state.

    Raises:
        TypeError:
            Raised if inputs or nested schema blocks have incompatible types.
        RuntimeError:
            Raised if the wiki root directory is absent or invalid.
        PermissionError:
            Propagated if the filesystem denies directory creation.
        OSError:
            Propagated for lower-level filesystem failures.

    Notes:
        This callable operationalizes the study's wiki substrate and write-scope
        policy without changing the original layout logic.
    """
    # Validate that the supplied contract is structurally compatible with the
    # immutable run contract interface.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because the callable requires the frozen study
        # contract.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected 'contract' to be an "
            "instance of FrozenRunContract."
        )

    # Validate that root_states is dictionary-like before semantic key lookup.
    if not isinstance(root_states, dict):
        # Raise a type error because root-state lookup requires a mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected 'root_states' to be a "
            "dictionary."
        )

    # Retrieve the wiki-directory state from the materialized root-state mapping.
    wiki_root_state: Optional[DirectoryState] = root_states.get("wiki_directory")

    # Validate that the wiki root exists and is a valid materialized directory.
    if not wiki_root_state or not wiki_root_state.is_valid:
        # Halt because the wiki layout cannot be created without a valid wiki
        # root directory.
        raise RuntimeError(
            "Wiki root directory is invalid or missing. Cannot build wiki "
            "layout."
        )

    # Validate that the retrieved wiki root state is of the expected dataclass
    # type.
    if not isinstance(wiki_root_state, DirectoryState):
        # Raise a type error because downstream access assumes DirectoryState.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected 'wiki_directory' to "
            "map to a DirectoryState."
        )

    # Extract the absolute wiki-root path from the validated directory state.
    wiki_root_path: pathlib.Path = wiki_root_state.absolute_path

    # Extract the filesystem-conventions block from the frozen configuration.
    fs_config: Dict[str, Any] = contract.original_config.get(
        "filesystem_conventions",
        {},
    )

    # Validate that the filesystem-conventions block is dictionary-like.
    if not isinstance(fs_config, dict):
        # Raise a type error because nested wiki-schema access requires a
        # mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'filesystem_conventions' to be a dictionary."
        )

    # Extract the compiled wiki representation schema.
    wiki_schema: Dict[str, Any] = fs_config.get(
        "compiled_wiki_representation",
        {},
    )

    # Validate that the wiki-schema block is dictionary-like.
    if not isinstance(wiki_schema, dict):
        # Raise a type error because subdirectory lookup requires a mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'compiled_wiki_representation' to be a dictionary."
        )

    # Extract the suggested wiki subdirectory mapping from the schema.
    suggested_subdirs: Dict[str, str] = wiki_schema.get("suggested_subdirs", {})

    # Validate that the suggested-subdirectory block is dictionary-like.
    if not isinstance(suggested_subdirs, dict):
        # Raise a type error because subdirectory materialization requires a
        # mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected 'suggested_subdirs' to "
            "be a dictionary."
        )

    # Initialize the dictionary that will hold materialized wiki subdirectory
    # paths.
    materialized_wiki_subdirs: Dict[str, pathlib.Path] = {}

    # Iterate over each configured wiki subdirectory and materialize it.
    for subdir_key, rel_path_str in suggested_subdirs.items():
        # Validate that the configured relative path is a string.
        if not isinstance(rel_path_str, str):
            # Raise a type error because path derivation requires a string.
            raise TypeError(
                f"initialize_wiki_layout_and_policy expected path for "
                f"wiki subdirectory '{subdir_key}' to be a string."
            )

        # Extract the basename so the path is anchored under the validated wiki
        # root without duplicating prefixes.
        subdir_name: str = pathlib.Path(rel_path_str).name

        # Construct the absolute path to the target wiki subdirectory.
        target_subdir: pathlib.Path = (wiki_root_path / subdir_name).resolve()

        # Idempotently create the wiki subdirectory exactly as in the original
        # implementation.
        target_subdir.mkdir(parents=True, exist_ok=True)

        # Record the materialized wiki subdirectory path.
        materialized_wiki_subdirs[subdir_key] = target_subdir

    # Extract the outputs-and-visualizations block from the frozen contract.
    outputs_config: Dict[str, Any] = contract.original_config.get(
        "outputs_and_visualizations",
        {},
    )

    # Validate that the outputs configuration block is dictionary-like.
    if not isinstance(outputs_config, dict):
        # Raise a type error because nested artifact-path lookup requires a
        # mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'outputs_and_visualizations' to be a dictionary."
        )

    # Extract the derived-artifact path mapping from the outputs configuration.
    artifact_paths: Dict[str, str] = outputs_config.get(
        "artifact_paths_design_proposal",
        {},
    )

    # Validate that the artifact-path mapping is dictionary-like.
    if not isinstance(artifact_paths, dict):
        # Raise a type error because artifact-path materialization requires a
        # mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'artifact_paths_design_proposal' to be a dictionary."
        )

    # Initialize the dictionary that will hold materialized derived-artifact
    # directories.
    materialized_derived_subdirs: Dict[str, pathlib.Path] = {}

    # Retrieve the absolute derived-root path created above, or fall back to
    # 'wiki_root/derived' if the key is absent.
    derived_root_path: pathlib.Path = materialized_wiki_subdirs.get(
        "derived",
        wiki_root_path / "derived",
    )

    # Iterate over each configured derived-artifact directory and materialize it.
    for artifact_key, rel_path_str in artifact_paths.items():
        # Validate that the artifact-relative path is a string.
        if not isinstance(rel_path_str, str):
            # Raise a type error because artifact-path derivation requires a
            # string.
            raise TypeError(
                f"initialize_wiki_layout_and_policy expected artifact path for "
                f"'{artifact_key}' to be a string."
            )

        # Extract the basename so the path is anchored under the derived root.
        artifact_name: str = pathlib.Path(rel_path_str).name

        # Construct the absolute path to the target derived-artifact directory.
        target_artifact_subdir: pathlib.Path = (
            derived_root_path / artifact_name
        ).resolve()

        # Idempotently create the derived-artifact directory.
        target_artifact_subdir.mkdir(parents=True, exist_ok=True)

        # Record the materialized derived-artifact directory path.
        materialized_derived_subdirs[artifact_key] = target_artifact_subdir

    # --- Establish Write Governance ---

    # Extract the scripted-calls framework block from the frozen configuration.
    scripted_calls: Dict[str, Any] = contract.original_config.get(
        "scripted_calls_framework_details",
        {},
    )

    # Validate that the scripted-calls block is dictionary-like.
    if not isinstance(scripted_calls, dict):
        # Raise a type error because nested write-rule lookup requires a mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'scripted_calls_framework_details' to be a dictionary."
        )

    # Extract the write-validation rules from the scripted-calls block.
    write_rules: Dict[str, Any] = scripted_calls.get(
        "validation_rules_before_writing_files",
        {},
    )

    # Validate that the write-rules block is dictionary-like.
    if not isinstance(write_rules, dict):
        # Raise a type error because write-policy construction requires a
        # mapping.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'validation_rules_before_writing_files' to be a dictionary."
        )

    # Extract the allowed write-root strings, preserving the original defaults.
    allowed_roots_str: List[str] = write_rules.get(
        "allowed_write_roots",
        ["wiki/", "assets/", "logs/"],
    )

    # Extract the forbidden write-root strings, preserving the original default.
    forbidden_roots_str: List[str] = write_rules.get(
        "forbidden_write_roots",
        ["raw/"],
    )

    # Validate that the allowed root specification is list-like.
    if not isinstance(allowed_roots_str, list):
        # Raise a type error because allowed-root resolution requires a list of
        # strings.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'allowed_write_roots' to be a list."
        )

    # Validate that the forbidden root specification is list-like.
    if not isinstance(forbidden_roots_str, list):
        # Raise a type error because forbidden-root resolution requires a list
        # of strings.
        raise TypeError(
            "initialize_wiki_layout_and_policy expected "
            "'forbidden_write_roots' to be a list."
        )

    # Resolve the allowed write roots to absolute pathlib.Path objects.
    allowed_roots_resolved: Set[pathlib.Path] = {
        pathlib.Path(p).resolve() for p in allowed_roots_str
    }

    # Resolve the forbidden write roots to absolute pathlib.Path objects.
    forbidden_roots_resolved: Set[pathlib.Path] = {
        pathlib.Path(p).resolve() for p in forbidden_roots_str
    }

    # Instantiate the centralized write-policy registry.
    write_policy: WritePolicyRegistry = WritePolicyRegistry(
        allowed_roots=allowed_roots_resolved,
        forbidden_roots=forbidden_roots_resolved,
    )

    # Construct and return the immutable wiki-layout manifest.
    return WikiLayoutManifest(
        wiki_subdirectories=materialized_wiki_subdirs,
        derived_subdirectories=materialized_derived_subdirs,
        write_policy=write_policy,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass to hold the complete filesystem substrate context.
@dataclass(frozen=True)
class FilesystemSubstrateContext:
    """
    Immutable aggregate record encapsulating the fully materialized filesystem
    substrate for the study.

    Purpose:
        Package the root directory states, the raw corpus layout, and the wiki
        layout plus write-governance policy into a single immutable object for
        downstream tasks.

    Inputs:
        This object is produced by ``task_3_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        root_states:
            Materialized root directory states.
        raw_layout:
            Raw corpus layout and manifest state.
        wiki_layout:
            Wiki layout and write-governance state.
    """

    # Store the materialized root directory states from Step 1.
    root_states: Dict[str, DirectoryState]

    # Store the raw corpus layout and manifest state from Step 2.
    raw_layout: RawLayoutManifest

    # Store the wiki layout and write-governance state from Step 3.
    wiki_layout: WikiLayoutManifest


# Define the orchestrator function for Task 3.
def task_3_orchestrator(
    contract: FrozenRunContract,
) -> FilesystemSubstrateContext:
    """
    Orchestrate Task 3 end-to-end by materializing the filesystem substrate and
    establishing centralized write governance.

    Purpose:
        This function is the hard-gating Task 3 entry point. It composes root
        namespace creation, raw corpus layout initialization, and wiki layout
        initialization in the required order.

    Inputs:
        contract:
            The frozen run contract produced by Task 1.

    Processes:
        1. Materialize the root persistent namespace:
           ``raw/``, ``wiki/``, ``assets/``, ``logs/``, and ``eval/``.
        2. Materialize the ``raw/`` subtree and initialize the append-only raw
           manifest.
        3. Materialize the ``wiki/`` subtree and instantiate the centralized
           write policy.
        4. Aggregate the resulting manifests into a single immutable context.

    Outputs:
        FilesystemSubstrateContext:
            Immutable filesystem substrate context used by downstream tasks.

    Raises:
        TypeError:
            Raised if ``contract`` is not a ``FrozenRunContract`` instance.
        FileExistsError:
            Propagated if a required directory path is occupied by a file.
        PermissionError:
            Propagated if required directories cannot be created.
        RuntimeError:
            Propagated if required root directories are missing or invalid.
        OSError:
            Propagated for lower-level filesystem failures.

    Notes:
        This callable preserves the original orchestration logic exactly while
        adding explicit validation and detailed documentation.
    """
    # Validate that the supplied contract is structurally compatible with the
    # immutable run contract interface.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error immediately because all downstream Task 3 callables
        # require the frozen contract.
        raise TypeError(
            "task_3_orchestrator expected 'contract' to be an instance of "
            "FrozenRunContract."
        )

    # Step 1: Materialize the root persistent namespace required by the study:
    # raw/, wiki/, assets/, logs/, and eval/.
    root_states: Dict[str, DirectoryState] = initialize_root_namespace(contract)

    # Step 2: Materialize the raw corpus layout and initialize the append-only
    # raw manifest under the validated raw/ root.
    raw_layout: RawLayoutManifest = initialize_raw_corpus_layout(
        contract,
        root_states,
    )

    # Step 3: Materialize the wiki layout and establish the centralized write
    # governance registry under the validated wiki/ root.
    wiki_layout: WikiLayoutManifest = initialize_wiki_layout_and_policy(
        contract,
        root_states,
    )

    # Aggregate the root states, raw layout, and wiki layout into a single
    # immutable filesystem substrate context.
    filesystem_context: FilesystemSubstrateContext = (
        FilesystemSubstrateContext(
            root_states=root_states,
            raw_layout=raw_layout,
            wiki_layout=wiki_layout,
        )
    )

    # Return the complete materialized filesystem substrate context.
    return filesystem_context
````

````python
# Task 4 — Ingest the raw source materials into the local corpus

# ==============================================================================
# Task 4: Ingest the raw source materials into the local corpus
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 1: Acquire and store all supported source types locally
# -------------------------------------------------------------------------------------------------------------------------------

# Define an enumeration for supported source types.
class SourceType(Enum):
    """
    Enumeration of supported raw source types.

    Purpose:
        Provide a constrained vocabulary for acquisition and ingestion logic.

    Inputs:
        This enumeration is used by RawAcquisitionRequest and ingestion
        validation.

    Processes:
        No computation is performed here.

    Outputs:
        WEB_ARTICLE_MARKDOWN:
            Web article clipped or stored as Markdown.
        PAPER_PDF:
            Paper stored as PDF.
        GIT_REPO_SNAPSHOT:
            Repository snapshot.
        DATASET:
            Dataset artifact.
        IMAGE:
            Image artifact.
        OTHER:
            Catch-all fallback type.
    """

    # Represent a web article stored as Markdown.
    WEB_ARTICLE_MARKDOWN = "web_article_markdown"

    # Represent a paper stored as PDF.
    PAPER_PDF = "paper_pdf"

    # Represent a git repository snapshot.
    GIT_REPO_SNAPSHOT = "git_repo_snapshot"

    # Represent a dataset artifact.
    DATASET = "dataset"

    # Represent an image artifact.
    IMAGE = "image"

    # Represent an uncategorized artifact.
    OTHER = "other"


# Define a dataclass for acquisition requests.
@dataclass(frozen=True)
class RawAcquisitionRequest:
    """
    Immutable record describing one raw acquisition request.

    Purpose:
        Store the acquisition intent and metadata overrides for one raw source.

    Inputs:
        This dataclass is supplied externally by ingestion orchestration.

    Processes:
        No computation is performed here.

    Outputs:
        source_type:
            Classified source type.
        title_or_name:
            Canonical title or name.
        source_url:
            Optional source URL.
        local_source_path:
            Optional local source path.
        revision:
            Optional repository revision.
        metadata_overrides:
            Optional metadata override dictionary.
    """

    # Store the source type.
    source_type: SourceType

    # Store the title or name.
    title_or_name: str

    # Store the optional source URL.
    source_url: Optional[str] = None

    # Store the optional local source path.
    local_source_path: Optional[pathlib.Path] = None

    # Store the optional revision identifier.
    revision: Optional[str] = None

    # Store optional metadata overrides.
    metadata_overrides: Dict[str, Any] = field(default_factory=dict)


# Define a dataclass for acquisition results.
@dataclass(frozen=True)
class RawAcquisitionResult:
    """
    Immutable record describing the result of one acquisition attempt.

    Purpose:
        Store whether acquisition succeeded and, if so, where the artifact was
        stored.

    Inputs:
        This dataclass is populated upstream by the ingestion stage.

    Processes:
        No computation is performed here.

    Outputs:
        request:
            Original acquisition request.
        success:
            Boolean success indicator.
        local_path:
            Absolute path of the acquired artifact if successful.
        error_message:
            Optional failure message.
    """

    # Store the original acquisition request.
    request: RawAcquisitionRequest

    # Store whether acquisition succeeded.
    success: bool

    # Store the local acquired path if successful.
    local_path: Optional[pathlib.Path] = None

    # Store the optional error message.
    error_message: Optional[str] = None



# Define the raw-source acquisition function.
def acquire_raw_sources(
    requests: List[RawAcquisitionRequest],
    raw_layout: RawLayoutManifest,
) -> List[RawAcquisitionResult]:
    """
    Acquire heterogeneous raw sources and store them under the appropriate
    ``raw/`` subdirectories.

    Purpose:
        Route each acquisition request to the correct storage branch based on
        source type and acquisition mode:
        1. git clone for repositories,
        2. URL download for files,
        3. local file copy for manually acquired artifacts.

        This callable operationalizes the first step in the raw-corpus ingestion
        pipeline, supporting the study's input universe:
        \[
        \{\text{articles}, \text{papers}, \text{repositories}, \text{datasets},
        \text{images}, \text{etc.}\}
        \]

    Inputs:
        requests:
            List of raw acquisition requests describing what to ingest.
        raw_layout:
            The materialized raw-layout manifest produced by Task 3, providing
            the absolute destination subdirectories under ``raw/``.

    Processes:
        1. Validate that ``requests`` is a list of ``RawAcquisitionRequest``
           objects.
        2. Validate that ``raw_layout`` is a ``RawLayoutManifest`` instance with
           a valid subdirectory mapping.
        3. Map each ``SourceType`` to its configured raw subdirectory key.
        4. For each acquisition request, in order:
           a. Validate the request structure and presence of required fields.
           b. Generate a filesystem-safe target name from ``title_or_name``.
           c. For git repositories:
              - clone the repository into the target subdirectory if it does not
                already exist,
              - optionally check out the requested revision.
           d. For downloadable artifacts:
              - infer an extension from the URL or source type,
              - download the remote file to the target subdirectory if it does
                not already exist.
           e. For local artifacts:
              - validate that the local source exists,
              - copy it into the target subdirectory if the target does not
                already exist.
           f. On success, record a ``RawAcquisitionResult`` with
              ``success=True``.
           g. On any exception, catch it and record a
              ``RawAcquisitionResult`` with ``success=False`` and a diagnostic
              message.
        5. Return the ordered list of acquisition outcomes.

    Outputs:
        List[RawAcquisitionResult]:
            Ordered list of acquisition outcomes aligned with the input request
            order.

    Raises:
        This callable intentionally catches all per-request exceptions and
        records them inside ``RawAcquisitionResult`` rather than raising them to
        the caller, ensuring that one failed acquisition does not halt the entire
        batch. However, it may still raise:
        TypeError:
            If top-level inputs have incompatible types before iteration begins.

    Notes:
        This function supports the raw-corpus acquisition stage preceding
        manifest construction:
        \[
        M_{\text{raw}} = \{ (\text{id}(d), \text{metadata}(d)) \mid
        d \in \mathcal{D}_{\text{raw}} \}
        \]
        It does not itself write manifest entries; it only materializes raw
        artifacts on disk.

        The original acquisition logic is preserved exactly for valid inputs.
    """
    # Validate that the acquisition request collection is a list because the
    # original logic iterates over it in order.
    if not isinstance(requests, list):
        # Raise a type error because batch acquisition requires a list of
        # requests.
        raise TypeError(
            "acquire_raw_sources expected 'requests' to be a list of "
            "RawAcquisitionRequest objects."
        )

    # Validate that the raw layout is structurally compatible with the Task 3
    # manifest interface.
    if not isinstance(raw_layout, RawLayoutManifest):
        # Raise a type error because destination lookup requires the raw layout
        # manifest.
        raise TypeError(
            "acquire_raw_sources expected 'raw_layout' to be an instance of "
            "RawLayoutManifest."
        )

    # Validate that the raw-layout subdirectory mapping is dictionary-like.
    if not isinstance(raw_layout.subdirectories, dict):
        # Raise a type error because subdirectory routing requires a mapping.
        raise TypeError(
            "acquire_raw_sources expected 'raw_layout.subdirectories' to be a "
            "dictionary."
        )

    # Initialize the ordered list that will accumulate acquisition outcomes.
    results: List[RawAcquisitionResult] = []

    # Map each supported source type to the corresponding raw-layout subdirectory
    # key produced in Task 3.
    subdir_mapping: Dict[SourceType, str] = {
        SourceType.WEB_ARTICLE_MARKDOWN: "web_articles_markdown",
        SourceType.PAPER_PDF: "papers_pdf_and_extracts",
        SourceType.GIT_REPO_SNAPSHOT: "repos_snapshots",
        SourceType.DATASET: "datasets",
        SourceType.IMAGE: "standalone_images",
        SourceType.OTHER: "web_articles_markdown",
    }

    # Iterate over each acquisition request in order.
    for req in requests:
        try:
            # Validate that the current request is of the expected dataclass type.
            if not isinstance(req, RawAcquisitionRequest):
                # Raise a type error because request routing depends on the
                # acquisition-request schema.
                raise TypeError(
                    "Each element of 'requests' must be a "
                    "RawAcquisitionRequest instance."
                )

            # Validate that the title_or_name field is a non-empty string because
            # it is used to generate a target filename.
            if not isinstance(req.title_or_name, str) or not req.title_or_name.strip():
                # Raise a value error because safe naming requires a non-empty
                # textual title.
                raise ValueError(
                    "Acquisition request 'title_or_name' must be a non-empty "
                    "string."
                )

            # Determine the raw-layout subdirectory key for the current source
            # type, falling back to the web-article area if needed.
            subdir_key = subdir_mapping.get(
                req.source_type,
                "web_articles_markdown",
            )

            # Resolve the absolute target raw subdirectory for the current
            # request.
            target_dir: pathlib.Path = raw_layout.subdirectories[subdir_key]

            # Validate that the resolved target directory is a pathlib.Path.
            if not isinstance(target_dir, pathlib.Path):
                # Raise a type error because path concatenation requires a Path.
                raise TypeError(
                    f"Raw layout subdirectory '{subdir_key}' must resolve to a "
                    "pathlib.Path."
                )

            # Generate a filesystem-safe lowercase name by replacing
            # non-alphanumeric characters with underscores.
            safe_name: str = "".join(
                [c if c.isalnum() else "_" for c in req.title_or_name]
            ).lower()

            # Initialize the target-path placeholder for the current acquisition.
            target_path: Optional[pathlib.Path] = None

            # Handle repository acquisition via git clone.
            if req.source_type == SourceType.GIT_REPO_SNAPSHOT:
                # Enforce that git repository acquisition requires a source URL.
                if not req.source_url:
                    # Raise a value error because clone cannot proceed without a
                    # repository URL.
                    raise ValueError(
                        "Git repo acquisition requires a source_url."
                    )

                # Construct the absolute target directory for the cloned
                # repository.
                target_path = target_dir / safe_name

                # Only clone if the target path does not already exist.
                if not target_path.exists():
                    # Construct the git clone command exactly as in the original
                    # implementation.
                    clone_cmd = ["git", "clone", req.source_url, str(target_path)]

                    # Execute the clone command and raise on non-zero exit.
                    subprocess.run(
                        clone_cmd,
                        check=True,
                        capture_output=True,
                        text=True,
                    )

                    # If a specific revision was requested, check it out after
                    # cloning.
                    if req.revision:
                        # Construct the git checkout command rooted at the
                        # cloned repository.
                        checkout_cmd = [
                            "git",
                            "-C",
                            str(target_path),
                            "checkout",
                            req.revision,
                        ]

                        # Execute the checkout command and raise on failure.
                        subprocess.run(
                            checkout_cmd,
                            check=True,
                            capture_output=True,
                            text=True,
                        )

                else:
                    # Preserve the original no-op behavior when the repository
                    # path already exists.
                    pass

            # Handle URL-based downloads for PDFs, images, datasets, and other
            # downloadable artifacts.
            elif req.source_url and not req.local_source_path:
                # Determine a default extension from the source URL suffix or
                # fall back to .bin if unavailable.
                ext: str = pathlib.Path(req.source_url).suffix or ".bin"

                # Override the extension to .pdf for paper sources exactly as in
                # the original logic.
                if req.source_type == SourceType.PAPER_PDF:
                    ext = ".pdf"

                # Override the extension to .md for web-article markdown sources
                # exactly as in the original logic.
                elif req.source_type == SourceType.WEB_ARTICLE_MARKDOWN:
                    ext = ".md"

                # Construct the target file path in the destination raw
                # subdirectory.
                target_path = target_dir / f"{safe_name}{ext}"

                # Only download if the target path does not already exist.
                if not target_path.exists():
                    # Construct a request object with a standard browser-like
                    # user agent header.
                    req_obj = urllib.request.Request(
                        req.source_url,
                        headers={"User-Agent": "Mozilla/5.0"},
                    )

                    # Open the remote resource and the local output file
                    # simultaneously for streamed transfer.
                    with urllib.request.urlopen(req_obj, timeout=30) as response, open(
                        target_path,
                        "wb",
                    ) as out_file:
                        # Read the entire response body and write it to the local
                        # target file.
                        out_file.write(response.read())

            # Handle manually acquired local files copied into the raw corpus.
            elif req.local_source_path:
                # Validate that the local source path exists before copying.
                if not req.local_source_path.exists():
                    # Raise a file-not-found error because the source file must
                    # exist locally.
                    raise FileNotFoundError(
                        f"Local source file not found: {req.local_source_path}"
                    )

                # Extract the suffix of the local source so the copied artifact
                # preserves its extension.
                ext = req.local_source_path.suffix

                # Construct the target path within the appropriate raw
                # subdirectory.
                target_path = target_dir / f"{safe_name}{ext}"

                # Only copy if the target path does not already exist.
                if not target_path.exists():
                    # Open the local source file for binary reading and the raw
                    # target file for binary writing.
                    with open(req.local_source_path, "rb") as src, open(
                        target_path,
                        "wb",
                    ) as dst:
                        # Copy the source bytes into the raw target file.
                        dst.write(src.read())

            else:
                # Raise a value error if neither a source URL nor a local source
                # path was supplied.
                raise ValueError(
                    "Must provide either source_url or local_source_path."
                )

            # Record a successful acquisition outcome for the current request.
            results.append(
                RawAcquisitionResult(
                    request=req,
                    success=True,
                    local_path=target_path,
                )
            )

        except Exception as e:
            # Record a failed acquisition outcome without crashing the full batch
            # ingestion process.
            results.append(
                RawAcquisitionResult(
                    request=req,
                    success=False,
                    error_message=str(e),
                )
            )

    # Return the ordered list of acquisition outcomes.
    return results


# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 2: Create metadata for every ingested item conforming to the manifest schema
# -------------------------------------------------------------------------------------------------------------------------------

# Define a frozen dataclass for one raw manifest item.
@dataclass(frozen=True)
class RawManifestItem:
    """
    Immutable metadata record representing one entry in ``raw/manifest.yaml``.

    Purpose:
        Store a schema-compliant manifest entry describing one ingested raw
        artifact.

    Inputs:
        This dataclass is populated by ``create_item_metadata``.

    Processes:
        No computation is performed here. It is a passive immutable manifest
        record.

    Outputs:
        id:
            Deterministic artifact identifier.
        type:
            Canonical raw artifact type string.
        title_or_name:
            Canonical title or name.
        ingested_at:
            UTC ISO 8601 ingestion timestamp.
        local_path:
            Repo-relative local path.
        source_url:
            Optional source URL.
        authors:
            Optional author list.
        year:
            Optional publication year.
        venue:
            Optional venue.
        revision:
            Optional revision metadata.
        tags:
            Optional tags.
        notes:
            Optional notes.
    """

    # Store the deterministic manifest identifier.
    id: str

    # Store the canonical raw artifact type string.
    type: str

    # Store the title or name of the artifact.
    title_or_name: str

    # Store the UTC ISO 8601 ingestion timestamp.
    ingested_at: str

    # Store the repo-relative local path.
    local_path: str

    # Store the optional source URL.
    source_url: Optional[str] = None

    # Store the optional author list.
    authors: Optional[List[str]] = None

    # Store the optional publication year.
    year: Optional[int] = None

    # Store the optional venue field.
    venue: Optional[str] = None

    # Store optional revision metadata.
    revision: Optional[Dict[str, str]] = None

    # Store optional tags.
    tags: Optional[List[str]] = None

    # Store optional notes.
    notes: Optional[List[str]] = None


# Define the metadata-construction function.
def create_item_metadata(
    acquisition_results: List[RawAcquisitionResult],
    repo_root: pathlib.Path,
) -> List[RawManifestItem]:
    """
    Generate schema-compliant raw-manifest items using full-file chunked
    hashing.

    Purpose:
        Convert successful acquisition results into structured raw-manifest
        entries while computing collision-resistant deterministic IDs from
        source URLs or full artifact bytes.

    Inputs:
        acquisition_results:
            Ordered list of acquisition results.
        repo_root:
            Project root directory used to derive repo-relative paths.

    Processes:
        1. Validate top-level input types.
        2. Iterate over acquisition results.
        3. Skip unsuccessful or pathless results.
        4. If a source URL exists:
           - hash the URL to derive the ID.
        5. Otherwise:
           - open the artifact in binary mode,
           - hash the full file in 4MB chunks,
           - derive the ID from the final digest.
        6. Compute the repo-relative local path.
        7. Generate the UTC ingestion timestamp.
        8. Construct a RawManifestItem using the request metadata overrides.
        9. Return the ordered manifest-item list.

    Outputs:
        List[RawManifestItem]:
            Ordered manifest-item records.

    Raises:
        TypeError:
            Raised if inputs are invalid.
        OSError:
            Propagated if artifact reads fail.

    Notes:
        This callable replaces heuristic partial hashing with full-file chunked
        hashing, providing stronger identifier fidelity.
    """
    # Validate that acquisition_results is a list.
    if not isinstance(acquisition_results, list):
        # Raise a type error because metadata generation requires a list.
        raise TypeError(
            "create_item_metadata expected 'acquisition_results' to be a list."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path arithmetic requires pathlib.Path.
        raise TypeError(
            "create_item_metadata expected 'repo_root' to be a pathlib.Path."
        )

    # Initialize the output manifest-item list.
    manifest_items: List[RawManifestItem] = []

    # Iterate over acquisition results.
    for result in acquisition_results:
        # Skip unsuccessful or pathless acquisition results.
        if not result.success or not result.local_path:
            continue

        # Retrieve the original acquisition request.
        req = result.request

        # Generate the deterministic identifier from source URL when present.
        if req.source_url:
            id_hash = hashlib.sha256(
                req.source_url.encode("utf-8")
            ).hexdigest()[:16]
        else:
            # Initialize the full-file SHA-256 hasher.
            file_hasher = hashlib.sha256()

            # Define the fixed 4MB chunk size.
            chunk_size = 4 * 1024 * 1024

            # Open the artifact for binary reading.
            with open(result.local_path, "rb") as f:
                # Read and hash the artifact in fixed-size chunks.
                while chunk := f.read(chunk_size):
                    file_hasher.update(chunk)

            # Extract the first 16 hex characters of the final digest.
            id_hash = file_hasher.hexdigest()[:16]

        # Attempt to derive a repo-relative path.
        try:
            rel_path = str(result.local_path.relative_to(repo_root))
        except ValueError:
            rel_path = str(result.local_path)

        # Import datetime locally exactly as in the provided implementation.
        import datetime

        # Generate the UTC ingestion timestamp.
        ingested_at = datetime.datetime.utcnow().isoformat() + "Z"

        # Construct the schema-compliant raw-manifest item.
        item = RawManifestItem(
            id=id_hash,
            type=req.source_type.value,
            title_or_name=req.title_or_name,
            ingested_at=ingested_at,
            local_path=rel_path,
            source_url=req.source_url,
            authors=req.metadata_overrides.get("authors"),
            year=req.metadata_overrides.get("year"),
            venue=req.metadata_overrides.get("venue"),
            revision=None,
            tags=req.metadata_overrides.get("tags"),
            notes=req.metadata_overrides.get("notes"),
        )

        # Append the manifest item to the output list.
        manifest_items.append(item)

    # Return the ordered manifest-item list.
    return manifest_items


# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 3: Validate ingestion quality item by item
# -------------------------------------------------------------------------------------------------------------------------------

# Define the enumeration of ingestion-quality outcomes.
class IngestionStatus(Enum):
    """
    Enumeration of item-level ingestion-quality outcomes.

    Purpose:
        Provide a constrained vocabulary for artifact-level ingestion audit
        results.

    Inputs:
        This enumeration is used by IngestionAuditRecord and ingestion
        validation.

    Processes:
        No computation is performed here.

    Outputs:
        ACCEPTED:
            Artifact passes validation.
        QUARANTINED:
            Artifact has non-fatal issues.
        REJECTED:
            Artifact fails critical validation.
    """

    # Represent an accepted artifact.
    ACCEPTED = "ACCEPTED"

    # Represent a quarantined artifact.
    QUARANTINED = "QUARANTINED"

    # Represent a rejected artifact.
    REJECTED = "REJECTED"


# Define the immutable ingestion-audit record.
@dataclass(frozen=True)
class IngestionAuditRecord:
    """
    Immutable audit record describing the outcome of ingestion validation.

    Purpose:
        Store the final status and failed-check details for one artifact.

    Inputs:
        This dataclass is populated by ``validate_ingestion_quality``.

    Processes:
        No computation is performed here.

    Outputs:
        item_id:
            Identifier of the audited manifest item.
        status:
            Final ingestion-quality status.
        failed_checks:
            Ordered list of failed-check diagnostic messages.
    """

    # Store the audited item identifier.
    item_id: str

    # Store the final validation status.
    status: IngestionStatus

    # Store the ordered list of failed checks.
    failed_checks: List[str]


# Define the ingestion-quality validation function.
def validate_ingestion_quality(
    manifest_items: List[RawManifestItem],
    repo_root: pathlib.Path,
) -> Tuple[List[RawManifestItem], Dict[str, IngestionAuditRecord]]:
    """
    Validate raw manifest items using binary magic-byte inspection.

    Purpose:
        Ensure that only physically valid artifacts enter the accepted raw
        corpus. For image artifacts, it replaces suffix-only checks with binary
        magic-byte inspection to detect spoofed or malformed files.

    Inputs:
        manifest_items:
            Candidate raw-manifest records.
        repo_root:
            Project root directory used to resolve repo-relative paths.

    Processes:
        1. Validate top-level input types.
        2. Initialize the accepted-item list and audit trail.
        3. Define the allowed magic-byte signatures for image formats.
        4. Iterate over manifest items.
        5. Resolve each artifact's absolute path.
        6. Reject missing paths.
        7. For IMAGE artifacts:
           - open the file in binary mode,
           - read the first 32 bytes,
           - compare against known signatures using startswith,
           - detect WebP via RIFF/WEBP header logic,
           - reject when no allowed signature matches.
        8. Record the audit outcome.
        9. Return accepted items and the full audit trail.

    Outputs:
        Tuple[List[RawManifestItem], Dict[str, IngestionAuditRecord]]:
            1. Ordered list of accepted manifest items.
            2. Mapping from item ID to audit record.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This callable hardens image validation by inspecting file headers rather
        than trusting filename extensions.
    """
    # Validate that manifest_items is a list.
    if not isinstance(manifest_items, list):
        # Raise a type error because validation requires a list.
        raise TypeError(
            "validate_ingestion_quality expected 'manifest_items' to be a list."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "validate_ingestion_quality expected 'repo_root' to be a "
            "pathlib.Path."
        )

    # Initialize the accepted-item list.
    accepted_items: List[RawManifestItem] = []

    # Initialize the audit-trail mapping.
    audit_trail: Dict[str, IngestionAuditRecord] = {}

    # Define the allowed binary magic-byte signatures.
    magic_signatures: Dict[bytes, str] = {
        b"\x89PNG\r\n\x1a\n": "png",
        b"\xFF\xD8\xFF": "jpeg",
        b"GIF87a": "gif",
        b"GIF89a": "gif",
    }

    # Iterate over each candidate manifest item.
    for item in manifest_items:
        # Initialize the failed-check list for this item.
        failed_checks: List[str] = []

        # Initialize the default status as ACCEPTED.
        status = IngestionStatus.ACCEPTED

        # Resolve the absolute artifact path.
        abs_path = (repo_root / item.local_path).resolve()

        # Reject the item if its path does not exist.
        if not abs_path.exists():
            failed_checks.append("File or directory does not exist.")
            status = IngestionStatus.REJECTED
        else:
            # Apply binary magic-byte inspection to IMAGE artifacts.
            if item.type == SourceType.IMAGE.value and abs_path.is_file():
                try:
                    # Read the first 32 bytes of the file header.
                    with open(abs_path, "rb") as f:
                        header_bytes = f.read(32)

                    # Initialize the valid-image flag.
                    is_valid_image = False

                    # Check against the standard signatures.
                    for sig in magic_signatures.keys():
                        if header_bytes.startswith(sig):
                            is_valid_image = True
                            break

                    # Apply the dedicated WebP header test.
                    if not is_valid_image and len(header_bytes) >= 12:
                        if (
                            header_bytes.startswith(b"RIFF")
                            and header_bytes[8:12] == b"WEBP"
                        ):
                            is_valid_image = True

                    # Reject when no valid signature matches.
                    if not is_valid_image:
                        failed_checks.append(
                            "Magic byte inspection failed: Unsupported or "
                            "spoofed image format."
                        )
                        status = IngestionStatus.REJECTED

                except OSError as e:
                    # Reject when the file header cannot be read.
                    failed_checks.append(f"Failed to read file header: {e}")
                    status = IngestionStatus.REJECTED

        # Record the final audit outcome.
        audit_trail[item.id] = IngestionAuditRecord(
            item_id=item.id,
            status=status,
            failed_checks=failed_checks,
        )

        # Include only ACCEPTED items in the accepted-item list.
        if status == IngestionStatus.ACCEPTED:
            accepted_items.append(item)

    # Return the accepted items and the full audit trail.
    return accepted_items, audit_trail


# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable ingestion-context aggregate returned by Task 4.
@dataclass(frozen=True)
class IngestionContext:
    """
    Immutable aggregate record encapsulating the outcome of the ingestion phase.

    Purpose:
        Package the accepted manifest items and the full audit trail into one
        immutable object for downstream tasks.

    Inputs:
        This dataclass is populated by ``task_4_orchestrator`` after executing
        the three ingestion steps and updating ``raw/manifest.yaml``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        accepted_items:
            Ordered list of manifest items admitted into the raw manifest after
            passing all critical quality gates.
        audit_trail:
            Mapping from item IDs to ingestion audit records, providing full
            traceability for accepted, quarantined, and rejected items.
    """

    # Store the ordered list of manifest items accepted into the raw corpus.
    accepted_items: List[RawManifestItem]

    # Store the full item-level audit trail for all validated candidates.
    audit_trail: Dict[str, IngestionAuditRecord]


# Define the Task 4 orchestrator.
def task_4_orchestrator(
    fs_context: FilesystemSubstrateContext,
    acquisition_requests: List[RawAcquisitionRequest],
    repo_root: pathlib.Path,
) -> IngestionContext:
    """
    Orchestrate Task 4 end-to-end by acquiring raw artifacts, generating raw
    metadata, validating ingestion quality, and appending accepted items to the
    raw manifest.

    Purpose:
        This function is the Task 4 entry point. It composes the three Task 4
        substeps in the required order and updates ``raw/manifest.yaml`` in an
        append-only, ID-deduplicated manner, thereby operationalizing the raw
        corpus ingestion layer of the study pipeline.

    Inputs:
        fs_context:
            Filesystem substrate context produced by Task 3, providing access to
            the raw layout and the raw manifest path.
        acquisition_requests:
            Ordered list of raw acquisition requests describing the sources to
            ingest during this run.
        repo_root:
            Project root directory used to compute repo-relative paths for
            stable cross-machine linking and to resolve local paths for physical
            validation.

    Processes:
        1. Validate that ``fs_context`` is a ``FilesystemSubstrateContext``.
        2. Validate that ``acquisition_requests`` is a list of
           ``RawAcquisitionRequest`` objects.
        3. Validate that ``repo_root`` is a ``pathlib.Path``.
        4. Step 1: Acquire raw sources into the materialized ``raw/`` corpus by
           calling ``acquire_raw_sources``.
        5. Step 2: Create schema-compliant metadata entries for successful
           acquisitions by calling ``create_item_metadata``.
        6. Step 3: Validate each metadata entry against the filesystem and
           ingestion quality rules by calling ``validate_ingestion_quality``.
        7. Read the existing ``raw/manifest.yaml`` if present; otherwise
           initialize a default manifest payload.
        8. Compute the set of existing manifest item IDs to prevent duplicates.
        9. Append newly accepted items to the manifest while avoiding ID
           collisions.
        10. Write the updated manifest payload back to ``raw/manifest.yaml``
            using deterministic YAML serialization.
        11. Return an immutable ``IngestionContext`` aggregating accepted items
            and the audit trail.

    Outputs:
        IngestionContext:
            Immutable aggregate containing accepted manifest items and the full
            audit trail.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.
        OSError:
            Propagated if manifest reads or writes fail.
        yaml.YAMLError:
            Could propagate if YAML parsing or serialization fails unexpectedly.

    Notes:
        This orchestrator implements the manifest-admission realization of:
        \[
        M_{\text{raw}} = \{ (\text{id}(d), \text{metadata}(d)) \mid
        d \in \mathcal{D}_{\text{raw}} \}
        \]
        by appending accepted manifest items into ``raw/manifest.yaml`` with
        ID-based deduplication.

        The original orchestration logic is preserved exactly for valid inputs.
    """
    # Validate that the filesystem substrate context has the expected structural
    # type.
    if not isinstance(fs_context, FilesystemSubstrateContext):
        # Raise a type error because downstream raw-layout access assumes the
        # filesystem-context schema.
        raise TypeError(
            "task_4_orchestrator expected 'fs_context' to be an instance of "
            "FilesystemSubstrateContext."
        )

    # Validate that the acquisition request collection is a list.
    if not isinstance(acquisition_requests, list):
        # Raise a type error because batch ingestion requires a list of requests.
        raise TypeError(
            "task_4_orchestrator expected 'acquisition_requests' to be a list "
            "of RawAcquisitionRequest objects."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_4_orchestrator expected 'repo_root' to be a pathlib.Path "
            "instance."
        )

    # Step 1: Acquire raw artifacts and store them inside the materialized raw
    # corpus layout.
    acq_results = acquire_raw_sources(
        acquisition_requests,
        fs_context.raw_layout,
    )

    # Step 2: Generate schema-compliant raw-manifest metadata for successfully
    # acquired artifacts.
    manifest_items = create_item_metadata(acq_results, repo_root)

    # Step 3: Validate the candidate manifest items against physical filesystem
    # and quality rules.
    accepted_items, audit_trail = validate_ingestion_quality(
        manifest_items,
        repo_root,
    )

    # --- Update the manifest.yaml file ---

    # Resolve the raw manifest path from the filesystem substrate context.
    manifest_path = fs_context.raw_layout.manifest_path

    # Validate that the manifest path is a pathlib.Path.
    if not isinstance(manifest_path, pathlib.Path):
        # Raise a type error because manifest I/O requires a Path object.
        raise TypeError(
            "task_4_orchestrator expected 'fs_context.raw_layout.manifest_path' "
            "to be a pathlib.Path instance."
        )

    # Initialize the default manifest payload used if the manifest does not yet
    # exist or loads as an empty document.
    existing_data = {
        "version": 1,
        "created_at": datetime.datetime.utcnow().isoformat() + "Z",
        "items": [],
    }

    # If the manifest file already exists, read and parse it.
    if manifest_path.exists():
        # Open the manifest for UTF-8 reading.
        with open(manifest_path, "r", encoding="utf-8") as f:
            # Parse the YAML payload safely.
            loaded_data = yaml.safe_load(f)

            # Preserve the original behavior: if parsed content is truthy, use
            # it as the current manifest payload.
            if loaded_data:
                existing_data = loaded_data

    # Compute the set of existing manifest IDs to preserve append-only semantics
    # with ID-based deduplication.
    existing_ids = {
        item.get("id") for item in existing_data.get("items", [])
    }

    # Iterate over accepted manifest items and append only new IDs.
    for item in accepted_items:
        # If the current item ID has not already been recorded in the manifest,
        # append it.
        if item.id not in existing_ids:
            # Convert the dataclass instance to a dictionary while excluding
            # fields whose values are None to keep the YAML compact.
            item_dict = {
                k: v for k, v in item.__dict__.items() if v is not None
            }

            # Append the new item dictionary to the manifest's item list.
            existing_data["items"].append(item_dict)

    # Open the manifest path for UTF-8 writing.
    with open(manifest_path, "w", encoding="utf-8") as f:
        # Serialize the updated manifest payload to YAML using block style and
        # preserved key order.
        yaml.dump(
            existing_data,
            f,
            default_flow_style=False,
            sort_keys=False,
        )

    # Construct and return the immutable ingestion context.
    return IngestionContext(
        accepted_items=accepted_items,
        audit_trail=audit_trail,
    )
````

````python
# Task 5 — Cleanse and normalize the ingested data

# ==============================================================================
# Task 5: Cleanse and normalize the ingested data
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 1 & 2: Cleanse text-based and non-text sources
# -------------------------------------------------------------------------------------------------------------------------------

# Define the enumeration of cleansing confidence scores.
class ConfidenceScore(Enum):
    """
    Confidence label attached to a normalized derivative.

    Purpose:
        Represent the normalization confidence assigned to a derivative artifact
        after text decoding, extraction, or other modality-specific processing.

    Members:
        HIGH:
            Indicates that normalization completed successfully without
            significant degradation or fallback failure.
        MEDIUM:
            Indicates that normalization completed, but some caution is
            warranted in downstream interpretation.
        LOW:
            Indicates that normalization degraded materially, failed partially,
            or required a placeholder output.

    Notes:
        These states support Task 5, Step 3, where low-quality extraction must
        be flagged so the compiler does not overclaim based on poor inputs.
    """

    # Define the high-confidence state for successful normalization outcomes.
    HIGH = "HIGH"

    # Define the medium-confidence state for partially degraded but usable outputs.
    MEDIUM = "MEDIUM"

    # Define the low-confidence state for failed or placeholder normalization outcomes.
    LOW = "LOW"

# Define the frozen dataclass for one normalized derivative record.
@dataclass(frozen=True)
class NormalizedDerivativeRecord:
    """
    Immutable record describing a normalized derivative produced from a raw artifact.

    Purpose:
        Capture the derivative file path, normalization provenance, confidence
        label, and content hash for downstream reproducibility, change
        detection, and audit.

    Attributes:
        raw_artifact_id:
            Stable identifier of the originating raw artifact.
        derivative_id:
            Stable derivative identifier, here derived from the content hash
            prefix.
        derivative_path:
            Repo-relative path to the normalized derivative.
        normalization_actions:
            Ordered list of normalization actions applied to produce the
            derivative.
        confidence_score:
            Confidence label indicating the reliability of the derivative.
        content_hash:
            SHA-256 hash of the derivative content used for reproducibility and
            change detection.

    Notes:
        This record supports the study's provenance-first workflow by making the
        relationship between raw and normalized artifacts explicit and hashable.
    """

    # Store the stable identifier of the originating raw artifact.
    raw_artifact_id: str

    # Store the derivative identifier, typically derived from the derivative content hash.
    derivative_id: str

    # Store the repo-relative derivative path written under the normalized derivative directory.
    derivative_path: str

    # Store the ordered normalization actions used to produce the derivative.
    normalization_actions: List[str]

    # Store the confidence score assigned to the derivative.
    confidence_score: ConfidenceScore

    # Store the SHA-256 content hash of the derivative for audit and change detection.
    content_hash: str



# Define the internal text-normalization helper.
# Define the helper that normalizes text bytes into a UTF-8 string while preserving content as much as possible.
def _normalize_text_content(
    raw_bytes: bytes,
) -> Tuple[str, List[str]]:
    """
    Decode raw bytes to UTF-8 and normalize line endings without destroying data.

    Purpose:
        Convert raw text bytes into a normalized UTF-8 string suitable for
        downstream indexing, parsing, validation, and LLM compilation. This
        callable is intentionally conservative: it performs encoding recovery and
        line-ending normalization, but it does not apply aggressive boilerplate
        deletion that could silently remove semantically relevant content.

    Inputs:
        raw_bytes:
            Raw bytes read from a text-like source artifact or from a
            normalized derivative produced by an upstream extractor.

    Processes:
        1. Validate that the supplied object is of type ``bytes``.
        2. Attempt strict UTF-8 decoding.
        3. If strict decoding fails, fall back to UTF-8 decoding with
           replacement characters so the pipeline remains operational.
        4. Normalize all line endings to Unix newlines (``\\n``).
        5. Record all normalization actions in a deterministic action log.

    Outputs:
        Tuple[str, List[str]]:
            A 2-tuple containing:
            1. The normalized text as a UTF-8 Python string.
            2. An ordered list of normalization actions applied.

    Raises:
        TypeError:
            Raised if ``raw_bytes`` is not a ``bytes`` object.

    Notes:
        This function supports Task 5's requirement to cleanse text-based
        sources without mutating original artifacts in ``raw/``. The output
        should be written to a derivative location rather than back to the raw
        source path.
    """
    # Validate that the input is a bytes object because decoding semantics require raw bytes.
    if not isinstance(raw_bytes, bytes):
        # Raise a type error immediately when the input contract is violated.
        raise TypeError(
            "_normalize_text_content expected 'raw_bytes' to be a bytes object."
        )

    # Initialize an ordered action log for provenance and downstream auditing.
    actions: List[str] = []

    # Attempt strict UTF-8 decoding first because this is the highest-integrity path.
    try:
        # Decode the raw bytes as strict UTF-8.
        text = raw_bytes.decode("utf-8", errors="strict")

        # Record that strict UTF-8 decoding succeeded.
        actions.append("Decoded as strict UTF-8")

    # Handle malformed byte sequences without aborting the entire pipeline.
    except UnicodeDecodeError:
        # Decode again using replacement mode so the output remains processable.
        text = raw_bytes.decode("utf-8", errors="replace")

        # Record that replacement characters were introduced because encoding errors were present.
        actions.append(
            "Decoded as UTF-8 with replacement characters (encoding errors detected)"
        )

    # Check whether the decoded text contains Windows or classic-Mac line endings.
    if "\r\n" in text or "\r" in text:
        # Normalize CRLF and CR line endings to LF, preserving content while standardizing separators.
        text = text.replace("\r\n", "\n").replace("\r", "\n")

        # Record that line-ending normalization was applied.
        actions.append("Normalized line endings to \\n")

    # Return the normalized text together with the ordered list of normalization actions.
    return text, actions

# Define the main normalization dispatcher callable.
def process_and_normalize_items(
    ingestion_context: Any,
    repo_root: pathlib.Path,
    normalized_dir: pathlib.Path,
) -> List[NormalizedDerivativeRecord]:
    """
    Dispatch accepted raw items to modality-specific normalization routines.

    Purpose:
        Convert raw artifacts into normalized derivatives while preserving the
        original callable's logic. This implementation handles all explicitly
        listed modalities:
        - web/text-like artifacts,
        - PDF papers,
        - repository snapshots,
        - datasets,
        - images.

        It also enforces failure isolation so one bad artifact does not halt the
        batch.

    Inputs:
        ingestion_context:
            Context object expected to expose ``accepted_items``. Each item is
            expected to provide at least:
            - ``id``
            - ``type``
            - ``local_path``
            - ``title_or_name`` for repository summaries
        repo_root:
            Repository root directory used to resolve raw paths and compute
            repo-relative derivative paths.
        normalized_dir:
            Directory in which normalized derivatives are to be written.

    Processes:
        1. Validate input types and required context attributes.
        2. Iterate over accepted items from the ingestion context.
        3. Route by modality:
           - WEB/OTHER:
             read bytes, normalize text, write normalized Markdown derivative.
           - PDF:
             run ``pdftotext -layout`` with a strict 60-second timeout, then
             normalize extracted text; on failure, emit a LOW-confidence
             placeholder.
           - REPOSITORY:
             build a deterministic summary including README, selected entry
             points, and a sampled source tree.
           - DATASET:
             locate and parse a dataset card, normalize key names, and write a
             normalized YAML derivative; on missing card, emit a LOW-confidence
             placeholder.
           - IMAGE:
             copy bytes directly to preserve binary integrity.
        4. Compute a SHA-256 content hash for each derivative.
        5. Emit ``NormalizedDerivativeRecord`` objects for all successfully
           processed artifacts.
        6. Log unexpected per-item failures without aborting the entire batch.

    Outputs:
        List[NormalizedDerivativeRecord]:
            Ordered list of normalized derivative records created during the
            batch.

    Raises:
        TypeError:
            Raised if ``repo_root`` or ``normalized_dir`` is not a
            ``pathlib.Path``.
        ValueError:
            Raised if ``ingestion_context`` does not expose ``accepted_items``.

    Error Handling:
        - PDF extractor timeouts produce LOW-confidence placeholder files.
        - PDF execution errors produce LOW-confidence placeholder files.
        - Missing dataset cards produce LOW-confidence placeholder files.
        - Unexpected per-item errors are logged and processing continues.

    Notes:
        This callable supports the study requirement that cleansing occurs on
        derivatives rather than mutating ``raw/``. Hashes emitted here feed
        later change-detection and reproducibility logic.
    """
    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path semantics.
        raise TypeError(
            "process_and_normalize_items expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that the normalized output directory is a pathlib.Path.
    if not isinstance(normalized_dir, pathlib.Path):
        # Raise a type error because derivative writing requires pathlib.Path semantics.
        raise TypeError(
            "process_and_normalize_items expected 'normalized_dir' to be a pathlib.Path."
        )

    # Validate that the ingestion context exposes the accepted_items collection.
    if not hasattr(ingestion_context, "accepted_items"):
        # Raise a value error because the batch dispatcher cannot iterate without accepted_items.
        raise ValueError(
            "process_and_normalize_items expected 'ingestion_context' to expose "
            "'accepted_items'."
        )

    # Initialize the ordered list that accumulates derivative records.
    derivatives: List[NormalizedDerivativeRecord] = []

    # Iterate over each accepted item in the ingestion context.
    for item in ingestion_context.accepted_items:
        # Resolve the absolute path of the raw artifact from the repo-relative local path.
        raw_abs_path = (repo_root / item.local_path).resolve()

        # Initialize the per-item action log.
        actions: List[str] = []

        # Initialize the per-item confidence score to HIGH and degrade only on failures.
        confidence = ConfidenceScore.HIGH

        # Capture the safe filename stem used for stable derivative naming.
        safe_stem = raw_abs_path.stem

        # Wrap per-item processing in a broad exception boundary so one artifact cannot fail the batch.
        try:
            # -----------------------------------------------------------------
            # Modality 1: Text and Markdown (WEB_ARTICLE_MARKDOWN, OTHER)
            # -----------------------------------------------------------------

            # Route web articles and generic text-like artifacts through text normalization.
            if item.type in [
                SourceType.WEB_ARTICLE_MARKDOWN.value,
                SourceType.OTHER.value,
            ]:
                # Restrict this branch to Markdown and plain-text suffixes.
                if raw_abs_path.suffix.lower() in [".md", ".txt"]:
                    # Read the raw artifact bytes from disk.
                    raw_bytes = raw_abs_path.read_bytes()

                    # Normalize the text encoding and line endings.
                    normalized_text, text_actions = _normalize_text_content(raw_bytes)

                    # Extend the action log with the normalization actions returned by the helper.
                    actions.extend(text_actions)

                    # Construct the normalized Markdown derivative output path.
                    out_path = normalized_dir / f"{item.id}_{safe_stem}.md"

                    # Write the normalized text to the derivative path as UTF-8.
                    out_path.write_text(normalized_text, encoding="utf-8")

                    # Compute the SHA-256 hash of the normalized derivative content.
                    content_hash = hashlib.sha256(
                        normalized_text.encode("utf-8")
                    ).hexdigest()

                    # Append the derivative record for the normalized text artifact.
                    derivatives.append(
                        NormalizedDerivativeRecord(
                            raw_artifact_id=item.id,
                            derivative_id=content_hash[:16],
                            derivative_path=str(out_path.relative_to(repo_root)),
                            normalization_actions=actions,
                            confidence_score=confidence,
                            content_hash=content_hash,
                        )
                    )

            # -----------------------------------------------------------------
            # Modality 2: PDF Papers (PAPER_PDF)
            # -----------------------------------------------------------------

            # Route PDF papers through external text extraction.
            elif item.type == SourceType.PAPER_PDF.value:
                # Construct the output path for the extracted text derivative.
                out_path = normalized_dir / f"{item.id}_{safe_stem}.txt"

                # Attempt PDF extraction and normalization inside a dedicated exception boundary.
                try:
                    # Construct the pdftotext command with ``-layout`` to preserve spatial structure.
                    cmd = ["pdftotext", "-layout", str(raw_abs_path), str(out_path)]

                    # Execute the extraction command with a strict 60-second timeout.
                    subprocess.run(
                        cmd,
                        check=True,
                        capture_output=True,
                        timeout=60,
                    )

                    # Record successful extraction.
                    actions.append(
                        "Extracted text using pdftotext -layout "
                        "(completed within timeout)"
                    )

                    # Read the extracted derivative as raw bytes.
                    raw_bytes = out_path.read_bytes()

                    # Normalize the extracted text using the shared text-normalization helper.
                    normalized_text, text_actions = _normalize_text_content(raw_bytes)

                    # Extend the action log with the returned normalization actions.
                    actions.extend(text_actions)

                    # Rewrite the normalized extracted text as UTF-8.
                    out_path.write_text(normalized_text, encoding="utf-8")

                # Handle extractor timeout without halting the run.
                except subprocess.TimeoutExpired:
                    # Record that extraction failed due to timeout.
                    actions.append(
                        "PDF extraction failed: subprocess timed out after 60 seconds."
                    )

                    # Degrade confidence because no valid extraction was produced.
                    confidence = ConfidenceScore.LOW

                    # Write a deterministic placeholder derivative.
                    out_path.write_text(
                        "PDF EXTRACTION FAILED: TIMEOUT.",
                        encoding="utf-8",
                    )

                    # Emit a warning for the audit trail.
                    logger.warning(f"Timeout extracting PDF: {raw_abs_path}")

                # Handle execution errors or missing external binaries.
                except (subprocess.CalledProcessError, FileNotFoundError) as e:
                    # Record the execution failure text in the action log.
                    actions.append(f"PDF extraction failed: {str(e)}")

                    # Degrade confidence because extraction did not complete successfully.
                    confidence = ConfidenceScore.LOW

                    # Write a deterministic placeholder derivative.
                    out_path.write_text(
                        "PDF EXTRACTION FAILED: EXECUTION ERROR.",
                        encoding="utf-8",
                    )

                    # Emit a warning for the audit trail.
                    logger.warning(f"Error extracting PDF: {raw_abs_path} - {e}")

                # Compute the SHA-256 hash of the resulting derivative file.
                content_hash = hashlib.sha256(out_path.read_bytes()).hexdigest()

                # Append the derivative record for the PDF-derived text artifact.
                derivatives.append(
                    NormalizedDerivativeRecord(
                        raw_artifact_id=item.id,
                        derivative_id=content_hash[:16],
                        derivative_path=str(out_path.relative_to(repo_root)),
                        normalization_actions=actions,
                        confidence_score=confidence,
                        content_hash=content_hash,
                    )
                )

            # -----------------------------------------------------------------
            # Modality 3: Git Repositories (GIT_REPO_SNAPSHOT)
            # -----------------------------------------------------------------

            # Route repository snapshots into a deterministic textual repository summary.
            elif item.type == SourceType.GIT_REPO_SNAPSHOT.value:
                # Construct the repository-summary derivative path.
                out_path = normalized_dir / f"{item.id}_{safe_stem}_summary.md"

                # Initialize the repository-summary payload with a title header.
                repo_summary: List[str] = [
                    f"# Repository Summary: {item.title_or_name}\n"
                ]

                # Construct the expected README path.
                readme_path = raw_abs_path / "README.md"

                # Include the README content if present.
                if readme_path.exists():
                    # Append a README section header.
                    repo_summary.append("## README\n")

                    # Append the README contents using replacement decoding for robustness.
                    repo_summary.append(
                        readme_path.read_text(
                            encoding="utf-8",
                            errors="replace",
                        )
                    )

                    # Record that the README was included.
                    actions.append("Included README.md")

                # Define the preferred repository entry-point files.
                entrypoints = ["pyproject.toml", "package.json", "setup.py"]

                # Iterate through the preferred entry points.
                for ep in entrypoints:
                    # Construct the absolute path to the entry-point file.
                    ep_path = raw_abs_path / ep

                    # Include the file if it exists.
                    if ep_path.exists():
                        # Append a section header and opening code fence for the entry point.
                        repo_summary.append(f"\n## {ep}\n```\n")

                        # Append the file content using replacement decoding for robustness.
                        repo_summary.append(
                            ep_path.read_text(
                                encoding="utf-8",
                                errors="replace",
                            )
                        )

                        # Append the closing code fence.
                        repo_summary.append("\n```\n")

                        # Record that the entry point was included.
                        actions.append(f"Included entrypoint {ep}")

                # Append the sampled source-tree section header and code fence.
                repo_summary.append("\n## Source Tree (Sampled)\n```text\n")

                # Initialize the sampled file counter.
                file_count = 0

                # Traverse the repository snapshot directory tree.
                for root, dirs, files in os.walk(raw_abs_path):
                    # Exclude hidden directories and common build-artifact directories.
                    dirs[:] = [
                        d
                        for d in dirs
                        if not d.startswith(".")
                        and d not in (
                            "__pycache__",
                            "node_modules",
                            "build",
                            "dist",
                        )
                    ]

                    # Compute the nesting level relative to the repository root.
                    level = str(root).replace(str(raw_abs_path), "").count(os.sep)

                    # Construct the indentation for the current directory level.
                    indent = " " * 4 * level

                    # Append the current directory name to the source-tree summary.
                    repo_summary.append(f"{indent}{os.path.basename(root)}/")

                    # Construct the indentation for child files.
                    subindent = " " * 4 * (level + 1)

                    # Iterate through files in the current directory.
                    for f in files:
                        # Truncate the sample at 50 files to honor the configured policy.
                        if file_count >= 50:
                            # Record the truncation marker in the summary.
                            repo_summary.append(
                                f"{subindent}... (truncated at 50 files)"
                            )

                            # Exit the inner file loop because the sample cap was reached.
                            break

                        # Append the current filename to the source-tree summary.
                        repo_summary.append(f"{subindent}{f}")

                        # Increment the sampled file count.
                        file_count += 1

                    # Exit the outer traversal loop once the sample cap has been reached.
                    if file_count >= 50:
                        # Break the directory traversal loop.
                        break

                # Append the closing code fence for the source-tree section.
                repo_summary.append("\n```\n")

                # Record that a sampled source tree was generated.
                actions.append(f"Generated source tree (sampled {file_count} files)")

                # Join the summary parts into a single normalized Markdown string.
                normalized_text = "\n".join(repo_summary)

                # Write the repository summary derivative to disk.
                out_path.write_text(normalized_text, encoding="utf-8")

                # Compute the SHA-256 hash of the repository summary derivative.
                content_hash = hashlib.sha256(
                    normalized_text.encode("utf-8")
                ).hexdigest()

                # Append the derivative record for the repository summary.
                derivatives.append(
                    NormalizedDerivativeRecord(
                        raw_artifact_id=item.id,
                        derivative_id=content_hash[:16],
                        derivative_path=str(out_path.relative_to(repo_root)),
                        normalization_actions=actions,
                        confidence_score=confidence,
                        content_hash=content_hash,
                    )
                )

            # -----------------------------------------------------------------
            # Modality 4: Datasets (DATASET)
            # -----------------------------------------------------------------

            # Route datasets into normalized dataset-card derivatives.
            elif item.type == SourceType.DATASET.value:
                # Construct the default sidecar dataset-card path.
                card_path = raw_abs_path / "DATASET_CARD.yaml"

                # Fall back to a sibling sidecar path when the raw artifact is a file.
                if not card_path.exists() and raw_abs_path.is_file():
                    # Construct the sibling dataset-card path based on the file stem.
                    card_path = (
                        raw_abs_path.parent
                        / f"{raw_abs_path.stem}_DATASET_CARD.yaml"
                    )

                # Construct the normalized dataset-card derivative path.
                out_path = (
                    normalized_dir
                    / f"{item.id}_{safe_stem}_NORMALIZED_CARD.yaml"
                )

                # Proceed with parsing when a dataset card exists.
                if card_path.exists():
                    # Open the dataset card for UTF-8 reading.
                    with open(card_path, "r", encoding="utf-8") as f:
                        # Parse the YAML card, defaulting to an empty mapping if the file is empty.
                        card_data = yaml.safe_load(f) or {}

                    # Normalize dataset-card field names by lowercasing and stripping whitespace.
                    normalized_card = {
                        str(k).strip().lower(): v
                        for k, v in card_data.items()
                    }

                    # Record that field names were standardized.
                    actions.append("Standardized dataset card field names")

                    # Open the normalized-card output path for UTF-8 writing.
                    with open(out_path, "w", encoding="utf-8") as f:
                        # Serialize the normalized card with sorted keys for determinism.
                        yaml.dump(normalized_card, f, sort_keys=True)

                # Emit a placeholder when no dataset card could be found.
                else:
                    # Record that no dataset card was found.
                    actions.append("No dataset card found; created empty placeholder")

                    # Degrade confidence because structural metadata is missing.
                    confidence = ConfidenceScore.LOW

                    # Write a deterministic placeholder schema payload.
                    out_path.write_text("schema: unknown\n", encoding="utf-8")

                # Compute the SHA-256 hash of the resulting normalized-card derivative.
                content_hash = hashlib.sha256(out_path.read_bytes()).hexdigest()

                # Append the derivative record for the dataset-card derivative.
                derivatives.append(
                    NormalizedDerivativeRecord(
                        raw_artifact_id=item.id,
                        derivative_id=content_hash[:16],
                        derivative_path=str(out_path.relative_to(repo_root)),
                        normalization_actions=actions,
                        confidence_score=confidence,
                        content_hash=content_hash,
                    )
                )

            # -----------------------------------------------------------------
            # Modality 5: Images (IMAGE)
            # -----------------------------------------------------------------

            # Route images by copying their bytes into stable derivative paths.
            elif item.type == SourceType.IMAGE.value:
                # Construct the derivative output path using a stable ID-prefixed filename.
                out_path = (
                    normalized_dir
                    / f"{item.id}_{safe_stem}{raw_abs_path.suffix}"
                )

                # Copy the source bytes directly to preserve binary integrity.
                out_path.write_bytes(raw_abs_path.read_bytes())

                # Record that the image was copied into the normalized directory.
                actions.append(
                    "Copied image to normalized directory with stable ID prefix"
                )

                # Compute the SHA-256 hash of the copied image derivative.
                content_hash = hashlib.sha256(out_path.read_bytes()).hexdigest()

                # Append the derivative record for the copied image.
                derivatives.append(
                    NormalizedDerivativeRecord(
                        raw_artifact_id=item.id,
                        derivative_id=content_hash[:16],
                        derivative_path=str(out_path.relative_to(repo_root)),
                        normalization_actions=actions,
                        confidence_score=confidence,
                        content_hash=content_hash,
                    )
                )

        # Catch unexpected per-item exceptions so the batch can continue.
        except Exception as e:
            # Log the unexpected normalization error with the raw artifact identifier.
            logger.error(f"Unexpected error normalizing item {item.id}: {e}")

    # Return the ordered list of derivative records produced during this batch.
    return derivatives


# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 3: Validate cleansing quality
# -------------------------------------------------------------------------------------------------------------------------------

# Define the derivative quality validation function.
def validate_derivative_quality(
    derivatives: List[NormalizedDerivativeRecord],
    repo_root: pathlib.Path,
) -> List[NormalizedDerivativeRecord]:
    """
    Validate the quality of normalized derivatives using deterministic heuristics.

    Purpose:
        Assess the quality of normalized derivatives by checking readability,
        structural preservation, and non-emptiness, thereby assigning final
        confidence scores before downstream use.

    Inputs:
        derivatives:
            Draft normalized derivative records produced by
            ``process_and_normalize_items``.
        repo_root:
            Project root directory used to resolve derivative paths.

    Processes:
        1. Validate top-level inputs.
        2. For each derivative record:
           a. Validate the record structure.
           b. Resolve the absolute path of the normalized derivative.
           c. For text-based derivatives (.md, .txt):
              - check non-emptiness (minimum 10 characters),
              - check alphanumeric density as a readability proxy (threshold:
                0.4),
              - for Markdown, check heading preservation,
              - downgrade confidence to LOW if critical checks fail,
              - downgrade confidence to MEDIUM if structural warnings are
                detected.
           d. For non-text derivatives (images, YAML), preserve the existing
              confidence score.
           e. Construct a new immutable record with updated actions and
              confidence.
        3. Return the ordered list of validated derivative records.

    Outputs:
        List[NormalizedDerivativeRecord]:
            Validated derivative records with finalized confidence scores.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.

    Notes:
        This callable enforces the study's quality-assurance discipline by
        applying deterministic heuristics to detect low-quality extractions and
        normalizations.

        The original validation logic is preserved exactly for valid inputs.
    """
    # Validate that the derivatives input is a list.
    if not isinstance(derivatives, list):
        # Raise a type error because quality validation requires a list of
        # derivative records.
        raise TypeError(
            "validate_derivative_quality expected 'derivatives' to be a list of "
            "NormalizedDerivativeRecord objects."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "validate_derivative_quality expected 'repo_root' to be a "
            "pathlib.Path instance."
        )

    # Initialize the ordered list that will hold validated derivative records.
    validated_derivatives: List[NormalizedDerivativeRecord] = []

    # Iterate over each draft derivative record.
    for record in derivatives:
        # Validate that the current record has the expected dataclass type.
        if not isinstance(record, NormalizedDerivativeRecord):
            # Raise a type error because downstream field access assumes the
            # normalized-derivative schema.
            raise TypeError(
                "Each element of 'derivatives' must be a "
                "NormalizedDerivativeRecord instance."
            )

        # Resolve the absolute path of the normalized derivative.
        abs_path = (repo_root / record.derivative_path).resolve()

        # Apply text-based quality heuristics only for .md and .txt files.
        if abs_path.suffix.lower() in [".md", ".txt"]:
            # Read the normalized derivative text, ignoring decoding errors.
            text = abs_path.read_text(encoding="utf-8", errors="ignore")

            # Create a mutable copy of the normalization actions list.
            new_actions = list(record.normalization_actions)

            # Initialize the new confidence score from the existing record.
            new_confidence = record.confidence_score

            # 1. Check non-emptiness (minimum 10 characters after stripping).
            if len(text.strip()) < 10:
                # Log the critical non-emptiness failure.
                new_actions.append("Validation Failed: File is essentially empty.")

                # Downgrade confidence to LOW because the derivative is empty.
                new_confidence = ConfidenceScore.LOW

            else:
                # 2. Check readability via alphanumeric density.
                alnum_count = sum(c.isalnum() for c in text)

                # Compute the density as the fraction of alphanumeric characters.
                density = alnum_count / len(text)

                # Check whether density is below the 0.4 threshold for "garbage"
                # text.
                if density < 0.4:
                    # Log the low-density warning.
                    new_actions.append(
                        f"Validation Warning: Low alphanumeric density "
                        f"({density:.2f}). Possible extraction error."
                    )

                    # Downgrade confidence to LOW because the derivative appears
                    # to be garbled.
                    new_confidence = ConfidenceScore.LOW

                # 3. Check heading preservation for Markdown files.
                if abs_path.suffix.lower() == ".md":
                    # Search for Markdown headings using a multiline regex.
                    has_headings = bool(re.search(r"^#+ ", text, re.MULTILINE))

                    # Log a structural warning if no headings were detected.
                    if not has_headings:
                        new_actions.append(
                            "Validation Warning: No Markdown headings detected. "
                            "Structure may be lost."
                        )

                        # Downgrade confidence to MEDIUM if it was previously HIGH.
                        if new_confidence == ConfidenceScore.HIGH:
                            new_confidence = ConfidenceScore.MEDIUM

            # Construct a new immutable record with updated validation actions and
            # confidence score.
            validated_derivatives.append(
                NormalizedDerivativeRecord(
                    raw_artifact_id=record.raw_artifact_id,
                    derivative_id=record.derivative_id,
                    derivative_path=record.derivative_path,
                    normalization_actions=new_actions,
                    confidence_score=new_confidence,
                    content_hash=record.content_hash,
                )
            )

        else:
            # For non-text files (images, YAML), assume prior steps handled
            # validation and preserve the existing record.
            validated_derivatives.append(record)

    # Return the ordered list of validated derivative records.
    return validated_derivatives


# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable normalization-context aggregate returned by Task 5.
@dataclass(frozen=True)
class NormalizationContext:
    """
    Immutable record encapsulating the results of the cleansing and normalization
    phase.

    Purpose:
        Package the validated normalized derivative records into one immutable
        object for downstream tasks.

    Inputs:
        This dataclass is populated by ``task_5_orchestrator`` after executing
        the normalization and validation steps.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        normalized_derivatives:
            Ordered list of validated normalized derivative records.
    """

    # Store the ordered list of validated normalized derivative records.
    normalized_derivatives: List[NormalizedDerivativeRecord]


# Define the Task 5 orchestrator.
def task_5_orchestrator(
    ingestion_context: IngestionContext,
    fs_context: FilesystemSubstrateContext,
    repo_root: pathlib.Path,
) -> NormalizationContext:
    """
    Orchestrate Task 5 end-to-end by cleansing, normalizing, and validating
    ingested data.

    Purpose:
        This function is the Task 5 entry point. It ensures the normalized
        derivatives directory exists, dispatches accepted raw items to modality-
        specific normalization routines, validates the resulting derivatives, and
        returns an immutable normalization context.

    Inputs:
        ingestion_context:
            Ingestion context produced by Task 4, containing accepted raw
            manifest items.
        fs_context:
            Filesystem substrate context produced by Task 3, providing access to
            the wiki layout and write policy.
        repo_root:
            Project root directory used to compute repo-relative paths and
            resolve absolute paths for validation.

    Processes:
        1. Validate top-level inputs.
        2. Resolve the target ``normalized/`` directory under
           ``wiki/derived/normalized/``.
        3. Ensure the normalized directory exists, creating it idempotently if
           absent.
        4. Validate that the normalized directory is permitted by the write
           policy.
        5. Step 1 & 2: Process and normalize accepted items by calling
           ``process_and_normalize_items``.
        6. Step 3: Validate derivative quality by calling
           ``validate_derivative_quality``.
        7. Return an immutable ``NormalizationContext`` aggregating the validated
           derivatives.

    Outputs:
        NormalizationContext:
            Immutable aggregate containing validated normalized derivative records.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.
        PermissionError:
            Raised if the write policy forbids writing to the normalized
            directory.
        OSError:
            Propagated if directory creation fails.

    Notes:
        This orchestrator operationalizes the study's cleansing and normalization
        pipeline without mutating the original append-only raw corpus.

        The original orchestration logic is preserved exactly for valid inputs.
    """
    # Validate that the ingestion context has the expected structural type.
    if not isinstance(ingestion_context, IngestionContext):
        # Raise a type error because normalization requires the ingestion
        # context schema.
        raise TypeError(
            "task_5_orchestrator expected 'ingestion_context' to be an instance "
            "of IngestionContext."
        )

    # Validate that the filesystem substrate context has the expected structural
    # type.
    if not isinstance(fs_context, FilesystemSubstrateContext):
        # Raise a type error because write-policy access requires the filesystem
        # context schema.
        raise TypeError(
            "task_5_orchestrator expected 'fs_context' to be an instance of "
            "FilesystemSubstrateContext."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_5_orchestrator expected 'repo_root' to be a pathlib.Path "
            "instance."
        )

    # Determine the target directory for normalized derivatives by resolving the
    # derived wiki subdirectory from the filesystem context.
    wiki_root = fs_context.wiki_layout.wiki_subdirectories.get(
        "derived",
        repo_root / "wiki" / "derived",
    )

    # Construct the normalized derivatives directory under the derived wiki root.
    normalized_dir = (wiki_root / "normalized").resolve()

    # Ensure the normalized directory exists, creating it idempotently if absent.
    normalized_dir.mkdir(parents=True, exist_ok=True)

    # Validate that the write policy permits writing to the normalized directory.
    if not fs_context.wiki_layout.write_policy.validate_write_path(
        normalized_dir
    ):
        # Raise a permission error because the write policy forbids writes to
        # the normalized directory.
        raise PermissionError(
            f"Write policy forbids writing to {normalized_dir}"
        )

    # Step 1 & 2: Process and normalize accepted items based on their modality.
    draft_derivatives = process_and_normalize_items(
        ingestion_context=ingestion_context,
        repo_root=repo_root,
        normalized_dir=normalized_dir,
    )

    # Step 3: Validate the quality of the generated normalized derivatives.
    validated_derivatives = validate_derivative_quality(
        derivatives=draft_derivatives,
        repo_root=repo_root,
    )

    # Construct and return the immutable normalization context.
    return NormalizationContext(
        normalized_derivatives=validated_derivatives,
    )
````

````python
# Task 6 — Build the raw manifest, compute hashes, and establish change detection

# ==============================================================================
# Task 6: Build the raw manifest, compute hashes, and establish change detection
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 1: Represent each raw artifact and define the manifest
# -------------------------------------------------------------------------------------------------------------------------------

# Define the frozen dataclass for the complete raw manifest.
@dataclass(frozen=True)
class RawManifest:
    """
    Immutable record representing the complete raw corpus manifest.

    Purpose:
        Store the deterministically sorted raw manifest items and provide O(1)
        lookup access by artifact ID.

    Inputs:
        This dataclass is populated by ``build_and_serialize_raw_manifest``.

    Processes:
        No computation is performed here. It is a passive immutable manifest
        container.

    Outputs:
        items:
            Ordered list of raw manifest items sorted deterministically by ID to
            minimize diff noise.
        item_index:
            O(1) lookup dictionary mapping artifact IDs to manifest items.
        manifest_path:
            Absolute path to the serialized ``raw/manifest.yaml`` file.

    Notes:
        This dataclass operationalizes the study's raw manifest formalization:
        \[
        M_{\text{raw}} = \{ (\text{id}(d), \text{metadata}(d)) \mid
        d \in \mathcal{D}_{\text{raw}} \}
        \]
    """

    # Store the deterministically sorted list of raw manifest items.
    items: List[RawManifestItem]

    # Store the O(1) lookup index from artifact ID to manifest item.
    item_index: Dict[str, RawManifestItem]

    # Store the absolute path to the serialized raw manifest file.
    manifest_path: pathlib.Path


# Define the manifest construction and serialization function.
def build_and_serialize_raw_manifest(
    ingestion_context: IngestionContext,
    fs_context: FilesystemSubstrateContext,
) -> RawManifest:
    """
    Construct the formal raw manifest and serialize it deterministically.

    Purpose:
        Build the unified raw manifest by merging newly accepted items with any
        existing manifest entries, sorting the result deterministically by ID,
        and serializing to ``raw/manifest.yaml`` using atomic write operations.

    Inputs:
        ingestion_context:
            Ingestion context containing newly accepted raw manifest items from
            Task 4.
        fs_context:
            Filesystem substrate context providing the absolute path to the raw
            manifest file.

    Processes:
        1. Validate top-level inputs.
        2. Resolve the raw manifest path from the filesystem context.
        3. Initialize the default manifest structure with version, creation time,
           and an empty items list.
        4. If the manifest file already exists:
           - load its contents,
           - preserve the existing version and creation timestamp,
           - extract existing items into a temporary list.
        5. Build an index of existing items keyed by ID.
        6. Merge newly accepted items into the index, avoiding ID collisions.
        7. Sort the merged items deterministically by ID to minimize diff noise.
        8. Serialize the sorted manifest to a temporary file using YAML.
        9. Atomically replace the existing manifest file via ``os.replace``.
        10. Reconstruct typed ``RawManifestItem`` objects from the serialized
            dictionaries for the in-memory representation.
        11. Return the immutable ``RawManifest`` object.

    Outputs:
        RawManifest:
            Immutable structured representation of the raw manifest containing:
            - deterministically sorted items,
            - O(1) lookup index,
            - absolute path to the manifest file.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.
        OSError:
            Propagated if manifest file read, write, or atomic replacement fails.
        yaml.YAMLError:
            Could propagate if YAML parsing or serialization fails unexpectedly.

    Notes:
        This callable implements the manifest-construction step required for the
        study's change-detection formalism:
        \[
        M_{\text{raw}} = \{ (\text{id}(d), \text{metadata}(d)) \mid
        d \in \mathcal{D}_{\text{raw}} \}
        \]

        The original manifest construction logic is preserved exactly for valid
        inputs.
    """
    # Validate that the ingestion context has the expected structural type.
    if not isinstance(ingestion_context, IngestionContext):
        # Raise a type error because manifest construction requires the ingestion
        # context schema.
        raise TypeError(
            "build_and_serialize_raw_manifest expected 'ingestion_context' to "
            "be an instance of IngestionContext."
        )

    # Validate that the filesystem substrate context has the expected structural
    # type.
    if not isinstance(fs_context, FilesystemSubstrateContext):
        # Raise a type error because manifest-path resolution requires the
        # filesystem context schema.
        raise TypeError(
            "build_and_serialize_raw_manifest expected 'fs_context' to be an "
            "instance of FilesystemSubstrateContext."
        )

    # Resolve the absolute path to the raw manifest file from the filesystem
    # context.
    manifest_path: pathlib.Path = fs_context.raw_layout.manifest_path

    # Validate that the manifest path is a pathlib.Path.
    if not isinstance(manifest_path, pathlib.Path):
        # Raise a type error because file I/O requires a Path object.
        raise TypeError(
            "build_and_serialize_raw_manifest expected "
            "'fs_context.raw_layout.manifest_path' to be a pathlib.Path "
            "instance."
        )

    # Initialize the base manifest structure with the default version, UTC
    # creation timestamp, and an empty items list.
    manifest_data: Dict[str, Any] = {
        "version": 1,
        "created_at": datetime.datetime.utcnow().isoformat() + "Z",
        "items": [],
    }

    # Load the existing manifest if it exists to preserve prior entries under the
    # append-only rule.
    if manifest_path.exists():
        # Open the existing manifest for UTF-8 reading.
        with open(manifest_path, "r", encoding="utf-8") as f:
            # Parse the YAML payload safely.
            loaded_data = yaml.safe_load(f)

            # Preserve the existing manifest structure if it is truthy and contains
            # items.
            if loaded_data and "items" in loaded_data:
                # Preserve the existing version.
                manifest_data["version"] = loaded_data.get(
                    "version",
                    1,
                )

                # Preserve the existing creation timestamp.
                manifest_data["created_at"] = loaded_data.get(
                    "created_at",
                    manifest_data["created_at"],
                )

                # Preserve the existing items list.
                manifest_data["items"] = loaded_data["items"]

    # Create an index of existing items keyed by ID to prevent duplicates during
    # merging.
    existing_items_index: Dict[str, Dict[str, Any]] = {
        item["id"]: item
        for item in manifest_data["items"]
        if "id" in item
    }

    # Merge newly accepted items from the ingestion context into the index.
    for new_item in ingestion_context.accepted_items:
        # Validate that each new item is of the expected dataclass type.
        if not isinstance(new_item, RawManifestItem):
            # Raise a type error because manifest merging assumes the manifest-item
            # schema.
            raise TypeError(
                "Each accepted item in 'ingestion_context' must be a "
                "RawManifestItem instance."
            )

        # Only add new items that do not already exist in the index.
        if new_item.id not in existing_items_index:
            # Convert the dataclass to a dictionary, excluding None values for
            # cleaner YAML serialization.
            item_dict = {
                k: v for k, v in asdict(new_item).items() if v is not None
            }

            # Insert the new item into the index.
            existing_items_index[new_item.id] = item_dict

    # Sort the merged items deterministically by ID to minimize diff noise across
    # runs.
    sorted_items_dicts: List[Dict[str, Any]] = sorted(
        existing_items_index.values(),
        key=lambda x: x["id"],
    )

    # Update the manifest payload with the sorted items list.
    manifest_data["items"] = sorted_items_dicts

    # Serialize the manifest to a temporary file to enable atomic replacement.
    temp_manifest_path = manifest_path.with_suffix(".yaml.tmp")

    # Open the temporary manifest path for UTF-8 writing.
    with open(temp_manifest_path, "w", encoding="utf-8") as f:
        # Serialize the manifest payload to YAML using block style and preserving
        # the deterministic key order established during sorting.
        yaml.dump(
            manifest_data,
            f,
            default_flow_style=False,
            sort_keys=False,
        )

    # Atomically replace the existing manifest file with the temporary file to
    # prevent corruption from interrupted writes.
    os.replace(temp_manifest_path, manifest_path)

    # Reconstruct typed RawManifestItem objects from the serialized dictionaries
    # for the in-memory representation.
    final_items: List[RawManifestItem] = []

    # Initialize the O(1) lookup index from artifact ID to manifest item.
    final_index: Dict[str, RawManifestItem] = {}

    # Iterate over the sorted manifest dictionaries to rebuild typed objects.
    for item_dict in sorted_items_dicts:
        # Reconstruct the RawManifestItem dataclass from the dictionary,
        # handling optional fields gracefully via ``.get()``.
        item = RawManifestItem(
            id=item_dict["id"],
            type=item_dict["type"],
            title_or_name=item_dict["title_or_name"],
            ingested_at=item_dict["ingested_at"],
            local_path=item_dict["local_path"],
            source_url=item_dict.get("source_url"),
            authors=item_dict.get("authors"),
            year=item_dict.get("year"),
            venue=item_dict.get("venue"),
            revision=item_dict.get("revision"),
            tags=item_dict.get("tags"),
            notes=item_dict.get("notes"),
        )

        # Append the typed item to the ordered list.
        final_items.append(item)

        # Index the typed item by its ID for O(1) lookup.
        final_index[item.id] = item

    # Construct and return the immutable raw manifest record.
    return RawManifest(
        items=final_items,
        item_index=final_index,
        manifest_path=manifest_path,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 2: Compute cryptographic hashes for raw files and normalized derivatives
# -------------------------------------------------------------------------------------------------------------------------------

# Define the frozen dataclass for one hash record.
@dataclass(frozen=True)
class HashRecord:
    """
    Immutable record containing the cryptographic hash of one file.

    Purpose:
        Store the SHA-256 hash and associated metadata for one raw artifact or
        normalized derivative, enabling change detection across runs.

    Inputs:
        This dataclass is populated by ``compute_cryptographic_hashes``.

    Processes:
        No computation is performed here. It is a passive immutable hash record.

    Outputs:
        artifact_id:
            Identifier of the artifact (raw or derivative).
        file_path:
            Repo-relative path to the file.
        sha256_hash:
            Computed SHA-256 hexadecimal digest.
        timestamp_utc:
            ISO 8601 timestamp of when the hash was computed.
        is_derivative:
            Whether this is a normalized derivative (True) or a raw artifact
            (False).

    Notes:
        This record supports the study's change-detection formalism by storing
        the hash values used in the delta-set computation:
        \[
        \Delta \mathcal{D} \leftarrow \{ d \in \mathcal{D}_{\text{raw}}^{(t)}
        \mid \text{hash}(d) \notin \text{hash}(\mathcal{D}_{\text{raw}}^{(t-1)})
        \}
        \]
    """

    # Store the identifier of the artifact (raw or derivative).
    artifact_id: str

    # Store the repo-relative path to the file.
    file_path: str

    # Store the computed SHA-256 hexadecimal digest.
    sha256_hash: str

    # Store the UTC ISO 8601 timestamp of the hash computation.
    timestamp_utc: str

    # Store whether this is a derivative (True) or raw artifact (False).
    is_derivative: bool


# Define the frozen dataclass for the hash registry.
@dataclass(frozen=True)
class HashRegistry:
    """
    Immutable registry of all computed cryptographic hashes.

    Purpose:
        Store the complete ledger of artifact hashes for the current run, indexed
        by artifact ID.

    Inputs:
        This dataclass is populated by ``compute_cryptographic_hashes``.

    Processes:
        No computation is performed here. It is a passive immutable hash registry.

    Outputs:
        records:
            Mapping from artifact IDs to hash records.
    """

    # Store the mapping from artifact IDs to hash records.
    records: Dict[str, HashRecord]


# Define the internal file-hashing helper function.
def _compute_file_sha256(
    file_path: pathlib.Path,
    chunk_size: int = 4 * 1024 * 1024,
) -> str:
    """
    Compute the SHA-256 hash of a file efficiently using chunked reading.

    Purpose:
        Deterministically compute the SHA-256 hexadecimal digest of a file
        without loading the entire file into memory.

    Inputs:
        file_path:
            Absolute path to the file to hash.
        chunk_size:
            Number of bytes to read per iteration.

    Processes:
        1. Validate that the file path exists and is a file.
        2. Initialize a SHA-256 hasher object.
        3. Open the file for binary reading.
        4. Read the file in chunks of ``chunk_size`` bytes.
        5. Update the hasher with each chunk.
        6. Return the hexadecimal digest.

    Outputs:
        str:
            The SHA-256 hexadecimal digest of the file.

    Raises:
        TypeError:
            Raised if ``file_path`` is not a ``pathlib.Path``.
        FileNotFoundError:
            Raised if the file does not exist or is not a file.

    Notes:
        The default chunk size of 4 MB balances memory efficiency and I/O
        performance.

        This helper operationalizes the hash component of the study's delta
        formalism.
    """
    # Validate that the file path is a pathlib.Path.
    if not isinstance(file_path, pathlib.Path):
        # Raise a type error because path operations require pathlib.Path.
        raise TypeError(
            "_compute_file_sha256 expected 'file_path' to be a pathlib.Path "
            "instance."
        )

    # Validate that the file exists and is a regular file.
    if not file_path.exists() or not file_path.is_file():
        # Raise a file-not-found error because the hash cannot be computed
        # otherwise.
        raise FileNotFoundError(
            f"Cannot compute hash; file missing or invalid: {file_path}"
        )

    # Initialize a SHA-256 hasher object.
    hasher = hashlib.sha256()

    # Open the file for binary reading.
    with open(file_path, "rb") as f:
        # Read the file in chunks and update the hasher iteratively.
        while chunk := f.read(chunk_size):
            # Update the hasher with the current chunk.
            hasher.update(chunk)

    # Return the hexadecimal digest of the complete file.
    return hasher.hexdigest()


# Define the hash-computation function.
def compute_cryptographic_hashes(
    raw_manifest: RawManifest,
    normalization_context: NormalizationContext,
    repo_root: pathlib.Path,
) -> HashRegistry:
    """
    Compute SHA-256 hashes for all raw files and normalized derivatives.

    Purpose:
        Build a complete ledger of cryptographic hashes for the current corpus
        state, enabling deterministic change detection across runs.

    Inputs:
        raw_manifest:
            The complete raw manifest containing all accepted raw items.
        normalization_context:
            The registry of normalized derivatives from Task 5.
        repo_root:
            Project root directory used to resolve repo-relative paths.

    Processes:
        1. Validate top-level inputs.
        2. Generate a UTC ISO 8601 timestamp for all hash records.
        3. For each raw artifact in the manifest:
           - resolve its absolute path,
           - for directories (e.g., git repos), hash a sidecar file
             (``REPO_SNAPSHOT.yaml`` or ``DATASET_CARD.yaml``) as a proxy,
           - for files, hash the file directly,
           - construct and store a ``HashRecord``.
        4. For each normalized derivative:
           - resolve its absolute path,
           - compute the SHA-256 hash of the derivative file,
           - construct and store a ``HashRecord``.
        5. Return the immutable ``HashRegistry``.

    Outputs:
        HashRegistry:
            Immutable structured ledger of all computed hashes.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.
        FileNotFoundError:
            Propagated from ``_compute_file_sha256`` if a file is missing.

    Notes:
        This callable operationalizes the hash component of the study's delta
        formalism:
        \[
        \text{hash}(d)
        \]
        used in:
        \[
        \Delta \mathcal{D} \leftarrow \{ d \in \mathcal{D}_{\text{raw}}^{(t)}
        \mid \text{hash}(d) \notin \text{hash}(\mathcal{D}_{\text{raw}}^{(t-1)})
        \}
        \]

        The original hash-computation logic is preserved exactly for valid inputs.
    """
    # Validate that the raw manifest has the expected structural type.
    if not isinstance(raw_manifest, RawManifest):
        # Raise a type error because hash computation requires the raw manifest
        # schema.
        raise TypeError(
            "compute_cryptographic_hashes expected 'raw_manifest' to be an "
            "instance of RawManifest."
        )

    # Validate that the normalization context has the expected structural type.
    if not isinstance(normalization_context, NormalizationContext):
        # Raise a type error because derivative hashing requires the normalization
        # context schema.
        raise TypeError(
            "compute_cryptographic_hashes expected 'normalization_context' to "
            "be an instance of NormalizationContext."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "compute_cryptographic_hashes expected 'repo_root' to be a "
            "pathlib.Path instance."
        )

    # Initialize the hash-record mapping keyed by artifact ID.
    records: Dict[str, HashRecord] = {}

    # Generate a single UTC ISO 8601 timestamp for all hash records in this run.
    current_time: str = datetime.datetime.utcnow().isoformat() + "Z"

    # 1. Hash Raw Artifacts.

    # Iterate over each raw manifest item.
    for item in raw_manifest.items:
        # Validate that the current item is of the expected dataclass type.
        if not isinstance(item, RawManifestItem):
            # Raise a type error because hash computation assumes the manifest-item
            # schema.
            raise TypeError(
                "Each item in 'raw_manifest.items' must be a RawManifestItem "
                "instance."
            )

        # Resolve the absolute path of the raw artifact.
        abs_path = (repo_root / item.local_path).resolve()

        # For directories (e.g., git repos, datasets), hash a sidecar file as a
        # proxy because hashing entire directory trees is complex and the
        # normalized derivative already captures the content.
        if abs_path.is_dir():
            # Attempt to locate the repository snapshot sidecar.
            target_file = abs_path / "REPO_SNAPSHOT.yaml"

            # Fall back to the dataset card sidecar if the repo snapshot is absent.
            if not target_file.exists():
                target_file = abs_path / "DATASET_CARD.yaml"

        else:
            # For files, hash the file directly.
            target_file = abs_path

        # Compute and record the hash if the target file exists and is a regular
        # file.
        if target_file.exists() and target_file.is_file():
            # Compute the SHA-256 hash of the target file.
            file_hash = _compute_file_sha256(target_file)

            # Construct and store the hash record for the current raw artifact.
            records[item.id] = HashRecord(
                artifact_id=item.id,
                file_path=item.local_path,
                sha256_hash=file_hash,
                timestamp_utc=current_time,
                is_derivative=False,
            )

    # 2. Hash Normalized Derivatives.

    # Iterate over each normalized derivative record.
    for deriv in normalization_context.normalized_derivatives:
        # Validate that the current derivative is of the expected dataclass type.
        if not isinstance(deriv, NormalizedDerivativeRecord):
            # Raise a type error because derivative hashing assumes the normalized-
            # derivative schema.
            raise TypeError(
                "Each derivative in 'normalization_context.normalized_derivatives' "
                "must be a NormalizedDerivativeRecord instance."
            )

        # Resolve the absolute path of the normalized derivative.
        abs_path = (repo_root / deriv.derivative_path).resolve()

        # Compute and record the hash if the derivative file exists and is a
        # regular file.
        if abs_path.exists() and abs_path.is_file():
            # Compute the SHA-256 hash of the derivative file to ensure absolute
            # filesystem fidelity at this exact moment.
            file_hash = _compute_file_sha256(abs_path)

            # Construct and store the hash record for the current derivative.
            records[deriv.derivative_id] = HashRecord(
                artifact_id=deriv.derivative_id,
                file_path=deriv.derivative_path,
                sha256_hash=file_hash,
                timestamp_utc=current_time,
                is_derivative=True,
            )

    # Construct and return the immutable hash registry.
    return HashRegistry(records=records)


# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 3: Define and operationalize the incremental delta set
# -------------------------------------------------------------------------------------------------------------------------------

# Define the frozen dataclass for the incremental delta set.
@dataclass(frozen=True)
class DeltaSet:
    """
    Immutable record representing the incremental changes detected in the corpus.

    Purpose:
        Store the categorized lists of artifact IDs that are new, changed, or
        unchanged compared to the previous hash snapshot.

    Inputs:
        This dataclass is populated by ``detect_incremental_deltas``.

    Processes:
        No computation is performed here. It is a passive immutable delta record.

    Outputs:
        changed_artifacts:
            Ordered list of artifact IDs whose hashes have changed since the
            previous snapshot.
        new_artifacts:
            Ordered list of artifact IDs that are newly observed in the current
            run.
        unchanged_artifacts:
            Ordered list of artifact IDs that remain identical to the previous
            snapshot.

    Notes:
        This dataclass operationalizes the study's delta-set formalism:
        \[
        \Delta \mathcal{D} \leftarrow \{ d \in \mathcal{D}_{\text{raw}}^{(t)}
        \mid \text{hash}(d) \notin \text{hash}(\mathcal{D}_{\text{raw}}^{(t-1)})
        \}
        \]
    """

    # Store the ordered list of artifact IDs whose hashes have changed.
    changed_artifacts: List[str]

    # Store the ordered list of newly observed artifact IDs.
    new_artifacts: List[str]

    # Store the ordered list of unchanged artifact IDs.
    unchanged_artifacts: List[str]


# Define the delta-detection function.
def detect_incremental_deltas(
    current_registry: HashRegistry,
    fs_context: FilesystemSubstrateContext,
) -> DeltaSet:
    """
    Compare current hashes against the previous snapshot to compute the delta set.

    Purpose:
        Operationalize the study's delta-detection formalism by comparing the
        current hash registry against the previous hash snapshot, categorizing
        artifacts into new, changed, and unchanged sets, and persisting the
        current registry for future runs.

    Inputs:
        current_registry:
            The newly computed hash registry for the current run.
        fs_context:
            Filesystem substrate context providing the logs/ directory path for
            snapshot storage.

    Processes:
        1. Validate top-level inputs.
        2. Resolve the logs/ directory and the symlink to the latest snapshot.
        3. Load the previous hash snapshot if it exists.
        4. For each artifact in the current registry:
           - if the artifact ID is absent from the previous snapshot, classify as
             new,
           - if the artifact hash differs from the previous hash, classify as
             changed,
           - otherwise, classify as unchanged.
        5. Persist the current hash registry as the new snapshot using a
           timestamped filename.
        6. Update the ``latest_hash_snapshot.json`` symlink to point to the new
           snapshot.
        7. Return the immutable ``DeltaSet``.

    Outputs:
        DeltaSet:
            Immutable delta set containing categorized artifact ID lists.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.
        OSError:
            Propagated if snapshot read, write, or symlink operations fail.
        json.JSONDecodeError:
            Could propagate if the previous snapshot is corrupted.

    Notes:
        This callable implements the study's delta-detection formalism exactly:
        \[
        \Delta \mathcal{D} \leftarrow \{ d \in \mathcal{D}_{\text{raw}}^{(t)}
        \mid \text{hash}(d) \notin \text{hash}(\mathcal{D}_{\text{raw}}^{(t-1)})
        \}
        \]

        The original delta-detection logic is preserved exactly for valid inputs.
    """
    # Validate that the current hash registry has the expected structural type.
    if not isinstance(current_registry, HashRegistry):
        # Raise a type error because delta detection requires the hash registry
        # schema.
        raise TypeError(
            "detect_incremental_deltas expected 'current_registry' to be an "
            "instance of HashRegistry."
        )

    # Validate that the filesystem substrate context has the expected structural
    # type.
    if not isinstance(fs_context, FilesystemSubstrateContext):
        # Raise a type error because logs-directory resolution requires the
        # filesystem context schema.
        raise TypeError(
            "detect_incremental_deltas expected 'fs_context' to be an instance "
            "of FilesystemSubstrateContext."
        )

    # Resolve the logs/ directory absolute path from the filesystem context.
    logs_dir: pathlib.Path = fs_context.root_states[
        "logs_directory"
    ].absolute_path

    # Validate that the logs directory is a pathlib.Path.
    if not isinstance(logs_dir, pathlib.Path):
        # Raise a type error because file I/O requires a Path object.
        raise TypeError(
            "detect_incremental_deltas expected logs directory to be a "
            "pathlib.Path instance."
        )

    # Construct the symlink path to the latest hash snapshot.
    snapshot_symlink: pathlib.Path = logs_dir / "latest_hash_snapshot.json"

    # Initialize the previous hash mapping as empty.
    previous_hashes: Dict[str, str] = {}

    # Load the previous hash snapshot if it exists.
    if snapshot_symlink.exists():
        try:
            # Open the previous snapshot for UTF-8 reading.
            with open(snapshot_symlink, "r", encoding="utf-8") as f:
                # Parse the JSON payload.
                prev_data = json.load(f)

                # Extract the hash mapping from the serialized registry.
                previous_hashes = {
                    k: v["sha256_hash"]
                    for k, v in prev_data.get("records", {}).items()
                }

        except (json.JSONDecodeError, KeyError) as e:
            # If the snapshot is corrupted, treat everything as new to be safe.
            print(
                f"Warning: Failed to load previous hash snapshot: {e}. "
                "Treating all as new."
            )

    # Initialize the categorized artifact-ID lists.
    changed: List[str] = []

    new: List[str] = []

    unchanged: List[str] = []

    # Compute the delta set by comparing current and previous hashes.
    for artifact_id, record in current_registry.records.items():
        # Validate that each record is of the expected dataclass type.
        if not isinstance(record, HashRecord):
            # Raise a type error because delta computation assumes the hash-record
            # schema.
            raise TypeError(
                "Each record in 'current_registry.records' must be a HashRecord "
                "instance."
            )

        # Extract the current hash for the artifact.
        current_hash = record.sha256_hash

        # Categorize the artifact based on presence and hash equality in the
        # previous snapshot.
        if artifact_id not in previous_hashes:
            # The artifact is newly observed in this run.
            new.append(artifact_id)

        elif previous_hashes[artifact_id] != current_hash:
            # The artifact hash has changed since the previous snapshot.
            changed.append(artifact_id)

        else:
            # The artifact hash remains unchanged.
            unchanged.append(artifact_id)

    # Persist the current hash registry as the new snapshot.

    # Generate a timestamped filename for the new snapshot.
    current_time_str = datetime.datetime.utcnow().strftime("%Y%m%dT%H%M%SZ")

    new_snapshot_path = logs_dir / f"hash_snapshot_{current_time_str}.json"

    # Serialize the dataclass records to a dictionary.
    registry_dict = {
        "records": {k: asdict(v) for k, v in current_registry.records.items()}
    }

    # Open the new snapshot path for UTF-8 writing.
    with open(new_snapshot_path, "w", encoding="utf-8") as f:
        # Serialize the registry to JSON with sorted keys for determinism.
        json.dump(registry_dict, f, indent=2, sort_keys=True)

    # Update the symlink to point to the new snapshot.
    try:
        # Remove the existing symlink if present.
        if snapshot_symlink.exists() or snapshot_symlink.is_symlink():
            snapshot_symlink.unlink()

        # Create a new symlink pointing to the new snapshot filename.
        snapshot_symlink.symlink_to(new_snapshot_path.name)

    except OSError:
        # Fallback for Windows without symlink privileges: copy the file instead.
        import shutil

        shutil.copy2(new_snapshot_path, snapshot_symlink)

    # Construct and return the immutable delta set.
    return DeltaSet(
        changed_artifacts=changed,
        new_artifacts=new,
        unchanged_artifacts=unchanged,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable change-detection context aggregate returned by Task 6.
@dataclass(frozen=True)
class ChangeDetectionContext:
    """
    Immutable record encapsulating the results of the change-detection phase.

    Purpose:
        Package the raw manifest, hash registry, and delta set into one immutable
        object for downstream tasks.

    Inputs:
        This dataclass is populated by ``task_6_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        raw_manifest:
            The complete raw manifest containing all accepted raw items.
        hash_registry:
            The complete hash registry for the current run.
        delta_set:
            The delta set categorizing artifacts into new, changed, and unchanged.
    """

    # Store the complete raw manifest.
    raw_manifest: RawManifest

    # Store the complete hash registry.
    hash_registry: HashRegistry

    # Store the delta set from change detection.
    delta_set: DeltaSet


# Define the Task 6 orchestrator.
def task_6_orchestrator(
    ingestion_context: IngestionContext,
    normalization_context: NormalizationContext,
    fs_context: FilesystemSubstrateContext,
    repo_root: pathlib.Path,
) -> ChangeDetectionContext:
    """
    Orchestrate Task 6 end-to-end by building the manifest, computing hashes, and
    detecting incremental deltas.

    Purpose:
        This function is the Task 6 entry point. It executes the three Task 6
        substeps in the required order and returns an immutable change-detection
        context.

    Inputs:
        ingestion_context:
            Ingestion context produced by Task 4, containing accepted raw items.
        normalization_context:
            Normalization context produced by Task 5, containing normalized
            derivatives.
        fs_context:
            Filesystem substrate context produced by Task 3, providing filesystem
            paths.
        repo_root:
            Project root directory used to resolve repo-relative paths.

    Processes:
        1. Validate top-level inputs.
        2. Step 1: Build and serialize the raw manifest by calling
           ``build_and_serialize_raw_manifest``.
        3. Step 2: Compute cryptographic hashes for all tracked files by calling
           ``compute_cryptographic_hashes``.
        4. Step 3: Detect incremental deltas against the previous snapshot by
           calling ``detect_incremental_deltas``.
        5. Return an immutable ``ChangeDetectionContext`` aggregating the
           manifest, hashes, and deltas.

    Outputs:
        ChangeDetectionContext:
            Immutable aggregate containing the raw manifest, hash registry, and
            delta set.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.

    Notes:
        This orchestrator implements the study's change-detection layer exactly,
        operationalizing the delta formalism:
        \[
        \Delta \mathcal{D} \leftarrow \{ d \in \mathcal{D}_{\text{raw}}^{(t)}
        \mid \text{hash}(d) \notin \text{hash}(\mathcal{D}_{\text{raw}}^{(t-1)})
        \}
        \]

        The original orchestration logic is preserved exactly for valid inputs.
    """
    # Validate that the ingestion context has the expected structural type.
    if not isinstance(ingestion_context, IngestionContext):
        # Raise a type error because manifest construction requires the ingestion
        # context schema.
        raise TypeError(
            "task_6_orchestrator expected 'ingestion_context' to be an instance "
            "of IngestionContext."
        )

    # Validate that the normalization context has the expected structural type.
    if not isinstance(normalization_context, NormalizationContext):
        # Raise a type error because hash computation requires the normalization
        # context schema.
        raise TypeError(
            "task_6_orchestrator expected 'normalization_context' to be an "
            "instance of NormalizationContext."
        )

    # Validate that the filesystem substrate context has the expected structural
    # type.
    if not isinstance(fs_context, FilesystemSubstrateContext):
        # Raise a type error because manifest serialization and delta detection
        # require the filesystem context schema.
        raise TypeError(
            "task_6_orchestrator expected 'fs_context' to be an instance of "
            "FilesystemSubstrateContext."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_6_orchestrator expected 'repo_root' to be a pathlib.Path "
            "instance."
        )

    # Step 1: Build and serialize the raw manifest.
    raw_manifest = build_and_serialize_raw_manifest(
        ingestion_context=ingestion_context,
        fs_context=fs_context,
    )

    # Step 2: Compute cryptographic hashes for all tracked files.
    hash_registry = compute_cryptographic_hashes(
        raw_manifest=raw_manifest,
        normalization_context=normalization_context,
        repo_root=repo_root,
    )

    # Step 3: Detect incremental deltas against the previous snapshot.
    delta_set = detect_incremental_deltas(
        current_registry=hash_registry,
        fs_context=fs_context,
    )

    # Construct and return the immutable change-detection context.
    return ChangeDetectionContext(
        raw_manifest=raw_manifest,
        hash_registry=hash_registry,
        delta_set=delta_set,
    )
````

````python
# Task 7 — Define the study architecture and the compiled wiki target state

# ==============================================================================
# Task 7: Define the study architecture and the compiled wiki target state
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 1: Instantiate the high-level system architecture
# -------------------------------------------------------------------------------------------------------------------------------

# Define Protocol for the incremental compiler operator.
class CompilerOperator(Protocol):
    """
    Protocol defining the interface for the incremental compilation operator.

    Purpose:
        Enforce a strict compile-operator contract without requiring a concrete
        implementation at this stage.

    Inputs:
        This protocol is satisfied by any callable matching the signature.

    Processes:
        No computation is performed here. It defines the expected interface.

    Outputs:
        __call__(delta_set, wiki_graph, conventions) -> Any:
            Signature of the incremental compiler operator.

    Notes:
        This protocol operationalizes the study's compiler operator:
        \[
        \Phi_{\text{compile}}: \mathcal{D}_{\text{raw}} \times \mathcal{K}_{t-1}
        \rightarrow \mathcal{K}_t
        \]
    """

    # Define the call signature for the incremental compiler operator.
    def __call__(
        self,
        delta_set: Any,
        wiki_graph: Any,
        conventions: Any,
    ) -> Any:
        ...


# Define Protocol for the grounded Q&A operator.
class QAOperator(Protocol):
    """
    Protocol defining the interface for the grounded question-answering operator.

    Purpose:
        Enforce a strict Q&A-operator contract without requiring a concrete
        implementation at this stage.

    Inputs:
        This protocol is satisfied by any callable matching the signature.

    Processes:
        No computation is performed here. It defines the expected interface.

    Outputs:
        __call__(query, wiki_graph, tools) -> Any:
            Signature of the grounded Q&A operator.

    Notes:
        This protocol operationalizes the study's Q&A operator:
        \[
        \Phi_{\text{qa}}: \mathcal{Q} \times \mathcal{K} \rightarrow \mathcal{A}
        \]
    """

    # Define the call signature for the grounded Q&A operator.
    def __call__(
        self,
        query: str,
        wiki_graph: Any,
        tools: Any,
    ) -> Any:
        ...


# Define Protocol for the artifact rendering operator.
class RenderOperator(Protocol):
    """
    Protocol defining the interface for the artifact rendering operator.

    Purpose:
        Enforce a strict render-operator contract without requiring a concrete
        implementation at this stage.

    Inputs:
        This protocol is satisfied by any callable matching the signature.

    Processes:
        No computation is performed here. It defines the expected interface.

    Outputs:
        __call__(qa_output, format_type) -> Any:
            Signature of the artifact rendering operator.

    Notes:
        This protocol operationalizes the study's rendering operator:
        \[
        \Phi_{\text{render}}: \mathcal{A}_{\text{intermediate}}
        \rightarrow \mathcal{A}_{\text{files}}
        \]
    """

    # Define the call signature for the artifact rendering operator.
    def __call__(
        self,
        qa_output: Any,
        format_type: str,
    ) -> Any:
        ...


# Define Protocol for the integrity verification operator.
class LintOperator(Protocol):
    """
    Protocol defining the interface for the integrity verification operator.

    Purpose:
        Enforce a strict lint-operator contract without requiring a concrete
        implementation at this stage.

    Inputs:
        This protocol is satisfied by any callable matching the signature.

    Processes:
        No computation is performed here. It defines the expected interface.

    Outputs:
        __call__(audit_set, wiki_graph) -> Any:
            Signature of the integrity verification operator.

    Notes:
        This protocol operationalizes the study's linting operator:
        \[
        \Phi_{\text{lint}}: \mathcal{K} \rightarrow \mathcal{L}_{\text{issues}}
        \times \mathcal{K}_{\text{patches}}
        \]
    """

    # Define the call signature for the integrity verification operator.
    def __call__(
        self,
        audit_set: Any,
        wiki_graph: Any,
    ) -> Any:
        ...


# Define Protocol for the external tool set.
class ToolSet(Protocol):
    """
    Protocol defining the interface for the external tool set.

    Purpose:
        Enforce a strict tool-set contract without requiring a concrete
        implementation at this stage.

    Inputs:
        This protocol is satisfied by any callable matching the signature.

    Processes:
        No computation is performed here. It defines the expected interface.

    Outputs:
        execute(tool_name, params) -> Any:
            Signature of the tool execution method.
    """

    # Define the execution signature for external tools.
    def execute(
        self,
        tool_name: str,
        params: Dict[str, Any],
    ) -> Any:
        ...


# Define Protocol for the logging repository.
class LoggingRepository(Protocol):
    """
    Protocol defining the interface for the logging repository.

    Purpose:
        Enforce a strict logging-repository contract without requiring a concrete
        implementation at this stage.

    Inputs:
        This protocol is satisfied by any callable matching the signature.

    Processes:
        No computation is performed here. It defines the expected interface.

    Outputs:
        append_log(entry) -> None:
            Signature of the log-append method.
    """

    # Define the log-append signature.
    def append_log(
        self,
        entry: Any,
    ) -> None:
        ...


# Define the mutable system architecture registry dataclass.
@dataclass
class SystemArchitectureRegistry:
    """
    Typed service container representing the formal system architecture tuple.

    Purpose:
        Store the study's formal system architecture:
        \[
        \mathcal{S} = \langle \mathcal{D}_{\text{raw}}, \mathcal{K},
        \Phi_{\text{compile}}, \Phi_{\text{qa}}, \Phi_{\text{render}},
        \Phi_{\text{lint}}, \mathcal{T}, \mathcal{L} \rangle
        \]

    Inputs:
        This dataclass is populated incrementally by the orchestrator and later
        tasks as operators are implemented.

    Processes:
        No computation is performed here. It is a passive mutable registry
        container.

    Outputs:
        D_raw:
            The raw corpus store (e.g., ``RawManifest``).
        K:
            The compiled knowledge base (e.g., ``WikiGraph``).
        Phi_compile:
            The incremental compiler operator interface.
        Phi_qa:
            The grounded Q&A operator interface.
        Phi_render:
            The artifact renderer operator interface.
        Phi_lint:
            The integrity verifier operator interface.
        T:
            The tool set interface.
        L:
            The logging repository interface.

    Notes:
        This dataclass is intentionally mutable because operators are bound
        incrementally as they are implemented in downstream tasks.
    """

    # Store the raw corpus store.
    D_raw: Any = None

    # Store the compiled knowledge base.
    K: Any = None

    # Store the incremental compiler operator.
    Phi_compile: Optional[CompilerOperator] = None

    # Store the grounded Q&A operator.
    Phi_qa: Optional[QAOperator] = None

    # Store the artifact rendering operator.
    Phi_render: Optional[RenderOperator] = None

    # Store the integrity verification operator.
    Phi_lint: Optional[LintOperator] = None

    # Store the external tool set.
    T: Optional[ToolSet] = None

    # Store the logging repository.
    L: Optional[LoggingRepository] = None


# Define the architecture registry initialization function.
def initialize_architecture_registry(
    change_context: ChangeDetectionContext,
) -> SystemArchitectureRegistry:
    """
    Initialize the system architecture registry by binding known state objects.

    Purpose:
        Partially populate the formal system architecture tuple by binding the
        raw corpus store from the change-detection context.

    Inputs:
        change_context:
            Change-detection context containing the raw manifest produced by
            Task 6.

    Processes:
        1. Validate that the change-detection context has the expected structural
           type.
        2. Instantiate a ``SystemArchitectureRegistry`` with all operators set to
           ``None``.
        3. Bind ``D_raw`` to the raw manifest from the change-detection context.
        4. Return the partially populated registry.

    Outputs:
        SystemArchitectureRegistry:
            The initialized architecture registry with ``D_raw`` bound and all
            operators unbound.

    Raises:
        TypeError:
            Raised if ``change_context`` is not a ``ChangeDetectionContext``
            instance.

    Notes:
        Operators and other state objects will be bound as they are implemented
        in later tasks.

        The original initialization logic is preserved exactly for valid inputs.
    """
    # Validate that the change-detection context has the expected structural type.
    if not isinstance(change_context, ChangeDetectionContext):
        # Raise a type error because registry binding requires the change-
        # detection context schema.
        raise TypeError(
            "initialize_architecture_registry expected 'change_context' to be "
            "an instance of ChangeDetectionContext."
        )

    # Instantiate the system architecture registry with all operators unbound.
    registry = SystemArchitectureRegistry()

    # Bind D_raw to the raw manifest from Task 6.
    registry.D_raw = change_context.raw_manifest

    # Operators and other state objects will be bound as they are implemented in
    # later tasks.

    # Return the partially populated architecture registry.
    return registry


# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 2: Implement the wiki as a directed labeled multigraph
# -------------------------------------------------------------------------------------------------------------------------------

# Define the enumeration of allowed wiki page types.
class PageType(Enum):
    """
    Enumeration of supported wiki page types.

    Purpose:
        Represent the node type in the compiled knowledge graph metadata, where
        the paper defines:

            type(v) ∈ {source_summary, concept, index, derived_artifact}

    Members:
        SOURCE_SUMMARY:
            Page summarizing a raw source artifact.
        CONCEPT:
            Page synthesizing a concept across multiple sources.
        INDEX:
            Navigational or map-of-content page.
        DERIVED_ARTIFACT:
            Filed report, figure, slide deck, or other generated output.
        UNKNOWN:
            Fallback type when no type has been inferred or assigned.
    """

    # Define the source-summary page type.
    SOURCE_SUMMARY = "source_summary"

    # Define the concept page type.
    CONCEPT = "concept"

    # Define the index page type.
    INDEX = "index"

    # Define the derived-artifact page type.
    DERIVED_ARTIFACT = "derived_artifact"

    # Define the fallback unknown page type.
    UNKNOWN = "unknown"



# Define the frozen dataclass for wiki node metadata.
@dataclass(frozen=True)
class WikiNodeMetadata:
    """
    Immutable metadata attached to a wiki node.

    Purpose:
        Represent the study's node-metadata tuple:

            meta(v) = <path(v), type(v), provenance(v), timestamp(v), tags(v)>

    Attributes:
        path:
            Repo-relative path of the page.
        type:
            Page category from the ``PageType`` enumeration.
        provenance:
            List of source identifiers or paths supporting the page.
        timestamp:
            Creation or last-modified timestamp string.
        tags:
            List of page tags.
    """

    # Store the repo-relative path of the page.
    path: str

    # Store the page type.
    type: PageType

    # Store the provenance identifiers associated with the page.
    provenance: List[str]

    # Store the creation or modification timestamp string.
    timestamp: str

    # Store the page tags.
    tags: List[str]


# Define the WikiGraph class implementing the study's knowledge graph formalism.
class WikiGraph:
    """
    In-memory directed labeled multigraph representation of the compiled wiki.

    Purpose:
        Implement the study's compiled knowledge-base formalism:

            K = (V, E, λ_V, λ_E)

        where:
            - V is the set of wiki pages,
            - E is the set of directed edges,
            - λ_V maps nodes to page content and metadata,
            - λ_E maps edges to labels such as references or broken references.

        This implementation uses AST-based Markdown parsing when available so
        that link extraction ignores code blocks and other unsafe contexts that
        regex-only approaches can mis-handle.
    """

    # Define the constructor for the wiki graph.
    def __init__(self, repo_root: pathlib.Path) -> None:
        """
        Initialize the wiki graph and, when available, the Markdown AST parser.

        Inputs:
            repo_root:
                Repository root directory from which repo-relative wiki paths
                are derived.

        Raises:
            TypeError:
                Raised when ``repo_root`` is not a ``pathlib.Path``.
        """
        # Validate that the repository root is a pathlib.Path before storing it.
        if not isinstance(repo_root, pathlib.Path):
            # Raise a type error immediately because path arithmetic depends on pathlib semantics.
            raise TypeError(
                "WikiGraph expected 'repo_root' to be a pathlib.Path."
            )

        # Import networkx lazily so the dependency is only required when the graph is instantiated.
        import networkx as nx

        # Instantiate the directed labeled multigraph representing K = (V, E, λ_V, λ_E).
        self.graph = nx.MultiDiGraph()

        # Store the repository root used for resolving and relativizing wiki page paths.
        self.repo_root = repo_root

        # Initialize the slug registry used for stable title-to-path resolution in O(1) average time.
        self.slug_registry: Dict[str, str] = {}

        # Check whether markdown-it-py is available for AST-based parsing.
        if MARKDOWN_IT_AVAILABLE:
            # Initialize a CommonMark parser for deterministic tokenization of Markdown files.
            self.md = MarkdownIt("commonmark")

        # Fall back gracefully when the AST parser dependency is unavailable.
        else:
            # Emit a critical log entry because AST-based parsing is part of the intended robust design.
            logger.critical(
                "markdown-it-py is not installed. AST parsing is unavailable."
            )

            # Record the absence of the parser so downstream methods can degrade safely.
            self.md = None

    # Define the slug-normalization helper used for stable internal linking.
    def _normalize_slug(self, text: str) -> str:
        """
        Convert a page title or wikilink target to a normalized slug.

        Inputs:
            text:
                Raw title or target text.

        Outputs:
            str:
                Lowercased, hyphen-normalized slug.

        Raises:
            TypeError:
                Raised when ``text`` is not a string.
        """
        # Validate that the slug source is a string.
        if not isinstance(text, str):
            # Raise a type error because regex-based normalization expects a string input.
            raise TypeError(
                "WikiGraph._normalize_slug expected 'text' to be a string."
            )

        # Normalize by lowercasing, replacing non-alphanumeric runs with hyphens, and stripping boundary hyphens.
        return re.sub(r"[^a-z0-9]+", "-", text.lower()).strip("-")

    # Define the AST-based wikilink extractor.
    def _extract_links_via_ast(self, content: str) -> List[str]:
        """
        Extract Obsidian-style wikilinks exclusively from safe text nodes using AST parsing.

        Purpose:
            Traverse the Markdown token stream and extract wikilinks only from
            ordinary text nodes. This avoids false positives from fenced code
            blocks, inline code, and other non-narrative contexts.

        Inputs:
            content:
                Markdown page content.

        Outputs:
            List[str]:
                Ordered list of extracted wikilink payloads.

        Notes:
            This helper supports construction of E and λ_E in the wiki graph.
        """
        # Return an empty link list if the Markdown parser is unavailable.
        if not self.md:
            # Degrade safely by returning no extracted links.
            return []

        # Parse the Markdown content into a token stream.
        tokens = self.md.parse(content)

        # Initialize the ordered list that will accumulate extracted link targets.
        extracted_links: List[str] = []

        # Compile the wikilink regex used to detect [[...]] patterns inside safe text nodes.
        wikilink_pattern = re.compile(r"\[\[(.*?)\]\]")

        # Iterate over each top-level Markdown token.
        for token in tokens:
            # Restrict extraction to inline tokens because they contain parsed narrative text.
            if token.type == "inline" and token.children:
                # Iterate over each child token inside the inline token.
                for child in token.children:
                    # Restrict extraction further to plain text nodes to avoid inline code segments.
                    if child.type == "text":
                        # Extract all wikilink payloads from the safe text content.
                        matches = wikilink_pattern.findall(child.content)

                        # Extend the ordered extracted-link list with the matches from this text node.
                        extracted_links.extend(matches)

        # Return the ordered list of extracted wikilink payloads.
        return extracted_links

    # Define the filesystem-to-graph rebuild routine.
    def build_from_filesystem(self, wiki_root: pathlib.Path) -> None:
        """
        Deterministically rebuild the graph view from the authoritative filesystem.

        Inputs:
            wiki_root:
                Root directory of the compiled wiki subtree.

        Processes:
            1. Clear prior in-memory graph state for idempotency.
            2. First pass:
               - discover Markdown files,
               - register graph nodes,
               - populate the slug registry.
            3. Second pass:
               - parse page content,
               - extract wikilinks using AST parsing,
               - add labeled edges for resolved and broken references.

        Raises:
            TypeError:
                Raised when ``wiki_root`` is not a ``pathlib.Path``.

        Notes:
            This method materializes the graph object corresponding to:

                K = (V, E, λ_V, λ_E)

            where nodes come from Markdown files and edges come from page links.
        """
        # Validate that the wiki root is a pathlib.Path.
        if not isinstance(wiki_root, pathlib.Path):
            # Raise a type error because filesystem traversal depends on pathlib semantics.
            raise TypeError(
                "WikiGraph.build_from_filesystem expected 'wiki_root' to be a pathlib.Path."
            )

        # Clear the existing graph so the rebuild is idempotent and fully determined by the filesystem.
        self.graph.clear()

        # Clear the slug registry so it is reconstructed from the current wiki snapshot.
        self.slug_registry.clear()

        # Return early if the wiki root does not exist, leaving the graph empty.
        if not wiki_root.exists():
            # Exit because there is no filesystem state to materialize.
            return

        # First pass: discover Markdown files and create graph nodes corresponding to V.
        for root, _, files in os.walk(wiki_root):
            # Iterate over every file discovered in the current directory.
            for file in files:
                # Restrict node discovery to Markdown files, which are the canonical wiki representation.
                if file.endswith(".md"):
                    # Construct the absolute path of the discovered Markdown file.
                    abs_path = pathlib.Path(root) / file

                    # Convert the absolute path to a repo-relative path for stable graph keys.
                    rel_path = str(abs_path.relative_to(self.repo_root))

                    # Derive the normalized slug from the page stem for link resolution.
                    slug = self._normalize_slug(abs_path.stem)

                    # Register the slug-to-path mapping in the lookup table.
                    self.slug_registry[slug] = rel_path

                    # Read the page content from disk so metadata could be parsed or inferred if needed.
                    content = abs_path.read_text(encoding="utf-8", errors="ignore")

                    # Construct placeholder metadata for the node, preserving the original callable's logic.
                    metadata = WikiNodeMetadata(
                        path=rel_path,
                        type=PageType.UNKNOWN,
                        provenance=[],
                        timestamp="",
                        tags=[],
                    )

                    # Add the node to the graph with its metadata payload.
                    self.graph.add_node(rel_path, metadata=metadata)

        # Second pass: parse pages and create edges corresponding to E and λ_E.
        for node_path in self.graph.nodes:
            # Construct the absolute file path for the current node.
            abs_path = self.repo_root / node_path

            # Read the Markdown content of the page.
            content = abs_path.read_text(encoding="utf-8", errors="ignore")

            # Extract wikilinks using the AST-safe extractor.
            links = self._extract_links_via_ast(content)

            # Iterate over each extracted wikilink payload.
            for link_text in links:
                # Resolve aliases by taking the target title before the '|' separator.
                target_title = link_text.split("|")[0]

                # Normalize the target title into a slug used for registry lookup.
                target_slug = self._normalize_slug(target_title)

                # Look up the repo-relative path of the resolved target, if any.
                target_path = self.slug_registry.get(target_slug)

                # Add a normal reference edge when the target resolves to an existing graph node.
                if target_path and target_path in self.graph:
                    # Add a directed edge labeled "references" from the source page to the resolved target page.
                    self.graph.add_edge(node_path, target_path, label="references")

                # Otherwise create or reuse an unresolved placeholder node and add a broken-reference edge.
                else:
                    # Construct a stable placeholder node identifier for the unresolved target.
                    placeholder_path = f"UNRESOLVED:{target_slug}"

                    # Create the placeholder node if it does not yet exist in the graph.
                    if placeholder_path not in self.graph:
                        # Add the unresolved placeholder node with unknown metadata.
                        self.graph.add_node(
                            placeholder_path,
                            metadata=WikiNodeMetadata(
                                path=placeholder_path,
                                type=PageType.UNKNOWN,
                                provenance=[],
                                timestamp="",
                                tags=[],
                            ),
                        )

                    # Add a directed broken-reference edge from the source page to the unresolved placeholder.
                    self.graph.add_edge(
                        node_path,
                        placeholder_path,
                        label="broken_reference",
                    )

    # Define the content accessor for a wiki node path.
    def get_content(self, node_path: str) -> str:
        """
        Retrieve the Markdown content of a wiki page.

        Inputs:
            node_path:
                Repo-relative path of the target node.

        Outputs:
            str:
                Markdown content of the page, or an empty string if the page is
                unresolved or absent.

        Raises:
            TypeError:
                Raised when ``node_path`` is not a string.

        Notes:
            In the study's notation, this method approximates access to λ_V(v)
            for a page node ``v`` represented by its repo-relative path.
        """
        # Validate that the node path is a string.
        if not isinstance(node_path, str):
            # Raise a type error because path-prefix checks and path joins expect a string.
            raise TypeError(
                "WikiGraph.get_content expected 'node_path' to be a string."
            )

        # Return an empty string for unresolved placeholder nodes because they have no backing file content.
        if node_path.startswith("UNRESOLVED:"):
            # Exit early for unresolved placeholders.
            return ""

        # Resolve the absolute filesystem path of the page from the repo-relative node path.
        abs_path = self.repo_root / node_path

        # Return the file content if the page exists on disk.
        if abs_path.exists():
            # Read and return the page text, ignoring invalid bytes.
            return abs_path.read_text(encoding="utf-8", errors="ignore")

        # Return an empty string when the page path does not exist.
        return ""


# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 3: Define page templates and link conventions
# -------------------------------------------------------------------------------------------------------------------------------

# Define the explicit epistemic disclaimer regarding the reasoning model.
GPT_5_2_EPISTEMIC_DISCLAIMER: str = (
    "No neural network architecture internals are available from the source "
    "for gpt-5.2; it must be treated as proprietary and unspecified."
)


# Define the frozen dataclass for one page template.
@dataclass(frozen=True)
class PageTemplate:
    """
    Immutable contract defining the required structure of a wiki page type.

    Purpose:
        Validate that a Markdown page contains all required sections for its
        page type using AST traversal rather than naive regex matching.

    Attributes:
        page_type:
            The page type to which the template applies.
        required_sections:
            Ordered list of required section headings.
        link_policy:
            Textual description of the link convention expected for the page.

    Notes:
        This template supports the study's requirement that source pages and
        concept pages conform to explicit section structures.
    """

    # Store the page type governed by this template.
    page_type: PageType

    # Store the required section-heading names for the page type.
    required_sections: List[str]

    # Store the link-policy description associated with the page type.
    link_policy: str

    # Define the AST-based content validator for a page template.
    def validate_content(self, content: str) -> Tuple[bool, List[str]]:
        """
        Validate that the provided Markdown content contains all required sections.

        Inputs:
            content:
                Raw Markdown text to validate.

        Processes:
            1. Validate the input type.
            2. Parse the Markdown into tokens using markdown-it-py when
               available.
            3. Traverse heading-open tokens and capture normalized heading text
               from the immediately following inline token.
            4. Compare the discovered headings against the template's required
               section list.
            5. Return both the validity flag and the missing sections list.

        Outputs:
            Tuple[bool, List[str]]:
                A tuple containing:
                1. ``True`` if all required sections are present, else ``False``.
                2. The list of missing required section names.

        Raises:
            TypeError:
                Raised when ``content`` is not a string.

        Error Handling:
            If markdown-it-py is unavailable, the original callable logs an
            error and returns ``False`` together with all required sections.
        """
        # Validate that the content is a string before parsing it.
        if not isinstance(content, str):
            # Raise a type error because Markdown tokenization expects a string input.
            raise TypeError(
                "PageTemplate.validate_content expected 'content' to be a string."
            )

        # Fall back immediately when the AST parser dependency is unavailable.
        if not MARKDOWN_IT_AVAILABLE:
            # Emit an error because AST-backed structural validation cannot be performed.
            logger.error(
                "markdown-it-py unavailable. Falling back to regex validation."
            )

            # Preserve the original callable's fallback behavior.
            return False, self.required_sections

        # Initialize the Markdown parser using the CommonMark grammar.
        md = MarkdownIt("commonmark")

        # Parse the Markdown content into a token stream.
        tokens = md.parse(content)

        # Initialize the set of discovered normalized headings.
        found_headings: Set[str] = set()

        # Iterate through the token stream by index so adjacent inline tokens can be inspected.
        for i, token in enumerate(tokens):
            # Identify opening heading tokens such as ``heading_open``.
            if token.type == "heading_open":
                # Confirm that the next token exists and is the inline heading-content token.
                if i + 1 < len(tokens) and tokens[i + 1].type == "inline":
                    # Normalize the heading text by lowercasing and stripping surrounding whitespace.
                    heading_text = tokens[i + 1].content.lower().strip()

                    # Record the normalized heading.
                    found_headings.add(heading_text)

        # Initialize the ordered list of missing required sections.
        missing_sections: List[str] = []

        # Check each required section name against the discovered headings.
        for req_section in self.required_sections:
            # Normalize the required section for a case-insensitive comparison.
            norm_req = req_section.lower().strip()

            # Record the required section if it was not found among the parsed headings.
            if norm_req not in found_headings:
                # Append the missing required section to the output list.
                missing_sections.append(req_section)

        # Return the overall validity flag together with the list of missing sections.
        return len(missing_sections) == 0, missing_sections


# Define the frozen dataclass for the page template registry.
@dataclass(frozen=True)
class PageTemplateRegistry:
    """
    Registry containing the structural contracts for all wiki page types.

    Purpose:
        Store the complete mapping from page types to their structural templates.

    Inputs:
        This dataclass is populated by ``initialize_page_templates``.

    Processes:
        No computation is performed here. It is a passive immutable registry
        container.

    Outputs:
        templates:
            Mapping from ``PageType`` to ``PageTemplate``.
    """

    # Store the mapping from page types to templates.
    templates: Dict[PageType, PageTemplate]


# Define the page template initialization function.
def initialize_page_templates(
    contract: FrozenRunContract,
) -> PageTemplateRegistry:
    """
    Initialize the page template registry based on the study configuration.

    Purpose:
        Define the strict section requirements for source summaries, concept
        pages, and derived artifacts exactly as specified by the study.

    Inputs:
        contract:
            The frozen run contract containing filesystem and link conventions.

    Processes:
        1. Validate that ``contract`` is a ``FrozenRunContract``.
        2. Extract link conventions from the configuration.
        3. Define the strict ``PageTemplate`` for source summary pages with the
           required sections specified by the study.
        4. Define the strict ``PageTemplate`` for concept pages with the required
           sections specified by the study.
        5. Define the strict ``PageTemplate`` for derived artifact pages with the
           required sections specified by the study.
        6. Return the immutable ``PageTemplateRegistry``.

    Outputs:
        PageTemplateRegistry:
            Immutable registry containing all page templates.

    Raises:
        TypeError:
            Raised if ``contract`` is not a ``FrozenRunContract`` instance.

    Notes:
        The original template initialization logic is preserved exactly for valid
        inputs.
    """
    # Validate that the contract has the expected structural type.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because template initialization requires the frozen
        # contract.
        raise TypeError(
            "initialize_page_templates expected 'contract' to be an instance "
            "of FrozenRunContract."
        )

    # Extract link conventions from the contract.
    link_conventions = (
        contract.original_config.get("filesystem_conventions", {}).get(
            "link_conventions",
            {},
        )
    )

    # Extract the preferred link style.
    link_style = link_conventions.get("wiki_link_style", "obsidian_wikilink")

    # Construct the link policy description.
    policy_desc = (
        f"Preferred style: {link_style}. Backlinks and cross-links are required."
    )

    # Define the strict template for Source Summary pages.
    source_summary_template = PageTemplate(
        page_type=PageType.SOURCE_SUMMARY,
        required_sections=[
            "Provenance",
            "Summary",
            "Key Claims",
            "Extracted Concepts",
            "Related Pages",
            "Open Questions",
        ],
        link_policy=policy_desc,
    )

    # Define the strict template for Concept pages.
    concept_template = PageTemplate(
        page_type=PageType.CONCEPT,
        required_sections=[
            "Definition",
            "Mechanisms",
            "Canonical References",
            "Related Concepts",
            "Open Questions",
        ],
        link_policy=policy_desc,
    )

    # Define the strict template for Derived Artifact pages.
    derived_artifact_template = PageTemplate(
        page_type=PageType.DERIVED_ARTIFACT,
        required_sections=[
            "Executive Summary",
            "Main Analysis",
            "Evidence & Citations",
            "Limitations",
            "Next Questions",
        ],
        link_policy=policy_desc,
    )

    # Construct the templates mapping.
    templates = {
        PageType.SOURCE_SUMMARY: source_summary_template,
        PageType.CONCEPT: concept_template,
        PageType.DERIVED_ARTIFACT: derived_artifact_template,
    }

    # Construct and return the immutable page template registry.
    return PageTemplateRegistry(templates=templates)


# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable wiki architecture context aggregate returned by Task 7.
@dataclass(frozen=True)
class WikiArchitectureContext:
    """
    Immutable record encapsulating the initialized system architecture and wiki
    state.

    Purpose:
        Package the system architecture registry, wiki graph, and page template
        registry into one immutable object for downstream tasks.

    Inputs:
        This dataclass is populated by ``task_7_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        architecture_registry:
            The initialized system architecture registry.
        wiki_graph:
            The wiki graph constructed from the filesystem.
        template_registry:
            The page template registry defining structural contracts.
    """

    # Store the initialized system architecture registry.
    architecture_registry: SystemArchitectureRegistry

    # Store the wiki graph.
    wiki_graph: WikiGraph

    # Store the page template registry.
    template_registry: PageTemplateRegistry


# Define the Task 7 orchestrator.
def task_7_orchestrator(
    change_context: ChangeDetectionContext,
    fs_context: FilesystemSubstrateContext,
    contract: FrozenRunContract,
    repo_root: pathlib.Path,
) -> WikiArchitectureContext:
    """
    Orchestrate Task 7 end-to-end by initializing the system architecture, wiki
    graph, and page templates.

    Purpose:
        This function is the Task 7 entry point. It executes the three Task 7
        substeps in the required order and returns an immutable wiki architecture
        context.

    Inputs:
        change_context:
            Change-detection context containing the raw manifest from Task 6.
        fs_context:
            Filesystem substrate context containing filesystem layouts from
            Task 3.
        contract:
            The frozen run contract from Task 1.
        repo_root:
            Project root directory used to resolve repo-relative paths.

    Processes:
        1. Validate top-level inputs.
        2. Step 1: Instantiate the high-level system architecture registry by
           calling ``initialize_architecture_registry``.
        3. Step 2: Construct the wiki graph as a directed labeled multigraph by
           instantiating ``WikiGraph`` and building it from the filesystem.
        4. Bind the wiki graph to the architecture registry.
        5. Step 3: Define page templates and link conventions by calling
           ``initialize_page_templates``.
        6. Return an immutable ``WikiArchitectureContext`` aggregating the
           architecture, graph, and templates.

    Outputs:
        WikiArchitectureContext:
            Immutable aggregate containing the system architecture, wiki graph,
            and template registry.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.

    Notes:
        This orchestrator operationalizes the study's system architecture
        formalization:
        \[
        \mathcal{S} = \langle \mathcal{D}_{\text{raw}}, \mathcal{K},
        \Phi_{\text{compile}}, \Phi_{\text{qa}}, \Phi_{\text{render}},
        \Phi_{\text{lint}}, \mathcal{T}, \mathcal{L} \rangle
        \]
        and the wiki graph formalization:
        \[
        \mathcal{K} = (V, E, \lambda_V, \lambda_E)
        \]

        The original orchestration logic is preserved exactly for valid inputs.
    """
    # Validate that the change-detection context has the expected structural type.
    if not isinstance(change_context, ChangeDetectionContext):
        # Raise a type error because architecture binding requires the change-
        # detection context schema.
        raise TypeError(
            "task_7_orchestrator expected 'change_context' to be an instance "
            "of ChangeDetectionContext."
        )

    # Validate that the filesystem substrate context has the expected structural
    # type.
    if not isinstance(fs_context, FilesystemSubstrateContext):
        # Raise a type error because wiki root resolution requires the filesystem
        # context schema.
        raise TypeError(
            "task_7_orchestrator expected 'fs_context' to be an instance of "
            "FilesystemSubstrateContext."
        )

    # Validate that the frozen run contract has the expected structural type.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because template initialization requires the frozen
        # contract.
        raise TypeError(
            "task_7_orchestrator expected 'contract' to be an instance of "
            "FrozenRunContract."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_7_orchestrator expected 'repo_root' to be a pathlib.Path "
            "instance."
        )

    # Step 1: Instantiate the high-level system architecture registry.
    arch_registry = initialize_architecture_registry(change_context)

    # Step 2: Implement the wiki as a directed labeled multigraph.

    # Instantiate the wiki graph.
    wiki_graph = WikiGraph(repo_root=repo_root)

    # Retrieve the absolute path to the wiki/ directory from the filesystem
    # context.
    wiki_root_state = fs_context.root_states.get("wiki_directory")

    # Check if the wiki root state is valid.
    if wiki_root_state and wiki_root_state.is_valid:
        # Deterministically build the graph from the current filesystem state.
        wiki_graph.build_from_filesystem(wiki_root_state.absolute_path)

    # Bind the initialized wiki graph to the architecture registry.
    arch_registry.K = wiki_graph

    # Step 3: Define page templates and link conventions.
    template_registry = initialize_page_templates(contract)

    # Construct and return the immutable wiki architecture context.
    return WikiArchitectureContext(
        architecture_registry=arch_registry,
        wiki_graph=wiki_graph,
        template_registry=template_registry,
    )
````

````python
# Task 8 — Initialize logging, validation gates, and run manifests

# ==============================================================================
# Task 8: Initialize logging, validation gates, and run manifests
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 1: Instantiate the per-call logging system
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable per-call log-entry record.
@dataclass(frozen=True)
class LogEntry:
    """
    Immutable record representing a single LLM invocation and its outcomes.

    Purpose:
        Store the full structured audit trail for one model call, including the
        call inputs, decoding controls, raw output, validated file operations,
        tool invocations, and warnings.

    Inputs:
        This dataclass is populated by downstream orchestration code whenever an
        LLM invocation occurs.

    Processes:
        No computation is performed here. It is a passive immutable logging
        record.

    Outputs:
        The dataclass fields implement the study's logging schema:
        \[
        \text{log\_entry} = \langle t, \tau, m, \theta, I, O, \Delta \rangle
        \]
        where:
        - \(t\) is the timestamp,
        - \(\tau\) is the stage name,
        - \(m\) is the model identifier,
        - \(\theta\) are the decoding settings,
        - \(I\) are the input file paths and hashes,
        - \(O\) is the raw model output,
        - \(\Delta\) is the set of validated file operations.

        Additional governance fields are retained from the implementation:
        - run_id,
        - call_id,
        - prompt_text,
        - tool_calls_and_results,
        - warnings.
    """

    # Store the unique identifier of the full pipeline run.
    run_id: str

    # Store the unique UUID for this specific LLM call.
    call_id: str

    # Store the UTC ISO 8601 timestamp of the call.
    timestamp_utc: str

    # Store the pipeline stage name corresponding to tau.
    stage_name: str

    # Store the concrete model identifier corresponding to m.
    model_id: str

    # Store the declared and effective decoding settings corresponding to theta.
    decoding_settings: Dict[str, Any]

    # Store the input file paths and hashes corresponding to I.
    input_file_paths_and_hashes: Dict[str, str]

    # Store the exact prompt string sent to the model.
    prompt_text: str

    # Store the exact raw string returned by the model corresponding to O.
    raw_model_output: str

    # Store the list of validated file operations corresponding to Delta.
    validated_file_operations: List[Dict[str, Any]]

    # Store any tool invocations performed during the call.
    tool_calls_and_results: List[Dict[str, Any]]

    # Store heuristic or validation warnings associated with the call.
    warnings: List[str]


# Define the custom JSON encoder used for log serialization.
class CustomJSONEncoder(json.JSONEncoder):
    """
    Custom JSON encoder that serializes pathlib.Path and set objects.

    Purpose:
        Extend the default JSON encoder so that structured log records
        containing path and set objects can be serialized without manual
        pre-processing.

    Inputs:
        This encoder is used by ``LoggingRepository.append_log``.

    Processes:
        1. If the object is a pathlib.Path, convert it to a string.
        2. If the object is a set, convert it to a list.
        3. Otherwise, defer to the superclass JSONEncoder implementation.

    Outputs:
        JSON-serializable representations of otherwise unsupported objects.
    """

    # Override the default JSON encoding method.
    def default(
        self,
        obj: Any,
    ) -> Any:
        """
        Convert unsupported Python objects into JSON-serializable forms.

        Purpose:
            Handle pathlib.Path and set objects during JSON serialization.

        Inputs:
            obj:
                Arbitrary object encountered during JSON serialization.

        Processes:
            1. Detect pathlib.Path and convert to string.
            2. Detect set and convert to list.
            3. Delegate all other cases to the superclass.

        Outputs:
            Any:
                A JSON-serializable representation of the input object.
        """
        # Check whether the object is a pathlib.Path.
        if isinstance(obj, pathlib.Path):
            # Convert the path to its string representation.
            return str(obj)

        # Check whether the object is a set.
        if isinstance(obj, set):
            # Convert the set to a list for JSON serialization.
            return list(obj)

        # Delegate all other object types to the base JSON encoder.
        return super().default(obj)


# Define the append-only JSONL logging repository.
class LoggingRepository:
    """
    Repository for appending structured log entries to a JSONL file.

    Purpose:
        Provide crash-resilient, append-only audit logging for all model calls in
        the pipeline.

    Inputs:
        log_dir:
            Absolute path to the logging directory.
        run_id:
            Unique identifier of the current pipeline run.

    Processes:
        1. Ensure the logging directory exists.
        2. Define a run-specific JSONL log file.
        3. Serialize LogEntry records to JSON strings.
        4. Append each serialized log entry as one line in the JSONL file.

    Outputs:
        log_dir:
            The resolved log directory.
        run_id:
            The current run identifier.
        log_file:
            The run-specific JSONL log file path.
    """

    # Define the repository constructor.
    def __init__(
        self,
        log_dir: pathlib.Path,
        run_id: str,
    ) -> None:
        """
        Initialize the logging repository.

        Purpose:
            Prepare the run-specific append-only JSONL log file.

        Inputs:
            log_dir:
                Absolute or relative path to the logging directory.
            run_id:
                Unique identifier for the current pipeline run.

        Processes:
            1. Validate the input types.
            2. Store the log directory and run identifier.
            3. Create the logging directory if it does not exist.
            4. Construct the run-specific JSONL file path.

        Outputs:
            None. State is stored on the instance.

        Raises:
            TypeError:
                Raised if input types are invalid.
            OSError:
                Propagated if directory creation fails.
        """
        # Validate that the log directory is a pathlib.Path.
        if not isinstance(log_dir, pathlib.Path):
            # Raise a type error because path operations require pathlib.Path.
            raise TypeError(
                "LoggingRepository expected 'log_dir' to be a pathlib.Path "
                "instance."
            )

        # Validate that the run identifier is a string.
        if not isinstance(run_id, str):
            # Raise a type error because run IDs must be strings.
            raise TypeError(
                "LoggingRepository expected 'run_id' to be a string."
            )

        # Store the logging directory on the instance.
        self.log_dir = log_dir

        # Store the run identifier on the instance.
        self.run_id = run_id

        # Ensure that the logging directory exists.
        self.log_dir.mkdir(parents=True, exist_ok=True)

        # Construct the run-specific JSONL log file path.
        self.log_file = self.log_dir / f"run_{self.run_id}_calls.jsonl"

    # Define the append-only log-writing method.
    def append_log(
        self,
        entry: LogEntry,
    ) -> None:
        """
        Serialize a LogEntry to JSON and append it to the JSONL log file.

        Purpose:
            Persist one structured model-call audit record in append-only JSONL
            format.

        Inputs:
            entry:
                The structured log entry to serialize and append.

        Processes:
            1. Validate that the supplied entry is a LogEntry.
            2. Convert the dataclass to a dictionary.
            3. Serialize the dictionary to JSON using the custom encoder.
            4. Append the serialized JSON string plus newline to the log file.

        Outputs:
            None. The side effect is one new line appended to the JSONL file.

        Raises:
            TypeError:
                Raised if the supplied entry is not a LogEntry.
            OSError:
                Propagated if file append fails.
        """
        # Validate that the supplied entry is a LogEntry instance.
        if not isinstance(entry, LogEntry):
            # Raise a type error because log serialization assumes the LogEntry
            # schema.
            raise TypeError(
                "LoggingRepository.append_log expected 'entry' to be a "
                "LogEntry instance."
            )

        # Convert the dataclass entry to a plain dictionary.
        entry_dict = asdict(entry)

        # Serialize the dictionary to a JSON string using the custom encoder.
        json_str = json.dumps(entry_dict, cls=CustomJSONEncoder)

        # Open the JSONL file in append mode with UTF-8 encoding.
        with open(self.log_file, "a", encoding="utf-8") as f:
            # Append the serialized log line plus newline terminator.
            f.write(json_str + "\n")


# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 2: Implement write-validation gates
# -------------------------------------------------------------------------------------------------------------------------------

# Define the enumeration of proposed file-operation types.
class OperationType(Enum):
    """
    Type of file operation proposed by the LLM.

    Purpose:
        Distinguish between full writes and patch-oriented updates.
    """

    # Define the full-write operation type.
    WRITE = "WRITE"

    # Define the patch/update operation type.
    PATCH = "PATCH"


# Define the enumeration of file-operation validation states.
class ValidationStatus(Enum):
    """
    Validation status for a proposed file operation.

    Purpose:
        Represent the lifecycle state of an LLM-proposed file operation as it
        moves through the write-governance pipeline.
    """

    # Define the initial pending state before validation has completed.
    PENDING = "PENDING"

    # Define the approved state for operations that passed all validation gates.
    APPROVED = "APPROVED"

    # Define the rejected state for operations that violated one or more hard rules.
    REJECTED = "REJECTED"

    # Define the review-required state for operations that need human or supervisory intervention.
    REQUIRES_REVIEW = "REQUIRES_REVIEW"


# Define the mutable file-operation plan record.
@dataclass
class FileOperationPlan:
    """
    Proposed file operation subject to validation prior to execution.

    Attributes:
        target_path:
            Absolute or repo-rooted filesystem path of the write target.
        operation_type:
            Operation mode, such as WRITE or PATCH.
        content:
            Proposed file content or patch payload.
        status:
            Validation status assigned during write validation.
        findings:
            Ordered list of validation findings accumulated during review.
    """

    # Store the target path of the proposed operation.
    target_path: pathlib.Path

    # Store the operation type for the proposed write.
    operation_type: OperationType

    # Store the proposed content payload for the operation.
    content: str

    # Store the current validation status, defaulting to PENDING before checks run.
    status: ValidationStatus = ValidationStatus.PENDING

    # Store the ordered validation findings associated with the operation.
    findings: List[str] = field(default_factory=list)



# Define the centralized write validator.
class WriteValidator:
    """
    Centralized gatekeeper for all file operations proposed by the LLM.

    Purpose:
        Enforce path governance, wiki file-extension rules, and AST-based link
        validation before any proposed operation is executed. This class
        supports Task 8's requirement that writes must be constrained and
        validated before touching the filesystem.
    """

    # Define the constructor for the write validator.
    def __init__(
        self,
        write_policy: Any,
        wiki_graph: WikiGraph,
        repo_root: pathlib.Path,
    ) -> None:
        """
        Initialize the write validator.

        Inputs:
            write_policy:
                Object expected to expose ``validate_write_path`` for path-policy
                enforcement.
            wiki_graph:
                Existing wiki graph used to resolve internal links.
            repo_root:
                Repository root directory used for relative-path derivation.
        """
        # Store the write-policy collaborator used to validate write paths.
        self.write_policy = write_policy

        # Store the wiki graph used for slug and link resolution.
        self.wiki_graph = wiki_graph

        # Store the repository root used for relative-path conversion.
        self.repo_root = repo_root

        # Store the unresolved-link marker permitted by the write-validation policy.
        self.unresolved_marker = "UNRESOLVED_LINK"

        # Check whether the AST parser dependency is available.
        if MARKDOWN_IT_AVAILABLE:
            # Initialize the CommonMark parser for safe Markdown tokenization.
            self.md = MarkdownIt("commonmark")

        # Fall back gracefully when AST parsing is unavailable.
        else:
            # Record the absence of a parser so link validation can degrade safely.
            self.md = None

    # Define the batch-validation routine for a list of file operations.
    def validate_batch(
        self,
        operations: List[FileOperationPlan],
    ) -> List[FileOperationPlan]:
        """
        Validate a batch of proposed file operations.

        Inputs:
            operations:
                Ordered list of ``FileOperationPlan`` objects proposed for
                validation.

        Processes:
            1. Validate the batch container type.
            2. Build the set of repo-relative candidate paths so links may
               resolve against pages being created in the same batch.
            3. Validate each operation individually using layered checks.
            4. Return the mutated operation plans with statuses and findings.

        Outputs:
            List[FileOperationPlan]:
                The same ordered list, now annotated with validation results.

        Raises:
            TypeError:
                Raised when ``operations`` is not a list.
        """
        # Validate that the supplied operations batch is a list.
        if not isinstance(operations, list):
            # Raise a type error because the validator expects an ordered list of operation plans.
            raise TypeError("WriteValidator.validate_batch expected a list.")

        # Build the set of repo-relative candidate paths so forward-declared links can resolve within the same batch.
        candidate_paths: Set[str] = {
            str(op.target_path.relative_to(self.repo_root))
            for op in operations
            if op.target_path.is_relative_to(self.repo_root)
        }

        # Validate each operation in the batch against path, extension, and link rules.
        for op in operations:
            # Apply the single-operation validation routine.
            self._validate_single_operation(op, candidate_paths)

        # Return the batch with updated validation statuses and findings.
        return operations

    # Define the layered validator for an individual file operation.
    def _validate_single_operation(
        self,
        op: FileOperationPlan,
        candidate_paths: Set[str],
    ) -> None:
        """
        Apply layered validation rules to a single proposed file operation.

        Inputs:
            op:
                Proposed file operation to validate.
            candidate_paths:
                Set of repo-relative paths proposed in the current batch.

        Raises:
            TypeError:
                Raised when ``op`` is not a ``FileOperationPlan``.

        Notes:
            The validation logic preserves the original callable's sequence:
            path governance, Markdown extension checks, then AST-based link
            validation.
        """
        # Validate that the operation object is a FileOperationPlan.
        if not isinstance(op, FileOperationPlan):
            # Raise a type error because downstream attribute access assumes the operation-plan schema.
            raise TypeError("Expected FileOperationPlan.")

        # 1. Path Governance
        # Validate the target path against the write-policy registry.
        if not self.write_policy.validate_write_path(op.target_path):
            # Record the path-governance violation in the operation findings.
            op.findings.append(f"Path {op.target_path} violates write policy.")

            # Reject the operation immediately because it attempts an invalid write location.
            op.status = ValidationStatus.REJECTED

            # Exit early because downstream checks are irrelevant once path governance fails.
            return

        # 2. Markdown Extension Check
        # Enforce that any file written under the wiki subtree must have a ``.md`` extension.
        if "wiki" in op.target_path.parts and op.target_path.suffix.lower() != ".md":
            # Record the extension violation in the operation findings.
            op.findings.append("Wiki files must have a .md extension.")

            # Reject the operation because wiki pages are contractually Markdown files.
            op.status = ValidationStatus.REJECTED

            # Exit early because extension failure is a hard validation error.
            return

        # 3. Link Resolution via AST
        # Restrict link validation to Markdown outputs when an AST parser is available.
        if op.target_path.suffix.lower() == ".md" and self.md:
            # Parse the proposed Markdown content into a token stream.
            tokens = self.md.parse(op.content)

            # Compile the regex used to extract Obsidian wikilinks from safe text nodes.
            wikilink_pattern = re.compile(r"\[\[(.*?)\]\]")

            # Traverse the token stream and inspect inline text nodes only.
            for token in tokens:
                # Restrict extraction to inline tokens that contain child nodes.
                if token.type == "inline" and token.children:
                    # Traverse each child token inside the inline token.
                    for child in token.children:
                        # Restrict extraction to plain text child tokens.
                        if child.type == "text":
                            # Extract all wikilink payloads from the safe text content.
                            links = wikilink_pattern.findall(child.content)

                            # Validate each extracted wikilink target.
                            for link in links:
                                # Resolve aliases by taking the title before the ``|`` separator.
                                target_title = link.split("|")[0]

                                # Normalize the target title to the slug form used by the wiki graph.
                                target_slug = self.wiki_graph._normalize_slug(target_title)

                                # Check whether the target slug already resolves in the current wiki graph.
                                exists_in_graph = target_slug in self.wiki_graph.slug_registry

                                # Check whether the target slug will be created by another operation in the same batch.
                                exists_in_batch = any(
                                    self.wiki_graph._normalize_slug(
                                        pathlib.Path(p).stem
                                    ) == target_slug
                                    for p in candidate_paths
                                )

                                # Reject unresolved and unmarked links.
                                if not (exists_in_graph or exists_in_batch):
                                    # Allow unresolved links only when the explicit unresolved marker is present.
                                    if self.unresolved_marker not in op.content:
                                        # Record the unresolved-link finding.
                                        op.findings.append(
                                            f"Unresolved link: [[{target_title}]]."
                                        )

                                        # Reject the operation because it violates link-resolution rules.
                                        op.status = ValidationStatus.REJECTED

                                        # Exit immediately after the first hard link-resolution failure.
                                        return

        # Approve the operation if it remained pending after all validation checks.
        if op.status == ValidationStatus.PENDING:
            # Promote the operation status from PENDING to APPROVED.
            op.status = ValidationStatus.APPROVED


# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 3: Implement no-fabrication heuristics and run summaries
# -------------------------------------------------------------------------------------------------------------------------------

# Define the heuristic scanner for unsupported factual content.
class FabricationHeuristicScanner:
    """
    Scan Markdown content for possible unsupported factual claims using NLP.

    Purpose:
        Emit heuristic warnings when a paragraph contains several formal named
        entities but no citations. This supports Task 8's no-fabrication
        governance layer without changing the underlying content.

    Notes:
        This scanner is intentionally heuristic. It does not prove fabrication;
        it flags potentially under-cited passages for review.
    """

    # Define the constructor for the heuristic scanner.
    def __init__(self) -> None:
        """
        Initialize the scanner and attempt to load the NLP pipeline.

        Processes:
            1. Compile the citation-detection regex.
            2. Initialize ``self.nlp`` to ``None``.
            3. If spaCy is installed, attempt to load ``en_core_web_sm``.
            4. If the model is unavailable, log a warning and disable NER.
        """
        # Compile the regex used to detect wikilink-style citations in Markdown.
        self.citation_pattern = re.compile(r"\[\[.*?\]\]")

        # Initialize the NLP pipeline reference to ``None`` until a model is loaded successfully.
        self.nlp = None

        # Proceed with NLP initialization only if spaCy is installed.
        if SPACY_AVAILABLE:
            # Attempt to load the lightweight English model used for NER.
            try:
                # Load the English small pipeline for paragraph-level entity detection.
                self.nlp = spacy.load("en_core_web_sm")

            # Disable NER scanning gracefully when the model package is missing.
            except OSError:
                # Emit a warning because the code package exists but the model weights are unavailable.
                logger.warning(
                    "spaCy model 'en_core_web_sm' not found. NER scanning disabled."
                )

    # Define the content scanner that emits paragraph-level warnings.
    def scan_content(self, content: str, file_path: str) -> List[str]:
        """
        Scan Markdown content paragraph by paragraph for unsupported entities.

        Inputs:
            content:
                Markdown content to inspect.
            file_path:
                Repo-relative path used in warning messages.

        Processes:
            1. Validate input types.
            2. Return immediately if the NLP pipeline is unavailable.
            3. Split content into paragraphs on double newlines.
            4. Skip structural Markdown blocks such as headings, fenced code,
               and list items.
            5. Skip paragraphs that already contain a citation.
            6. Run NER on uncited paragraphs.
            7. Emit a warning if more than two formal named entities are found.

        Outputs:
            List[str]:
                Ordered list of warning messages for potentially unsupported
                paragraphs.

        Raises:
            TypeError:
                Raised when ``content`` or ``file_path`` is not a string.
        """
        # Validate that both inputs are strings before paragraph scanning begins.
        if not isinstance(content, str) or not isinstance(file_path, str):
            # Raise a type error because string operations and diagnostics require string inputs.
            raise TypeError("scan_content expected string inputs.")

        # Initialize the ordered warning list.
        warnings: List[str] = []

        # Return immediately if the NLP pipeline is unavailable.
        if not self.nlp:
            # Degrade safely by emitting no NER-based warnings.
            return warnings

        # Split the Markdown content into paragraphs on double newlines.
        paragraphs = content.split("\n\n")

        # Iterate over the paragraphs with 1-based human-readable numbering derived from the zero-based index.
        for i, para in enumerate(paragraphs):
            # Skip structural Markdown blocks such as headings, fenced code, and top-level bullet items.
            if para.startswith("#") or para.startswith("```") or para.startswith("- "):
                # Continue because structural blocks are not scanned with this heuristic.
                continue

            # Detect whether the paragraph already contains at least one citation.
            has_citation = bool(self.citation_pattern.search(para))

            # Only run the more expensive NLP pipeline if no citation is present.
            if not has_citation:
                # Process the paragraph text through the NLP pipeline.
                doc = self.nlp(para)

                # Define the set of entity labels interpreted here as formal factual entities.
                target_labels = {"PERSON", "ORG", "GPE", "EVENT"}

                # Collect the text of all formal named entities found in the paragraph.
                formal_entities = [
                    ent.text
                    for ent in doc.ents
                    if ent.label_ in target_labels
                ]

                # Emit a warning if the paragraph contains multiple formal entities but no citations.
                if len(formal_entities) > 2:
                    # Append a warning containing the file path, paragraph number, and a truncated entity preview.
                    warnings.append(
                        f"[{file_path}] Paragraph {i + 1} contains formal entities "
                        f"{formal_entities[:3]}... but lacks citations."
                    )

        # Return the ordered warning list.
        return warnings


# Define the mutable run-summary dataclass.
@dataclass
class RunSummary:
    """
    Mutable accumulator for run-level statistics.

    Purpose:
        Store coarse-grained counts and timing metadata summarizing the pipeline
        run.

    Inputs:
        This dataclass is populated and updated by RunSummaryAccumulator.

    Processes:
        No computation is performed here. It is a mutable state carrier.

    Outputs:
        run_id:
            Unique identifier of the pipeline run.
        start_time_utc:
            UTC ISO 8601 start timestamp.
        end_time_utc:
            UTC ISO 8601 end timestamp.
        wiki_files_written:
            Count of wiki files newly written.
        wiki_files_edited:
            Count of wiki files edited.
        lint_issues_found_count:
            Count of lint issues found.
        questions_answered_count:
            Count of questions answered.
        artifacts_generated_count:
            Count of derived artifacts generated.
        total_warnings:
            Total count of warnings accumulated across the run.
    """

    # Store the run identifier.
    run_id: str

    # Store the UTC start timestamp.
    start_time_utc: str

    # Store the UTC end timestamp.
    end_time_utc: str = ""

    # Store the count of wiki files written.
    wiki_files_written: int = 0

    # Store the count of wiki files edited.
    wiki_files_edited: int = 0

    # Store the count of lint issues found.
    lint_issues_found_count: int = 0

    # Store the count of questions answered.
    questions_answered_count: int = 0

    # Store the count of artifacts generated.
    artifacts_generated_count: int = 0

    # Store the total warning count.
    total_warnings: int = 0


# Define the centralized run-summary accumulator.
class RunSummaryAccumulator:
    """
    Centralized counter for pipeline events.

    Purpose:
        Maintain and update run-level counters for writes, edits, lint issues,
        answered questions, generated artifacts, and warnings.

    Inputs:
        run_id:
            Unique identifier of the current run.

    Processes:
        1. Initialize the summary with a UTC start timestamp.
        2. Increment counters as events occur.
        3. Finalize the summary with an end timestamp.

    Outputs:
        summary:
            The mutable RunSummary object tracked by the accumulator.
    """

    # Define the accumulator constructor.
    def __init__(
        self,
        run_id: str,
    ) -> None:
        """
        Initialize the run-summary accumulator.

        Purpose:
            Construct a mutable RunSummary object for the current run.

        Inputs:
            run_id:
                Unique identifier of the current run.

        Processes:
            1. Validate the input type.
            2. Create a RunSummary with the current UTC start time.

        Outputs:
            None. State is stored on the instance.

        Raises:
            TypeError:
                Raised if run_id is not a string.
        """
        # Validate that the run identifier is a string.
        if not isinstance(run_id, str):
            # Raise a type error because run IDs must be strings.
            raise TypeError(
                "RunSummaryAccumulator expected 'run_id' to be a string."
            )

        # Initialize the mutable run summary with the start timestamp.
        self.summary = RunSummary(
            run_id=run_id,
            start_time_utc=datetime.datetime.utcnow().isoformat() + "Z"
        )

    # Define the written-file counter increment method.
    def increment_written(
        self,
        count: int = 1,
    ) -> None:
        """
        Increment the count of wiki files written.

        Inputs:
            count:
                Number of writes to add.

        Processes:
            Add count to wiki_files_written.

        Outputs:
            None. State is mutated in place.
        """
        # Increment the written-file counter.
        self.summary.wiki_files_written += count

    # Define the edited-file counter increment method.
    def increment_edited(
        self,
        count: int = 1,
    ) -> None:
        """
        Increment the count of wiki files edited.

        Inputs:
            count:
                Number of edits to add.

        Processes:
            Add count to wiki_files_edited.

        Outputs:
            None. State is mutated in place.
        """
        # Increment the edited-file counter.
        self.summary.wiki_files_edited += count

    # Define the lint-issue counter increment method.
    def increment_lint_issues(
        self,
        count: int = 1,
    ) -> None:
        """
        Increment the count of lint issues found.

        Inputs:
            count:
                Number of lint issues to add.

        Processes:
            Add count to lint_issues_found_count.

        Outputs:
            None. State is mutated in place.
        """
        # Increment the lint-issue counter.
        self.summary.lint_issues_found_count += count

    # Define the answered-question counter increment method.
    def increment_qa(
        self,
        count: int = 1,
    ) -> None:
        """
        Increment the count of answered questions.

        Inputs:
            count:
                Number of answered questions to add.

        Processes:
            Add count to questions_answered_count.

        Outputs:
            None. State is mutated in place.
        """
        # Increment the answered-question counter.
        self.summary.questions_answered_count += count

    # Define the artifact counter increment method.
    def increment_artifacts(
        self,
        count: int = 1,
    ) -> None:
        """
        Increment the count of generated artifacts.

        Inputs:
            count:
                Number of artifacts to add.

        Processes:
            Add count to artifacts_generated_count.

        Outputs:
            None. State is mutated in place.
        """
        # Increment the artifact counter.
        self.summary.artifacts_generated_count += count

    # Define the warning counter increment method.
    def increment_warnings(
        self,
        count: int = 1,
    ) -> None:
        """
        Increment the total warning count.

        Inputs:
            count:
                Number of warnings to add.

        Processes:
            Add count to total_warnings.

        Outputs:
            None. State is mutated in place.
        """
        # Increment the warning counter.
        self.summary.total_warnings += count

    # Define the run-finalization method.
    def finalize(self) -> RunSummary:
        """
        Finalize the run summary by setting the end timestamp.

        Purpose:
            Mark the end of the run and return the completed summary object.

        Inputs:
            None.

        Processes:
            1. Set the UTC end timestamp.
            2. Return the summary.

        Outputs:
            RunSummary:
                The completed run summary.
        """
        # Set the UTC end timestamp for the run.
        self.summary.end_time_utc = (
            datetime.datetime.utcnow().isoformat() + "Z"
        )

        # Return the finalized summary.
        return self.summary


# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable governance-context aggregate returned by Task 8.
@dataclass(frozen=True)
class GovernanceContext:
    """
    Immutable record encapsulating the initialized governance and auditing tools.

    Purpose:
        Package the logging repository, write validator, heuristic scanner,
        run-summary accumulator, and run ID into one immutable object for
        downstream tasks.

    Inputs:
        This dataclass is populated by ``task_8_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        logging_repository:
            Append-only JSONL logging repository.
        write_validator:
            Centralized file-operation validator.
        heuristic_scanner:
            No-fabrication heuristic scanner.
        summary_accumulator:
            Mutable run-summary accumulator.
        run_id:
            Unique identifier of the current run.
    """

    # Store the logging repository.
    logging_repository: LoggingRepository

    # Store the write validator.
    write_validator: WriteValidator

    # Store the heuristic scanner.
    heuristic_scanner: FabricationHeuristicScanner

    # Store the run-summary accumulator.
    summary_accumulator: RunSummaryAccumulator

    # Store the unique run identifier.
    run_id: str


# Define the Task 8 orchestrator.
def task_8_orchestrator(
    contract: FrozenRunContract,
    wiki_context: WikiArchitectureContext,
    write_policy: WritePolicyRegistry,
    repo_root: pathlib.Path
) -> GovernanceContext:
    """
    Orchestrate Task 8 end-to-end by initializing logging, write validation,
    heuristic scanning, and run summaries.

    Purpose:
        This function is the Task 8 entry point. It constructs the governance
        toolchain required to support safe LLM file production and auditable run
        execution.

    Inputs:
        contract:
            The frozen run contract containing filesystem configuration.
        wiki_context:
            The wiki architecture context containing the wiki graph.
        write_policy:
            The write-policy registry from Task 3.
        repo_root:
            Project root directory used to resolve the log directory.

    Processes:
        1. Validate top-level inputs.
        2. Generate a unique run ID using UUID4.
        3. Resolve the configured logs directory relative to the repository root.
        4. Step 1: Instantiate the LoggingRepository.
        5. Step 2: Instantiate the WriteValidator.
        6. Step 3: Instantiate the FabricationHeuristicScanner and
           RunSummaryAccumulator.
        7. Return an immutable GovernanceContext aggregating the governance
           toolset.

    Outputs:
        GovernanceContext:
            Immutable aggregate containing the governance toolchain.

    Raises:
        TypeError:
            Raised if top-level inputs have incompatible types.
        OSError:
            Propagated if log directory creation fails.

    Notes:
        This orchestrator initializes the governance layer that supports the
        study's logging schema:
        \[
        \text{log\_entry} = \langle t, \tau, m, \theta, I, O, \Delta \rangle
        \]
        and its file-safety and no-fabrication controls.

        The original orchestration logic is preserved exactly for valid inputs.
    """
    # Validate that the run contract has the expected structural type.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because governance initialization requires the
        # frozen contract.
        raise TypeError(
            "task_8_orchestrator expected 'contract' to be an instance of "
            "FrozenRunContract."
        )

    # Validate that the wiki context has the expected structural type.
    if not isinstance(wiki_context, WikiArchitectureContext):
        # Raise a type error because write validation requires the wiki context.
        raise TypeError(
            "task_8_orchestrator expected 'wiki_context' to be an instance of "
            "WikiArchitectureContext."
        )

    # Validate that the write policy exposes the expected validation method.
    if not hasattr(write_policy, "validate_write_path"):
        # Raise a type error because governance initialization requires a
        # write-policy object with the expected interface.
        raise TypeError(
            "task_8_orchestrator expected 'write_policy' to expose "
            "'validate_write_path'."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_8_orchestrator expected 'repo_root' to be a pathlib.Path "
            "instance."
        )

    # Generate a unique run identifier for this execution.
    run_id: str = str(uuid.uuid4())

    # Extract the logs-directory string from the frozen configuration.
    log_dir_str = (
        contract.original_config
        .get("filesystem_conventions", {})
        .get("logs_directory", "logs/")
    )

    # Resolve the logs directory relative to the repository root.
    log_dir = (repo_root / log_dir_str).resolve()

    # Step 1: Instantiate the append-only logging repository.
    logging_repo = LoggingRepository(log_dir=log_dir, run_id=run_id)

    # Step 2: Instantiate the centralized write validator.
    write_validator = WriteValidator(
        write_policy=write_policy,
        wiki_graph=wiki_context.wiki_graph,
        repo_root=repo_root
    )

    # Step 3: Instantiate the no-fabrication heuristic scanner.
    heuristic_scanner = FabricationHeuristicScanner()

    # Step 3: Instantiate the run-summary accumulator.
    summary_accumulator = RunSummaryAccumulator(run_id=run_id)

    # Construct and return the immutable governance context.
    return GovernanceContext(
        logging_repository=logging_repo,
        write_validator=write_validator,
        heuristic_scanner=heuristic_scanner,
        summary_accumulator=summary_accumulator,
        run_id=run_id
    )
````

````python
# Task 9 — Register the prompt templates and task-specific LLM settings

# ==============================================================================
# Task 9: Register the prompt templates and task-specific LLM settings
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 1: Register the global system prompt and the incremental compilation prompt
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable prompt-registry entry record.
@dataclass(frozen=True)
class PromptRegistryEntry:
    """
    Immutable record binding a prompt text to its runtime settings and schema contract.

    Attributes:
        task_name:
            Stable task identifier for registry lookup.
        prompt_text:
            Final rendered prompt text supplied to the model.
        prompt_hash:
            SHA-256 hash of the prompt text for reproducibility and auditing.
        declared_settings:
            Settings as declared in the original configuration.
        effective_settings:
            Effective runtime settings after any upstream semantic translation.
        expected_output_schema:
            Schema contract expected from the model output for this prompt.
    """

    # Store the stable task identifier.
    task_name: str

    # Store the final prompt text to be sent to the model.
    prompt_text: str

    # Store the SHA-256 hash of the prompt text.
    prompt_hash: str

    # Store the settings declared in the source configuration.
    declared_settings: Dict[str, Any]

    # Store the effective settings that should govern the runtime invocation.
    effective_settings: Dict[str, Any]

    # Store the expected output schema associated with the prompt.
    expected_output_schema: Any


# Define the central prompt registry.
class PromptRegistry:
    """
    Centralized repository for validated and hashed prompt templates.

    Purpose:
        Store prompt definitions by task name, prevent duplicate registration,
        and allow the registry to be frozen once initialization is complete.
    """

    # Define the registry constructor.
    def __init__(self) -> None:
        """
        Initialize the mutable prompt registry.

        Processes:
            1. Create the internal task-to-entry mapping.
            2. Mark the registry as mutable until ``freeze`` is called.
        """
        # Initialize the internal dictionary that maps task names to prompt entries.
        self._entries: Dict[str, PromptRegistryEntry] = {}

        # Initialize the immutability flag that prevents registration after freezing.
        self._is_frozen: bool = False

    # Define the prompt-registration method.
    def register(self, entry: PromptRegistryEntry) -> None:
        """
        Register a new prompt entry.

        Inputs:
            entry:
                Prompt entry to register.

        Raises:
            RuntimeError:
                Raised when the registry has already been frozen.
            TypeError:
                Raised when ``entry`` is not a ``PromptRegistryEntry``.
            ValueError:
                Raised when a prompt for the same task is already registered.
        """
        # Reject registration once the registry has been frozen.
        if self._is_frozen:
            # Raise a runtime error because the registry is immutable after freezing.
            raise RuntimeError(
                "Cannot register prompts: PromptRegistry is frozen."
            )

        # Validate that the supplied entry is a PromptRegistryEntry.
        if not isinstance(entry, PromptRegistryEntry):
            # Raise a type error because the registry stores only typed prompt entries.
            raise TypeError("Expected PromptRegistryEntry.")

        # Reject duplicate task-name registration.
        if entry.task_name in self._entries:
            # Raise a value error because duplicate task registrations would create ambiguity.
            raise ValueError(
                f"Prompt for task '{entry.task_name}' is already registered."
            )

        # Store the new prompt entry under its task name.
        self._entries[entry.task_name] = entry

    # Define the prompt-lookup method.
    def get(self, task_name: str) -> PromptRegistryEntry:
        """
        Retrieve a registered prompt entry by task name.

        Inputs:
            task_name:
                Stable task identifier.

        Outputs:
            PromptRegistryEntry:
                The registered prompt entry for the task.

        Raises:
            KeyError:
                Raised when no prompt is registered for ``task_name``.
        """
        # Raise a key error if the requested task name is absent from the registry.
        if task_name not in self._entries:
            # Abort lookup because the registry has no entry for the requested task.
            raise KeyError(f"Prompt for task '{task_name}' not found.")

        # Return the registered prompt entry.
        return self._entries[task_name]

    # Define the registry-freeze method.
    def freeze(self) -> None:
        """
        Freeze the registry to prevent further mutation.

        Purpose:
            Mark prompt registration as complete so downstream orchestration can
            rely on prompt immutability.
        """
        # Set the immutability flag to True so subsequent registrations fail.
        self._is_frozen = True

# Define the internal prompt-hashing helper.
def _hash_prompt(text: str) -> str:
    """
    Compute the SHA-256 hash of a prompt string.

    Inputs:
        text:
            Prompt text to hash.

    Outputs:
        str:
            Hex-encoded SHA-256 digest of the prompt text.

    Notes:
        Prompt hashing is part of reproducibility control because it allows
        exact prompt identity to be recorded in run logs.
    """
    # Compute and return the SHA-256 hash of the UTF-8 encoded prompt text.
    return hashlib.sha256(text.encode("utf-8")).hexdigest()


# Define the registration function for the system and compilation prompts.
def register_system_and_compilation_prompts(
    contract: Any,
    registry: PromptRegistry,
) -> None:
    """
    Validate, hash, and register prompts using structured configuration objects.

    Purpose:
        Populate the prompt registry by reading prompt-related configuration from
        the frozen run contract, constructing final prompt strings, hashing them,
        and associating them with both declared and effective runtime settings.

    Inputs:
        contract:
            Frozen run-contract-like object expected to provide:
            - ``original_config``
            - ``effective_runtime_settings``
            - ``effective_runtime_settings.task_settings``
        registry:
            Mutable ``PromptRegistry`` into which entries will be registered.

    Processes:
        1. Extract prompt templates and runtime defaults from the contract.
        2. Read the structured system-prompt object.
        3. Construct the final system prompt from its structured fields.
        4. Hash and register the system prompt.
        5. Read the structured compilation-prompt object.
        6. Validate its required-output-format field.
        7. Construct the final compilation prompt.
        8. Hash and register the compilation prompt with its declared and
           effective settings.

    Outputs:
        None:
            The callable mutates ``registry`` by registering prompt entries.

    Raises:
        ValueError:
            Raised when the expected structured prompt objects are missing or
            malformed.

    Notes:
        The original callable's logic is preserved exactly. In particular, it
        expects configuration keys named:
            - ``system_prompt_structured``
            - ``wiki_compile_incremental_structured``

        If the frozen contract instead contains only raw string prompts, this
        callable will raise because the original logic depends on structured
        prompt dictionaries.
    """
    # Extract the prompt-templates block from the original frozen configuration.
    prompts_config = contract.original_config.get("prompt_templates", {})

    # Extract the declared task-default decoding settings from the original configuration.
    declared_defaults = (
        contract.original_config
        .get("llm_decoding_settings", {})
        .get("task_defaults", {})
    )

    # Extract the effective task-level runtime settings from the run contract.
    effective_settings_map = contract.effective_runtime_settings.task_settings

    # 1. Register System Prompt

    # Read the structured system-prompt object from the prompt configuration.
    sys_prompt_obj = prompts_config.get("system_prompt_structured", {})

    # Validate that the structured system-prompt object is a dictionary.
    if not isinstance(sys_prompt_obj, dict):
        # Raise a value error because the original logic expects a structured mapping.
        raise ValueError("system_prompt_structured must be a dictionary.")

    # Extract the base instruction text from the structured system prompt.
    base_instruction = sys_prompt_obj.get("instruction_text", "")

    # Extract the boolean flag that governs whether the no-fabrication rule is emitted.
    enforce_no_fabrication = sys_prompt_obj.get(
        "enforce_no_fabrication",
        False,
    )

    # Extract the boolean flag that governs whether the grounding rule is emitted.
    enforce_grounding = sys_prompt_obj.get("enforce_grounding", False)

    # Initialize the prompt-part list with the base instruction and the non-negotiable-rules heading.
    sys_prompt_parts = [base_instruction, "\nNon-negotiable rules:"]

    # Append the no-fabrication rule if requested by the structured system-prompt object.
    if enforce_no_fabrication:
        # Add the no-fabrication instruction to the final system-prompt assembly.
        sys_prompt_parts.append("1) Do not fabricate facts.")

    # Append the grounding rule if requested by the structured system-prompt object.
    if enforce_grounding:
        # Add the grounding instruction to the final system-prompt assembly.
        sys_prompt_parts.append(
            "2) Every nontrivial claim must be grounded in provided local sources."
        )

    # Join the system-prompt parts with newline separators to produce the final prompt text.
    final_sys_prompt = "\n".join(sys_prompt_parts)

    # Register the fully constructed and hashed system-prompt entry.
    registry.register(
        PromptRegistryEntry(
            task_name="system_prompt",
            prompt_text=final_sys_prompt,
            prompt_hash=_hash_prompt(final_sys_prompt),
            declared_settings={},
            effective_settings={},
            expected_output_schema=None,
        )
    )

    # 2. Register Incremental Compilation Prompt

    # Define the task name used to look up compilation settings.
    compile_task_name = "wiki_compile_incremental"

    # Read the structured compilation-prompt object from the prompt configuration.
    compile_obj = prompts_config.get("wiki_compile_incremental_structured", {})

    # Validate that the structured compilation prompt is a dictionary.
    if not isinstance(compile_obj, dict):
        # Raise a value error because the original logic expects a structured mapping.
        raise ValueError(
            "wiki_compile_incremental_structured must be a dictionary."
        )

    # Extract the main instruction text for the compilation prompt.
    compile_instruction = compile_obj.get("instruction_text", "")

    # Extract the list of required output keys for the structured JSON output.
    required_format = compile_obj.get("required_output_format", [])

    # Validate that the required-output-format field is a non-empty list.
    if not isinstance(required_format, list) or not required_format:
        # Raise a value error because the compilation prompt must declare a schema contract.
        raise ValueError(
            "Compilation prompt must specify a non-empty required_output_format list."
        )

    # Construct the final compilation prompt as specified by the original callable.
    compile_prompt_parts = [
        compile_instruction,
        "\nOUTPUT SCHEMA (JSON):",
        "Must contain the following keys: " + ", ".join(required_format),
    ]

    # Join the compilation-prompt parts into the final prompt string.
    final_compile_prompt = "\n".join(compile_prompt_parts)

    # Register the fully constructed and hashed compilation-prompt entry.
    registry.register(
        PromptRegistryEntry(
            task_name=compile_task_name,
            prompt_text=final_compile_prompt,
            prompt_hash=_hash_prompt(final_compile_prompt),
            declared_settings=declared_defaults.get(compile_task_name, {}),
            effective_settings=effective_settings_map.get(
                compile_task_name,
                {},
            ),
            expected_output_schema={
                "type": "json",
                "required_keys": required_format,
            },
        )
    )

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 2: Register the Q&A and artifact-rendering prompts
# -------------------------------------------------------------------------------------------------------------------------------

# Define the registration function for Q&A and rendering prompts.
def register_qa_and_rendering_prompts(
    contract: Any,
    registry: PromptRegistry,
) -> None:
    """
    Validate, hash, and register the grounded Q&A prompt and rendering prompts.

    Purpose:
        Populate the prompt registry with the grounded Q&A prompt, the Markdown
        report-rendering prompt, and the Marp slide-rendering prompt using
        structured configuration objects rather than brittle substring matching.

    Inputs:
        contract:
            Frozen run contract expected to expose:
            - ``original_config``
            - ``effective_runtime_settings``
            - ``effective_runtime_settings.task_settings``
        registry:
            Central prompt registry to populate.

    Processes:
        1. Validate the registry type and contract shape.
        2. Extract prompt templates and runtime settings from the contract.
        3. Register the Q&A prompt:
           - validate the structured dictionary,
           - extract ``instruction_text`` and ``required_headers``,
           - construct the final prompt string,
           - hash and register the prompt.
        4. Register the Markdown report prompt:
           - validate the structured dictionary,
           - extract ``instruction_text`` and ``required_headers``,
           - construct the final prompt string,
           - hash and register the prompt.
        5. Register the Marp prompt:
           - validate the structured dictionary,
           - extract boolean constraint flags,
           - construct the final prompt string,
           - hash and register the prompt.

    Outputs:
        None:
            The prompt registry is mutated in place.

    Raises:
        TypeError:
            Raised if ``registry`` is not a ``PromptRegistry``.
        ValueError:
            Raised if the required structured prompts are missing or malformed.

    Notes:
        This callable preserves the original logic exactly while adding explicit
        validation and documentation.
    """
    # Validate that the registry is a PromptRegistry instance.
    if not isinstance(registry, PromptRegistry):
        # Raise a type error because downstream registration depends on registry methods and invariants.
        raise TypeError(
            "register_qa_and_rendering_prompts expected 'registry' to be a PromptRegistry."
        )

    # Validate that the contract exposes the original configuration.
    if not hasattr(contract, "original_config"):
        # Raise a value error because prompt extraction depends on contract.original_config.
        raise ValueError(
            "register_qa_and_rendering_prompts expected 'contract' to expose "
            "'original_config'."
        )

    # Validate that the contract exposes effective runtime settings.
    if not hasattr(contract, "effective_runtime_settings"):
        # Raise a value error because effective prompt settings are drawn from the runtime contract.
        raise ValueError(
            "register_qa_and_rendering_prompts expected 'contract' to expose "
            "'effective_runtime_settings'."
        )

    # Extract the prompt templates block from the frozen configuration.
    prompts_config = contract.original_config.get("prompt_templates", {})

    # Extract the declared task defaults from the configuration.
    declared_defaults = (
        contract.original_config
        .get("llm_decoding_settings", {})
        .get("task_defaults", {})
    )

    # Extract the effective task-level runtime settings map.
    effective_settings_map = contract.effective_runtime_settings.task_settings

    # 1. Register Q&A Prompt

    # Define the task name for grounded Q&A.
    qa_task_name = "wiki_qa_synthesis"

    # Read the structured Q&A prompt object from the configuration.
    qa_obj = prompts_config.get("wiki_qa_structured", {})

    # Validate that the structured Q&A prompt is a non-empty dictionary.
    if not isinstance(qa_obj, dict) or not qa_obj:
        # Raise a value error because the original logic requires a structured mapping.
        raise ValueError("wiki_qa_structured must be a non-empty dictionary.")

    # Extract the Q&A instruction text.
    qa_instruction = qa_obj.get("instruction_text", "")

    # Extract the list of required Markdown headers.
    qa_required_headers = qa_obj.get("required_headers", [])

    # Validate that the required headers list is non-empty.
    if not isinstance(qa_required_headers, list) or not qa_required_headers:
        # Raise a value error because grounded Q&A requires a fixed section schema.
        raise ValueError(
            "QA prompt must specify a non-empty required_headers list."
        )

    # Construct the final Q&A prompt string from its structured components.
    qa_prompt_parts = [
        qa_instruction,
        "\nOUTPUT FORMAT (Markdown):",
        "Must contain the following exact headers: "
        + ", ".join(qa_required_headers),
    ]

    # Join the Q&A prompt parts into the final prompt text.
    final_qa_prompt = "\n".join(qa_prompt_parts)

    # Register the grounded Q&A prompt entry.
    registry.register(
        PromptRegistryEntry(
            task_name=qa_task_name,
            prompt_text=final_qa_prompt,
            prompt_hash=_hash_prompt(final_qa_prompt),
            declared_settings=declared_defaults.get(qa_task_name, {}),
            effective_settings=effective_settings_map.get(qa_task_name, {}),
            expected_output_schema={
                "type": "markdown_sections",
                "required_headers": qa_required_headers,
            },
        )
    )

    # 2. Register Markdown Report Prompt

    # Define the task name for Markdown report rendering.
    report_task_name = "artifact_render_markdown_report"

    # Read the structured Markdown report prompt object from the configuration.
    report_obj = prompts_config.get(
        "artifact_render_markdown_report_structured",
        {},
    )

    # Validate that the structured report prompt is a non-empty dictionary.
    if not isinstance(report_obj, dict) or not report_obj:
        # Raise a value error because the report renderer requires a structured prompt mapping.
        raise ValueError(
            "artifact_render_markdown_report_structured must be a non-empty "
            "dictionary."
        )

    # Extract the report instruction text.
    report_instruction = report_obj.get("instruction_text", "")

    # Extract the required report section headers.
    report_required_headers = report_obj.get("required_headers", [])

    # Validate that the report header list is non-empty.
    if not isinstance(report_required_headers, list) or not report_required_headers:
        # Raise a value error because report rendering requires a fixed section structure.
        raise ValueError(
            "Report prompt must specify a non-empty required_headers list."
        )

    # Construct the final Markdown report prompt string.
    report_prompt_parts = [
        report_instruction,
        "\nSTRUCTURE:",
        "Must contain the following exact headers: "
        + ", ".join(report_required_headers),
    ]

    # Join the report prompt parts into the final prompt text.
    final_report_prompt = "\n".join(report_prompt_parts)

    # Register the Markdown report prompt entry.
    registry.register(
        PromptRegistryEntry(
            task_name=report_task_name,
            prompt_text=final_report_prompt,
            prompt_hash=_hash_prompt(final_report_prompt),
            declared_settings=declared_defaults.get(report_task_name, {}),
            effective_settings=effective_settings_map.get(report_task_name, {}),
            expected_output_schema={
                "type": "markdown_sections",
                "required_headers": report_required_headers,
            },
        )
    )

    # 3. Register Marp Slides Prompt

    # Define the task name for Marp slide rendering.
    marp_task_name = "artifact_render_marp_slides"

    # Read the structured Marp prompt object from the configuration.
    marp_obj = prompts_config.get("artifact_render_marp_structured", {})

    # Validate that the structured Marp prompt is a non-empty dictionary.
    if not isinstance(marp_obj, dict) or not marp_obj:
        # Raise a value error because the slide renderer requires a structured prompt mapping.
        raise ValueError(
            "artifact_render_marp_structured must be a non-empty dictionary."
        )

    # Extract the Marp instruction text.
    marp_instruction = marp_obj.get("instruction_text", "")

    # Extract the boolean separator constraint flag.
    requires_separators = marp_obj.get("requires_separators", False)

    # Extract the boolean references-slide constraint flag.
    requires_references = marp_obj.get("requires_references", False)

    # Initialize the Marp prompt parts with the instruction text and constraints heading.
    marp_prompt_parts = [marp_instruction, "\nCONSTRAINTS:"]

    # Append the slide-separator constraint when requested.
    if requires_separators:
        # Add the slide-separator instruction.
        marp_prompt_parts.append("- Use '---' slide separators.")

    # Append the references-slide constraint when requested.
    if requires_references:
        # Add the references-slide instruction.
        marp_prompt_parts.append("- Include a final 'References' slide.")

    # Join the Marp prompt parts into the final prompt text.
    final_marp_prompt = "\n".join(marp_prompt_parts)

    # Register the Marp slide prompt entry.
    registry.register(
        PromptRegistryEntry(
            task_name=marp_task_name,
            prompt_text=final_marp_prompt,
            prompt_hash=_hash_prompt(final_marp_prompt),
            declared_settings=declared_defaults.get(marp_task_name, {}),
            effective_settings=effective_settings_map.get(marp_task_name, {}),
            expected_output_schema={
                "type": "marp_markdown",
                "requires_separators": requires_separators,
                "requires_references": requires_references,
            },
        )
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 3: Register the linting and synthetic-data prompts
# -------------------------------------------------------------------------------------------------------------------------------

# Define the registration function for linting and synthetic-data prompts.
def register_linting_and_synthetic_prompts(
    contract: Any,
    registry: PromptRegistry,
) -> None:
    """
    Validate, hash, and register the linting and synthetic-data generation prompts.

    Purpose:
        Populate the prompt registry with the linting prompt and the synthetic
        data generation prompt using structured configuration dictionaries rather
        than brittle substring matching.

    Inputs:
        contract:
            Frozen run contract expected to expose:
            - ``original_config``
            - ``effective_runtime_settings``
            - ``effective_runtime_settings.task_settings``
        registry:
            Central prompt registry to populate.

    Processes:
        1. Validate the registry type and contract shape.
        2. Extract prompt templates and runtime settings from the contract.
        3. Register the linting prompt:
           - validate the structured dictionary,
           - extract ``instruction_text`` and ``required_issue_classes``,
           - construct the final prompt string,
           - hash and register the prompt.
        4. Register the synthetic-data prompt:
           - validate the structured dictionary,
           - extract ``instruction_text`` and ``required_json_fields``,
           - construct the final prompt string,
           - hash and register the prompt.

    Outputs:
        None:
            The prompt registry is mutated in place.

    Raises:
        TypeError:
            Raised if ``registry`` is not a ``PromptRegistry``.
        ValueError:
            Raised if any required structured prompt is missing or malformed.

    Notes:
        The original logic is preserved exactly, including the enforced output
        schema types:
        - linting -> ``json``
        - synthetic data -> ``jsonl``
    """
    # Validate that the registry is a PromptRegistry instance.
    if not isinstance(registry, PromptRegistry):
        # Raise a type error because downstream registration depends on registry methods and invariants.
        raise TypeError(
            "register_linting_and_synthetic_prompts expected 'registry' to be "
            "a PromptRegistry."
        )

    # Validate that the contract exposes the original configuration.
    if not hasattr(contract, "original_config"):
        # Raise a value error because prompt extraction depends on contract.original_config.
        raise ValueError(
            "register_linting_and_synthetic_prompts expected 'contract' to "
            "expose 'original_config'."
        )

    # Validate that the contract exposes effective runtime settings.
    if not hasattr(contract, "effective_runtime_settings"):
        # Raise a value error because effective prompt settings are drawn from the runtime contract.
        raise ValueError(
            "register_linting_and_synthetic_prompts expected 'contract' to "
            "expose 'effective_runtime_settings'."
        )

    # Extract the prompt templates block from the frozen configuration.
    prompts_config = contract.original_config.get("prompt_templates", {})

    # Extract the declared task defaults from the configuration.
    declared_defaults = (
        contract.original_config
        .get("llm_decoding_settings", {})
        .get("task_defaults", {})
    )

    # Extract the effective task-level runtime settings map.
    effective_settings_map = contract.effective_runtime_settings.task_settings

    # 1. Register Linting Prompt

    # Define the task name for the linting stage.
    lint_task_name = "wiki_lint_health_check"

    # Read the structured linting prompt object from the configuration.
    lint_obj = prompts_config.get("wiki_lint_structured", {})

    # Validate that the structured linting prompt is a non-empty dictionary.
    if not isinstance(lint_obj, dict) or not lint_obj:
        # Raise a value error because the linting stage requires a structured prompt mapping.
        raise ValueError("wiki_lint_structured must be a non-empty dictionary.")

    # Extract the linting instruction text.
    lint_instruction = lint_obj.get("instruction_text", "")

    # Extract the list of required issue classes.
    lint_required_issues = lint_obj.get("required_issue_classes", [])

    # Validate that the required issue-class list is non-empty.
    if not isinstance(lint_required_issues, list) or not lint_required_issues:
        # Raise a value error because the linting contract must explicitly specify required issue classes.
        raise ValueError(
            "Lint prompt must specify a non-empty required_issue_classes list."
        )

    # Construct the final linting prompt string.
    lint_prompt_parts = [
        lint_instruction,
        "\nFIND AND REPORT:",
        "Must detect the following issue classes: "
        + ", ".join(lint_required_issues),
    ]

    # Join the linting prompt parts into the final prompt text.
    final_lint_prompt = "\n".join(lint_prompt_parts)

    # Register the linting prompt entry.
    registry.register(
        PromptRegistryEntry(
            task_name=lint_task_name,
            prompt_text=final_lint_prompt,
            prompt_hash=_hash_prompt(final_lint_prompt),
            declared_settings=declared_defaults.get(lint_task_name, {}),
            effective_settings=effective_settings_map.get(lint_task_name, {}),
            expected_output_schema={
                "type": "json",
                "required_issue_classes": lint_required_issues,
            },
        )
    )

    # 2. Register Synthetic Data Generation Prompt

    # Define the task name for synthetic data generation.
    synth_task_name = "synthetic_data_generation"

    # Read the structured synthetic-data prompt object from the configuration.
    synth_obj = prompts_config.get(
        "synthetic_data_generation_structured",
        {},
    )

    # Validate that the structured synthetic-data prompt is a non-empty dictionary.
    if not isinstance(synth_obj, dict) or not synth_obj:
        # Raise a value error because synthetic-data generation requires a structured prompt mapping.
        raise ValueError(
            "synthetic_data_generation_structured must be a non-empty "
            "dictionary."
        )

    # Extract the synthetic-data instruction text.
    synth_instruction = synth_obj.get("instruction_text", "")

    # Extract the required JSONL field names.
    synth_required_fields = synth_obj.get("required_json_fields", [])

    # Validate that the required field list is non-empty.
    if not isinstance(synth_required_fields, list) or not synth_required_fields:
        # Raise a value error because the synthetic-data schema must be explicit.
        raise ValueError(
            "Synthetic data prompt must specify a non-empty "
            "required_json_fields list."
        )

    # Construct the final synthetic-data prompt string.
    synth_prompt_parts = [
        synth_instruction,
        "\nOUTPUT: JSONL (one JSON object per line) with fields:",
        ", ".join(synth_required_fields),
    ]

    # Join the synthetic-data prompt parts into the final prompt text.
    final_synth_prompt = "\n".join(synth_prompt_parts)

    # Register the synthetic-data prompt entry.
    registry.register(
        PromptRegistryEntry(
            task_name=synth_task_name,
            prompt_text=final_synth_prompt,
            prompt_hash=_hash_prompt(final_synth_prompt),
            declared_settings=declared_defaults.get(synth_task_name, {}),
            effective_settings=effective_settings_map.get(synth_task_name, {}),
            expected_output_schema={
                "type": "jsonl",
                "required_keys": synth_required_fields,
            },
        )
    )

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable prompt-context aggregate returned by Task 9.
@dataclass(frozen=True)
class PromptContext:
    """
    Immutable record encapsulating the fully populated and validated prompt
    registry.

    Purpose:
        Package the complete prompt registry into one immutable object for
        downstream tasks.

    Inputs:
        This dataclass is populated by ``task_9_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        prompt_registry:
            The populated prompt registry containing all validated prompt
            entries.
    """

    # Store the populated prompt registry.
    prompt_registry: PromptRegistry


# Define the Task 9 orchestrator.
def task_9_orchestrator(
    contract: FrozenRunContract,
) -> PromptContext:
    """
    Orchestrate Task 9 end-to-end by validating, hashing, and registering all
    prompt templates.

    Purpose:
        This function is the Task 9 entry point. It executes all Task 9 prompt-
        registration steps in sequence and returns the unified prompt context.

    Inputs:
        contract:
            Frozen run contract containing prompt templates and task-specific
            runtime settings.

    Processes:
        1. Validate that the contract has the expected structural type.
        2. Instantiate an empty PromptRegistry.
        3. Step 1: Register the system and compilation prompts.
        4. Step 2: Register the Q&A and rendering prompts.
        5. Step 3: Register the linting and synthetic-data prompts.
        6. Return an immutable PromptContext containing the populated registry.

    Outputs:
        PromptContext:
            Immutable aggregate containing the validated prompt registry.

    Raises:
        TypeError:
            Raised if the contract type is invalid.
        ValueError:
            Raised if any required prompt is missing or violates structural
            constraints.
        KeyError:
            May propagate if downstream registry access is misused in future
            extensions.

    Notes:
        This orchestrator acts as a hard gate for prompt integrity. If any prompt
        is missing, malformed, or structurally inconsistent with its required
        schema, execution halts.

        The original orchestration logic is preserved exactly for valid inputs.
    """
    # Validate that the contract has the expected structural type.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because prompt registration requires the frozen
        # contract.
        raise TypeError(
            "task_9_orchestrator expected 'contract' to be an instance of "
            "FrozenRunContract."
        )

    # Instantiate the empty prompt registry.
    registry = PromptRegistry()

    # Step 1: Register the system and compilation prompts.
    register_system_and_compilation_prompts(contract, registry)

    # Step 2: Register the Q&A and rendering prompts.
    register_qa_and_rendering_prompts(contract, registry)

    # Step 3: Register the linting and synthetic-data prompts.
    register_linting_and_synthetic_prompts(contract, registry)

    # Construct and return the immutable prompt context.
    return PromptContext(prompt_registry=registry)
````

````python
# Task 10 — Implement the incremental wiki compilation callable

# ==============================================================================
# Task 10: Implement the incremental wiki compilation callable
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 1: Execute Algorithm 1: Incremental Wiki Compilation (Context Assembly)
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable chunk record.
@dataclass(frozen=True)
class ChunkRecord:
    """
    Immutable record representing one text chunk prepared for compilation.

    Attributes:
        ordinal:
            Zero-based chunk index preserving chunk order.
        content:
            Chunk text content.
        char_span:
            Inclusive-exclusive character-span tuple ``(start, end)`` into the
            original source text.
    """

    # Store the zero-based chunk ordinal.
    ordinal: int

    # Store the chunk text content.
    content: str

    # Store the character-span tuple in the original source text.
    char_span: Tuple[int, int]


# Define the immutable compilation work-package record.
@dataclass(frozen=True)
class ChunkRecord:
    """
    Immutable record representing one text chunk prepared for compilation.

    Attributes:
        ordinal:
            Zero-based chunk index preserving chunk order.
        content:
            Chunk text content.
        char_span:
            Inclusive-exclusive character-span tuple ``(start, end)`` into the
            original source text.
    """

    # Store the zero-based chunk ordinal.
    ordinal: int

    # Store the chunk text content.
    content: str

    # Store the character-span tuple in the original source text.
    char_span: Tuple[int, int]


# Declare the immutable work package supplied to one compilation LLM call.
@dataclass(frozen=True)
class CompilationWorkPackage:
    """
    Immutable record encapsulating the context required for one compilation call.

    Purpose:
        Represent the input package assembled for the compilation operator
        described in Algorithm 1 of the study.

    Attributes:
        artifact_id:
            Stable identifier of the changed raw artifact.
        artifact_type:
            Raw artifact type string.
        chunks:
            Ordered list of text chunks derived from the normalized artifact.
        existing_summary_content:
            Existing source-summary page content, if present.
        relevant_concept_contents:
            Mapping from concept-page paths to their contents.
        relevant_index_contents:
            Mapping from index-page paths to their contents.
    """

    # Store the changed raw artifact identifier.
    artifact_id: str

    # Store the changed raw artifact type string.
    artifact_type: str

    # Store the ordered chunk list for the normalized artifact content.
    chunks: List[ChunkRecord]

    # Store the existing source-summary content when available.
    existing_summary_content: Optional[str]

    # Store relevant concept-page contents keyed by page path.
    relevant_concept_contents: Dict[str, str]

    # Store relevant index-page contents keyed by page path.
    relevant_index_contents: Dict[str, str]


# Define the internal semantic chunking helper.
def _semantic_chunk_text(
    text: str,
    max_chars: int = 12000,
    overlap: int = 800,
) -> List[ChunkRecord]:
    """
    Split text into chunks using AST-aware heading boundaries.

    Purpose:
        Partition long source text into model-manageable chunks while preserving
        semantic boundaries as much as possible. This supports the context
        assembly portion of Algorithm 1 by implementing the configured chunking
        policy:
        - ``max_chars = 12000``
        - ``overlap = 800``

    Inputs:
        text:
            Source text to split.
        max_chars:
            Maximum target chunk size in characters.
        overlap:
            Character overlap between adjacent chunks.

    Processes:
        1. Validate input types.
        2. Return an empty list for empty input.
        3. If markdown-it-py is available, parse the Markdown AST and derive
           heading-aligned character boundaries.
        4. Otherwise, fall back to regex-based heading detection.
        5. Iteratively select chunk boundaries, preferring the last heading
           boundary inside the current window.
        6. Fall back to a hard character split when no boundary exists.
        7. Apply the configured overlap and emit ordered ``ChunkRecord`` objects.

    Outputs:
        List[ChunkRecord]:
            Ordered list of chunk records.

    Raises:
        TypeError:
            Raised if ``text`` is not a string or if ``max_chars`` / ``overlap``
            are not integers.
    """
    # Validate that the input text is a string.
    if not isinstance(text, str):
        # Raise a type error because chunking logic expects a string.
        raise TypeError("_semantic_chunk_text expected 'text' to be a string.")

    # Validate that max_chars is an integer.
    if not isinstance(max_chars, int) or not isinstance(overlap, int):
        # Raise a type error because chunk-window arithmetic requires integers.
        raise TypeError("max_chars and overlap must be integers.")

    # Initialize the ordered chunk accumulator.
    chunks: List[ChunkRecord] = []

    # Return early for empty input text.
    if not text:
        # Emit an empty chunk list for empty source text.
        return chunks

    # Initialize the boundary list with the start-of-text boundary.
    boundaries: List[int] = [0]

    # Attempt AST-aware heading-boundary detection when markdown-it-py is available.
    if MARKDOWN_IT_AVAILABLE:
        # Initialize the CommonMark parser.
        md = MarkdownIt("commonmark")

        # Parse the text into a Markdown token stream.
        tokens = md.parse(text)

        # Initialize the line-number to character-index map with the start of the text.
        line_to_char_idx = [0]

        # Traverse the text character by character to map line starts to character indices.
        for i, char in enumerate(text):
            # Record the next character index whenever a newline is encountered.
            if char == "\n":
                # Append the character index immediately after the newline.
                line_to_char_idx.append(i + 1)

        # Traverse the Markdown tokens to locate true heading boundaries.
        for token in tokens:
            # Restrict extraction to heading-open tokens that expose source line spans.
            if token.type == "heading_open" and token.map:
                # Read the starting source line of the heading.
                line_num = token.map[0]

                # Convert the source line number to a character boundary when possible.
                if line_num < len(line_to_char_idx):
                    # Append the heading-start character boundary.
                    boundaries.append(line_to_char_idx[line_num])

    # Fall back to regex-based heading detection when AST parsing is unavailable.
    else:
        # Emit a warning because regex chunking is less robust than AST chunking.
        logger.warning(
            "markdown-it-py unavailable. Falling back to regex chunking."
        )

        # Compile the regex that approximates Markdown heading starts.
        heading_pattern = re.compile(r"\n#+\s+")

        # Extend the boundary list with regex-derived heading positions.
        boundaries.extend([m.start() for m in heading_pattern.finditer(text)])

    # Append the end-of-text boundary.
    boundaries.append(len(text))

    # Sort and deduplicate the boundaries for stable chunk computation.
    boundaries = sorted(list(set(boundaries)))

    # Initialize the current chunk-start pointer.
    current_chunk_start = 0

    # Initialize the chunk ordinal.
    ordinal = 0

    # Iterate until the chunk-start pointer reaches the end of the text.
    while current_chunk_start < len(text):
        # Compute the ideal hard upper bound for the current chunk.
        ideal_end = current_chunk_start + max_chars

        # If the ideal end passes the end of the text, clamp to the text length.
        if ideal_end >= len(text):
            # Use the end of the text as the actual chunk end.
            actual_end = len(text)

        # Otherwise, attempt to align the chunk end to a heading boundary.
        else:
            # Collect heading boundaries that fall strictly داخل the current chunk window.
            valid_boundaries = [
                b
                for b in boundaries
                if current_chunk_start < b <= ideal_end
            ]

            # Prefer the last valid heading boundary within the window.
            if valid_boundaries:
                # Use the last in-window heading boundary as the chunk end.
                actual_end = valid_boundaries[-1]

            # Fall back to a hard character split when no boundary exists.
            else:
                # Use the ideal hard window boundary.
                actual_end = ideal_end

        # Extract the chunk text for the computed character span.
        chunk_content = text[current_chunk_start:actual_end]

        # Append the chunk record for the current span.
        chunks.append(
            ChunkRecord(
                ordinal=ordinal,
                content=chunk_content,
                char_span=(current_chunk_start, actual_end),
            )
        )

        # Increment the chunk ordinal.
        ordinal += 1

        # Advance the chunk-start pointer using the configured overlap rule.
        current_chunk_start = (
            actual_end - overlap
            if actual_end < len(text)
            else len(text)
        )

        # Prevent infinite loops in pathological overlap configurations.
        if current_chunk_start <= chunks[-1].char_span[0]:
            # Force forward progress by advancing to the current actual_end.
            current_chunk_start = actual_end

    # Return the ordered list of chunk records.
    return chunks


# Define the compilation-context assembly function.
def assemble_compilation_context(
    artifact_id: str,
    change_context: Any,
    normalization_context: Any,
    wiki_context: Any,
    repo_root: pathlib.Path,
) -> CompilationWorkPackage:
    """
    Gather and chunk the necessary context for a changed artifact.

    Purpose:
        Implement the context-assembly stage of Algorithm 1:
            1. detect changed items,
            2. gather relevant context,
            3. prepare model-ready payloads.

        This implementation correctly reads the normalized derivative rather
        than bypassing the normalization layer.

    Inputs:
        artifact_id:
            Identifier of the changed raw artifact.
        change_context:
            Context expected to expose ``raw_manifest.item_index``.
        normalization_context:
            Context expected to expose ``normalized_derivatives``.
        wiki_context:
            Context expected to expose ``wiki_graph``.
        repo_root:
            Repository root used to resolve derivative paths.

    Processes:
        1. Validate input types and required context attributes.
        2. Retrieve the raw artifact metadata from the raw manifest.
        3. Locate the normalized derivative record corresponding to the
           artifact.
        4. Read the normalized derivative from disk.
        5. Chunk the normalized text using the configured chunking policy.
        6. Search the wiki graph for an existing source summary.
        7. Gather successor concept-page contents related to the source summary.
        8. Return the immutable ``CompilationWorkPackage``.

    Outputs:
        CompilationWorkPackage:
            Immutable context package for one compilation LLM call.

    Raises:
        TypeError:
            Raised when ``artifact_id`` or ``repo_root`` has the wrong type.
        ValueError:
            Raised when the artifact is absent from the raw manifest or has no
            normalized derivative.
        RuntimeError:
            Raised when the derivative file is missing from disk.

    Notes:
        This callable directly supports the study's incremental compilation
        operator:

            Φ_compile : D_raw × K_(t-1) → K_t
    """
    # Validate that the artifact identifier is a string.
    if not isinstance(artifact_id, str):
        # Raise a type error because manifest lookup expects a string identifier.
        raise TypeError(
            "assemble_compilation_context expected 'artifact_id' to be a string."
        )

    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because derivative path resolution requires pathlib semantics.
        raise TypeError(
            "assemble_compilation_context expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that the change context exposes the raw manifest.
    if not hasattr(change_context, "raw_manifest"):
        # Raise a value error because artifact metadata lookup depends on the raw manifest.
        raise ValueError(
            "assemble_compilation_context expected 'change_context' to expose "
            "'raw_manifest'."
        )

    # Validate that the normalization context exposes normalized derivatives.
    if not hasattr(normalization_context, "normalized_derivatives"):
        # Raise a value error because derivative lookup depends on normalized_derivatives.
        raise ValueError(
            "assemble_compilation_context expected 'normalization_context' to "
            "expose 'normalized_derivatives'."
        )

    # Validate that the wiki context exposes a wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a value error because source-summary and concept lookup depend on the wiki graph.
        raise ValueError(
            "assemble_compilation_context expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # 1. Retrieve the raw artifact metadata from the raw-manifest index.
    raw_item = change_context.raw_manifest.item_index.get(artifact_id)

    # Halt if the artifact is absent from the raw manifest.
    if not raw_item:
        # Raise a value error because the changed artifact is not represented in the manifest.
        raise ValueError(f"Artifact {artifact_id} not found in raw manifest.")

    # 2. Retrieve the normalized derivative record corresponding to the artifact.
    derivative_record = None

    # Scan the normalized derivatives for the one originating from this raw artifact.
    for record in normalization_context.normalized_derivatives:
        # Match by raw artifact identifier.
        if record.raw_artifact_id == artifact_id:
            # Store the matching derivative record.
            derivative_record = record

            # Exit the search loop after the first match.
            break

    # Halt if no derivative record exists for the artifact.
    if not derivative_record:
        # Raise a value error because the normalization layer was bypassed or failed.
        raise ValueError(
            f"No normalized derivative found for artifact {artifact_id}."
        )

    # 3. Resolve the absolute path of the normalized derivative.
    abs_path = (repo_root / derivative_record.derivative_path).resolve()

    # Validate that the derivative exists on disk and is a file.
    if not abs_path.exists() or not abs_path.is_file():
        # Raise a runtime error because the pipeline state is inconsistent with its metadata.
        raise RuntimeError(f"Derivative file missing from disk: {abs_path}")

    # Read the normalized derivative content using strict UTF-8 decoding.
    content = abs_path.read_text(encoding="utf-8", errors="strict")

    # 4. Apply the configured semantic chunking policy.
    chunks = _semantic_chunk_text(content)

    # 5. Initialize containers for relevant existing wiki context.
    existing_summary = None
    relevant_concepts: Dict[str, str] = {}
    relevant_indices: Dict[str, str] = {}

    # Retrieve the wiki graph collaborator from the wiki context.
    wiki_graph = wiki_context.wiki_graph

    # Compute the expected slug of the source summary from the artifact title.
    expected_slug = wiki_graph._normalize_slug(raw_item.title_or_name)

    # Attempt to resolve the expected summary path from the slug registry.
    summary_path = wiki_graph.slug_registry.get(expected_slug)

    # Gather existing summary content and neighboring concept pages when the summary exists.
    if summary_path:
        # Read the existing source-summary content.
        existing_summary = wiki_graph.get_content(summary_path)

        # Proceed only if the summary path exists as a graph node.
        if summary_path in wiki_graph.graph:
            # Traverse outgoing neighbors from the summary page.
            for neighbor in wiki_graph.graph.successors(summary_path):
                # Retrieve the metadata payload for the neighboring page.
                neighbor_meta = wiki_graph.graph.nodes[neighbor].get("metadata")

                # Restrict gathered neighbors to concept pages.
                if (
                    neighbor_meta
                    and getattr(
                        neighbor_meta.type,
                        "value",
                        str(neighbor_meta.type),
                    ) == "concept"
                ):
                    # Read and store the concept-page content.
                    relevant_concepts[neighbor] = wiki_graph.get_content(neighbor)

    # Construct and return the immutable compilation work package.
    return CompilationWorkPackage(
        artifact_id=artifact_id,
        artifact_type=raw_item.type,
        chunks=chunks,
        existing_summary_content=existing_summary,
        relevant_concept_contents=relevant_concepts,
        relevant_index_contents=relevant_indices,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 2: Construct and issue the compilation LLM call
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable compilation LLM response record.
@dataclass(frozen=True)
class CompilationLLMResponse:
    """
    Immutable record representing the parsed output of one compilation model call.

    Attributes:
        raw_json_string:
            Raw textual model response captured before structured parsing.
        parsed_dict:
            Parsed structured payload extracted from the model response.
        token_usage:
            Token-usage accounting returned by the model wrapper.
    """

    # Store the raw response text captured from the model wrapper.
    raw_json_string: str

    # Store the parsed structured payload.
    parsed_dict: Dict[str, Any]

    # Store token-usage accounting information.
    token_usage: Dict[str, int]



# Define the compilation model-call function.
def issue_compilation_call(
    work_package: CompilationWorkPackage,
    prompt_context: Any,
    governance_context: Any,
) -> CompilationLLMResponse:
    """
    Construct the compilation prompt payload and invoke the GPT-5.2 reasoning model.

    Purpose:
        Implement the model-invocation stage of Algorithm 1. In the paper's
        notation, this corresponds to:

            o ← m(π; θ_compile)

        where:
            - ``π`` is the assembled prompt payload,
            - ``m`` is the configured model,
            - ``θ_compile`` are the compile-time runtime settings.

        This implementation also performs robust JSON extraction using a
        stack-based bracket-matching procedure rather than fragile fence-based
        regex parsing.

    Inputs:
        work_package:
            Assembled compilation context for a single changed artifact.
        prompt_context:
            Context expected to expose ``prompt_registry``.
        governance_context:
            Context expected to expose ``summary_accumulator`` for warning
            counting.

    Processes:
        1. Validate input types and required collaborators.
        2. Retrieve the registered system and compilation prompts.
        3. Construct the user payload from the work package.
        4. Read effective runtime settings from the compilation prompt entry.
        5. Invoke ``query_gpt52`` with up to two attempts.
        6. Extract the first balanced JSON object using stack-based bracket
           matching.
        7. Parse the JSON payload and validate required keys.
        8. On repeated parse failure, increment warnings and return a structured
           empty fallback.

    Outputs:
        CompilationLLMResponse:
            Immutable parsed response record.

    Raises:
        TypeError:
            Raised if ``work_package`` has the wrong type.
        ValueError:
            Propagated if the prompt registry is missing required entries.

    Error Handling:
        - JSON parse failures and schema validation failures trigger retries.
        - After exhausting retries, the callable returns an empty structured
          fallback rather than raising.
    """
    # Validate that the work package is of the expected type.
    if not isinstance(work_package, CompilationWorkPackage):
        # Raise a type error because prompt construction depends on the work-package schema.
        raise TypeError(
            "issue_compilation_call expected 'work_package' to be a "
            "CompilationWorkPackage."
        )

    # Validate that the prompt context exposes the prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a value error because prompt retrieval depends on prompt_context.prompt_registry.
        raise ValueError(
            "issue_compilation_call expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a value error because warning accounting depends on summary_accumulator.
        raise ValueError(
            "issue_compilation_call expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Retrieve the registered system prompt entry.
    sys_prompt_entry = prompt_context.prompt_registry.get("system_prompt")

    # Retrieve the registered compilation prompt entry.
    compile_prompt_entry = prompt_context.prompt_registry.get(
        "wiki_compile_incremental"
    )

    # Initialize the user payload with artifact identity fields.
    payload_parts = [
        f"Artifact ID: {work_package.artifact_id}",
        f"Type: {work_package.artifact_type}",
    ]

    # Include the existing summary content if present.
    if work_package.existing_summary_content:
        # Append the existing summary block to the payload.
        payload_parts.append(
            f"--- EXISTING SUMMARY ---\n"
            f"{work_package.existing_summary_content}"
        )

    # Append each chunk block in order so the model receives the full normalized source context.
    for chunk in work_package.chunks:
        # Append the current chunk with its ordinal label.
        payload_parts.append(f"--- CHUNK {chunk.ordinal} ---\n{chunk.content}")

    # Construct the final user text by concatenating the compilation prompt and payload sections.
    user_text = compile_prompt_entry.prompt_text + "\n\n" + "\n".join(payload_parts)

    # Extract the effective reasoning-effort setting, defaulting to "high".
    effort = compile_prompt_entry.effective_settings.get(
        "reasoning_effort",
        "high",
    )

    # Extract the effective max-output-token budget, defaulting to 12000.
    max_tokens = compile_prompt_entry.effective_settings.get(
        "max_output_tokens",
        12000,
    )

    # Set the maximum number of parse-retry attempts.
    max_retries = 2

    # Attempt the model call up to the configured retry limit.
    for attempt in range(max_retries):
        try:
            # Invoke the GPT-5.2 query wrapper with the constructed prompt payload.
            result = query_gpt52(
                prompt=user_text,
                system_prompt=sys_prompt_entry.prompt_text,
                reasoning_effort=effort,
                max_output_tokens=max_tokens,
                log_run_manifest=False,
            )

            # Extract the raw response text from the wrapper result.
            raw_text = str(result.get("response_text", ""))

            # Extract the usage accounting from the wrapper result.
            usage = dict(result.get("usage", {}))

            # Initialize the extracted JSON substring accumulator.
            json_str = ""

            # Initialize the stack used for balanced-brace detection.
            stack = []

            # Initialize the start index of the current candidate JSON object.
            start_idx = -1

            # Traverse the raw response character by character to find the first balanced JSON object.
            for i, char in enumerate(raw_text):
                # Start or deepen a JSON-object stack when an opening brace is encountered.
                if char == "{":
                    # Record the start index when entering the outermost object.
                    if not stack:
                        # Store the starting index of the outermost JSON object.
                        start_idx = i

                    # Push the opening brace onto the stack.
                    stack.append(char)

                # Pop the stack on closing braces.
                elif char == "}":
                    # Proceed only if the stack is non-empty.
                    if stack:
                        # Pop one opening brace from the stack.
                        stack.pop()

                        # When the stack becomes empty, a balanced JSON object has been found.
                        if not stack:
                            # Extract the balanced JSON substring.
                            json_str = raw_text[start_idx:i + 1]

                            # Exit the scan after the first balanced object.
                            break

            # Fall back to using the raw response text when no balanced object was found.
            if not json_str:
                # Use the full raw response as the candidate JSON string.
                json_str = raw_text

            # Parse the candidate JSON string into a dictionary.
            parsed_dict = json.loads(json_str)

            # Read the required output keys from the compilation prompt schema contract.
            expected_keys = compile_prompt_entry.expected_output_schema.get(
                "required_keys",
                [],
            )

            # Validate that every required key is present in the parsed payload.
            for key in expected_keys:
                # Raise a value error when a required key is absent.
                if key not in parsed_dict:
                    # Abort this attempt because the output schema contract was violated.
                    raise ValueError(
                        f"Missing required key '{key}' in JSON output."
                    )

            # Return the successfully parsed compilation response.
            return CompilationLLMResponse(
                raw_json_string=raw_text,
                parsed_dict=parsed_dict,
                token_usage=usage,
            )

        # Retry on JSON parsing failures or schema-validation failures.
        except (json.JSONDecodeError, ValueError) as e:
            # If this was the final attempt, emit the structured fallback response.
            if attempt == max_retries - 1:
                # Increment the run-warning counter because the model output was unusable.
                governance_context.summary_accumulator.increment_warnings()

                # Return the empty structured fallback payload required for safe downstream execution.
                return CompilationLLMResponse(
                    raw_json_string=str(e),
                    parsed_dict={
                        "files_to_write": [],
                        "files_to_edit": [],
                        "new_concepts": [],
                        "open_questions": [],
                    },
                    token_usage={},
                )

    # Preserve the original fallback return, although normal control flow should return earlier.
    return CompilationLLMResponse("", {}, {})


# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 3: Validate and apply model outputs
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable compilation transaction record.
@dataclass(frozen=True)
class CompilationTransaction:
    """
    Immutable record of applied file operations for one compiled artifact.

    Attributes:
        artifact_id:
            Identifier of the raw artifact whose compilation output was applied.
        applied_operations:
            Ordered list of applied file-operation records.
    """

    # Store the raw artifact identifier.
    artifact_id: str

    # Store the ordered list of applied operations.
    applied_operations: List[Dict[str, str]]


# Define the compilation-output application function.
def apply_compilation_outputs(
    llm_response: CompilationLLMResponse,
    work_package: CompilationWorkPackage,
    governance_context: Any,
    wiki_context: Any,
    repo_root: pathlib.Path,
) -> CompilationTransaction:
    """
    Validate proposed compilation outputs, apply approved writes, and update the wiki graph.

    Purpose:
        Implement the final stages of Algorithm 1:
            1. parse structured output,
            2. validate candidate writes,
            3. write approved files,
            4. update the graph.

        This callable also enforces the rewrite guard described in the design
        proposal: files exceeding 20,000 characters require review before a full
        rewrite is allowed.

    Inputs:
        llm_response:
            Parsed structured response returned by the compilation model call.
        work_package:
            Context package used to generate the model response.
        governance_context:
            Context expected to expose:
            - ``write_validator``
            - ``summary_accumulator``
        wiki_context:
            Context expected to expose ``wiki_graph``.
        repo_root:
            Repository root directory.

    Processes:
        1. Validate input types and required collaborators.
        2. Translate ``files_to_write`` payloads into ``FileOperationPlan``
           objects.
        3. For existing targets, enforce the 20,000-character full-rewrite guard
           using strict UTF-8 decoding.
        4. Validate proposed operations via the centralized write validator.
        5. Apply only operations with status ``APPROVED``.
        6. Update summary counters for written and edited files.
        7. Rebuild the wiki graph from the updated filesystem state.
        8. Return the immutable ``CompilationTransaction``.

    Outputs:
        CompilationTransaction:
            Immutable transaction record describing the applied operations.

    Raises:
        TypeError:
            Raised if required typed inputs are invalid.
        ValueError:
            Raised if required governance or wiki collaborators are missing.
        OSError:
            Propagated if an approved filesystem write fails.

    Notes:
        This callable helps realize the study's compilation update step:

            D_raw^(t) --Φ_compile--> K^(t)

        and then rebuilds the graph representation
        ``K = (V, E, λ_V, λ_E)`` from the authoritative filesystem.
    """
    # Validate that the model response is of the expected type.
    if not isinstance(llm_response, CompilationLLMResponse):
        # Raise a type error because downstream parsing depends on the response schema.
        raise TypeError(
            "apply_compilation_outputs expected 'llm_response' to be a "
            "CompilationLLMResponse."
        )

    # Validate that the work package is of the expected type.
    if not isinstance(work_package, CompilationWorkPackage):
        # Raise a type error because transaction reporting depends on the work-package schema.
        raise TypeError(
            "apply_compilation_outputs expected 'work_package' to be a "
            "CompilationWorkPackage."
        )

    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib semantics.
        raise TypeError(
            "apply_compilation_outputs expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that the governance context exposes the write validator.
    if not hasattr(governance_context, "write_validator"):
        # Raise a value error because write validation depends on governance_context.write_validator.
        raise ValueError(
            "apply_compilation_outputs expected 'governance_context' to expose "
            "'write_validator'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a value error because write-count accounting depends on summary_accumulator.
        raise ValueError(
            "apply_compilation_outputs expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate that the wiki context exposes a wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a value error because graph rebuild depends on wiki_context.wiki_graph.
        raise ValueError(
            "apply_compilation_outputs expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Initialize the list of proposed file-operation plans.
    proposed_ops: List[FileOperationPlan] = []

    # Parse candidate writes from the model response.
    for file_obj in llm_response.parsed_dict.get("files_to_write", []):
        # Extract the target path string from the candidate write object.
        path_str = file_obj.get("path")

        # Extract the Markdown content payload from the candidate write object.
        content = file_obj.get("content_markdown", "")

        # Proceed only when both the path and content are present.
        if path_str and content:
            # Resolve the target path against the repository root.
            target_path = (repo_root / path_str).resolve()

            # Initialize the manual-review flag for the current operation.
            requires_review = False

            # Enforce the full-rewrite guard only when the target file already exists.
            if target_path.exists():
                try:
                    # Read the existing file using strict UTF-8 decoding.
                    existing_content = target_path.read_text(
                        encoding="utf-8",
                        errors="strict",
                    )

                    # Force review when the existing file exceeds the rewrite-size threshold.
                    if len(existing_content) > 20000:
                        # Mark the operation as requiring manual review.
                        requires_review = True

                # Force review when strict UTF-8 decoding fails.
                except UnicodeDecodeError:
                    # Emit a critical log entry because a malformed or binary file cannot be safely rewritten.
                    logger.critical(
                        f"UnicodeDecodeError reading {target_path}. "
                        f"Forcing patch review."
                    )

                    # Mark the operation as requiring manual review.
                    requires_review = True

            # Construct a review-gated operation when review is required.
            if requires_review:
                # Create the operation plan with REQUIRES_REVIEW status.
                op = FileOperationPlan(
                    target_path,
                    OperationType.WRITE,
                    content,
                    status=ValidationStatus.REQUIRES_REVIEW,
                )

                # Record the reason manual review is required.
                op.findings.append(
                    "File exceeds 20,000 characters or is malformed; full "
                    "rewrite requires manual review."
                )

            # Otherwise construct a normal pending write operation.
            else:
                # Create the operation plan with the default PENDING status.
                op = FileOperationPlan(
                    target_path,
                    OperationType.WRITE,
                    content,
                )

            # Append the proposed operation to the batch.
            proposed_ops.append(op)

    # Validate the proposed operations via the centralized write validator.
    validated_ops = governance_context.write_validator.validate_batch(proposed_ops)

    # Initialize the ordered list of applied-operation records.
    applied_records: List[Dict[str, str]] = []

    # Apply only operations that were approved by validation.
    for op in validated_ops:
        # Restrict execution to APPROVED operations.
        if op.status == ValidationStatus.APPROVED:
            # Ensure the target parent directory exists before writing.
            op.target_path.parent.mkdir(parents=True, exist_ok=True)

            # Write the approved content to disk as UTF-8 text.
            op.target_path.write_text(op.content, encoding="utf-8")

            # Record the applied operation in repo-relative form.
            applied_records.append(
                {
                    "path": str(op.target_path.relative_to(repo_root)),
                    "type": op.operation_type.value,
                }
            )

            # Increment the written-files counter for WRITE operations.
            if op.operation_type == OperationType.WRITE:
                # Update the written-file counter.
                governance_context.summary_accumulator.increment_written()

            # Increment the edited-files counter for non-WRITE operations.
            else:
                # Update the edited-file counter.
                governance_context.summary_accumulator.increment_edited()

    # Resolve the canonical wiki root directory.
    wiki_root = repo_root / "wiki"

    # Rebuild the in-memory wiki graph from the authoritative filesystem state.
    wiki_context.wiki_graph.build_from_filesystem(wiki_root)

    # Construct and return the immutable compilation transaction.
    return CompilationTransaction(
        artifact_id=work_package.artifact_id,
        applied_operations=applied_records,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable compilation-context aggregate returned by Task 10.
# Declare the immutable compilation-context record returned by the compilation orchestrator.
@dataclass(frozen=True)
class CompilationContext:
    """
    Immutable record encapsulating the results of the compilation phase.

    Purpose:
        Package the ordered list of per-artifact compilation transactions into a
        single immutable object for downstream tasks such as index refresh,
        linting, and run-manifest serialization.

    Attributes:
        transactions:
            Ordered list of per-artifact compilation transactions. In the full
            study pipeline, these correspond to the validated and applied output
            of Algorithm 1 for each processed artifact.

    Notes:
        This context object is the aggregate output of the compilation phase
        implementing the paper's incremental compiler operator:

            Φ_compile : D_raw × K_(t-1) → K_t
    """

    # Store the ordered list of compilation transactions.
    transactions: List[Any]


# Define the Task 10 orchestrator.
def task_10_orchestrator(
    change_context: Any,
    normalization_context: Any,
    wiki_context: Any,
    prompt_context: Any,
    governance_context: Any,
    repo_root: pathlib.Path,
) -> CompilationContext:
    """
    Orchestrate the incremental compilation of changed raw sources.

    Purpose:
        Execute the end-to-end incremental compilation loop over all changed and
        newly observed artifacts. This orchestrator is the concrete pipeline
        driver for Algorithm 1 from the paper:

            1. Detect changes.
            2. Gather context.
            3. Construct prompt.
            4. Invoke model.
            5. Parse structured output.
            6. Validate.
            7. Write files.
            8. Log transaction.

        This refactored version explicitly requires the ``normalization_context``
        so that compilation always consumes normalized derivatives rather than
        raw bytes.

    Inputs:
        change_context:
            Change-detection context expected to expose ``delta_set`` with:
            - ``changed_artifacts``
            - ``new_artifacts``
        normalization_context:
            Normalization context expected to expose ``normalized_derivatives``.
        wiki_context:
            Wiki architecture context expected to expose ``wiki_graph``.
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        governance_context:
            Governance context expected to expose ``summary_accumulator``.
        repo_root:
            Repository root directory used for path resolution.

    Processes:
        1. Validate top-level inputs and required interfaces.
        2. Concatenate changed and new artifact identifiers into one processing list.
        3. For each artifact:
           a. Assemble its compilation context.
           b. Issue the compilation model call.
           c. Validate and apply the returned file operations.
           d. Append the resulting transaction.
        4. On per-artifact failure:
           - increment the warning counter,
           - emit a warning log,
           - continue processing remaining artifacts.
        5. Return the immutable ``CompilationContext``.

    Outputs:
        CompilationContext:
            Immutable aggregate of all successful per-artifact compilation
            transactions.

    Raises:
        TypeError:
            Raised if:
            - ``repo_root`` is not a ``pathlib.Path``,
            - required collaborator attributes are missing.

    Notes:
        This orchestrator preserves the original logic exactly. It does not halt
        the entire batch on one artifact failure; it degrades gracefully and
        continues processing the remaining delta set.
    """
    # Validate that repo_root is a pathlib.Path because path resolution depends on pathlib semantics.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error immediately when the repository-root contract is violated.
        raise TypeError(
            "task_10_orchestrator expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that change_context exposes a delta set because artifact scheduling depends on it.
    if not hasattr(change_context, "delta_set"):
        # Raise a type error because compilation scheduling requires change_context.delta_set.
        raise TypeError(
            "task_10_orchestrator expected 'change_context' to expose 'delta_set'."
        )

    # Validate that normalization_context exposes normalized derivatives because the compiler must read normalized text.
    if not hasattr(normalization_context, "normalized_derivatives"):
        # Raise a type error because normalized derivatives are required for safe compilation.
        raise TypeError(
            "task_10_orchestrator expected 'normalization_context' to expose "
            "'normalized_derivatives'."
        )

    # Validate that wiki_context exposes the wiki graph because compilation mutates and queries the wiki structure.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because compilation requires wiki_context.wiki_graph.
        raise TypeError(
            "task_10_orchestrator expected 'wiki_context' to expose 'wiki_graph'."
        )

    # Validate that prompt_context exposes a prompt registry because model prompt lookup depends on it.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a type error because compilation prompt retrieval requires prompt_registry.
        raise TypeError(
            "task_10_orchestrator expected 'prompt_context' to expose 'prompt_registry'."
        )

    # Validate that governance_context exposes a summary accumulator because warning accounting depends on it.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a type error because warning tracking requires summary_accumulator.
        raise TypeError(
            "task_10_orchestrator expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Initialize the ordered list of successful compilation transactions.
    transactions: List[Any] = []

    # Combine changed and new artifacts into one ordered processing sequence.
    artifacts_to_process = (
        change_context.delta_set.changed_artifacts
        + change_context.delta_set.new_artifacts
    )

    # Iterate over every artifact scheduled for incremental compilation.
    for artifact_id in artifacts_to_process:
        try:
            # Step 1 of Algorithm 1: assemble compilation context for the current artifact.
            work_package = assemble_compilation_context(
                artifact_id=artifact_id,
                change_context=change_context,
                normalization_context=normalization_context,
                wiki_context=wiki_context,
                repo_root=repo_root,
            )

            # Step 2 of Algorithm 1: issue the compilation model call for the current artifact.
            llm_response = issue_compilation_call(
                work_package=work_package,
                prompt_context=prompt_context,
                governance_context=governance_context,
            )

            # Step 3 of Algorithm 1: validate and apply the model outputs.
            transaction = apply_compilation_outputs(
                llm_response=llm_response,
                work_package=work_package,
                governance_context=governance_context,
                wiki_context=wiki_context,
                repo_root=repo_root,
            )

            # Append the successful compilation transaction to the aggregate list.
            transactions.append(transaction)

        # Catch any per-artifact failure so one bad artifact does not abort the entire compilation batch.
        except Exception as e:
            # Increment the run-level warning counter for the failed artifact.
            governance_context.summary_accumulator.increment_warnings()

            # Emit a warning log while continuing with the remaining artifacts.
            logger.warning(f"Compilation failed for artifact {artifact_id}: {e}")

    # Construct and return the immutable compilation context.
    return CompilationContext(transactions=transactions)
````

````python
# Task 11 — Execute Algorithm 2 (Index Refresh)

# ==============================================================================
# Task 11: Implement the index refresh callable
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 1: Enumerate wiki pages by index type and extract metadata
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable candidate-record dataclass used for indexing.
@dataclass(frozen=True)
class IndexCandidateRecord:
    """
    Immutable record representing one indexable wiki page.

    Purpose:
        Capture the metadata required to build deterministic index pages and
        maps of content.

    Attributes:
        canonical_path:
            Repo-relative canonical path of the page.
        page_type:
            Classified page type.
        title:
            Human-readable page title.
        summary:
            Deterministic one-sentence summary used in index rendering.
        timestamp_iso:
            ISO timestamp used for recency sorting.
        tags:
            List of raw page tags used for MOC generation.
    """

    # Store the canonical repo-relative path of the page.
    canonical_path: str

    # Store the page type classification.
    page_type: PageType

    # Store the page title.
    title: str

    # Store the page summary.
    summary: str

    # Store the ISO timestamp string for sorting and recency display.
    timestamp_iso: str

    # Store the page tags.
    tags: List[str]


# Define the immutable page-catalog view dataclass.
@dataclass(frozen=True)
class PageCatalogView:
    """
    Structured, queryable view of the wiki state for indexing purposes.

    Purpose:
        Package the full set of indexable page candidates extracted from the
        wiki graph into one immutable object.

    Inputs:
        This dataclass is populated by ``build_page_catalog_view``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        candidates:
            Ordered list of index candidate records derived from the current
            wiki state.
    """

    # Store the ordered list of index candidate records.
    candidates: List[IndexCandidateRecord]


# Define the internal helper that extracts a display title from Markdown.
def _extract_title_from_markdown(
    content: str,
    fallback_path: str,
) -> str:
    """
    Extract the first level-1 Markdown header as the page title, falling back to
    the filename stem when no explicit title exists.

    Purpose:
        Produce a stable display title for index rendering.

    Inputs:
        content:
            Markdown content of the page.
        fallback_path:
            Repo-relative page path used as a fallback naming source.

    Processes:
        1. Validate input types.
        2. Search for the first line matching ``^#\\s+(.+)$``.
        3. If found, return the stripped header text.
        4. Otherwise, derive a title from the filename stem.

    Outputs:
        str:
            Display title for the page.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This helper supports Step 2 of Algorithm 2:
        "For each page v, extract title and generate a one-sentence summary."
    """
    # Validate that content is a string.
    if not isinstance(content, str):
        # Raise a type error because Markdown parsing requires a string.
        raise TypeError(
            "_extract_title_from_markdown expected 'content' to be a string."
        )

    # Validate that fallback_path is a string.
    if not isinstance(fallback_path, str):
        # Raise a type error because fallback title derivation requires a string.
        raise TypeError(
            "_extract_title_from_markdown expected 'fallback_path' to be a "
            "string."
        )

    # Search for the first level-1 Markdown header.
    match = re.search(r"^#\s+(.+)$", content, re.MULTILINE)

    # If a title header exists, return its stripped text.
    if match:
        return match.group(1).strip()

    # Otherwise, derive a title from the filename stem.
    return pathlib.Path(fallback_path).stem.replace("-", " ").title()


# Define the internal helper that extracts a deterministic one-sentence summary.
def _extract_deterministic_summary(
    content: str,
) -> str:
    """
    Deterministically extract a one-sentence summary from Markdown content using
    sentence boundary disambiguation.

    Purpose:
        Produce a stable one-sentence summary for index entries without requiring
        an LLM call. This supports Algorithm 2 (Index Refresh), where each page
        contributes a concise description to an index.

    Inputs:
        content:
            Markdown page content from which a summary sentence should be
            extracted.

    Processes:
        1. Validate the input type.
        2. Prefer prose under an explicit ``## Summary`` or ``## Definition``
           heading if available.
        3. Remove front matter, headers, fenced code blocks, blockquotes, and
           list items to isolate prose.
        4. If NLTK is available, apply ``nltk.sent_tokenize`` for robust
           sentence boundary disambiguation.
        5. Otherwise, fall back to regex-based sentence splitting and log a
           warning.
        6. Return the first sufficiently substantive sentence longer than
           20 characters.
        7. If no suitable sentence exists, return ``"No summary available."``.

    Outputs:
        str:
            Deterministically extracted summary sentence.

    Raises:
        TypeError:
            Raised if ``content`` is not a string.
    """
    # Validate that the supplied content is a string.
    if not isinstance(content, str):
        # Raise a type error because text cleaning and sentence extraction require a string.
        raise TypeError(
            "_extract_deterministic_summary expected 'content' to be a string."
        )

    # Attempt to isolate an explicit Summary or Definition section using regex.
    section_match = re.search(
        r"^##\s+(?:Summary|Definition)\s*\n(.*?)(?=^##|\Z)",
        content,
        re.MULTILINE | re.DOTALL | re.IGNORECASE,
    )

    # Prefer the isolated section body when present; otherwise search the full content.
    text_to_search = section_match.group(1) if section_match else content

    # Remove YAML front matter from the candidate text.
    text_to_search = re.sub(
        r"^---\n.*?\n---",
        "",
        text_to_search,
        flags=re.DOTALL,
    )

    # Remove Markdown headers from the candidate text.
    text_to_search = re.sub(
        r"^#+ .*$",
        "",
        text_to_search,
        flags=re.MULTILINE,
    )

    # Remove fenced code blocks from the candidate text.
    text_to_search = re.sub(
        r"```.*?```",
        "",
        text_to_search,
        flags=re.DOTALL,
    )

    # Remove blockquotes from the candidate text.
    text_to_search = re.sub(
        r"^> .*$",
        "",
        text_to_search,
        flags=re.MULTILINE,
    )

    # Remove unordered-list items from the candidate text.
    text_to_search = re.sub(
        r"^\s*[-*+]\s+.*$",
        "",
        text_to_search,
        flags=re.MULTILINE,
    )

    # Strip leading and trailing whitespace from the cleaned text.
    clean_text = text_to_search.strip()

    # Initialize the ordered sentence list.
    sentences: List[str] = []

    # Prefer NLTK-backed sentence segmentation when available.
    if NLTK_AVAILABLE and clean_text:
        # Tokenize the cleaned prose into sentences using Punkt.
        sentences = nltk.sent_tokenize(clean_text)

    # Fall back to regex sentence splitting when NLTK is unavailable.
    elif clean_text:
        # Emit a warning because regex sentence splitting is less robust than Punkt.
        logger.warning(
            "NLTK unavailable. Falling back to regex for sentence boundary "
            "disambiguation."
        )

        # Split on punctuation followed by whitespace as a coarse fallback heuristic.
        sentences = re.split(r"(?<=[.!?])\s+", clean_text)

    # Iterate over the candidate sentences in order.
    for sentence in sentences:
        # Normalize intra-sentence newlines and trim surrounding whitespace.
        clean_sentence = sentence.strip().replace("\n", " ")

        # Return the first sentence exceeding the substantive-length threshold.
        if len(clean_sentence) > 20:
            # Return the first substantive summary sentence.
            return clean_sentence

    # Return the deterministic fallback when no substantive sentence is found.
    return "No summary available."


# Define the page-catalog builder.
def build_page_catalog_view(
    wiki_context: WikiArchitectureContext,
    repo_root: pathlib.Path,
) -> PageCatalogView:
    """
    Enumerate wiki pages and extract indexable metadata.

    Purpose:
        Build the structured page catalog used by Algorithm 2 by traversing the
        current wiki graph and extracting title, summary, timestamp, and tag
        metadata for indexable pages.

    Inputs:
        wiki_context:
            Wiki architecture context containing the in-memory wiki graph.
        repo_root:
            Project root directory used for timestamp fallback via filesystem
            metadata.

    Processes:
        1. Validate top-level inputs.
        2. Iterate over all nodes in the wiki graph.
        3. Skip unresolved placeholder nodes.
        4. Skip nodes without metadata.
        5. Restrict candidates to source summaries and concept pages.
        6. Retrieve page content.
        7. Extract title and deterministic summary.
        8. Resolve timestamp from metadata or filesystem mtime fallback.
        9. Construct IndexCandidateRecord objects.
        10. Return the immutable PageCatalogView.

    Outputs:
        PageCatalogView:
            Immutable structured catalog of indexable wiki pages.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This callable implements Step 1 of Algorithm 2:
        "Enumerate all pages of relevant type in K^(t)."
    """
    # Validate that the wiki context exposes a wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because catalog construction requires the wiki
        # graph.
        raise TypeError(
            "build_page_catalog_view expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because filesystem timestamp fallback requires a
        # pathlib.Path.
        raise TypeError(
            "build_page_catalog_view expected 'repo_root' to be a pathlib.Path "
            "instance."
        )

    # Initialize the ordered list of index candidate records.
    candidates: List[IndexCandidateRecord] = []

    # Retrieve the wiki graph from the context.
    wiki_graph = wiki_context.wiki_graph

    # Iterate over every node path in the wiki graph.
    for node_path in wiki_graph.graph.nodes:
        # Skip unresolved placeholder nodes by convention.
        if node_path.startswith("UNRESOLVED:"):
            continue

        # Retrieve the node data payload from the graph.
        node_data = wiki_graph.graph.nodes[node_path]

        # Retrieve the metadata object from the node payload.
        metadata = node_data.get("metadata")

        # Skip nodes without metadata.
        if not metadata:
            continue

        # Restrict the primary indices to source summaries and concepts.
        if metadata.type not in [PageType.SOURCE_SUMMARY, PageType.CONCEPT]:
            continue

        # Retrieve the page content from the wiki graph.
        content = wiki_graph.get_content(node_path)

        # Skip nodes whose content cannot be retrieved.
        if not content:
            continue

        # Extract the display title from Markdown content.
        title = _extract_title_from_markdown(content, node_path)

        # Extract a deterministic one-sentence summary.
        summary = _extract_deterministic_summary(content)

        # Prefer the timestamp stored in metadata.
        timestamp = metadata.timestamp

        # Fall back to filesystem modification time when no metadata timestamp is
        # available.
        if not timestamp:
            # Resolve the absolute path of the page.
            abs_path = repo_root / node_path

            # Use filesystem metadata when the page exists.
            if abs_path.exists():
                # Import datetime locally exactly as in the original
                # implementation.
                import datetime

                # Retrieve the modification time in seconds since the epoch.
                mtime = abs_path.stat().st_mtime

                # Convert the modification time to an ISO 8601 UTC timestamp.
                timestamp = (
                    datetime.datetime.utcfromtimestamp(mtime).isoformat()
                    + "Z"
                )

        # Append the constructed candidate record to the catalog.
        candidates.append(
            IndexCandidateRecord(
                canonical_path=node_path,
                page_type=metadata.type,
                title=title,
                summary=summary,
                timestamp_iso=timestamp,
                tags=metadata.tags,
            )
        )

    # Construct and return the immutable page catalog.
    return PageCatalogView(candidates=candidates)


# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 2: Apply Algorithm 2: Index Refresh
# -------------------------------------------------------------------------------------------------------------------------------

# Define the sorting policy used during index generation.
class SortPolicy(Enum):
    """
    Enumeration of supported sort policies for index pages.

    Purpose:
        Control deterministic ordering of index entries during Algorithm 2
        (Index Refresh).

    Members:
        RECENCY:
            Sort entries by timestamp in descending recency order.
        ALPHABETICAL:
            Sort entries lexicographically by title.
    """

    # Define recency-based sorting.
    RECENCY = "recency"

    # Define alphabetical sorting.
    ALPHABETICAL = "alphabetical"


# Define the immutable index-refresh plan.
@dataclass(frozen=True)
class IndexRefreshPlan:
    """
    Definition of one index page to be generated.

    Purpose:
        Store the plan needed to generate one index page, including title,
        target path, sort policy, filtering predicate, and description.

    Inputs:
        This dataclass is populated by ``execute_batch_index_refresh``.

    Processes:
        No computation is performed here. It is a passive immutable planning
        record.

    Outputs:
        index_title:
            Display title of the index page.
        target_path:
            Absolute target path of the rendered index file.
        sort_policy:
            Ordering policy for the filtered entries.
        filter_predicate:
            Callable deciding whether a candidate belongs in the index.
        description:
            Human-readable description placed near the top of the index page.
    """

    # Store the display title of the index page.
    index_title: str

    # Store the absolute target path of the index file.
    target_path: pathlib.Path

    # Store the sort policy controlling entry order.
    sort_policy: SortPolicy

    # Store the predicate used to filter catalog candidates.
    filter_predicate: Callable[[IndexCandidateRecord], bool]

    # Store the human-readable description of the index page.
    description: str


# Define the index-page renderer implementing Algorithm 2.
def render_index_page(
    plan: IndexRefreshPlan,
    catalog: PageCatalogView,
) -> FileOperationPlan:
    """
    Execute Algorithm 2 for one index page by filtering, sorting, and rendering
    Markdown.

    Purpose:
        Implement the core index-refresh transformation:
        1. enumerate relevant pages,
        2. sort them,
        3. render them as a bulleted wikilink list,
        4. return a write plan for downstream validation and execution.

    Inputs:
        plan:
            Definition of the index page to generate.
        catalog:
            Structured page catalog containing all index candidates.

    Processes:
        1. Validate top-level inputs.
        2. Filter catalog candidates using the plan predicate.
        3. Sort the filtered candidates according to the plan sort policy.
        4. Render the index page header, description, counts, and entry list.
        5. Return a pending FileOperationPlan.

    Outputs:
        FileOperationPlan:
            Proposed write operation for the rendered index file.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This callable implements Steps 2 through 5 of Algorithm 2:
        - extract summaries,
        - sort,
        - render as a bulleted list with wikilinks,
        - produce the writeable page artifact.
    """
    # Validate that the plan has the expected structural type.
    if not isinstance(plan, IndexRefreshPlan):
        # Raise a type error because rendering requires an IndexRefreshPlan.
        raise TypeError(
            "render_index_page expected 'plan' to be an IndexRefreshPlan "
            "instance."
        )

    # Validate that the catalog has the expected structural type.
    if not isinstance(catalog, PageCatalogView):
        # Raise a type error because rendering requires a PageCatalogView.
        raise TypeError(
            "render_index_page expected 'catalog' to be a PageCatalogView "
            "instance."
        )

    # 1. Enumerate and filter pages using the plan predicate.
    filtered_candidates = [
        c for c in catalog.candidates if plan.filter_predicate(c)
    ]

    # 2 & 3. Sort pages based on the selected policy.
    if plan.sort_policy == SortPolicy.RECENCY:
        # Sort candidates in descending order by timestamp.
        sorted_candidates = sorted(
            filtered_candidates,
            key=lambda x: x.timestamp_iso,
            reverse=True,
        )
    else:
        # Sort candidates in ascending alphabetical order by lowercase title.
        sorted_candidates = sorted(
            filtered_candidates,
            key=lambda x: x.title.lower(),
        )

    # 4. Render the bulleted list with wikilinks.
    lines: List[str] = [
        f"# {plan.index_title}",
        "",
        f"*{plan.description}*",
        "",
        f"**Total Entries:** {len(sorted_candidates)}",
        "",
        "## Index",
        "",
    ]

    # Iterate over the sorted candidates to render each index entry.
    for candidate in sorted_candidates:
        # Initialize the clean path as the canonical page path.
        clean_path = candidate.canonical_path

        # Strip the .md suffix for standard wikilink compatibility when present.
        if clean_path.endswith(".md"):
            clean_path = clean_path[:-3]

        # Append the rendered wikilink entry with the deterministic summary.
        lines.append(
            f"- [[{clean_path}|{candidate.title}]] - {candidate.summary}"
        )

    # Join all rendered lines into the final Markdown content string.
    content = "\n".join(lines) + "\n"

    # 5. Return the proposed file operation as a pending write.
    return FileOperationPlan(
        target_path=plan.target_path,
        operation_type=OperationType.WRITE,
        content=content,
        status=ValidationStatus.PENDING,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 3: Refresh required index structures (Sources, Concepts, MOCs)
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable orphan-report record.
@dataclass(frozen=True)
class OrphanReport:
    """
    Immutable record of pages that do not appear in any generated index.

    Purpose:
        Record index coverage gaps detected after batch index generation.

    Inputs:
        This dataclass is populated by ``execute_batch_index_refresh``.

    Processes:
        No computation is performed here. It is a passive immutable report.

    Outputs:
        orphan_paths:
            Ordered list of canonical page paths not included in any generated
            index.
    """

    # Store the ordered list of orphan page paths.
    orphan_paths: List[str]


# Define the batch index-refresh executor.
def execute_batch_index_refresh(
    catalog: PageCatalogView,
    fs_context: Any,
    governance_context: Any,
    repo_root: pathlib.Path,
) -> Tuple[List[Any], OrphanReport]:
    """
    Define all required index plans, render them, validate them, write approved
    indices, and report orphan pages.

    Purpose:
        Execute the batch version of Algorithm 2 (Index Refresh). The logic
        creates:
        - a global source index,
        - a global concept index,
        - topic-specific MOCs derived from normalized concept tags.

    Inputs:
        catalog:
            Structured page catalog of indexable pages.
        fs_context:
            Filesystem context, optionally exposing ``wiki_layout`` and
            ``wiki_subdirectories``.
        governance_context:
            Governance context expected to expose:
            - ``write_validator``
            - ``summary_accumulator``
        repo_root:
            Repository root directory used to resolve repo-relative paths.

    Processes:
        1. Validate top-level inputs.
        2. Resolve the canonical index directory.
        3. Construct index plans for all sources and all concepts.
        4. Count concept tags after strict lowercase slug normalization to avoid
           fragmentation.
        5. Generate one MOC plan for each normalized tag whose frequency meets
           the threshold.
        6. Render all plans into file-operation proposals.
        7. Validate the proposals through the centralized write validator.
        8. Apply approved writes and record batch-level transaction metadata.
        9. Compute orphan pages as candidates not covered by any plan predicate.
        10. Return the batch transaction list and orphan report.

    Outputs:
        Tuple[List[Any], OrphanReport]:
            A tuple containing:
            1. A list containing one batch transaction dictionary.
            2. The orphan-page report.

    Raises:
        TypeError:
            Raised if ``catalog`` or ``repo_root`` is of the wrong type.
        ValueError:
            Raised if required governance collaborators are absent.

    Notes:
        This callable implements the paper's index-refresh method, where the
        required process is:
            1. enumerate pages,
            2. generate one-sentence summaries,
            3. sort,
            4. render bullet lists with wikilinks,
            5. write the index page.
    """
    # Validate that the catalog is a PageCatalogView.
    if not isinstance(catalog, PageCatalogView):
        # Raise a type error because batch refresh requires the catalog view schema.
        raise TypeError(
            "execute_batch_index_refresh expected 'catalog' to be a "
            "PageCatalogView."
        )

    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib semantics.
        raise TypeError(
            "execute_batch_index_refresh expected 'repo_root' to be a "
            "pathlib.Path."
        )

    # Validate that the governance context exposes a write validator.
    if not hasattr(governance_context, "write_validator"):
        # Raise a value error because validation of proposed writes depends on the write validator.
        raise ValueError(
            "execute_batch_index_refresh expected 'governance_context' to "
            "expose 'write_validator'."
        )

    # Validate that the governance context exposes a summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a value error because run-level counters depend on the summary accumulator.
        raise ValueError(
            "execute_batch_index_refresh expected 'governance_context' to "
            "expose 'summary_accumulator'."
        )

    # Resolve the default wiki root directory as repo_root / "wiki".
    wiki_root = repo_root / "wiki"

    # Attempt to retrieve the wiki-layout object from the filesystem context.
    index_dir = getattr(fs_context, "wiki_layout", None)

    # Use the configured index directory if a wiki layout object is available.
    if index_dir:
        # Resolve the configured index subdirectory, falling back to wiki/index.
        index_dir = index_dir.wiki_subdirectories.get(
            "index",
            wiki_root / "index",
        )

    # Fall back to the default wiki/index directory when no layout object exists.
    else:
        # Use the default index directory under wiki/.
        index_dir = wiki_root / "index"

    # Ensure that the index directory exists before writing.
    index_dir.mkdir(parents=True, exist_ok=True)

    # Initialize the ordered list of index-refresh plans.
    plans: List[IndexRefreshPlan] = []

    # Append the global "All Sources" index plan sorted by recency.
    plans.append(
        IndexRefreshPlan(
            index_title="Master Index: All Sources",
            target_path=index_dir / "all_sources.md",
            sort_policy=SortPolicy.RECENCY,
            filter_predicate=lambda c: c.page_type == PageType.SOURCE_SUMMARY,
            description="Chronological ledger of all ingested source materials.",
        )
    )

    # Append the global "All Concepts" index plan sorted alphabetically.
    plans.append(
        IndexRefreshPlan(
            index_title="Master Index: All Concepts",
            target_path=index_dir / "all_concepts.md",
            sort_policy=SortPolicy.ALPHABETICAL,
            filter_predicate=lambda c: c.page_type == PageType.CONCEPT,
            description="Alphabetical directory of all synthesized concepts.",
        )
    )

    # Initialize the normalized tag-frequency counter for MOC generation.
    tag_counter = Counter()

    # Count normalized concept tags across all candidates.
    for c in catalog.candidates:
        # Restrict tag counting to concept pages only.
        if c.page_type == PageType.CONCEPT:
            # Iterate over the raw tags attached to the concept page.
            for raw_tag in c.tags:
                # Apply strict lowercase slug normalization to prevent tag fragmentation.
                normalized_tag = re.sub(
                    r"[^a-z0-9]+",
                    "-",
                    raw_tag.lower(),
                ).strip("-")

                # Count only non-empty normalized tags.
                if normalized_tag:
                    # Increment the normalized tag count.
                    tag_counter[normalized_tag] += 1

    # Define the MOC generation threshold exactly as in the original logic.
    moc_threshold = 5

    # Generate one MOC plan for each normalized tag meeting or exceeding the threshold.
    for norm_tag, count in tag_counter.items():
        # Restrict MOC creation to tags satisfying the threshold.
        if count >= moc_threshold:
            # Append the MOC index plan, capturing the normalized tag in the predicate default argument.
            plans.append(
                IndexRefreshPlan(
                    index_title=f"MOC: {norm_tag.replace('-', ' ').title()}",
                    target_path=index_dir / f"moc_{norm_tag}.md",
                    sort_policy=SortPolicy.ALPHABETICAL,
                    filter_predicate=lambda c, nt=norm_tag: any(
                        re.sub(r"[^a-z0-9]+", "-", t.lower()).strip("-") == nt
                        for t in c.tags
                    ),
                    description=f"Map of Content for the topic: {norm_tag}.",
                )
            )

    # Render every plan into a proposed file-operation object.
    proposed_ops = [render_index_page(plan, catalog) for plan in plans]

    # Validate all proposed index writes through the centralized write validator.
    validated_ops = governance_context.write_validator.validate_batch(proposed_ops)

    # Initialize the ordered list of applied write records.
    applied_records: List[Dict[str, str]] = []

    # Apply only approved operations.
    for op in validated_ops:
        # Check whether the operation status equals APPROVED.
        if getattr(op, "status", None) == getattr(
            ValidationStatus,
            "APPROVED",
            None,
        ):
            # Write the approved index content to disk as UTF-8 text.
            op.target_path.write_text(op.content, encoding="utf-8")

            # Record the applied operation as a repo-relative path and operation type.
            applied_records.append(
                {
                    "path": str(op.target_path.relative_to(repo_root)),
                    "type": op.operation_type.value,
                }
            )

            # Increment the run-level written-file counter.
            governance_context.summary_accumulator.increment_written()

    # Construct the batch-level transaction dictionary exactly as in the original logic.
    transaction = {
        "artifact_id": "batch_index_refresh",
        "applied_operations": applied_records,
    }

    # Initialize the set of canonical paths covered by at least one index plan.
    indexed_paths: Set[str] = set()

    # Recompute coverage by applying every plan predicate to every catalog candidate.
    for plan in plans:
        # Traverse every catalog candidate for the current plan.
        for c in catalog.candidates:
            # Add the candidate path when the plan predicate selects it.
            if plan.filter_predicate(c):
                indexed_paths.add(c.canonical_path)

    # Compute orphan pages as candidates not covered by any index plan.
    orphans = [
        c.canonical_path
        for c in catalog.candidates
        if c.canonical_path not in indexed_paths
    ]

    # Increment warnings when orphan pages exist.
    if orphans:
        # Increment the warning count by the number of orphan pages.
        governance_context.summary_accumulator.increment_warnings(len(orphans))

    # Return the batch transaction list and the orphan report.
    return [transaction], OrphanReport(orphan_paths=orphans)


# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable index-refresh context aggregate returned by Task 11.
@dataclass(frozen=True)
class IndexRefreshContext:
    """
    Immutable record encapsulating the results of the index-refresh phase.

    Purpose:
        Package the applied index-refresh transactions and the orphan report into
        one immutable object for downstream tasks.

    Inputs:
        This dataclass is populated by ``task_11_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        transactions:
            Ordered list of applied batch index-refresh transactions.
        orphan_report:
            Coverage report identifying pages not included in any generated
            index.
    """

    # Store the applied batch index-refresh transactions.
    transactions: List[CompilationTransaction]

    # Store the orphan-page coverage report.
    orphan_report: OrphanReport


# Define the Task 11 orchestrator.
def task_11_orchestrator(
    wiki_context: WikiArchitectureContext,
    fs_context: Any,
    governance_context: GovernanceContext,
    repo_root: pathlib.Path,
) -> IndexRefreshContext:
    """
    Orchestrate the enumeration, rendering, validation, writing, and graph
    refresh of all index structures.

    Purpose:
        Execute the full index-refresh workflow by:
        1. building the page catalog,
        2. generating all required indices,
        3. validating and applying approved writes,
        4. rebuilding the wiki graph to reflect newly written index files.

    Inputs:
        wiki_context:
            Wiki architecture context containing the wiki graph.
        fs_context:
            Filesystem substrate context or compatible object exposing wiki
            layout.
        governance_context:
            Governance context containing validators, counters, and logging
            infrastructure.
        repo_root:
            Project root directory used for path resolution and wiki-root
            reconstruction.

    Processes:
        1. Validate top-level inputs.
        2. Step 1: Enumerate pages and extract metadata by calling
           ``build_page_catalog_view``.
        3. Step 2 & 3: Define plans, render them, validate them, and write the
           approved indices by calling ``execute_batch_index_refresh``.
        4. Rebuild the wiki graph from the authoritative filesystem so that
           newly written index pages are represented in the graph.
        5. Return the immutable IndexRefreshContext.

    Outputs:
        IndexRefreshContext:
            Immutable aggregate of index-refresh transactions and orphan
            coverage report.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.

    Notes:
        This orchestrator realizes Algorithm 2 end-to-end:
        1. enumerate relevant pages,
        2. extract titles and one-sentence summaries,
        3. sort,
        4. render linked bullet lists,
        5. write updated index pages.
    """
    # Validate that the wiki context exposes a wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because index refresh requires the wiki graph.
        raise TypeError(
            "task_11_orchestrator expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that governance_context exposes a write validator.
    if not hasattr(governance_context, "write_validator"):
        # Raise a type error because index refresh requires centralized write
        # validation.
        raise TypeError(
            "task_11_orchestrator expected 'governance_context' to expose "
            "'write_validator'."
        )

    # Validate that governance_context exposes a summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a type error because index refresh updates run-level counters.
        raise TypeError(
            "task_11_orchestrator expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_11_orchestrator expected 'repo_root' to be a pathlib.Path "
            "instance."
        )

    # Step 1: Enumerate pages and extract metadata into a page catalog.
    catalog = build_page_catalog_view(wiki_context, repo_root)

    # Step 2 & 3: Define index plans, render them, validate them, write them,
    # and compute orphan coverage.
    transactions, orphan_report = execute_batch_index_refresh(
        catalog=catalog,
        fs_context=fs_context,
        governance_context=governance_context,
        repo_root=repo_root,
    )

    # Rebuild the wiki graph to reflect the newly written index files.
    wiki_root = repo_root / "wiki"

    # Reconstruct the graph from the authoritative filesystem state.
    wiki_context.wiki_graph.build_from_filesystem(wiki_root)

    # Construct and return the immutable index-refresh context.
    return IndexRefreshContext(
        transactions=transactions,
        orphan_report=orphan_report,
    )
````

````python
# Task 12 — Implement the local search engine and retrieval interface

# ==============================================================================
# Task 12: Implement the local search engine and retrieval interface
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 1: Build a deterministic local search index over wiki Markdown
# -------------------------------------------------------------------------------------------------------------------------------

# Define a deterministic set of basic English stopwords so tokenization remains
# Define the deterministic stopword set used in tokenization.
ENGLISH_BASIC_STOPWORDS: Set[str] = {
    "a",
    "an",
    "and",
    "are",
    "as",
    "at",
    "be",
    "but",
    "by",
    "for",
    "if",
    "in",
    "into",
    "is",
    "it",
    "no",
    "not",
    "of",
    "on",
    "or",
    "such",
    "that",
    "the",
    "their",
    "then",
    "there",
    "these",
    "they",
    "this",
    "to",
    "was",
    "will",
    "with",
}


# Declare the per-document per-term field-frequency statistics.
@dataclass(frozen=True)
class DocumentStatistics:
    """
    Immutable field-frequency statistics for one term in one document.

    Purpose:
        Store the field-specific term frequencies required to compute the
        weighted term-frequency quantity:

            tf_w(t, v) =
                3.0 * tf_title(t, v)
              + 2.0 * tf_headings(t, v)
              + 1.0 * tf_body(t, v)
              + 2.0 * tf_tags(t, v)

    Attributes:
        tf_title:
            Term frequency in the title field.
        tf_headings:
            Term frequency in heading fields.
        tf_body:
            Term frequency in the main body.
        tf_tags:
            Term frequency in the tags/front-matter field.
    """

    # Store the title-field term frequency.
    tf_title: int = 0

    # Store the headings-field term frequency.
    tf_headings: int = 0

    # Store the body-field term frequency.
    tf_body: int = 0

    # Store the tags-field term frequency.
    tf_tags: int = 0



# Define the immutable dataclass for the full inverted index.
@dataclass(frozen=True)
class InvertedIndex:
    """
    Immutable record representing the deterministic local search index.

    Purpose:
        Store the postings lists and corpus-level statistics required for BM25
        retrieval over the wiki.

    Inputs:
        This dataclass is populated by ``build_search_index``.

    Processes:
        No computation is performed here. It is a passive immutable index
        record.

    Outputs:
        postings:
            Mapping:
            ``term -> {document_path -> DocumentStatistics}``
        doc_lengths:
            Mapping from document path to total token length \(|v|\).
        total_documents:
            Total number of indexed documents \(|V|\).
        average_document_length:
            Corpus-average document length:
            \[
            \mathrm{avgdl}
            =
            \frac{1}{|V|}
            \sum_{v \in V} |v|
            \]
    """

    # Store the inverted postings index.
    postings: Dict[str, Dict[str, DocumentStatistics]]

    # Store document lengths |v| keyed by canonical document path.
    doc_lengths: Dict[str, int]

    # Store the number of indexed documents |V|.
    total_documents: int

    # Store the average document length avgdl.
    average_document_length: float


# Define the deterministic tokenization function.
def tokenize_text(
    text: str,
) -> List[str]:
    """
    Tokenize text according to the strict study configuration, preserving hyphens.

    Purpose:
        Implement a deterministic tokenization pipeline consistent with the
        configured retrieval rules:
        - lowercase: yes
        - strip punctuation: yes
        - split on whitespace / token boundaries: yes
        - stopwords: english_basic
        - stemming: no

        This implementation preserves intra-word hyphens so technical terms such
        as ``fine-tuning`` are not corrupted.

    Inputs:
        text:
            Raw input string to tokenize.

    Processes:
        1. Validate that the input is a string.
        2. Convert the string to lowercase.
        3. Extract tokens with the regex ``\\b[\\w-]+\\b``.
        4. Remove stopwords using the deterministic stopword set.

    Outputs:
        List[str]:
            Ordered list of normalized tokens.

    Raises:
        TypeError:
            Raised if ``text`` is not a string.
    """
    # Validate that the input is a string.
    if not isinstance(text, str):
        # Raise a type error because tokenization requires string input.
        raise TypeError("tokenize_text expected 'text' to be a string.")

    # Lowercase the input string to normalize case deterministically.
    text = text.lower()

    # Extract tokens while preserving intra-word hyphens and dropping terminal punctuation.
    tokens = re.findall(r"\b[\w-]+\b", text)

    # Remove stopwords while preserving original token order.
    return [t for t in tokens if t not in ENGLISH_BASIC_STOPWORDS]


# Define the Markdown field parser.
def parse_markdown_fields(
    content: str,
    metadata_tags: List[str],
) -> Dict[str, str]:
    """
    Parse raw Markdown into the four configured fields:
    1. title,
    2. headings,
    3. body,
    4. tags.

    Purpose:
        Separate a Markdown page into semantically meaningful retrieval fields
        so that field-aware BM25 boosting can be applied.

    Inputs:
        content:
            Raw Markdown content of the document.
        metadata_tags:
            Tags extracted from page metadata or front matter.

    Processes:
        1. Validate the input types.
        2. Initialize the field dictionary.
        3. Extract the first level-1 heading as the title.
        4. Extract all headings (levels 1 to 6) as the heading field.
        5. Remove front matter and headings to isolate the body field.
        6. Join tags into a whitespace-separated tag field.

    Outputs:
        Dict[str, str]:
            Dictionary with keys:
            - ``title``
            - ``headings``
            - ``body``
            - ``tags``

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        These fields feed the boosted term-frequency computation:
        \[
        \mathrm{tf}_w
        = 3.0\mathrm{tf}_{title}
        + 2.0\mathrm{tf}_{headings}
        + 1.0\mathrm{tf}_{body}
        + 2.0\mathrm{tf}_{tags}
        \]
    """
    # Validate that the Markdown content is a string.
    if not isinstance(content, str):
        # Raise a type error because Markdown parsing requires a string.
        raise TypeError(
            "parse_markdown_fields expected 'content' to be a string."
        )

    # Validate that metadata_tags is a list.
    if not isinstance(metadata_tags, list):
        # Raise a type error because tag joining requires a list.
        raise TypeError(
            "parse_markdown_fields expected 'metadata_tags' to be a list of "
            "strings."
        )

    # Initialize the field dictionary with empty strings and the joined tag
    # field.
    fields: Dict[str, str] = {
        "title": "",
        "headings": "",
        "body": "",
        "tags": " ".join(metadata_tags),
    }

    # Extract the first level-1 heading as the document title.
    title_match = re.search(r"^#\s+(.+)$", content, re.MULTILINE)

    # If a title heading is found, store it in the title field.
    if title_match:
        fields["title"] = title_match.group(1)

    # Extract all Markdown headings across all heading levels.
    headings = re.findall(r"^#+\s+(.+)$", content, re.MULTILINE)

    # Join all headings into a single whitespace-separated field.
    fields["headings"] = " ".join(headings)

    # Remove YAML front matter from the body candidate text.
    body_text = re.sub(r"^---\n.*?\n---", "", content, flags=re.DOTALL)

    # Remove all Markdown headings to isolate the body field.
    body_text = re.sub(r"^#+\s+.*$", "", body_text, flags=re.MULTILINE)

    # Store the cleaned body field.
    fields["body"] = body_text

    # Return the parsed field dictionary.
    return fields


# Define the deterministic search-index builder.
def build_search_index(
    wiki_context: WikiArchitectureContext,
) -> InvertedIndex:
    """
    Build the deterministic local search index over wiki Markdown.

    Purpose:
        Traverse all valid wiki pages, parse semantic fields, tokenize them, and
        construct the inverted index and corpus-level statistics needed for BM25
        retrieval.

    Inputs:
        wiki_context:
            Wiki architecture context exposing the wiki graph and page-content
            retrieval function.

    Processes:
        1. Validate that ``wiki_context`` exposes the required wiki graph.
        2. Initialize postings and document-length structures.
        3. Iterate over all graph nodes.
        4. Skip unresolved placeholders.
        5. Retrieve metadata and content for each node.
        6. Skip nodes with missing content or metadata.
        7. Parse the Markdown into title, headings, body, and tags.
        8. Tokenize each field using the deterministic tokenizer.
        9. Compute total document length:
           \[
           |v|
           =
           |tokens_{title}| + |tokens_{headings}| + |tokens_{body}| +
           |tokens_{tags}|
           \]
        10. Count field-specific term frequencies.
        11. Populate the postings list for each unique term.
        12. Compute:
            - total documents \(|V|\)
            - average document length \(\mathrm{avgdl}\)
        13. Return the immutable ``InvertedIndex``.

    Outputs:
        InvertedIndex:
            Fully constructed deterministic search index.

    Raises:
        TypeError:
            Raised if the input context is invalid.

    Notes:
        This callable implements Step 1 of the local retrieval architecture:
        constructing a deterministic index over the wiki corpus.
    """
    # Validate that the wiki context exposes a wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because index construction requires a wiki graph.
        raise TypeError(
            "build_search_index expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that the wiki graph exposes an underlying graph object.
    if not hasattr(wiki_context.wiki_graph, "graph"):
        # Raise a type error because postings construction requires node
        # iteration over the graph.
        raise TypeError(
            "build_search_index expected 'wiki_context.wiki_graph' to expose "
            "'graph'."
        )

    # Validate that the wiki graph exposes content retrieval.
    if not hasattr(wiki_context.wiki_graph, "get_content"):
        # Raise a type error because search indexing requires document content.
        raise TypeError(
            "build_search_index expected 'wiki_context.wiki_graph' to expose "
            "'get_content'."
        )

    # Initialize the postings mapping.
    postings: Dict[str, Dict[str, DocumentStatistics]] = {}

    # Initialize the document-length mapping.
    doc_lengths: Dict[str, int] = {}

    # Retrieve the underlying graph object.
    graph = wiki_context.wiki_graph.graph

    # Iterate over every node path in the wiki graph.
    for node_path in graph.nodes:
        # Skip unresolved placeholder nodes.
        if node_path.startswith("UNRESOLVED:"):
            continue

        # Retrieve node metadata from the graph node payload.
        metadata = graph.nodes[node_path].get("metadata")

        # Retrieve the raw page content from the wiki graph.
        content = wiki_context.wiki_graph.get_content(node_path)

        # Skip nodes with missing content or metadata.
        if not content or not metadata:
            continue

        # Parse the Markdown into semantic retrieval fields.
        fields = parse_markdown_fields(content, metadata.tags)

        # Tokenize the title field.
        tokens_title = tokenize_text(fields["title"])

        # Tokenize the headings field.
        tokens_headings = tokenize_text(fields["headings"])

        # Tokenize the body field.
        tokens_body = tokenize_text(fields["body"])

        # Tokenize the tags field.
        tokens_tags = tokenize_text(fields["tags"])

        # Compute the total document length |v| as the unboosted total token
        # count across all fields.
        total_length = (
            len(tokens_title)
            + len(tokens_headings)
            + len(tokens_body)
            + len(tokens_tags)
        )

        # Skip empty documents that produce no searchable tokens.
        if total_length == 0:
            continue

        # Store the document length keyed by canonical document path.
        doc_lengths[node_path] = total_length

        # Count title term frequencies.
        tf_title = dict(Counter(tokens_title))

        # Count heading term frequencies.
        tf_headings = dict(Counter(tokens_headings))

        # Count body term frequencies.
        tf_body = dict(Counter(tokens_body))

        # Count tag term frequencies.
        tf_tags = dict(Counter(tokens_tags))

        # Compute the set of all unique terms occurring in this document.
        unique_terms = (
            set(tf_title.keys())
            | set(tf_headings.keys())
            | set(tf_body.keys())
            | set(tf_tags.keys())
        )

        # Populate the postings list for each unique term.
        for term in unique_terms:
            # Initialize the postings bucket for the term when first seen.
            if term not in postings:
                postings[term] = {}

            # Store the field-specific term statistics for this term-document
            # pair.
            postings[term][node_path] = DocumentStatistics(
                tf_title=tf_title.get(term, 0),
                tf_headings=tf_headings.get(term, 0),
                tf_body=tf_body.get(term, 0),
                tf_tags=tf_tags.get(term, 0),
            )

    # Compute the total number of indexed documents |V|.
    total_documents = len(doc_lengths)

    # Compute the average document length avgdl, guarding against division by
    # zero.
    average_document_length = sum(doc_lengths.values()) / max(1, total_documents)

    # Return the immutable inverted index.
    return InvertedIndex(
        postings=postings,
        doc_lengths=doc_lengths,
        total_documents=total_documents,
        average_document_length=average_document_length,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 2: Implement BM25 scoring with field-aware boosted term frequency
# -------------------------------------------------------------------------------------------------------------------------------

# Define the BM25 scoring function.
def score_bm25(
    query: str,
    index: InvertedIndex,
    k1: float = 1.5,
    b: float = 0.75,
) -> List[Tuple[str, float]]:
    """
    Score documents against a query using field-aware BM25.

    Purpose:
        Implement the paper's BM25-style retrieval equations exactly. The paper
        specifies:

            idf(t_i) = log((|V| - n(t_i) + 0.5) / (n(t_i) + 0.5))

        and

            score(v, q) =
                Σ_i idf(t_i) *
                [ tf_w(t_i, v) * (k1 + 1) ] /
                [ tf_w(t_i, v) + k1 * (1 - b + b * |v| / avgdl) ]

        with the field-aware weighted term frequency:

            tf_w(t, v) =
                3.0 * tf_title(t, v)
              + 2.0 * tf_headings(t, v)
              + 1.0 * tf_body(t, v)
              + 2.0 * tf_tags(t, v)

    Inputs:
        query:
            Raw user or rewritten search query.
        index:
            Prebuilt inverted index over the wiki corpus.
        k1:
            BM25 term-frequency saturation parameter.
        b:
            BM25 document-length normalization parameter.

    Processes:
        1. Validate input types.
        2. Return early for an empty corpus.
        3. Tokenize the query deterministically.
        4. Initialize a score accumulator for all indexed documents.
        5. For each unique query term:
           - compute document frequency,
           - compute IDF,
           - compute boosted term frequency,
           - compute the BM25 contribution,
           - accumulate the score.
        6. Filter out zero-score documents.
        7. Sort by descending score and return the ranking.

    Outputs:
        List[Tuple[str, float]]:
            Ranked list of ``(document_path, score)`` tuples.

    Raises:
        TypeError:
            Raised if inputs are invalid.
    """
    # Validate that the query is a string.
    if not isinstance(query, str):
        # Raise a type error because tokenization requires string input.
        raise TypeError("score_bm25 expected 'query' to be a string.")

    # Validate that the index is an InvertedIndex instance.
    if not isinstance(index, InvertedIndex):
        # Raise a type error because postings and document statistics require the index schema.
        raise TypeError(
            "score_bm25 expected 'index' to be an InvertedIndex instance."
        )

    # Validate that k1 and b are numeric.
    if not isinstance(k1, (int, float)) or not isinstance(b, (int, float)):
        # Raise a type error because BM25 arithmetic requires numeric parameters.
        raise TypeError(
            "score_bm25 expected 'k1' and 'b' to be numeric."
        )

    # Return early when the corpus is empty to avoid division by zero and meaningless ranking.
    if index.total_documents == 0:
        # Emit an empty ranking because no documents exist.
        return []

    # Tokenize the query deterministically.
    query_tokens = tokenize_text(query)

    # Initialize the score accumulator with zero scores for every indexed document.
    scores: Dict[str, float] = {
        doc_path: 0.0
        for doc_path in index.doc_lengths.keys()
    }

    # Read the corpus size |V|.
    V = index.total_documents

    # Read the average document length avgdl.
    avgdl = index.average_document_length

    # Iterate over each unique query term.
    for term in set(query_tokens):
        # Skip terms absent from the vocabulary.
        if term not in index.postings:
            # Continue because out-of-vocabulary terms contribute nothing.
            continue

        # Retrieve the postings for the current term.
        term_postings = index.postings[term]

        # Compute the document frequency n(t_i).
        n_t = len(term_postings)

        # Compute the raw inverse document frequency exactly as specified in the paper.
        raw_idf = math.log((V - n_t + 0.5) / (n_t + 0.5))

        # Floor the IDF to a small positive epsilon to avoid unstable negative weighting.
        idf = max(0.01, raw_idf)

        # Iterate over all documents containing the current term.
        for doc_path, stats in term_postings.items():
            # Compute the weighted term frequency tf_w(t, v) using the configured field boosts.
            tf_w = (
                3.0 * stats.tf_title
                + 2.0 * stats.tf_headings
                + 1.0 * stats.tf_body
                + 2.0 * stats.tf_tags
            )

            # Retrieve the document length |v|.
            doc_len = index.doc_lengths[doc_path]

            # Compute the BM25 numerator:
            # tf_w(t, v) * (k1 + 1).
            numerator = tf_w * (k1 + 1.0)

            # Compute the BM25 denominator:
            # tf_w(t, v) + k1 * (1 - b + b * |v| / avgdl).
            denominator = tf_w + k1 * (1.0 - b + b * (doc_len / avgdl))

            # Compute the BM25 contribution for the current term.
            term_score = idf * (numerator / denominator)

            # Accumulate the contribution into the document score.
            scores[doc_path] += term_score

    # Filter out zero-score documents and sort descending by score.
    ranked_results = sorted(
        [
            (doc, score)
            for doc, score in scores.items()
            if score > 0.0
        ],
        key=lambda x: x[1],
        reverse=True,
    )

    # Return the ranked document-score tuples.
    return ranked_results


# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 3: Expose the search engine as a web UI and CLI/tool interface
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable search-result record.
@dataclass(frozen=True)
class SearchResult:
    """
    Immutable record representing one ranked search result.

    Purpose:
        Package the canonical path, BM25 score, and highlighted snippet for
        downstream consumption by either:
        1. the human-facing web interface, or
        2. the machine-facing LLM tool interface.

    Inputs:
        This dataclass is populated by the SearchInterface.

    Processes:
        No computation is performed here. It is a passive immutable result
        record.

    Outputs:
        path:
            Canonical repo-relative page path.
        score:
            BM25 relevance score.
        snippet:
            Highlighted snippet centered around the first matched query term.
    """

    # Store the canonical repo-relative path.
    path: str

    # Store the BM25 relevance score.
    score: float

    # Store the highlighted snippet.
    snippet: str


# Define the snippet-generation helper.
def generate_snippet(
    content: str,
    query_tokens: List[str],
    snippet_length: int = 240,
) -> str:
    """
    Generate a highlighted snippet centered around the first query-term match.

    Purpose:
        Produce a bounded, human-readable snippet suitable for UI display or LLM
        tool output. The implementation highlights matched query tokens while
        avoiding corrupted Markdown from overlapping highlight spans by merging
        intervals before injection.

    Inputs:
        content:
            Raw Markdown or text content of the document.
        query_tokens:
            Normalized query tokens used for highlighting.
        snippet_length:
            Maximum target snippet length in characters.

    Processes:
        1. Validate input types.
        2. Normalize repeated whitespace in the content.
        3. Find the first query-token occurrence and center the snippet around
           it.
        4. Extract the bounded snippet window.
        5. Find all token-match intervals inside the snippet.
        6. Sort and merge overlapping intervals.
        7. Inject Markdown bold markers in reverse interval order.
        8. Return the highlighted snippet.

    Outputs:
        str:
            Highlighted snippet string.

    Raises:
        TypeError:
            Raised if any input is invalid.
    """
    # Validate that the content is a string.
    if not isinstance(content, str):
        # Raise a type error because snippet extraction requires string content.
        raise TypeError("generate_snippet expected 'content' to be a string.")

    # Validate that query_tokens is a list.
    if not isinstance(query_tokens, list):
        # Raise a type error because the snippet highlighter expects a list of tokens.
        raise TypeError(
            "generate_snippet expected 'query_tokens' to be a list of strings."
        )

    # Validate that snippet_length is an integer.
    if not isinstance(snippet_length, int):
        # Raise a type error because window sizing requires an integer.
        raise TypeError(
            "generate_snippet expected 'snippet_length' to be an integer."
        )

    # Normalize repeated whitespace for cleaner snippets.
    clean_content = re.sub(r"\s+", " ", content).strip()

    # Initialize the match index to indicate "not found".
    match_index = -1

    # Search for the first case-insensitive occurrence of any query token.
    for token in query_tokens:
        # Search for the current token as a whole word.
        match = re.search(
            rf"\b{re.escape(token)}\b",
            clean_content,
            re.IGNORECASE,
        )

        # Stop once the first match is found.
        if match:
            # Record the match start index.
            match_index = match.start()

            # Exit the search loop.
            break

    # Default to the start of the document when no query token match exists.
    if match_index == -1:
        # Use the beginning of the content as the centering point.
        match_index = 0

    # Compute half the snippet window width.
    half_window = snippet_length // 2

    # Compute the snippet start boundary.
    start = max(0, match_index - half_window)

    # Compute the snippet end boundary.
    end = min(len(clean_content), match_index + half_window)

    # Extract the raw snippet window.
    snippet = clean_content[start:end]

    # Add a leading ellipsis when the snippet does not begin at the document start.
    if start > 0:
        # Prefix the snippet with a leading ellipsis.
        snippet = "..." + snippet

        # Preserve the original offset assignment even though it is not used later.
        offset = 3

    # Preserve the original else branch and offset variable.
    else:
        # Set the offset to zero when no leading ellipsis is added.
        offset = 0

    # Add a trailing ellipsis when the snippet does not end at the document end.
    if end < len(clean_content):
        # Suffix the snippet with a trailing ellipsis.
        snippet = snippet + "..."

    # Initialize the list of match intervals inside the snippet.
    intervals: List[List[int]] = []

    # Find all occurrences of all query tokens within the snippet.
    for token in query_tokens:
        # Iterate over every whole-word match of the current token.
        for match in re.finditer(
            rf"\b{re.escape(token)}\b",
            snippet,
            re.IGNORECASE,
        ):
            # Append the [start, end] interval for the current token match.
            intervals.append([match.start(), match.end()])

    # Return the unmodified snippet if no matches were found within it.
    if not intervals:
        # Emit the raw snippet without highlighting.
        return snippet

    # Sort the intervals by ascending start index.
    intervals.sort(key=lambda x: x[0])

    # Initialize the merged interval list with the first interval.
    merged_intervals: List[List[int]] = [intervals[0]]

    # Merge overlapping or touching intervals.
    for current in intervals[1:]:
        # Read the most recently merged interval.
        previous = merged_intervals[-1]

        # Merge intervals that overlap or touch.
        if current[0] <= previous[1]:
            # Extend the previous interval's end if needed.
            previous[1] = max(previous[1], current[1])

        # Otherwise append a new disjoint interval.
        else:
            # Add the current interval as a new merged segment.
            merged_intervals.append(current)

    # Inject Markdown bold markers in reverse order to avoid shifting future indices.
    for interval in reversed(merged_intervals):
        # Read the merged interval boundaries.
        span_start, span_end = interval[0], interval[1]

        # Insert the closing bold marker first.
        snippet = snippet[:span_end] + "**" + snippet[span_end:]

        # Insert the opening bold marker second.
        snippet = snippet[:span_start] + "**" + snippet[span_start:]

    # Return the highlighted snippet.
    return snippet


# Define the dual-purpose search interface.
class SearchInterface:
    """
    Dual-purpose interface exposing the search engine to humans and machines.

    Purpose:
        Wrap the core BM25 scoring function with policy-enforcing interfaces
        that:
        1. truncate results to configured limits,
        2. generate snippets,
        3. return either SearchResult objects for humans or JSON-serializable
           dictionaries for LLM tools.

    Inputs:
        index:
            Deterministic inverted index over the wiki.
        wiki_context:
            Wiki architecture context providing document-content access.
        k1:
            BM25 saturation parameter.
        b:
            BM25 length-normalization parameter.

    Processes:
        1. Store the retrieval state and BM25 constants.
        2. Execute internal search via BM25 scoring.
        3. Generate snippets and truncate according to interface policy.
        4. Expose human-facing and LLM-facing methods.

    Outputs:
        SearchResult objects or JSON-serializable search-result dictionaries.
    """

    # Define the search-interface constructor.
    def __init__(
        self,
        index: InvertedIndex,
        wiki_context: WikiArchitectureContext,
        k1: float,
        b: float,
    ) -> None:
        """
        Initialize the search interface.

        Purpose:
            Bind the inverted index, wiki context, and BM25 constants used for
            retrieval.

        Inputs:
            index:
                Deterministic inverted index.
            wiki_context:
                Wiki architecture context containing the wiki graph.
            k1:
                BM25 saturation parameter.
            b:
                BM25 length-normalization parameter.

        Processes:
            1. Validate input types.
            2. Store the inputs on the instance.

        Outputs:
            None. State is stored on the instance.

        Raises:
            TypeError:
                Raised if inputs are invalid.
        """
        # Validate that the index is an InvertedIndex instance.
        if not isinstance(index, InvertedIndex):
            # Raise a type error because retrieval requires the index schema.
            raise TypeError(
                "SearchInterface expected 'index' to be an InvertedIndex "
                "instance."
            )

        # Validate that the wiki context exposes a wiki graph.
        if not hasattr(wiki_context, "wiki_graph"):
            # Raise a type error because snippet generation requires content
            # retrieval via the wiki graph.
            raise TypeError(
                "SearchInterface expected 'wiki_context' to expose "
                "'wiki_graph'."
            )

        # Validate that k1 is numeric.
        if not isinstance(k1, (int, float)):
            # Raise a type error because BM25 arithmetic requires numeric k1.
            raise TypeError(
                "SearchInterface expected 'k1' to be numeric."
            )

        # Validate that b is numeric.
        if not isinstance(b, (int, float)):
            # Raise a type error because BM25 arithmetic requires numeric b.
            raise TypeError(
                "SearchInterface expected 'b' to be numeric."
            )

        # Store the inverted index on the instance.
        self.index = index

        # Store the wiki architecture context on the instance.
        self.wiki_context = wiki_context

        # Store the BM25 k1 parameter on the instance.
        self.k1 = k1

        # Store the BM25 b parameter on the instance.
        self.b = b

    # Define the internal search executor shared by both external interfaces.
    def _execute_search(
        self,
        query: str,
        top_k: int,
    ) -> List[SearchResult]:
        """
        Execute BM25 search, generate snippets, and truncate the result list.

        Purpose:
            Provide the common retrieval logic underlying both the human-facing
            and LLM-facing search interfaces.

        Inputs:
            query:
                Raw search query.
            top_k:
                Maximum number of results to return.

        Processes:
            1. Validate input types.
            2. Score documents using BM25.
            3. Truncate to the top_k highest-scoring documents.
            4. Tokenize the query for snippet highlighting.
            5. Retrieve page content for each top result.
            6. Generate a snippet for each result.
            7. Construct SearchResult objects.

        Outputs:
            List[SearchResult]:
                Ordered list of top-k search results.

        Raises:
            TypeError:
                Raised if inputs are invalid.
        """
        # Validate that the query is a string.
        if not isinstance(query, str):
            # Raise a type error because scoring requires a textual query.
            raise TypeError(
                "SearchInterface._execute_search expected 'query' to be a "
                "string."
            )

        # Validate that top_k is an integer.
        if not isinstance(top_k, int):
            # Raise a type error because result truncation requires an integer.
            raise TypeError(
                "SearchInterface._execute_search expected 'top_k' to be an "
                "integer."
            )

        # 1. Score all documents using BM25.
        ranked_docs = score_bm25(query, self.index, self.k1, self.b)

        # 2. Truncate the ranked result list to top_k entries.
        top_docs = ranked_docs[:top_k]

        # 3. Tokenize the query for snippet highlighting.
        query_tokens = tokenize_text(query)

        # Initialize the ordered search-result list.
        results: List[SearchResult] = []

        # Construct a SearchResult for each top-ranked document.
        for path, score in top_docs:
            # Retrieve the raw page content from the wiki graph.
            content = self.wiki_context.wiki_graph.get_content(path)

            # Generate a highlighted snippet centered on the first query match.
            snippet = generate_snippet(
                content,
                query_tokens,
                snippet_length=240,
            )

            # Append the structured search result.
            results.append(
                SearchResult(
                    path=path,
                    score=score,
                    snippet=snippet,
                )
            )

        # Return the ordered search results.
        return results

    # Define the human-facing search method.
    def search_for_web(
        self,
        query: str,
    ) -> List[SearchResult]:
        """
        Execute search for the human-facing web interface.

        Purpose:
            Apply the standard top-k policy for interactive human search.

        Inputs:
            query:
                Raw search query.

        Processes:
            1. Validate that the query is a string.
            2. Execute the shared search logic with ``top_k = 20``.

        Outputs:
            List[SearchResult]:
                Ordered list of up to 20 search results.

        Raises:
            TypeError:
                Raised if query is invalid.
        """
        # Validate that the query is a string.
        if not isinstance(query, str):
            # Raise a type error because search queries must be textual.
            raise TypeError(
                "SearchInterface.search_for_web expected 'query' to be a "
                "string."
            )

        # Execute the shared search logic with the human-facing result cap.
        return self._execute_search(query, top_k=20)

    # Define the LLM-facing search method.
    def search_for_llm(
        self,
        query: str,
    ) -> List[Dict[str, Any]]:
        """
        Execute search for the LLM tool interface.

        Purpose:
            Apply the LLM-facing truncation policy and return JSON-serializable
            result dictionaries.

        Inputs:
            query:
                Raw search query.

        Processes:
            1. Validate that the query is a string.
            2. Execute the shared search logic with ``top_k = 10``.
            3. Convert each SearchResult to a JSON-serializable dictionary with
               rounded relevance score.

        Outputs:
            List[Dict[str, Any]]:
                Ordered list of search-result dictionaries suitable for tool
                payloads.

        Raises:
            TypeError:
                Raised if query is invalid.
        """
        # Validate that the query is a string.
        if not isinstance(query, str):
            # Raise a type error because search queries must be textual.
            raise TypeError(
                "SearchInterface.search_for_llm expected 'query' to be a "
                "string."
            )

        # Execute the shared search logic with the LLM-facing result cap.
        results = self._execute_search(query, top_k=10)

        # Convert the results to JSON-serializable dictionaries.
        return [
            {
                "path": r.path,
                "relevance_score": round(r.score, 3),
                "snippet": r.snippet,
            }
            for r in results
        ]


# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable search-engine context aggregate.
@dataclass(frozen=True)
class SearchEngineContext:
    """
    Immutable record encapsulating the initialized search engine.

    Purpose:
        Package the deterministic inverted index and the instantiated search
        interface into one immutable object for downstream Q&A and routing
        tasks.

    Inputs:
        This dataclass is populated by ``task_12_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        inverted_index:
            The constructed deterministic inverted index.
        search_interface:
            The dual-purpose search interface exposing human and LLM retrieval
            methods.
    """

    # Store the built inverted index.
    inverted_index: InvertedIndex

    # Store the instantiated search interface.
    search_interface: SearchInterface


# Define the Task 12 orchestrator.
def task_12_orchestrator(
    wiki_context: WikiArchitectureContext,
    contract: FrozenRunContract,
) -> SearchEngineContext:
    """
    Orchestrate Task 12 end-to-end by building the deterministic search index
    and exposing the retrieval interfaces.

    Purpose:
        Execute the full local retrieval initialization workflow by:
        1. building the deterministic inverted index,
        2. extracting BM25 constants from the frozen run contract,
        3. instantiating the dual-purpose search interface.

    Inputs:
        wiki_context:
            Wiki architecture context containing the current wiki graph.
        contract:
            Frozen run contract containing resolved BM25 defaults.

    Processes:
        1. Validate top-level inputs.
        2. Step 1: Build the deterministic local search index by calling
           ``build_search_index``.
        3. Extract BM25 constants from:
           ``contract.resolved_defaults["retrieval_bm25_constants"]``.
        4. Step 2 & 3: Instantiate the SearchInterface wrapping BM25 scoring and
           snippet generation.
        5. Return the immutable SearchEngineContext.

    Outputs:
        SearchEngineContext:
            Immutable aggregate containing the built inverted index and search
            interface.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.

    Notes:
        This orchestrator realizes the deterministic local retrieval subsystem
        that supports downstream grounded Q&A.
    """
    # Validate that the wiki context has the expected structural type.
    if not isinstance(wiki_context, WikiArchitectureContext):
        # Raise a type error because search-index construction requires the wiki
        # architecture context.
        raise TypeError(
            "task_12_orchestrator expected 'wiki_context' to be a "
            "WikiArchitectureContext."
        )

    # Validate that the contract has the expected structural type.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error because BM25 constant extraction requires the
        # frozen run contract.
        raise TypeError(
            "task_12_orchestrator expected 'contract' to be a "
            "FrozenRunContract."
        )

    # Validate that the contract exposes resolved defaults.
    if not hasattr(contract, "resolved_defaults"):
        # Raise a type error because BM25 constant extraction requires resolved
        # defaults.
        raise TypeError(
            "task_12_orchestrator expected 'contract' to expose "
            "'resolved_defaults'."
        )

    # Step 1: Build the deterministic local search index.
    inverted_index = build_search_index(wiki_context)

    # Extract the BM25 constants injected during Task 1 freezing.
    bm25_constants = contract.resolved_defaults.get(
        "retrieval_bm25_constants",
        {},
    )

    # Resolve the BM25 k1 parameter with the configured default.
    k1 = bm25_constants.get("k_1", 1.5)

    # Resolve the BM25 b parameter with the configured default.
    b = bm25_constants.get("b", 0.75)

    # Step 2 & 3: Instantiate the dual-purpose search interface.
    search_interface = SearchInterface(
        index=inverted_index,
        wiki_context=wiki_context,
        k1=k1,
        b=b,
    )

    # Return the immutable search-engine context.
    return SearchEngineContext(
        inverted_index=inverted_index,
        search_interface=search_interface,
    )
````

````python
# Task 13 — Implement the grounded Q&A callable

# ==============================================================================
# Task 13: Implement the grounded Q&A callable
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 1: Run Algorithm 3 (Routing Phase)
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable routing-state record.
@dataclass(frozen=True)
class QARoutingState:
    """
    Immutable record capturing the provenance of the Q&A routing phase.

    Purpose:
        Store the complete, auditable state of the routing phase of Algorithm 3,
        including:
        1. the original user query,
        2. whether the query was classified as broad,
        3. the exact sequence of search queries issued,
        4. the final shortlist of wiki pages selected for reading,
        5. routing diagnostics explaining each decision.

    Inputs:
        This dataclass is populated by ``route_qa_query``.

    Processes:
        No computation is performed here. It is a passive immutable provenance
        record for the routing phase.

    Outputs:
        original_query:
            The raw user question \(q \in \mathcal{Q}\).
        is_broad_query:
            Boolean indicating whether the query was heuristically classified as
            broad.
        search_queries_issued:
            Ordered list of all search queries executed, including rewrites.
        shortlisted_paths:
            Ordered list of canonical repo-relative wiki page paths selected for
            reading.
        diagnostics:
            Ordered list of human-readable diagnostic messages explaining the
            routing decisions.

    Notes:
        This record documents the routing phase of Algorithm 3:
        \[
        \text{Retrieve indices } I,\quad
        P_{\text{shortlist}} \leftarrow m(\text{RoutingPrompt}(q, I))
        \]
        with local-search refinement when needed.
    """

    # Store the raw user question.
    original_query: str

    # Store the broad-query classification flag.
    is_broad_query: bool

    # Store the ordered list of search queries issued during routing.
    search_queries_issued: List[str]

    # Store the final ordered list of shortlisted canonical page paths.
    shortlisted_paths: List[str]

    # Store human-readable routing diagnostics.
    diagnostics: List[str]


# Define the routing-phase callable.
def route_qa_query(
    query: str,
    search_context: Any,
    prompt_context: Any,
    contract: Any,
    max_pages: int = 8,
    max_rewrites: int = 2,
) -> QARoutingState:
    """
    Execute the routing phase of Algorithm 3 to shortlist relevant wiki pages.

    Purpose:
        Implement the routing phase of the paper's grounded Q&A procedure.
        In the paper's notation, this is the stage that constructs
        ``P_shortlist`` before reading page contents.

    Inputs:
        query:
            Raw user question.
        search_context:
            Search context expected to expose ``search_interface`` with
            ``search_for_llm``.
        prompt_context:
            Prompt context preserved in the signature for orchestration
            compatibility.
        contract:
            Frozen run contract preserved in the signature for orchestration
            compatibility.
        max_pages:
            Hard cap on the number of shortlisted pages.
        max_rewrites:
            Maximum number of semantic search-query rewrites.

    Processes:
        1. Validate input types.
        2. Classify the query as broad or specific using length and keyword
           heuristics.
        3. Issue the initial search query.
        4. If results are weak, iteratively rewrite the query through an LLM
           auxiliary call.
        5. Re-issue search after each rewrite.
        6. Aggregate, deduplicate, and cap shortlisted paths.
        7. Return the immutable routing state.

    Outputs:
        QARoutingState:
            Immutable routing provenance record.

    Raises:
        TypeError:
            Raised if inputs are invalid.
    """
    # Validate that the query is a string.
    if not isinstance(query, str):
        # Raise a type error because routing begins from a raw query string.
        raise TypeError("route_qa_query expected 'query' to be a string.")

    # Validate that max_pages and max_rewrites are integers.
    if not isinstance(max_pages, int) or not isinstance(max_rewrites, int):
        # Raise a type error because routing caps must be integers.
        raise TypeError("max_pages and max_rewrites must be integers.")

    # Validate that the search context exposes the search interface.
    if not hasattr(search_context, "search_interface"):
        # Raise a type error because routing depends on search_context.search_interface.
        raise TypeError(
            "route_qa_query expected 'search_context' to expose "
            "'search_interface'."
        )

    # Initialize the diagnostics list.
    diagnostics: List[str] = []

    # Initialize the ordered list of search queries issued.
    search_queries: List[str] = []

    # Initialize the ordered shortlist of paths.
    shortlisted_paths: List[str] = []

    # Initialize the set used to deduplicate shortlisted paths.
    seen_paths: Set[str] = set()

    # Lowercase the query for heuristic broadness checks.
    query_lower = query.lower()

    # Compute the query word count.
    word_count = len(query_lower.split())

    # Define the broad-query keyword set.
    broad_keywords = {"explain", "summarize", "overview", "what is", "how does"}

    # Classify the query as broad if short or containing broad-summary language.
    is_broad = (
        word_count < 5
        or any(kw in query_lower for kw in broad_keywords)
    )

    # Record the broad-query classification diagnostic.
    if is_broad:
        # Record that the query was classified as broad.
        diagnostics.append(
            "Query classified as BROAD based on length or keywords."
        )

    # Record the specific-query classification diagnostic.
    else:
        # Record that the query was classified as specific.
        diagnostics.append("Query classified as SPECIFIC.")

    # Initialize the current search query with the original query.
    current_search_query = query

    # Record the initial search query.
    search_queries.append(current_search_query)

    # Execute the initial search against the search interface.
    results = search_context.search_interface.search_for_llm(
        current_search_query
    )

    # Initialize the rewrite counter.
    rewrite_count = 0

    # Enter the rewrite loop when results are empty or weak.
    while (
        rewrite_count < max_rewrites
        and (
            not results
            or results[0].get("relevance_score", 0.0) < 1.0
        )
    ):
        # Increment the rewrite counter.
        rewrite_count += 1

        # Construct the semantic rewrite prompt for the LLM.
        rewrite_prompt = (
            f"Given the user query '{query}', extract the core search keywords. "
            "Return ONLY a space-separated list of keywords, with no quotes or "
            "punctuation."
        )

        # Define the strict system prompt for the rewrite task.
        rewrite_sys_prompt = (
            "You are a search query optimizer. Output only keywords."
        )

        # Attempt to obtain an LLM-based semantic rewrite.
        try:
            # Invoke the GPT-5.2 wrapper in low-effort mode for this auxiliary rewrite task.
            llm_result = query_gpt52(
                prompt=rewrite_prompt,
                system_prompt=rewrite_sys_prompt,
                reasoning_effort="low",
                max_output_tokens=50,
                enable_web_search=False,
                log_run_manifest=False,
            )

            # Extract and sanitize the rewritten query string.
            rewritten_query = (
                str(llm_result.get("response_text", ""))
                .strip()
                .replace('"', "")
                .replace("'", "")
            )

            # Reject empty rewrite outputs.
            if not rewritten_query:
                # Raise a value error so the rewrite loop can halt cleanly.
                raise ValueError("LLM returned empty rewrite.")

            # Promote the rewritten query to the current search query.
            current_search_query = rewritten_query

        # Halt rewrites on any rewrite failure.
        except Exception as e:
            # Record the rewrite failure diagnostic.
            diagnostics.append(
                f"Rewrite {rewrite_count} failed: {e}. Halting rewrites."
            )

            # Break out of the rewrite loop.
            break

        # Record the successful rewrite diagnostic.
        diagnostics.append(
            f"Results poor. Rewriting query via LLM to: "
            f"'{current_search_query}'"
        )

        # Record the rewritten search query.
        search_queries.append(current_search_query)

        # Re-issue the search using the rewritten query.
        results = search_context.search_interface.search_for_llm(
            current_search_query
        )

    # Aggregate and deduplicate canonical result paths.
    for res in results:
        # Read the result path from the search result object.
        path = res.get("path")

        # Add the path only if it is a new valid string path.
        if path and isinstance(path, str) and path not in seen_paths:
            # Record the path in the deduplication set.
            seen_paths.add(path)

            # Append the path to the shortlist.
            shortlisted_paths.append(path)

    # Enforce the strict maximum number of pages to read.
    if len(shortlisted_paths) > max_pages:
        # Record the truncation diagnostic.
        diagnostics.append(
            f"Truncating shortlist from {len(shortlisted_paths)} to "
            f"{max_pages} pages."
        )

        # Truncate the shortlist to the configured cap.
        shortlisted_paths = shortlisted_paths[:max_pages]

    # Construct and return the immutable routing state.
    return QARoutingState(
        original_query=query,
        is_broad_query=is_broad,
        search_queries_issued=search_queries,
        shortlisted_paths=shortlisted_paths,
        diagnostics=diagnostics,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 2: Read pages and assemble context (Reading & Synthesis Phase)
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable synthesis-package record.
@dataclass(frozen=True)
class QASynthesisPackage:
    """
    Immutable record capturing the exact context and response of the Q&A
    synthesis phase.

    Purpose:
        Store the complete synthesis-phase provenance, including:
        1. the exact assembled context string shown to the model,
        2. the effective runtime settings used,
        3. the raw model response text,
        4. token-usage statistics.

    Inputs:
        This dataclass is populated by ``synthesize_qa_answer``.

    Processes:
        No computation is performed here. It is a passive immutable synthesis
        record.

    Outputs:
        assembled_context:
            Exact concatenated context shown to the model.
        effective_settings:
            Effective runtime settings used during the model call.
        raw_response_text:
            Exact textual response returned by the model.
        token_usage:
            Token-usage statistics returned by the governed LLM wrapper.

    Notes:
        This record documents the reading and synthesis phases of Algorithm 3:
        \[
        C_q \leftarrow \{\lambda_V(p) \mid p \in P_{\text{shortlist}}\}
        \]
        \[
        a \leftarrow m(\text{QAPrompt}(q, C_q); \theta_{\text{qa}})
        \]
    """

    # Store the exact concatenated context shown to the model.
    assembled_context: str

    # Store the effective runtime settings used for the call.
    effective_settings: Dict[str, Any]

    # Store the exact raw response text returned by the model.
    raw_response_text: str

    # Store token-usage statistics returned by the wrapper.
    token_usage: Dict[str, int]


# Define the synthesis-phase callable.
def synthesize_qa_answer(
    routing_state: QARoutingState,
    wiki_context: Any,
    prompt_context: Any,
    contract: Any,
) -> QASynthesisPackage:
    """
    Assemble the bounded context and invoke the LLM to synthesize a grounded answer.

    Purpose:
        Implement the reading and synthesis phases of Algorithm 3. In the
        paper's notation, this constructs the bounded context

            C_q = { λ_V(p) | p ∈ P_shortlist }

        and then performs the synthesis call:

            a ← m(QAPrompt(q, C_q); θ_qa)

    Inputs:
        routing_state:
            Routing state containing the shortlisted wiki page paths.
        wiki_context:
            Wiki context expected to expose ``wiki_graph`` with ``get_content``.
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        contract:
            Frozen run contract preserved in the signature for orchestration
            compatibility.

    Processes:
        1. Validate input types.
        2. Iterate over shortlisted page paths.
        3. Retrieve each page's content from the wiki graph.
        4. Apply semantic truncation at paragraph boundaries when content exceeds
           the safety threshold.
        5. Wrap each page inside explicit provenance delimiters.
        6. Concatenate all wrapped pages into the assembled context.
        7. Retrieve the registered Q&A prompt and effective settings.
        8. Format the final user prompt.
        9. Invoke the GPT-5.2 wrapper with web search disabled.
        10. Return the immutable synthesis package.

    Outputs:
        QASynthesisPackage:
            Immutable record describing the synthesis call and assembled context.

    Raises:
        TypeError:
            Raised if required inputs are invalid.
    """
    # Validate that the routing state is a QARoutingState.
    if not isinstance(routing_state, QARoutingState):
        # Raise a type error because synthesis depends on the routing-state schema.
        raise TypeError(
            "synthesize_qa_answer expected 'routing_state' to be a "
            "QARoutingState."
        )

    # Validate that the wiki context exposes a wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because content retrieval depends on wiki_context.wiki_graph.
        raise TypeError(
            "synthesize_qa_answer expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that the prompt context exposes a prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a type error because Q&A prompt lookup depends on prompt_context.prompt_registry.
        raise TypeError(
            "synthesize_qa_answer expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Initialize the ordered list of formatted page blocks.
    context_parts: List[str] = []

    # Define the per-page safety cap for inlined content.
    MAX_CHARS_PER_PAGE = 15000

    # Iterate over the shortlisted canonical page paths.
    for path in routing_state.shortlisted_paths:
        # Retrieve the raw Markdown content for the current page.
        content = wiki_context.wiki_graph.get_content(path)

        # Skip empty or missing content.
        if not content:
            # Continue to the next shortlisted path.
            continue

        # Apply semantic truncation if the page content exceeds the safety threshold.
        if len(content) > MAX_CHARS_PER_PAGE:
            # Find the last paragraph boundary before the truncation limit.
            truncate_idx = content.rfind("\n\n", 0, MAX_CHARS_PER_PAGE)

            # Fall back to a hard cut when no paragraph boundary exists.
            if truncate_idx == -1:
                # Use the hard character limit as the truncation boundary.
                truncate_idx = MAX_CHARS_PER_PAGE

            # Truncate at the semantic boundary and append the truncation marker.
            content = (
                content[:truncate_idx]
                + "\n\n...[CONTENT TRUNCATED FOR LENGTH]..."
            )

        # Wrap the page content with explicit provenance delimiters.
        formatted_page = (
            f"--- BEGIN FILE: {path} ---\n"
            f"{content}\n"
            f"--- END FILE: {path} ---\n"
        )

        # Append the formatted page block to the assembled context parts.
        context_parts.append(formatted_page)

    # Concatenate the formatted page blocks into a single context string.
    assembled_context = "\n".join(context_parts)

    # Provide a deterministic fallback when no relevant content was assembled.
    if not assembled_context.strip():
        # Emit the fallback context message.
        assembled_context = "No relevant wiki pages found."

    # Retrieve the registered Q&A prompt entry.
    qa_prompt_entry = prompt_context.prompt_registry.get("wiki_qa_synthesis")

    # Read the effective settings from the prompt entry.
    effective_settings = qa_prompt_entry.effective_settings

    # Extract the reasoning-effort setting, defaulting to "high".
    reasoning_effort = effective_settings.get("reasoning_effort", "high")

    # Extract the maximum output-token budget, defaulting to 6000.
    max_tokens = effective_settings.get("max_output_tokens", 6000)

    # Format the user prompt using the registered Q&A prompt template.
    user_prompt = qa_prompt_entry.prompt_text.format(
        question=routing_state.original_query,
        index_excerpts=assembled_context,
        read_tool_name="N/A (Context provided inline)",
    )

    # Define the strict anti-fabrication system prompt.
    system_prompt = (
        "You are a rigorous research assistant. Do not fabricate facts. "
        "Cite sources explicitly using Obsidian wikilinks. "
        "If you do not know, say so."
    )

    # Invoke the GPT-5.2 wrapper for grounded answer synthesis.
    llm_result = query_gpt52(
        prompt=user_prompt,
        system_prompt=system_prompt,
        reasoning_effort=reasoning_effort,
        max_output_tokens=max_tokens,
        enable_web_search=False,
        log_run_manifest=False,
    )

    # Extract the raw response text from the model wrapper result.
    raw_response = str(llm_result.get("response_text", ""))

    # Extract token-usage metadata from the model wrapper result.
    usage = dict(llm_result.get("usage", {}))

    # Construct and return the immutable synthesis package.
    return QASynthesisPackage(
        assembled_context=assembled_context,
        effective_settings=effective_settings,
        raw_response_text=raw_response,
        token_usage=usage,
    )



# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 3: Enforce grounding invariant and required sections
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable grounding-audit record.
@dataclass(frozen=True)
class GroundingAuditResult:
    """
    Immutable record capturing structural and grounding validation of the Q&A response.

    Purpose:
        Store the outcome of:
        1. required-section validation,
        2. citation extraction,
        3. grounding-invariant enforcement,
        4. Unknown-style fallback detection.

        The relevant grounding invariant from the paper is:

            ∀ claim c ∈ a, ∃ p ∈ P_shortlist : c is entailed by λ_V(p)

        This callable approximates that invariant operationally by validating
        that citations extracted from the Evidence section belong only to the
        routed shortlist.

    Attributes:
        is_structurally_valid:
            Whether all required Q&A sections are present.
        missing_sections:
            Ordered list of missing required sections.
        extracted_citations:
            Set of citations extracted from the Evidence section.
        hallucinated_citations:
            Set of citations that do not belong to the shortlisted paths.
        is_grounding_valid:
            Whether all extracted citations are grounded in the shortlisted set.
        unknown_fallback_used:
            Whether the answer used an Unknown-style unsupported-information
            fallback.
    """

    # Store whether the answer satisfies the required section structure.
    is_structurally_valid: bool

    # Store the list of missing required sections.
    missing_sections: List[str]

    # Store the set of extracted citations from the Evidence section.
    extracted_citations: Set[str]

    # Store the set of citations not found in the routed shortlist.
    hallucinated_citations: Set[str]

    # Store whether the grounding check succeeded.
    is_grounding_valid: bool

    # Store whether the model used an Unknown-style lexical fallback.
    unknown_fallback_used: bool


# Define the grounding-validation callable.
def validate_qa_grounding(
    synthesis_package: QASynthesisPackage,
    routing_state: QARoutingState,
) -> GroundingAuditResult:
    """
    Validate the LLM response against the required structural contract and the
    grounding invariant.

    Purpose:
        Enforce that the generated answer:
        1. contains all required Q&A sections, and
        2. cites only pages from the routed shortlist in its Evidence section.

        This function operationalizes the paper's grounding requirement by
        checking that citations in the answer's Evidence section belong to the
        shortlisted pages selected during routing.

    Inputs:
        synthesis_package:
            Q&A synthesis package containing the raw model response text.
        routing_state:
            Routing state containing the valid shortlisted wiki paths.

    Processes:
        1. Validate input types.
        2. Define the required output sections.
        3. Validate section existence using robust header matching.
        4. If AST parsing is available:
           - parse the Markdown token stream,
           - isolate the Evidence section,
           - extract wikilinks only from the Evidence subtree.
        5. Otherwise, fall back to regex-based Evidence section extraction.
        6. Normalize extracted citations.
        7. Compare citations against the valid shortlist to identify
           hallucinated citations.
        8. Detect lexical Unknown-style fallback usage.
        9. Return the immutable audit result.

    Outputs:
        GroundingAuditResult:
            Immutable record of structural and grounding validation.

    Raises:
        TypeError:
            Raised if the supplied inputs are of invalid types.

    Notes:
        This function does not prove semantic entailment. Rather, it enforces the
        strongest deterministic citation-path constraint available from the
        routed shortlist.
    """
    # Validate that the synthesis package has the expected structural type.
    if not isinstance(synthesis_package, QASynthesisPackage):
        # Raise a type error because structural and citation validation require
        # the QASynthesisPackage schema.
        raise TypeError(
            "validate_qa_grounding expected 'synthesis_package' to be a "
            "QASynthesisPackage."
        )

    # Validate that the routing state has the expected structural type.
    if not isinstance(routing_state, QARoutingState):
        # Raise a type error because grounding validation requires routed
        # shortlist provenance.
        raise TypeError(
            "validate_qa_grounding expected 'routing_state' to be a "
            "QARoutingState."
        )

    # Extract the raw response text from the synthesis package.
    response_text = synthesis_package.raw_response_text

    # Define the required Q&A sections expected by the prompt contract.
    required_sections = [
        "Conclusion",
        "Evidence",
        "Caveats",
        "Pages Consulted",
        "Recommended Wiki Improvements",
    ]

    # Initialize the list of missing required sections.
    missing_sections: List[str] = []

    # Check for the presence of each required section using header-matching regex.
    for section in required_sections:
        # Compile a case-insensitive multiline regex that matches a Markdown
        # heading containing the required section text.
        pattern = re.compile(
            rf"^#+\s+.*{re.escape(section)}.*$",
            re.MULTILINE | re.IGNORECASE,
        )

        # Record the section as missing if the pattern does not match.
        if not pattern.search(response_text):
            # Append the missing section name.
            missing_sections.append(section)

    # Compute whether the structural contract is satisfied.
    is_structurally_valid = len(missing_sections) == 0

    # Initialize the set of extracted citations.
    extracted_citations: Set[str] = set()

    # Prefer AST-based Evidence section extraction when markdown-it-py is available.
    if MARKDOWN_IT_AVAILABLE:
        # Initialize the CommonMark parser.
        md = MarkdownIt("commonmark")

        # Parse the full response text into a token stream.
        tokens = md.parse(response_text)

        # Initialize the Evidence-section state flag.
        in_evidence_section = False

        # Initialize the heading level at which the Evidence section starts.
        evidence_level = 0

        # Compile the wikilink extraction regex.
        wikilink_pattern = re.compile(r"\[\[(.*?)\]\]")

        # Traverse the parsed token stream.
        for i, token in enumerate(tokens):
            # Detect Markdown heading openings.
            if token.type == "heading_open":
                # Check whether the following inline token contains the heading text.
                if i + 1 < len(tokens) and tokens[i + 1].type == "inline":
                    # Normalize the heading text for case-insensitive matching.
                    heading_text = tokens[i + 1].content.lower()

                    # Enter the Evidence section when its heading is encountered.
                    if "evidence" in heading_text:
                        # Mark that subsequent inline tokens belong to the
                        # Evidence section.
                        in_evidence_section = True

                        # Record the heading level, e.g. h2 -> 2.
                        evidence_level = int(token.tag[1:])

                        # Continue because the section entry logic is complete.
                        continue

                # Exit the Evidence section when a same-level or higher-level
                # heading is encountered after entering it.
                if in_evidence_section:
                    # Read the current heading level, e.g. h1 -> 1, h2 -> 2.
                    current_level = int(token.tag[1:])

                    # A lower or equal heading number indicates a section boundary.
                    if current_level <= evidence_level:
                        # Mark the end of the Evidence section.
                        in_evidence_section = False

                        # Break because the relevant section has ended.
                        break

            # Extract citations only from inline text inside the Evidence section.
            if in_evidence_section and token.type == "inline" and token.children:
                # Traverse child tokens inside the inline token.
                for child in token.children:
                    # Restrict extraction to text nodes.
                    if child.type == "text":
                        # Extract all wikilink matches from the child content.
                        matches = wikilink_pattern.findall(child.content)

                        # Normalize and record each extracted citation path.
                        for match in matches:
                            # Remove alias suffixes and surrounding whitespace.
                            clean_path = match.split("|")[0].strip()

                            # Add the normalized citation path to the set.
                            extracted_citations.add(clean_path)

    # Fall back to regex section extraction when AST parsing is unavailable.
    else:
        # Emit a warning because regex section extraction is less robust than AST parsing.
        logger.warning(
            "markdown-it-py unavailable. Falling back to regex for section extraction."
        )

        # Capture the Evidence section content until the next heading or end of file.
        evidence_match = re.search(
            r"^#+\s+.*Evidence.*$(.*?)(?:^#+\s+|\Z)",
            response_text,
            re.MULTILINE | re.IGNORECASE | re.DOTALL,
        )

        # Read the matched Evidence section body, or use an empty string if absent.
        evidence_text = evidence_match.group(1) if evidence_match else ""

        # Extract raw wikilink payloads from the Evidence body.
        raw_links = re.findall(r"\[\[(.*?)\]\]", evidence_text)

        # Normalize each extracted wikilink payload.
        for link in raw_links:
            # Remove alias suffixes and surrounding whitespace.
            clean_path = link.split("|")[0].strip()

            # Add the normalized citation path to the set.
            extracted_citations.add(clean_path)

    # Convert the routed shortlist to a set for O(1) membership testing.
    valid_paths = set(routing_state.shortlisted_paths)

    # Initialize the set of hallucinated citations.
    hallucinated_citations: Set[str] = set()

    # Detect any citations that fall outside the routed shortlist.
    for citation in extracted_citations:
        # Mark the citation as hallucinated if it is not in the valid shortlist.
        if citation not in valid_paths:
            # Add the out-of-shortlist citation to the hallucination set.
            hallucinated_citations.add(citation)

    # Compute whether grounding validation succeeded.
    is_grounding_valid = len(hallucinated_citations) == 0

    # Detect lexical Unknown-style unsupported-information fallback usage.
    unknown_fallback_used = bool(
        re.search(
            r"\b(unknown|missing|not provided)\b",
            response_text,
            re.IGNORECASE,
        )
    )

    # Construct and return the immutable grounding audit result.
    return GroundingAuditResult(
        is_structurally_valid=is_structurally_valid,
        missing_sections=missing_sections,
        extracted_citations=extracted_citations,
        hallucinated_citations=hallucinated_citations,
        is_grounding_valid=is_grounding_valid,
        unknown_fallback_used=unknown_fallback_used,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable end-to-end Q&A context.
@dataclass(frozen=True)
class QAContext:
    """
    Immutable record encapsulating the complete end-to-end grounded Q&A
    transaction.

    Purpose:
        Package the routing provenance, synthesis provenance, and grounding
        audit result into one immutable object for downstream rendering,
        filing, and evaluation.

    Attributes:
        routing_state:
            Routing-phase provenance, corresponding to the shortlist-generation
            phase of Algorithm 3.
        synthesis_package:
            Synthesis-phase provenance, corresponding to the bounded-context
            answer-generation call.
        audit_result:
            Structural and grounding validation result for the generated answer.

    Notes:
        This object is the concrete aggregate representation of the paper's
        grounded Q&A operator:

            Φ_qa : Q × K → A
    """

    # Store the routing-phase provenance record.
    routing_state: Any

    # Store the synthesis-phase provenance record.
    synthesis_package: Any

    # Store the grounding-audit result.
    audit_result: Any


# Define the Task 13 orchestrator.
def task_13_orchestrator(
    query: str,
    wiki_context: Any,
    search_context: Any,
    prompt_context: Any,
    governance_context: Any,
    contract: Any,
) -> QAContext:
    """
    Orchestrate grounded Q&A end-to-end by routing the query, synthesizing the
    answer, and validating grounding.

    Purpose:
        Execute the three major phases of Algorithm 3 in the correct order:

            1. Routing phase
            2. Reading and synthesis phase
            3. Grounding validation phase

        Concretely, this orchestrator implements the paper's flow:

            Retrieve shortlist P_shortlist
            Read λ_V(p) for p in P_shortlist
            Assemble C_q
            Invoke a ← m(QAPrompt(q, C_q); θ_qa)
            Validate grounding

    Inputs:
        query:
            Raw user question string.
        wiki_context:
            Wiki architecture context expected to expose ``wiki_graph``.
        search_context:
            Search engine context expected to expose ``search_interface``.
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        governance_context:
            Governance context expected to expose ``summary_accumulator``.
        contract:
            Frozen run contract expected to expose ``original_config``.

    Processes:
        1. Validate top-level inputs and required interfaces.
        2. Extract retrieval-policy limits from the frozen contract.
        3. Route the query via ``route_qa_query``.
        4. Synthesize the answer via ``synthesize_qa_answer``.
        5. Validate structural and grounding constraints via
           ``validate_qa_grounding``.
        6. Increment warning counters when structural or grounding validation fails.
        7. Increment the global Q&A counter.
        8. Return the immutable ``QAContext``.

    Outputs:
        QAContext:
            Immutable aggregate of routing, synthesis, and validation provenance.

    Raises:
        TypeError:
            Raised if:
            - ``query`` is not a string,
            - required collaborator attributes are missing.

    Notes:
        This orchestrator preserves the original logic exactly, including:
        - extraction of ``pages_to_read_max``,
        - extraction of ``search_query_rewrite_rounds_max``,
        - warning-counter increments on invalid structure or grounding.
    """
    # Validate that the query is a string because routing and synthesis begin from raw textual input.
    if not isinstance(query, str):
        # Raise a type error immediately if the query contract is violated.
        raise TypeError("task_13_orchestrator expected 'query' to be a string.")

    # Validate that wiki_context exposes the wiki graph because content retrieval depends on it.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because wiki traversal requires wiki_context.wiki_graph.
        raise TypeError(
            "task_13_orchestrator expected 'wiki_context' to expose 'wiki_graph'."
        )

    # Validate that search_context exposes the search interface because routing depends on it.
    if not hasattr(search_context, "search_interface"):
        # Raise a type error because query routing requires search_context.search_interface.
        raise TypeError(
            "task_13_orchestrator expected 'search_context' to expose "
            "'search_interface'."
        )

    # Validate that prompt_context exposes the prompt registry because prompt lookup depends on it.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a type error because Q&A prompt retrieval requires prompt_registry.
        raise TypeError(
            "task_13_orchestrator expected 'prompt_context' to expose 'prompt_registry'."
        )

    # Validate that governance_context exposes the summary accumulator because warning and throughput accounting depend on it.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a type error because run-accounting requires summary_accumulator.
        raise TypeError(
            "task_13_orchestrator expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate that the contract exposes the original configuration because retrieval-policy extraction depends on it.
    if not hasattr(contract, "original_config"):
        # Raise a type error because contract.original_config is required for policy lookup.
        raise TypeError(
            "task_13_orchestrator expected 'contract' to expose 'original_config'."
        )

    # Extract the retrieval-and-search policy block from the frozen contract.
    qa_policy = (
        contract.original_config
        .get("retrieval_and_search_parameters", {})
        .get("tool_assisted_qa_policy", {})
    )

    # Extract the maximum number of pages to read, defaulting to 8 as specified in the study configuration.
    max_pages = qa_policy.get("pages_to_read_max", 8)

    # Extract the maximum number of search-query rewrite rounds, defaulting to 2.
    max_rewrites = qa_policy.get("search_query_rewrite_rounds_max", 2)

    # Step 1 of Algorithm 3: route the query to shortlist relevant wiki pages.
    routing_state = route_qa_query(
        query=query,
        search_context=search_context,
        prompt_context=prompt_context,
        contract=contract,
        max_pages=max_pages,
        max_rewrites=max_rewrites,
    )

    # Step 2 of Algorithm 3: assemble bounded context and synthesize the answer.
    synthesis_package = synthesize_qa_answer(
        routing_state=routing_state,
        wiki_context=wiki_context,
        prompt_context=prompt_context,
        contract=contract,
    )

    # Step 3 of Algorithm 3: validate structural contract and grounding invariant.
    audit_result = validate_qa_grounding(
        synthesis_package=synthesis_package,
        routing_state=routing_state,
    )

    # Increment the warning counter when the answer fails required structural checks.
    if not audit_result.is_structurally_valid:
        # Record a warning for structural invalidity.
        governance_context.summary_accumulator.increment_warnings()

    # Increment the warning counter when the answer fails grounding validation.
    if not audit_result.is_grounding_valid:
        # Record a warning for grounding invalidity.
        governance_context.summary_accumulator.increment_warnings()

    # Increment the global Q&A counter to track successful end-to-end Q&A executions.
    governance_context.summary_accumulator.increment_qa()

    # Construct and return the immutable QA context.
    return QAContext(
        routing_state=routing_state,
        synthesis_package=synthesis_package,
        audit_result=audit_result,
    )
````

````python
# Task 14 — Implement the artifact rendering callables

# ==============================================================================
# Task 14: Implement the artifact rendering callables
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 1: Render Markdown reports
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable Markdown-report rendering provenance record.
@dataclass(frozen=True)
class ReportRenderingPackage:
    """
    Immutable provenance record for Markdown report rendering.

    Purpose:
        Capture the output path, effective runtime settings, raw model response,
        token usage, and structural-validity result for a rendered report.

    Attributes:
        target_path:
            Absolute filesystem path of the rendered report.
        effective_settings:
            Effective runtime settings used for rendering.
        raw_response_text:
            Raw model output text written as the report body.
        token_usage:
            Token-usage accounting returned by the model wrapper.
        is_structurally_valid:
            Whether the rendered report satisfies the required report sections.
    """

    # Store the absolute report file path.
    target_path: pathlib.Path

    # Store the effective rendering settings.
    effective_settings: Dict[str, Any]

    # Store the raw rendered report text.
    raw_response_text: str

    # Store token-usage accounting for the rendering call.
    token_usage: Dict[str, int]

    # Store the structural-validity flag.
    is_structurally_valid: bool



# Define the Markdown-report rendering callable.
def render_markdown_report(
    qa_context: Any,
    prompt_context: Any,
    fs_context: Any,
    governance_context: Any,
    repo_root: pathlib.Path,
) -> ReportRenderingPackage:
    """
    Render a grounded intermediate answer into a durable Markdown report.

    Purpose:
        Implement the Markdown-report branch of the paper's Artifact Rendering
        operator:

            Φ_render : A_intermediate → A_files

        This callable takes grounded Q&A findings and formats them into a
        durable report suitable for storage under ``wiki/derived/reports/``.

    Inputs:
        qa_context:
            Q&A context expected to expose:
            - ``routing_state.original_query``
            - ``routing_state.shortlisted_paths``
            - ``synthesis_package.raw_response_text``
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        fs_context:
            Filesystem substrate context expected to expose
            ``wiki_layout.derived_subdirectories``.
        governance_context:
            Governance context expected to expose ``summary_accumulator``.
        repo_root:
            Repository root directory.

    Processes:
        1. Validate input types and required collaborators.
        2. Retrieve the registered Markdown report prompt.
        3. Extract effective settings.
        4. Derive a deterministic SHA-256-based filename from the original query.
        5. Resolve and create the reports directory.
        6. Format the user prompt using grounded findings and cited paths.
        7. Invoke the model with web search disabled.
        8. Validate the required report sections.
        9. Write the rendered report to disk.
        10. Update governance counters and return the rendering package.

    Outputs:
        ReportRenderingPackage:
            Immutable rendering provenance record.

    Raises:
        TypeError:
            Raised if ``repo_root`` is not a ``pathlib.Path``.
        ValueError:
            Raised if required collaborator attributes are missing.
    """
    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because filesystem path resolution requires pathlib semantics.
        raise TypeError(
            "render_markdown_report expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that the prompt context exposes the prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a value error because prompt lookup depends on prompt_registry.
        raise ValueError(
            "render_markdown_report expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the filesystem context exposes the wiki layout.
    if not hasattr(fs_context, "wiki_layout"):
        # Raise a value error because derived directory resolution depends on wiki_layout.
        raise ValueError(
            "render_markdown_report expected 'fs_context' to expose 'wiki_layout'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a value error because artifact and warning counters depend on summary_accumulator.
        raise ValueError(
            "render_markdown_report expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate that the QA context exposes the routing state.
    if not hasattr(qa_context, "routing_state"):
        # Raise a value error because query and citations are drawn from routing_state.
        raise ValueError(
            "render_markdown_report expected 'qa_context' to expose 'routing_state'."
        )

    # Validate that the QA context exposes the synthesis package.
    if not hasattr(qa_context, "synthesis_package"):
        # Raise a value error because grounded findings are drawn from the synthesis package.
        raise ValueError(
            "render_markdown_report expected 'qa_context' to expose "
            "'synthesis_package'."
        )

    # Retrieve the registered Markdown report-rendering prompt entry.
    report_prompt_entry = prompt_context.prompt_registry.get(
        "artifact_render_markdown_report"
    )

    # Read the effective runtime settings for report rendering.
    effective_settings = report_prompt_entry.effective_settings

    # Extract the reasoning-effort setting, defaulting to "high".
    reasoning_effort = effective_settings.get("reasoning_effort", "high")

    # Extract the max-output-token budget, defaulting to 10000.
    max_tokens = effective_settings.get("max_output_tokens", 10000)

    # Extract the original user query from the QA context.
    original_query = qa_context.routing_state.original_query

    # Extract the grounded findings from the synthesis package.
    grounded_findings = qa_context.synthesis_package.raw_response_text

    # Join the shortlisted page paths into a citation string.
    citations_str = ", ".join(qa_context.routing_state.shortlisted_paths)

    # Create a filesystem-safe slug from the original query.
    query_slug = re.sub(
        r"[^a-z0-9]+",
        "-",
        original_query.lower()[:50],
    ).strip("-")

    # Compute the deterministic SHA-256-based query hash prefix.
    query_hash = hashlib.sha256(
        original_query.encode("utf-8")
    ).hexdigest()[:6]

    # Construct the report filename.
    filename = f"report_{query_slug}_{query_hash}.md"

    # Resolve the reports directory from the filesystem context, falling back to wiki/derived/reports.
    reports_dir = fs_context.wiki_layout.derived_subdirectories.get(
        "reports_dir",
        repo_root / "wiki" / "derived" / "reports",
    )

    # Ensure the reports directory exists.
    reports_dir.mkdir(parents=True, exist_ok=True)

    # Resolve the absolute target path of the report file.
    target_path = (reports_dir / filename).resolve()

    # Format the rendering prompt using the grounded QA findings.
    user_prompt = report_prompt_entry.prompt_text.format(
        Title=f"Research Report: {original_query[:50]}...",
        objective=original_query,
        findings=grounded_findings,
        citations=citations_str,
        output_path=str(target_path.relative_to(repo_root)),
    )

    # Define the strict anti-fabrication system prompt for report rendering.
    system_prompt = (
        "You are a rigorous research assistant. Do not fabricate facts. "
        "You must strictly format the provided findings into the requested "
        "Markdown structure. Do not introduce any new information not present "
        "in the findings."
    )

    # Invoke the model wrapper for report rendering.
    llm_result = query_gpt52(
        prompt=user_prompt,
        system_prompt=system_prompt,
        reasoning_effort=reasoning_effort,
        max_output_tokens=max_tokens,
        enable_web_search=False,
        log_run_manifest=False,
    )

    # Extract the raw response text from the model wrapper result.
    raw_response = str(llm_result.get("response_text", ""))

    # Extract token-usage metadata from the model wrapper result.
    usage = dict(llm_result.get("usage", {}))

    # Define the required report sections from the rendering contract.
    required_sections = [
        "Executive Summary",
        "Main Analysis",
        "Evidence & Citations",
        "Limitations",
        "Next Questions",
    ]

    # Initialize the list of missing required sections.
    missing_sections: List[str] = []

    # Check whether each required section is present as a Markdown heading.
    for section in required_sections:
        # Compile a heading-matching regex for the required section.
        pattern = re.compile(
            rf"^#+\s+.*{re.escape(section)}.*$",
            re.MULTILINE | re.IGNORECASE,
        )

        # Record the section as missing when the pattern does not match.
        if not pattern.search(raw_response):
            # Append the missing section to the list.
            missing_sections.append(section)

    # Compute whether the rendered report satisfies the structural contract.
    is_structurally_valid = len(missing_sections) == 0

    # Emit a warning and increment warnings when the structure is invalid.
    if not is_structurally_valid:
        # Increment the warning counter because the report failed structural validation.
        governance_context.summary_accumulator.increment_warnings()

        # Emit a warning with the missing sections.
        logger.warning(
            f"Rendered report missing required sections: {missing_sections}"
        )

    # Write the rendered report text to disk as UTF-8.
    target_path.write_text(raw_response, encoding="utf-8")

    # Increment the artifact counter because a report artifact was generated.
    governance_context.summary_accumulator.increment_artifacts()

    # Construct and return the immutable report-rendering package.
    return ReportRenderingPackage(
        target_path=target_path,
        effective_settings=effective_settings,
        raw_response_text=raw_response,
        token_usage=usage,
        is_structurally_valid=is_structurally_valid,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 2: Render Marp slide decks
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable slide-rendering provenance record.
@dataclass(frozen=True)
class SlideRenderingPackage:
    """
    Immutable record capturing the provenance and output of the Marp slide
    rendering phase.

    Purpose:
        Store the output path, runtime settings, raw model output, token usage,
        and structural-validation result for a rendered Marp slide deck.

    Inputs:
        This dataclass is populated by ``render_marp_slides``.

    Processes:
        No computation is performed here. It is a passive immutable rendering
        record.

    Outputs:
        target_path:
            Absolute filesystem path where the slide deck was written.
        effective_settings:
            Runtime settings used for the rendering call.
        raw_response_text:
            Exact Marp-flavored Markdown returned by the LLM.
        token_usage:
            Token-usage statistics returned by the model wrapper.
        is_structurally_valid:
            Boolean indicating whether Marp separators and a References slide
            were found.
    """

    # Store the absolute path to the written slide deck.
    target_path: pathlib.Path

    # Store the effective runtime settings used for the LLM call.
    effective_settings: Dict[str, Any]

    # Store the raw textual response returned by the model.
    raw_response_text: str

    # Store token-usage statistics.
    token_usage: Dict[str, int]

    # Store the structural validation result.
    is_structurally_valid: bool


# Define the Marp-slide rendering callable.
def render_marp_slides(
    qa_context: Any,
    prompt_context: Any,
    fs_context: Any,
    governance_context: Any,
    repo_root: pathlib.Path,
) -> SlideRenderingPackage:
    """
    Render a grounded intermediate answer into a Marp-flavored Markdown slide deck.

    Purpose:
        Implement the Marp-slide branch of the paper's Artifact Rendering
        operator. The output is written under ``wiki/derived/slides/`` and must
        satisfy the Marp prompt contract, including separator and References
        checks.

    Inputs:
        qa_context:
            Q&A context expected to expose the routed query and grounded findings.
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        fs_context:
            Filesystem substrate context expected to expose
            ``wiki_layout.derived_subdirectories``.
        governance_context:
            Governance context expected to expose ``summary_accumulator``.
        repo_root:
            Repository root directory.

    Processes:
        1. Validate input types and required collaborators.
        2. Retrieve the registered Marp prompt and effective settings.
        3. Generate a deterministic SHA-256-based filename.
        4. Resolve and create the slides directory.
        5. Format the rendering prompt using grounded findings and citations.
        6. Invoke the model with web search disabled.
        7. Strip enclosing Markdown fences if present.
        8. Validate strict Marp separators and References slide presence.
        9. Write the deck to disk.
        10. Update counters and return the rendering package.

    Outputs:
        SlideRenderingPackage:
            Immutable rendering provenance record.

    Raises:
        TypeError:
            Raised if ``repo_root`` is not a ``pathlib.Path``.
        ValueError:
            Raised if required collaborator attributes are missing.
    """
    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because filesystem path resolution requires pathlib semantics.
        raise TypeError(
            "render_marp_slides expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that the prompt context exposes the prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a value error because prompt lookup depends on prompt_registry.
        raise ValueError(
            "render_marp_slides expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the filesystem context exposes the wiki layout.
    if not hasattr(fs_context, "wiki_layout"):
        # Raise a value error because directory resolution depends on wiki_layout.
        raise ValueError(
            "render_marp_slides expected 'fs_context' to expose 'wiki_layout'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a value error because warnings and artifact counters depend on summary_accumulator.
        raise ValueError(
            "render_marp_slides expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate that the QA context exposes the routing state.
    if not hasattr(qa_context, "routing_state"):
        # Raise a value error because the original query and citations come from routing_state.
        raise ValueError(
            "render_marp_slides expected 'qa_context' to expose 'routing_state'."
        )

    # Validate that the QA context exposes the synthesis package.
    if not hasattr(qa_context, "synthesis_package"):
        # Raise a value error because grounded findings come from the synthesis package.
        raise ValueError(
            "render_marp_slides expected 'qa_context' to expose "
            "'synthesis_package'."
        )

    # Retrieve the registered Marp slide-rendering prompt entry.
    marp_prompt_entry = prompt_context.prompt_registry.get(
        "artifact_render_marp_slides"
    )

    # Read the effective runtime settings for slide rendering.
    effective_settings = marp_prompt_entry.effective_settings

    # Extract the reasoning-effort setting, defaulting to "high".
    reasoning_effort = effective_settings.get("reasoning_effort", "high")

    # Extract the max-output-token budget, defaulting to 10000.
    max_tokens = effective_settings.get("max_output_tokens", 10000)

    # Extract the original user query.
    original_query = qa_context.routing_state.original_query

    # Extract the grounded findings from the synthesis package.
    grounded_findings = qa_context.synthesis_package.raw_response_text

    # Join the shortlisted citation paths into a single string.
    citations_str = ", ".join(qa_context.routing_state.shortlisted_paths)

    # Create a filesystem-safe slug from the original query.
    query_slug = re.sub(
        r"[^a-z0-9]+",
        "-",
        original_query.lower()[:50],
    ).strip("-")

    # Compute the deterministic SHA-256-based query hash prefix.
    query_hash = hashlib.sha256(
        original_query.encode("utf-8")
    ).hexdigest()[:6]

    # Construct the slide-deck filename.
    filename = f"slides_{query_slug}_{query_hash}.md"

    # Resolve the slides directory from the filesystem context, falling back to wiki/derived/slides.
    slides_dir = fs_context.wiki_layout.derived_subdirectories.get(
        "slides_dir",
        repo_root / "wiki" / "derived" / "slides",
    )

    # Ensure the slides directory exists.
    slides_dir.mkdir(parents=True, exist_ok=True)

    # Resolve the absolute target path of the slide-deck file.
    target_path = (slides_dir / filename).resolve()

    # Format the slide-rendering prompt using the grounded findings.
    user_prompt = marp_prompt_entry.prompt_text.format(
        talk_title=f"Presentation: {original_query[:50]}...",
        audience="Research Peers",
        duration_min="10",
        key_points=grounded_findings,
        citations=citations_str,
        output_path=str(target_path.relative_to(repo_root)),
    )

    # Define the strict anti-fabrication system prompt for slide rendering.
    system_prompt = (
        "You are a rigorous research assistant. Do not fabricate facts. "
        "You must strictly format the provided findings into Marp-flavored "
        "Markdown slides. Do not introduce any new information not present "
        "in the findings."
    )

    # Invoke the model wrapper for slide rendering.
    llm_result = query_gpt52(
        prompt=user_prompt,
        system_prompt=system_prompt,
        reasoning_effort=reasoning_effort,
        max_output_tokens=max_tokens,
        enable_web_search=False,
        log_run_manifest=False,
    )

    # Extract the raw response text from the model wrapper result.
    raw_response = str(llm_result.get("response_text", ""))

    # Extract token-usage metadata from the model wrapper result.
    usage = dict(llm_result.get("usage", {}))

    # Strip enclosing Markdown code fences if the model wrapped the entire response.
    raw_response = re.sub(
        r"^```markdown\n|\n```$",
        "",
        raw_response,
        flags=re.IGNORECASE,
    ).strip()

    # Validate strict block-level Marp separators using exact three-hyphen lines.
    separator_matches = re.findall(r"(?m)^---$", raw_response)

    # Require at least two separators to indicate multiple slide boundaries.
    has_separators = len(separator_matches) >= 2

    # Validate that a References slide heading is present.
    has_references = bool(
        re.search(
            r"^#+\s+.*References.*$",
            raw_response,
            re.MULTILINE | re.IGNORECASE,
        )
    )

    # Compute the overall structural-validity flag.
    is_structurally_valid = has_separators and has_references

    # Emit a warning and increment warnings when structure is invalid.
    if not is_structurally_valid:
        # Increment the warning counter because the deck failed structural validation.
        governance_context.summary_accumulator.increment_warnings()

        # Emit a warning describing the structural issue.
        logger.warning(
            "Rendered slides missing strict Marp separators or References slide."
        )

    # Write the rendered slide deck to disk as UTF-8.
    target_path.write_text(raw_response, encoding="utf-8")

    # Increment the artifact counter because a slide artifact was generated.
    governance_context.summary_accumulator.increment_artifacts()

    # Construct and return the immutable slide-rendering package.
    return SlideRenderingPackage(
        target_path=target_path,
        effective_settings=effective_settings,
        raw_response_text=raw_response,
        token_usage=usage,
        is_structurally_valid=is_structurally_valid,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 3: Handle matplotlib figures
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable figure-rendering provenance record.
@dataclass(frozen=True)
class FigureRenderingPackage:
    """
    Immutable record capturing the provenance and output of the matplotlib
    figure-rendering phase.

    Purpose:
        Store the output image path, script path, raw generated Python code,
        execution status, and token usage for a rendered figure.

    Inputs:
        This dataclass is populated by ``render_matplotlib_figure``.

    Processes:
        No computation is performed here. It is a passive immutable rendering
        record.

    Outputs:
        image_path:
            Absolute filesystem path to the rendered figure image.
        script_path:
            Absolute filesystem path to the generated Python script.
        raw_python_code:
            Exact Python code returned by the LLM after extraction.
        execution_success:
            Boolean indicating whether script execution succeeded and produced
            the target image.
        token_usage:
            Token-usage statistics returned by the model wrapper.
    """

    # Store the absolute path to the rendered image.
    image_path: pathlib.Path

    # Store the absolute path to the generated Python script.
    script_path: pathlib.Path

    # Store the raw Python code extracted from the model response.
    raw_python_code: str

    # Store whether script execution succeeded.
    execution_success: bool

    # Store token-usage statistics.
    token_usage: Dict[str, int]


# Define the matplotlib-figure rendering callable.
def render_matplotlib_figure(
    qa_context: Any,
    prompt_context: Any,
    fs_context: Any,
    governance_context: Any,
    contract: Any,
    repo_root: pathlib.Path,
) -> FigureRenderingPackage:
    """
    Generate and execute a Python script to render a matplotlib figure securely.

    Purpose:
        Implement the figure branch of the paper's Artifact Rendering operator.
        The LLM stage generates plotting code or a plotting plan, while the
        Python runtime stage executes the generated script in a sandboxed Docker
        container.

    Inputs:
        qa_context:
            Q&A context expected to expose the routed query and grounded findings.
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        fs_context:
            Filesystem substrate context expected to expose
            ``wiki_layout.derived_subdirectories``.
        governance_context:
            Governance context expected to expose ``summary_accumulator``.
        contract:
            Frozen run contract expected to expose
            ``original_config.outputs_and_visualizations``.
        repo_root:
            Repository root directory.

    Processes:
        1. Validate input types and required collaborators.
        2. Retrieve the registered figure prompt and effective settings.
        3. Resolve plot style defaults from the contract.
        4. Generate deterministic SHA-256-based script and image filenames.
        5. Resolve and create the figures directory.
        6. Format the code-generation prompt.
        7. Invoke the model with web search disabled.
        8. Extract Python code from fenced output when present.
        9. Write the generated script to disk.
        10. If Docker is available:
            - create a client,
            - mount only the figures directory,
            - disable network access,
            - run the script in an ephemeral container,
            - verify the PNG output,
            - clean up the container.
        11. Otherwise, refuse execution on the host for security reasons.
        12. Return the immutable figure-rendering package.

    Outputs:
        FigureRenderingPackage:
            Immutable rendering provenance record.

    Raises:
        TypeError:
            Raised if ``repo_root`` is not a ``pathlib.Path``.
        ValueError:
            Raised if required collaborator attributes are missing.

    Notes:
        This function intentionally refuses bare-metal execution of LLM-generated
        code when Docker sandboxing is unavailable.
    """
    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because filesystem path resolution requires pathlib semantics.
        raise TypeError(
            "render_matplotlib_figure expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that the prompt context exposes the prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a value error because prompt lookup depends on prompt_registry.
        raise ValueError(
            "render_matplotlib_figure expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the filesystem context exposes the wiki layout.
    if not hasattr(fs_context, "wiki_layout"):
        # Raise a value error because figures directory resolution depends on wiki_layout.
        raise ValueError(
            "render_matplotlib_figure expected 'fs_context' to expose "
            "'wiki_layout'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a value error because warnings and artifact counters depend on summary_accumulator.
        raise ValueError(
            "render_matplotlib_figure expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate that the contract exposes the original configuration.
    if not hasattr(contract, "original_config"):
        # Raise a value error because plotting configuration is read from original_config.
        raise ValueError(
            "render_matplotlib_figure expected 'contract' to expose "
            "'original_config'."
        )

    # Validate that the QA context exposes the routing state.
    if not hasattr(qa_context, "routing_state"):
        # Raise a value error because the original query is read from routing_state.
        raise ValueError(
            "render_matplotlib_figure expected 'qa_context' to expose "
            "'routing_state'."
        )

    # Validate that the QA context exposes the synthesis package.
    if not hasattr(qa_context, "synthesis_package"):
        # Raise a value error because grounded findings are read from the synthesis package.
        raise ValueError(
            "render_matplotlib_figure expected 'qa_context' to expose "
            "'synthesis_package'."
        )

    # Retrieve the registered figure-rendering prompt entry.
    figure_prompt_entry = prompt_context.prompt_registry.get(
        "artifact_render_matplotlib_plan_or_code"
    )

    # Read the effective runtime settings for figure rendering.
    effective_settings = figure_prompt_entry.effective_settings

    # Extract the reasoning-effort setting, defaulting to "high".
    reasoning_effort = effective_settings.get("reasoning_effort", "high")

    # Extract the max-output-token budget, defaulting to 6000.
    max_tokens = effective_settings.get("max_output_tokens", 6000)

    # Read the outputs-and-visualizations configuration block.
    outputs_config = contract.original_config.get(
        "outputs_and_visualizations",
        {},
    )

    # Read the plot-construction parameters block.
    plot_params = outputs_config.get("plot_construction_parameters", {})

    # Read the style defaults, falling back to the configured article defaults.
    style_defaults = plot_params.get(
        "default_style",
        {
            "figure_size_inches": [10, 6],
            "dpi": 200,
            "grid": True,
        },
    )

    # Convert the style defaults to a string for prompt injection.
    style_str = str(style_defaults)

    # Extract the original query from the QA context.
    original_query = qa_context.routing_state.original_query

    # Extract the grounded findings from the synthesis package.
    grounded_findings = qa_context.synthesis_package.raw_response_text

    # Create a filesystem-safe slug from the original query.
    query_slug = re.sub(
        r"[^a-z0-9]+",
        "-",
        original_query.lower()[:50],
    ).strip("-")

    # Compute the deterministic SHA-256-based query hash prefix.
    query_hash = hashlib.sha256(
        original_query.encode("utf-8")
    ).hexdigest()[:6]

    # Construct the figure image filename.
    image_filename = f"figure_{query_slug}_{query_hash}.png"

    # Construct the Python script filename.
    script_filename = f"script_{query_slug}_{query_hash}.py"

    # Resolve the figures directory from the filesystem context, falling back to wiki/derived/figures.
    figures_dir = fs_context.wiki_layout.derived_subdirectories.get(
        "figures_dir",
        repo_root / "wiki" / "derived" / "figures",
    )

    # Ensure the figures directory exists.
    figures_dir.mkdir(parents=True, exist_ok=True)

    # Resolve the absolute figure image path.
    image_path = (figures_dir / image_filename).resolve()

    # Resolve the absolute Python script path.
    script_path = (figures_dir / script_filename).resolve()

    # Format the code-generation prompt for the figure renderer.
    user_prompt = figure_prompt_entry.prompt_text.format(
        figure_goal=f"Visualize data related to: {original_query}",
        data_spec=grounded_findings,
        style_requirements=style_str,
        image_path=f"/workspace/{image_filename}",
        script_path=f"/workspace/{script_filename}",
    )

    # Define the strict anti-fabrication and exact-save-path system prompt.
    system_prompt = (
        "You are an expert Python data scientist. Generate a single, "
        "self-contained Python script using matplotlib to create the requested "
        "figure. Do not fabricate data. The script MUST save the figure to the "
        "exact provided image_path."
    )

    # Invoke the model wrapper for Python plotting code generation.
    llm_result = query_gpt52(
        prompt=user_prompt,
        system_prompt=system_prompt,
        reasoning_effort=reasoning_effort,
        max_output_tokens=max_tokens,
        enable_web_search=False,
        log_run_manifest=False,
    )

    # Extract the raw response text from the model wrapper result.
    raw_response = str(llm_result.get("response_text", ""))

    # Extract token-usage metadata from the model wrapper result.
    usage = dict(llm_result.get("usage", {}))

    # Attempt to extract fenced Python code from the model response.
    code_match = re.search(
        r"```python\s*(.*?)\s*```",
        raw_response,
        re.DOTALL | re.IGNORECASE,
    )

    # Use the fenced code when present; otherwise fall back to the raw response.
    raw_python_code = code_match.group(1) if code_match else raw_response

    # Write the generated Python script to disk under the mounted figures directory.
    script_path.write_text(raw_python_code, encoding="utf-8")

    # Initialize the execution-success flag.
    execution_success = False

    # Use Docker sandbox execution when Docker SDK is available.
    if DOCKER_AVAILABLE:
        try:
            # Create a Docker client from the host environment.
            client = docker.from_env()

            # Define the bind mount mapping: host figures_dir -> container /workspace.
            volumes = {
                str(figures_dir): {
                    "bind": "/workspace",
                    "mode": "rw",
                }
            }

            # Launch an ephemeral, network-isolated container to execute the script.
            container = client.containers.run(
                "python:3.10-slim",
                command=(
                    f"sh -c 'pip install matplotlib && "
                    f"python /workspace/{script_filename}'"
                ),
                volumes=volumes,
                network_disabled=True,
                detach=True,
                mem_limit="512m",
            )

            # Wait for container completion with a strict timeout.
            result = container.wait(timeout=30)

            # Mark success only when the container exits successfully and the image file exists.
            if result["StatusCode"] == 0 and image_path.exists() and image_path.is_file():
                # Record successful execution.
                execution_success = True

                # Increment the artifact counter because the figure artifact exists.
                governance_context.summary_accumulator.increment_artifacts()

            # Otherwise collect logs, emit a warning, and increment warnings.
            else:
                # Read the container logs for diagnosis.
                logs = container.logs().decode("utf-8")

                # Emit a warning containing the execution logs.
                logger.warning(
                    f"Docker execution failed or image not created. Logs: {logs}"
                )

                # Increment the warning counter.
                governance_context.summary_accumulator.increment_warnings()

            # Remove the container forcefully after execution.
            container.remove(force=True)

        # Handle any Docker sandbox failure.
        except Exception as e:
            # Emit an error describing the sandbox failure.
            logger.error(f"Docker sandbox execution failed: {e}")

            # Increment the warning counter because figure execution failed.
            governance_context.summary_accumulator.increment_warnings()

    # Refuse unsafe host execution when Docker is unavailable.
    else:
        # Emit a critical security warning because host execution is intentionally forbidden.
        logger.critical(
            "Docker SDK is unavailable. REFUSING to execute LLM-generated code "
            "on the host machine due to severe security risks. Figure generation failed."
        )

        # Increment the warning counter because secure execution was unavailable.
        governance_context.summary_accumulator.increment_warnings()

    # Construct and return the immutable figure-rendering package.
    return FigureRenderingPackage(
        image_path=image_path,
        script_path=script_path,
        raw_python_code=raw_python_code,
        execution_success=execution_success,
        token_usage=usage,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable artifact-rendering context aggregate.
@dataclass(frozen=True)
class ArtifactRenderingContext:
    """
    Immutable record encapsulating the complete artifact-rendering transaction.

    Purpose:
        Package the optional report, slide, and figure rendering outputs into one
        immutable object for downstream filing tasks.

    Inputs:
        This dataclass is populated by ``task_14_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        report_package:
            Markdown report rendering package if requested and successful.
        slide_package:
            Marp slide rendering package if requested and successful.
        figure_package:
            Matplotlib figure rendering package if requested and successful.
    """

    # Store the optional Markdown report package.
    report_package: Optional[ReportRenderingPackage] = None

    # Store the optional Marp slide-deck package.
    slide_package: Optional[SlideRenderingPackage] = None

    # Store the optional matplotlib figure package.
    figure_package: Optional[FigureRenderingPackage] = None


# Define the Task 14 orchestrator.
def task_14_orchestrator(
    qa_context: "QAContext",
    prompt_context: "PromptContext",
    fs_context: "FilesystemSubstrateContext",
    governance_context: "GovernanceContext",
    contract: "FrozenRunContract",
    repo_root: pathlib.Path,
    requested_formats: List[str]
) -> ArtifactRenderingContext:
    """
    Orchestrate Task 14 end-to-end by rendering the requested artifact formats.

    Purpose:
        Execute the artifact-rendering stage conditionally by format, ensuring
        that each requested output is:
        1. generated from grounded findings,
        2. validated,
        3. persisted to disk,
        4. isolated so that one rendering failure does not cascade.

    Inputs:
        qa_context:
            Q&A context containing grounded findings.
        prompt_context:
            Prompt context exposing rendering prompts.
        fs_context:
            Filesystem substrate context exposing derived artifact directories.
        governance_context:
            Governance context exposing warning and artifact counters.
        contract:
            Frozen run contract containing plotting defaults and other runtime
            configuration.
        repo_root:
            Project root directory used for path resolution.
        requested_formats:
            List of requested output modalities, such as:
            - ``"markdown"``
            - ``"marp"``
            - ``"figure"``

    Processes:
        1. Validate top-level inputs.
        2. Initialize all optional package outputs to ``None``.
        3. If ``"markdown"`` is requested:
           - call ``render_markdown_report`` inside a try-except block.
        4. If ``"marp"`` is requested:
           - call ``render_marp_slides`` inside a try-except block.
        5. If ``"figure"`` is requested:
           - call ``render_matplotlib_figure`` inside a try-except block.
        6. On each rendering failure:
           - increment the warning counter,
           - print a diagnostic,
           - continue processing the remaining formats.
        7. Return the immutable aggregate context.

    Outputs:
        ArtifactRenderingContext:
            Immutable aggregate containing any successfully rendered artifacts.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.

    Notes:
        This orchestrator realizes the multi-format branch of Algorithm 4 by
        conditionally applying the rendering operator for each requested output
        modality.
    """
    # Validate that the QA context is of the correct type.
    if not isinstance(qa_context, QAContext):
        # Raise a type error if the QA context is invalid.
        raise TypeError(
            "task_14_orchestrator expected 'qa_context' to be a QAContext."
        )

    # Validate that the prompt context is of the correct type.
    if not isinstance(prompt_context, PromptContext):
        # Raise a type error if the prompt context is invalid.
        raise TypeError(
            "task_14_orchestrator expected 'prompt_context' to be a "
            "PromptContext."
        )

    # Validate that the filesystem context is of the correct type.
    if not isinstance(fs_context, FilesystemSubstrateContext):
        # Raise a type error if the filesystem context is invalid.
        raise TypeError(
            "task_14_orchestrator expected 'fs_context' to be a "
            "FilesystemSubstrateContext."
        )

    # Validate that the governance context is of the correct type.
    if not isinstance(governance_context, GovernanceContext):
        # Raise a type error if the governance context is invalid.
        raise TypeError(
            "task_14_orchestrator expected 'governance_context' to be a "
            "GovernanceContext."
        )

    # Validate that the contract is of the correct type.
    if not isinstance(contract, FrozenRunContract):
        # Raise a type error if the contract is invalid.
        raise TypeError(
            "task_14_orchestrator expected 'contract' to be a "
            "FrozenRunContract."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_14_orchestrator expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that requested_formats is a list.
    if not isinstance(requested_formats, list):
        # Raise a type error because conditional dispatch requires a list.
        raise TypeError(
            "task_14_orchestrator expected 'requested_formats' to be a list "
            "of strings."
        )

    # Initialize the optional report package to None.
    report_package: Optional[ReportRenderingPackage] = None

    # Initialize the optional slide package to None.
    slide_package: Optional[SlideRenderingPackage] = None

    # Initialize the optional figure package to None.
    figure_package: Optional[FigureRenderingPackage] = None

    # Step 1: Render a Markdown report if requested.
    if "markdown" in requested_formats:
        try:
            # Invoke the Markdown-report rendering callable.
            report_package = render_markdown_report(
                qa_context=qa_context,
                prompt_context=prompt_context,
                fs_context=fs_context,
                governance_context=governance_context,
                repo_root=repo_root
            )
        except Exception as e:
            # Increment the warning counter on rendering failure.
            governance_context.summary_accumulator.increment_warnings()

            # Emit a console error preserving the original behavior.
            print(f"ERROR: Failed to render Markdown report: {e}")

    # Step 2: Render a Marp slide deck if requested.
    if "marp" in requested_formats:
        try:
            # Invoke the Marp-slide rendering callable.
            slide_package = render_marp_slides(
                qa_context=qa_context,
                prompt_context=prompt_context,
                fs_context=fs_context,
                governance_context=governance_context,
                repo_root=repo_root
            )
        except Exception as e:
            # Increment the warning counter on rendering failure.
            governance_context.summary_accumulator.increment_warnings()

            # Emit a console error preserving the original behavior.
            print(f"ERROR: Failed to render Marp slides: {e}")

    # Step 3: Render a matplotlib figure if requested.
    if "figure" in requested_formats:
        try:
            # Invoke the matplotlib-figure rendering callable.
            figure_package = render_matplotlib_figure(
                qa_context=qa_context,
                prompt_context=prompt_context,
                fs_context=fs_context,
                governance_context=governance_context,
                contract=contract,
                repo_root=repo_root
            )
        except Exception as e:
            # Increment the warning counter on rendering failure.
            governance_context.summary_accumulator.increment_warnings()

            # Emit a console error preserving the original behavior.
            print(f"ERROR: Failed to render matplotlib figure: {e}")

    # Construct and return the immutable artifact-rendering context.
    return ArtifactRenderingContext(
        report_package=report_package,
        slide_package=slide_package,
        figure_package=figure_package
    )
````

````python
# Task 15 — File generated outputs back into the wiki

# ==============================================================================
# Task 15: File generated outputs back into the wiki
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 1: Create derived artifact nodes
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable derived-artifact registration record.
@dataclass(frozen=True)
class DerivedArtifactRegistrationRecord:
    """
    Immutable record capturing the metadata required to integrate a derived
    artifact into the wiki graph.

    Purpose:
        Formalize the metadata required to treat generated reports, slide decks,
        and figures as first-class knowledge objects in the compiled wiki.

    Inputs:
        This dataclass is populated by ``create_derived_artifact_nodes``.

    Processes:
        No computation is performed here. It is a passive immutable registration
        record.

    Outputs:
        artifact_id:
            Deterministic SHA-256-derived identifier for the artifact.
        artifact_type:
            Artifact format label, such as:
            - ``markdown_report``
            - ``marp_slides``
            - ``matplotlib_figure``
        canonical_path:
            Repo-relative path to the artifact node or its companion metadata
            page.
        title:
            Human-readable title derived from the originating query.
        originating_query:
            Exact user query that triggered the artifact generation.
        consulted_paths:
            Ordered list of wiki pages consulted during Q&A synthesis.
        render_timestamp_utc:
            UTC ISO 8601 timestamp of the filing operation.
        related_concept_paths:
            Related concept pages inferred from the consulted paths.
        related_source_paths:
            Related source-summary pages inferred from the consulted paths.

    Notes:
        This record supports the filing operation:
        \[
        \mathcal{K}^{(t+1)} \leftarrow \mathcal{K}^{(t)} \cup
        \{v_{\text{artifact}}\}
        \]
        by materializing the metadata required to insert
        \(v_{\text{artifact}}\) into the wiki graph.
    """

    # Store the deterministic artifact identifier.
    artifact_id: str

    # Store the artifact type label.
    artifact_type: str

    # Store the canonical repo-relative node path.
    canonical_path: str

    # Store the human-readable artifact title.
    title: str

    # Store the originating user query.
    originating_query: str

    # Store the ordered list of consulted wiki paths.
    consulted_paths: List[str]

    # Store the UTC ISO 8601 filing timestamp.
    render_timestamp_utc: str

    # Store the related concept-page paths.
    related_concept_paths: List[str]

    # Store the related source-summary page paths.
    related_source_paths: List[str]


# Define the helper that injects YAML front matter into a Markdown file.
def _inject_yaml_front_matter(
    file_path: pathlib.Path,
    metadata: Dict[str, Any],
) -> None:
    """
    Inject or update YAML front matter in a Markdown file robustly.

    Purpose:
        Embed structured provenance metadata directly into a Markdown file
        without silently corrupting existing content. This callable prefers
        ``python-frontmatter`` when available and falls back to a constrained
        regex approach when it is not.

    Inputs:
        file_path:
            Absolute path to the Markdown file to update.
        metadata:
            Dictionary of metadata fields to serialize into front matter.

    Processes:
        1. Validate input types.
        2. Read the target file using strict UTF-8 decoding.
        3. If ``python-frontmatter`` is available:
           - parse the file into a Post object,
           - update its metadata,
           - dump it back to a string.
        4. If YAML parsing fails:
           - emit a warning,
           - strip any broken existing front matter,
           - write a clean metadata block followed by the body.
        5. If ``python-frontmatter`` is unavailable:
           - emit a warning,
           - inject or replace front matter via a constrained regex anchored at
             the absolute start of the string.
        6. Write the updated content back to disk.

    Outputs:
        None:
            The target Markdown file is updated in place.

    Raises:
        TypeError:
            Raised if ``file_path`` or ``metadata`` has the wrong type.
        UnicodeDecodeError:
            Propagated if strict UTF-8 decoding fails on the target file.
    """
    # Validate that the file path is a pathlib.Path.
    if not isinstance(file_path, pathlib.Path):
        # Raise a type error because filesystem updates require a pathlib path.
        raise TypeError(
            "_inject_yaml_front_matter expected 'file_path' to be a pathlib.Path."
        )

    # Validate that the metadata payload is a dictionary.
    if not isinstance(metadata, dict):
        # Raise a type error because YAML serialization requires a mapping.
        raise TypeError(
            "_inject_yaml_front_matter expected 'metadata' to be a dictionary."
        )

    # Read the target file using strict UTF-8 decoding to avoid silent corruption.
    content = file_path.read_text(encoding="utf-8", errors="strict")

    # Prefer frontmatter-based parsing when the dependency is available.
    if FRONTMATTER_AVAILABLE:
        try:
            # Parse the Markdown file into a frontmatter Post object.
            post = frontmatter.loads(content)

            # Update the parsed metadata in place.
            post.metadata.update(metadata)

            # Serialize the updated post back to a string.
            new_content = frontmatter.dumps(post)

        # Handle malformed existing YAML front matter gracefully.
        except yaml.YAMLError:
            # Emit a warning because the existing front matter was malformed.
            logger.warning(
                f"Malformed YAML in {file_path}. Overwriting frontmatter."
            )

            # Serialize the replacement metadata to YAML.
            yaml_str = yaml.dump(
                metadata,
                default_flow_style=False,
                sort_keys=False,
            )

            # Remove any existing broken front-matter attempt at the start of the file.
            clean_content = re.sub(
                r"\A---\n.*?\n---\n",
                "",
                content,
                flags=re.DOTALL,
            )

            # Construct the new content with a clean front-matter block.
            new_content = f"---\n{yaml_str}---\n{clean_content}"

    # Fall back to constrained regex injection when python-frontmatter is unavailable.
    else:
        # Emit a warning because the robust front-matter library is unavailable.
        logger.warning(
            "python-frontmatter unavailable. Falling back to strict regex injection."
        )

        # Serialize the metadata dictionary to YAML.
        yaml_str = yaml.dump(
            metadata,
            default_flow_style=False,
            sort_keys=False,
        )

        # Construct the front-matter block string.
        front_matter_block = f"---\n{yaml_str}---\n"

        # Match an existing front-matter block only at the absolute start of the file.
        match = re.match(r"\A---\n.*?\n---\n", content, re.DOTALL)

        # Replace the existing front matter when present.
        if match:
            # Build the updated content by replacing the initial front matter only.
            new_content = front_matter_block + content[match.end():]

        # Otherwise prepend a new front-matter block.
        else:
            # Build the updated content by prepending the metadata block.
            new_content = front_matter_block + content

    # Write the updated Markdown content back to disk as UTF-8.
    file_path.write_text(new_content, encoding="utf-8")


# Define the helper that creates derived-artifact registration records.
def create_derived_artifact_nodes(
    rendering_context: Any,
    qa_context: Any,
    wiki_context: Any,
    repo_root: pathlib.Path,
) -> List[DerivedArtifactRegistrationRecord]:
    """
    Extract provenance metadata and prepare derived artifacts for graph integration.

    Purpose:
        Process successfully rendered artifacts, inject provenance metadata into
        their Markdown representations, and produce registration records ready
        for graph integration. This callable supports the paper's filing rule by
        preparing ``v_artifact`` nodes to be added back into the wiki.

    Inputs:
        rendering_context:
            Artifact-rendering context expected to expose:
            - ``report_package``
            - ``slide_package``
            - ``figure_package``
        qa_context:
            Q&A context expected to expose ``routing_state``.
        wiki_context:
            Wiki context expected to expose ``wiki_graph.graph``.
        repo_root:
            Repository root directory.

    Processes:
        1. Validate input types and required collaborators.
        2. Read the originating query and consulted paths.
        3. Traverse consulted paths in the current graph to classify related
           concept and source pages.
        4. Generate a UTC render timestamp.
        5. Define the nested artifact-processing helper.
        6. For each structurally valid or successfully executed artifact:
           - compute a deterministic artifact id from file bytes,
           - construct structured metadata,
           - inject YAML front matter into Markdown artifacts,
           - create companion Markdown pages for non-Markdown artifacts,
           - append a registration record.
        7. Return the ordered registration records.

    Outputs:
        List[DerivedArtifactRegistrationRecord]:
            Ordered list of artifact registration records ready for graph filing.

    Raises:
        TypeError:
            Raised if ``repo_root`` is not a ``pathlib.Path``.
        ValueError:
            Raised if required collaborator attributes are missing.
    """
    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path derivation requires pathlib semantics.
        raise TypeError(
            "create_derived_artifact_nodes expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that the QA context exposes the routing state.
    if not hasattr(qa_context, "routing_state"):
        # Raise a value error because query provenance is read from routing_state.
        raise ValueError(
            "create_derived_artifact_nodes expected 'qa_context' to expose "
            "'routing_state'."
        )

    # Validate that the wiki context exposes the wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a value error because graph metadata is read from wiki_graph.
        raise ValueError(
            "create_derived_artifact_nodes expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Extract the originating query from the Q&A routing state.
    originating_query = qa_context.routing_state.original_query

    # Extract the consulted page paths from the Q&A routing state.
    consulted_paths = qa_context.routing_state.shortlisted_paths

    # Initialize the related concept-page path accumulator.
    related_concepts: List[str] = []

    # Initialize the related source-summary path accumulator.
    related_sources: List[str] = []

    # Retrieve the underlying graph object.
    graph = wiki_context.wiki_graph.graph

    # Traverse each consulted path to classify related nodes.
    for path in consulted_paths:
        # Only inspect nodes that exist in the graph.
        if path in graph:
            # Retrieve the node metadata payload.
            node_meta = graph.nodes[path].get("metadata")

            # Proceed only when metadata exists.
            if node_meta:
                # Read the normalized page-type value.
                type_val = getattr(node_meta.type, "value", str(node_meta.type))

                # Classify concept pages.
                if type_val == "concept":
                    # Record the related concept path.
                    related_concepts.append(path)

                # Classify source-summary pages.
                elif type_val == "source_summary":
                    # Record the related source path.
                    related_sources.append(path)

    # Import datetime locally exactly as in the original code.
    import datetime

    # Generate the UTC registration timestamp.
    timestamp_utc = datetime.datetime.utcnow().isoformat() + "Z"

    # Initialize the registration-record accumulator.
    records: List[DerivedArtifactRegistrationRecord] = []

    # Define the nested helper that processes one artifact path and type.
    def process_artifact(
        target_path: pathlib.Path,
        artifact_type: str,
        title_prefix: str,
    ) -> None:
        """
        Process one generated artifact into a graph-registration record.

        Inputs:
            target_path:
                Absolute path to the artifact or artifact note.
            artifact_type:
                Logical artifact type label.
            title_prefix:
                Human-readable title prefix.

        Outputs:
            None:
                The helper appends to the enclosing ``records`` list.

        Notes:
            This nested helper preserves the original callable's logic exactly.
        """
        # Read the artifact file bytes from disk.
        content_bytes = target_path.read_bytes()

        # Compute a deterministic artifact identifier from the artifact bytes.
        artifact_id = hashlib.sha256(content_bytes).hexdigest()[:16]

        # Construct the artifact title.
        title = f"{title_prefix}: {originating_query[:50]}..."

        # Build the structured provenance metadata block.
        metadata = {
            "type": "derived_artifact",
            "artifact_type": artifact_type,
            "title": title,
            "originating_query": originating_query,
            "consulted_paths": consulted_paths,
            "related_concepts": related_concepts,
            "related_sources": related_sources,
            "timestamp": timestamp_utc,
        }

        # Derive the canonical repo-relative artifact path string.
        canonical_path_str = str(target_path.relative_to(repo_root))

        # Inject front matter directly into Markdown artifacts.
        if target_path.suffix.lower() == ".md":
            # Update the Markdown artifact in place with structured metadata.
            _inject_yaml_front_matter(target_path, metadata)

        # Create a Markdown companion note for non-Markdown artifacts.
        else:
            # Derive the companion Markdown path by replacing the suffix with .md.
            companion_path = target_path.with_suffix(".md")

            # Use the companion note as the canonical path.
            canonical_path_str = str(companion_path.relative_to(repo_root))

            # Derive the image-relative path for the Markdown embed.
            image_rel_path = target_path.name

            # Construct the companion Markdown content.
            companion_content = (
                f"# {title}\n\n"
                f"![{title}]({image_rel_path})\n\n"
                f"## Provenance\n"
                f"Generated from query: *{originating_query}*\n"
            )

            # Write the companion Markdown note to disk.
            companion_path.write_text(companion_content, encoding="utf-8")

            # Inject structured front matter into the companion note.
            _inject_yaml_front_matter(companion_path, metadata)

        # Append the derived-artifact registration record to the output list.
        records.append(
            DerivedArtifactRegistrationRecord(
                artifact_id=artifact_id,
                artifact_type=artifact_type,
                canonical_path=canonical_path_str,
                title=title,
                originating_query=originating_query,
                consulted_paths=consulted_paths,
                render_timestamp_utc=timestamp_utc,
                related_concept_paths=related_concepts,
                related_source_paths=related_sources,
            )
        )

    # Process the report artifact only when it exists and passed structural validation.
    if (
        rendering_context.report_package
        and rendering_context.report_package.is_structurally_valid
    ):
        # Process the structurally valid Markdown report artifact.
        process_artifact(
            rendering_context.report_package.target_path,
            "markdown_report",
            "Report",
        )

    # Process the slide artifact only when it exists and passed structural validation.
    if (
        rendering_context.slide_package
        and rendering_context.slide_package.is_structurally_valid
    ):
        # Process the structurally valid Marp slide artifact.
        process_artifact(
            rendering_context.slide_package.target_path,
            "marp_slides",
            "Slides",
        )

    # Process the figure artifact only when it exists and executed successfully.
    if (
        rendering_context.figure_package
        and rendering_context.figure_package.execution_success
    ):
        # Process the successfully rendered figure artifact.
        process_artifact(
            rendering_context.figure_package.image_path,
            "matplotlib_figure",
            "Figure",
        )

    # Return the ordered list of registration records.
    return records


# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 2: Apply the filing rule and update the graph
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable graph-update transaction record.
@dataclass(frozen=True)
class GraphUpdateTransaction:
    """
    Immutable record capturing the auditable ledger of the filing-induced graph
    mutation.

    Purpose:
        Store:
        1. the pre-update graph hash,
        2. the list of newly inserted nodes,
        3. the list of injected backlinks,
        4. the list of stale indices requiring refresh.

    Inputs:
        This dataclass is populated by
        ``apply_filing_rule_and_update_graph``.

    Processes:
        No computation is performed here. It is a passive immutable mutation
        ledger.

    Outputs:
        pre_update_graph_hash:
            Deterministic hash of the graph state before filing.
        new_nodes:
            Canonical paths of the newly inserted derived artifacts.
        injected_backlinks:
            Ordered list of backlink insertions as
            ``(target_page_path, artifact_path)`` tuples.
        stale_indices:
            Ordered list of index-page paths that should be refreshed.
    """

    # Store the graph hash before mutation.
    pre_update_graph_hash: str

    # Store the canonical paths of newly inserted artifact nodes.
    new_nodes: List[str]

    # Store the injected backlinks as target-artifact tuples.
    injected_backlinks: List[Tuple[str, str]]

    # Store the paths of stale indices requiring refresh.
    stale_indices: List[str]


# Define the deterministic graph-hash helper.
def _compute_graph_hash(wiki_graph: Any) -> str:
    """
    Compute a deterministic hash of the current WikiGraph state.

    Purpose:
        Produce a cryptographic fingerprint of the in-memory wiki graph topology
        for audit, diff-based verification, and transaction tracking. The hash
        is computed from a canonical JSON serialization of:
        - sorted node identifiers,
        - sorted edge endpoint pairs.

    Inputs:
        wiki_graph:
            Wiki context-like object expected to expose ``wiki_graph.graph``.

    Processes:
        1. Validate that the supplied object exposes ``wiki_graph``.
        2. Alias the supplied object to the internally referenced name used by
           the original callable.
        3. Extract the graph object.
        4. Build a canonical Python dictionary containing sorted nodes and edges.
        5. Serialize the dictionary using canonical JSON.
        6. Compute and return the SHA-256 digest of the canonical JSON string.

    Outputs:
        str:
            SHA-256 hexadecimal digest of the graph topology state.

    Raises:
        TypeError:
            Raised if the supplied object does not expose ``wiki_graph``.
    """
    # Validate that the supplied object exposes a wiki_graph attribute.
    if not hasattr(wiki_graph, "wiki_graph"):
        # Raise a type error because graph hashing requires access to the in-memory wiki graph.
        raise TypeError(
            "_compute_graph_hash expected 'wiki_graph' to expose 'wiki_graph'."
        )

    # Alias the supplied object to the internal name used by the original callable.
    wiki_context = wiki_graph

    # Retrieve the underlying graph object from the wiki context.
    graph = wiki_context.wiki_graph.graph

    # Extract nodes and edges into a standard canonical dictionary.
    graph_dict = {
        "nodes": sorted(list(graph.nodes)),
        "edges": sorted([[str(u), str(v)] for u, v, k in graph.edges(keys=True)]),
    }

    # Serialize the graph dictionary using canonical JSON formatting.
    canonical_json_str = json.dumps(
        graph_dict,
        sort_keys=True,
        separators=(",", ":"),
    )

    # Compute and return the SHA-256 digest of the canonical JSON string.
    return hashlib.sha256(
        canonical_json_str.encode("utf-8")
    ).hexdigest()


# Define the backlink-injection helper.
def _inject_backlink(
    target_path: pathlib.Path,
    artifact_path_str: str,
    artifact_title: str
) -> bool:
    """
    Idempotently inject a backlink into a target Markdown page.

    Purpose:
        Append an Obsidian-style wikilink to the ``Derived Artifacts`` or
        related section of a source or concept page without creating duplicate
        backlinks.

    Inputs:
        target_path:
            Absolute path to the target Markdown page receiving the backlink.
        artifact_path_str:
            Repo-relative path of the artifact node or companion page.
        artifact_title:
            Human-readable title of the artifact used as the link alias.

    Processes:
        1. Validate input types.
        2. Return False immediately if the target page does not exist.
        3. Read the target Markdown content.
        4. Construct the backlink wikilink.
        5. Return False if the backlink already exists.
        6. Search for a suitable insertion section:
           - Derived Artifacts
           - Related Pages
           - Related Concepts
        7. Insert the backlink after the section header when found; otherwise
           append a new ``## Derived Artifacts`` section.
        8. Write the updated page back to disk.
        9. Return True on successful injection.

    Outputs:
        bool:
            True if a backlink was injected, False otherwise.

    Raises:
        TypeError:
            Raised if inputs are invalid.
        OSError:
            Propagated if file reads or writes fail.

    Notes:
        This helper preserves idempotency by checking for the exact backlink
        line before writing.
    """
    # Validate that the target path is a pathlib.Path.
    if not isinstance(target_path, pathlib.Path):
        # Raise a type error because file operations require pathlib.Path.
        raise TypeError(
            "_inject_backlink expected 'target_path' to be a pathlib.Path."
        )

    # Validate that the artifact path is a string.
    if not isinstance(artifact_path_str, str):
        # Raise a type error because wikilink construction requires a string.
        raise TypeError(
            "_inject_backlink expected 'artifact_path_str' to be a string."
        )

    # Validate that the artifact title is a string.
    if not isinstance(artifact_title, str):
        # Raise a type error because alias construction requires a string.
        raise TypeError(
            "_inject_backlink expected 'artifact_title' to be a string."
        )

    # Return False immediately if the target page does not exist.
    if not target_path.exists():
        return False

    # Read the existing target-page content.
    content = target_path.read_text(encoding="utf-8", errors="ignore")

    # Strip the .md suffix from the artifact path for cleaner wikilinks.
    clean_path = (
        artifact_path_str[:-3]
        if artifact_path_str.endswith(".md")
        else artifact_path_str
    )

    # Construct the exact backlink wikilink line.
    wikilink = f"- [[{clean_path}|{artifact_title}]]"

    # Preserve idempotency by returning False when the exact backlink already
    # exists.
    if wikilink in content:
        return False

    # Search for a suitable backlink insertion section.
    section_match = re.search(
        r"^(##\s+(?:Derived Artifacts|Related Pages|Related Concepts))\s*$",
        content,
        re.MULTILINE | re.IGNORECASE
    )

    # Insert the backlink after the matched section header when available.
    if section_match:
        # Compute the insertion position immediately after the section header.
        insert_pos = section_match.end()

        # Insert the backlink line after the section header.
        new_content = (
            content[:insert_pos]
            + f"\n{wikilink}"
            + content[insert_pos:]
        )
    else:
        # Append a new Derived Artifacts section at the end of the file.
        new_content = content + f"\n\n## Derived Artifacts\n{wikilink}\n"

    # Write the updated content back to disk.
    target_path.write_text(new_content, encoding="utf-8")

    # Indicate that a backlink was successfully injected.
    return True


# Define the filing-rule application and graph-update callable.
# Define the callable that applies the filing rule and updates the in-memory graph.
def apply_filing_rule_and_update_graph(
    records: List[Any],
    wiki_context: Any,
    repo_root: pathlib.Path,
) -> GraphUpdateTransaction:
    """
    Enforce the filing rule and incrementally update the wiki graph and filesystem.

    Purpose:
        Apply the study's filing rule exactly:

            K^(t+1) ← K^(t) ∪ {v_artifact}

        This callable incrementally mutates the in-memory graph by adding the
        new artifact node and its associated edges directly, rather than forcing
        an unnecessary full graph rebuild.

    Inputs:
        records:
            Ordered list of derived-artifact registration records, expected to
            contain objects compatible with ``DerivedArtifactRegistrationRecord``.
        wiki_context:
            Wiki architecture context exposing:
            - ``wiki_graph.graph``
            - ``wiki_graph.slug_registry``
            - ``wiki_graph._normalize_slug(...)``
        repo_root:
            Project root directory used to resolve absolute filesystem paths.

    Processes:
        1. Validate the top-level input types.
        2. Compute the pre-update graph hash using the deterministic helper.
        3. Initialize transaction ledgers for new nodes and injected backlinks.
        4. Extract the underlying graph and slug registry.
        5. For each registration record:
           a. Construct a ``WikiNodeMetadata`` object.
           b. Insert the artifact node into the graph.
           c. Update the slug registry for O(1) link resolution.
           d. Add forward reference edges to consulted pages.
           e. Inject filesystem backlinks into related concept and source pages.
           f. Add corresponding backlink edges to the in-memory graph when the
              filesystem injection succeeds.
        6. Mark the affected global index pages as stale.
        7. Return the immutable graph-update transaction.

    Outputs:
        GraphUpdateTransaction:
            Immutable ledger of the filing-induced mutation.

    Raises:
        TypeError:
            Raised if:
            - ``records`` is not a list,
            - ``wiki_context`` does not expose ``wiki_graph``,
            - ``repo_root`` is not a ``pathlib.Path``.
        OSError:
            Propagated if filesystem backlink writes fail inside the helper.

    Notes:
        This implementation preserves the original logic exactly, including:
        - incremental in-memory node and edge insertion,
        - slug-registry mutation,
        - idempotent backlink injection mediated by ``_inject_backlink``,
        - fixed stale-index list.
    """
    # Validate that records is a list.
    if not isinstance(records, list):
        # Raise a type error because filing requires an ordered list of records.
        raise TypeError(
            "apply_filing_rule_and_update_graph expected 'records' to be a list."
        )

    # Validate that the wiki context exposes the wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because graph mutation requires the wiki context.
        raise TypeError(
            "apply_filing_rule_and_update_graph expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "apply_filing_rule_and_update_graph expected 'repo_root' to be a "
            "pathlib.Path."
        )

    # Compute the pre-update graph hash using the deterministic helper.
    # This captures the state of K^(t) before applying:
    # K^(t+1) ← K^(t) ∪ {v_artifact}
    pre_update_hash = _compute_graph_hash(wiki_context)

    # Initialize the ordered list of newly inserted node paths.
    new_nodes: List[str] = []

    # Initialize the ordered ledger of injected backlinks.
    injected_backlinks: List[Tuple[str, str]] = []

    # Extract the underlying networkx graph object.
    graph = wiki_context.wiki_graph.graph

    # Extract the slug registry for O(1) path-resolution updates.
    slug_registry = wiki_context.wiki_graph.slug_registry

    # Iterate over all registration records to apply the filing rule incrementally.
    for record in records:
        # Record the canonical path of the new artifact node in the transaction ledger.
        new_nodes.append(record.canonical_path)

        # Construct the metadata object for the new derived-artifact node.
        # The page type is mapped to the formal PageType.DERIVED_ARTIFACT enum.
        node_meta = WikiNodeMetadata(
            path=record.canonical_path,
            type=PageType.DERIVED_ARTIFACT,
            provenance=record.consulted_paths,
            timestamp=record.render_timestamp_utc,
            tags=[],
        )

        # Add the new artifact node directly to the in-memory graph.
        graph.add_node(record.canonical_path, metadata=node_meta)

        # Normalize the artifact title into a stable slug for wikilink resolution.
        normalized_slug = wiki_context.wiki_graph._normalize_slug(record.title)

        # Update the slug registry so the new artifact is resolvable in O(1) time.
        slug_registry[normalized_slug] = record.canonical_path

        # Add directed forward reference edges from the artifact to its consulted pages.
        for consulted_path in record.consulted_paths:
            # Add the edge only when the consulted page already exists in the graph.
            if consulted_path in graph:
                # Insert the forward reference edge with the "references" label.
                graph.add_edge(
                    record.canonical_path,
                    consulted_path,
                    label="references",
                )

        # Combine the related concept and related source paths into a single backlink traversal list.
        all_related_paths = (
            record.related_concept_paths + record.related_source_paths
        )

        # Inject backlinks into each related page on both the filesystem and the in-memory graph.
        for related_path_str in all_related_paths:
            # Resolve the absolute filesystem path of the related page.
            target_abs_path = (repo_root / related_path_str).resolve()

            # Inject the backlink idempotently into the Markdown file.
            injection_success = _inject_backlink(
                target_path=target_abs_path,
                artifact_path_str=record.canonical_path,
                artifact_title=record.title,
            )

            # Update the transaction ledger and in-memory graph only if the filesystem injection succeeded.
            if injection_success:
                # Record the backlink injection in the transaction ledger.
                injected_backlinks.append(
                    (related_path_str, record.canonical_path)
                )

                # Add the reverse reference edge into the in-memory graph when the related page exists.
                if related_path_str in graph:
                    # Insert the backlink edge with the "references" label.
                    graph.add_edge(
                        related_path_str,
                        record.canonical_path,
                        label="references",
                    )

    # Identify stale indices requiring refresh after the filing-induced graph mutation.
    stale_indices = [
        "wiki/index/All Concepts.md",
        "wiki/index/All Sources.md",
        "wiki/index/Derived Artifacts.md",
    ]

    # Construct and return the immutable graph-update transaction.
    return GraphUpdateTransaction(
        pre_update_graph_hash=pre_update_hash,
        new_nodes=new_nodes,
        injected_backlinks=injected_backlinks,
        stale_indices=stale_indices,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 3: Close the main operational loop
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable cycle-state record.
@dataclass(frozen=True)
class CycleStateRecord:
    """
    Immutable record formalizing the compounding cycle state transition.

    Purpose:
        Provide a complete cryptographic and operational audit trail of the
        filing transition from pre-filing graph state to post-filing graph
        state.

    Inputs:
        This dataclass is populated by ``task_15_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable cycle-state
        record.

    Outputs:
        cycle_id:
            Deterministic identifier for this cycle iteration.
        run_id:
            Global run identifier from the governance context.
        pre_filing_graph_hash:
            Graph hash before filing.
        post_filing_graph_hash:
            Graph hash after filing.
        artifacts_filed_count:
            Number of artifacts successfully integrated.
        next_stage_required:
            String flag enforcing the next stage in the pipeline sequence.

    Notes:
        This record closes the operational loop:
        \[
        \mathcal{D}_{\text{raw}}^{(t)}
        \xrightarrow{\Phi_{\text{compile}}}
        \mathcal{K}^{(t)}
        \xrightarrow{\Phi_{\text{qa}}}
        \mathcal{A}^{(t)}
        \xrightarrow{\text{file}}
        \mathcal{K}^{(t+1)}
        \xrightarrow{\Phi_{\text{lint}}}
        \mathcal{K}^{(t+1)}_{\text{corrected}}
        \]
        by explicitly recording the transition into the post-filing knowledge
        state.
    """

    # Store the deterministic cycle identifier.
    cycle_id: str

    # Store the global run identifier.
    run_id: str

    # Store the graph hash before filing.
    pre_filing_graph_hash: str

    # Store the graph hash after filing.
    post_filing_graph_hash: str

    # Store the number of artifacts successfully filed.
    artifacts_filed_count: int

    # Store the enforced next-stage flag.
    next_stage_required: str


# Define the Task 15 orchestrator.
def task_15_orchestrator(
    rendering_context: "ArtifactRenderingContext",
    qa_context: "QAContext",
    wiki_context: "WikiArchitectureContext",
    governance_context: "GovernanceContext",
    repo_root: pathlib.Path
) -> CycleStateRecord:
    """
    Orchestrate Task 15 end-to-end by creating artifact nodes, applying the
    filing rule, and closing the operational loop.

    Purpose:
        Execute the full artifact-filing workflow by:
        1. preparing artifact registration records,
        2. injecting backlinks and rebuilding the graph,
        3. computing pre- and post-filing graph hashes,
        4. producing a cryptographically auditable cycle-state record.

    Inputs:
        rendering_context:
            Context containing the generated artifact packages.
        qa_context:
            Q&A context containing the originating query and consulted paths.
        wiki_context:
            Wiki architecture context exposing the WikiGraph.
        governance_context:
            Governance context exposing the summary accumulator and run ID.
        repo_root:
            Project root directory used for path resolution.

    Processes:
        1. Validate top-level input types.
        2. Step 1: Extract metadata and prepare artifact nodes by calling
           ``create_derived_artifact_nodes``.
        3. If no artifacts are fileable:
           - compute the current graph hash,
           - construct a no-op cycle record with zero filed artifacts.
        4. Step 2: Apply the filing rule and update the graph by calling
           ``apply_filing_rule_and_update_graph``.
        5. Step 3: Compute the post-filing graph hash.
        6. Construct a deterministic cycle ID:
           - for the no-op branch: from run ID and current hash,
           - for the filing branch: from run ID and current UTC timestamp.
        7. Return the immutable cycle-state record with ``next_stage_required =
           "Phi_lint"``.

    Outputs:
        CycleStateRecord:
            Immutable aggregate of the filing-cycle state transition.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.

    Notes:
        This orchestrator closes the main operational loop by converting derived
        artifacts into filed wiki nodes and explicitly directing the pipeline to
        the linting stage.
    """
    # Validate that the rendering context is of the correct type.
    if not isinstance(rendering_context, ArtifactRenderingContext):
        # Raise a type error if the rendering context is invalid.
        raise TypeError(
            "task_15_orchestrator expected 'rendering_context' to be an "
            "ArtifactRenderingContext."
        )

    # Validate that the QA context is of the correct type.
    if not isinstance(qa_context, QAContext):
        # Raise a type error if the QA context is invalid.
        raise TypeError(
            "task_15_orchestrator expected 'qa_context' to be a QAContext."
        )

    # Validate that the wiki architecture context is of the correct type.
    if not isinstance(wiki_context, WikiArchitectureContext):
        # Raise a type error if the wiki context is invalid.
        raise TypeError(
            "task_15_orchestrator expected 'wiki_context' to be a "
            "WikiArchitectureContext."
        )

    # Validate that the governance context is of the correct type.
    if not isinstance(governance_context, GovernanceContext):
        # Raise a type error if the governance context is invalid.
        raise TypeError(
            "task_15_orchestrator expected 'governance_context' to be a "
            "GovernanceContext."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_15_orchestrator expected 'repo_root' to be a pathlib.Path."
        )

    # Step 1: Extract metadata and prepare artifact nodes.
    registration_records = create_derived_artifact_nodes(
        rendering_context=rendering_context,
        qa_context=qa_context,
        wiki_context=wiki_context,
        repo_root=repo_root
    )

    # If no artifacts were generated, produce a no-op cycle-state record.
    if not registration_records:
        # Compute the current graph hash.
        current_hash = _compute_graph_hash(wiki_context)

        # Construct a deterministic no-op cycle ID from the run ID and current
        # graph hash.
        cycle_id = hashlib.md5(
            f"{governance_context.run_id}_{current_hash}".encode("utf-8")
        ).hexdigest()[:12]

        # Return the no-op cycle-state record.
        return CycleStateRecord(
            cycle_id=cycle_id,
            run_id=governance_context.run_id,
            pre_filing_graph_hash=current_hash,
            post_filing_graph_hash=current_hash,
            artifacts_filed_count=0,
            next_stage_required="Phi_lint"
        )

    # Step 2: Inject backlinks and rebuild the graph.
    transaction = apply_filing_rule_and_update_graph(
        records=registration_records,
        wiki_context=wiki_context,
        repo_root=repo_root
    )

    # Step 3: Compute the post-filing graph hash.
    post_update_hash = _compute_graph_hash(wiki_context)

    # Generate a UTC timestamp string for the cycle identifier.
    timestamp_str = datetime.datetime.utcnow().isoformat()

    # Construct the deterministic cycle ID from the run ID and current timestamp.
    cycle_id = hashlib.md5(
        f"{governance_context.run_id}_{timestamp_str}".encode("utf-8")
    ).hexdigest()[:12]

    # Construct and return the immutable cycle-state record.
    return CycleStateRecord(
        cycle_id=cycle_id,
        run_id=governance_context.run_id,
        pre_filing_graph_hash=transaction.pre_update_graph_hash,
        post_filing_graph_hash=post_update_hash,
        artifacts_filed_count=len(registration_records),
        next_stage_required="Phi_lint"
    )
````

````python
# Task 16 — Implement the linting and health-check callable

# ==============================================================================
# Task 16: Implement the linting and health-check callable
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 1: Select Audit Set and Precompute Structural Diagnostics
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable structural-diagnostics record.
@dataclass(frozen=True)
class StructuralDiagnostics:
    """
    Immutable record containing deterministically precomputed structural issues.

    Purpose:
        Store graph-topology issues that can be derived without an LLM, thereby
        conserving tokens and increasing precision for:
        1. broken links,
        2. orphan pages.

    Inputs:
        This dataclass is populated by ``prepare_lint_audit_batch``.

    Processes:
        No computation is performed here. It is a passive immutable diagnostics
        record.

    Outputs:
        broken_links:
            Ordered list of formatted warnings for links targeting unresolved
            placeholder nodes.
        orphan_pages:
            Ordered list of formatted warnings for pages with in-degree 0, after
            excluding explicitly allowed orphan indices.

    Notes:
        These deterministic diagnostics complement Algorithm 5 by offloading
        purely structural checks from the LLM.
    """

    # Store formatted warnings describing broken links.
    broken_links: List[str]

    # Store formatted warnings describing orphan pages.
    orphan_pages: List[str]


# Define the immutable lint-audit batch record.
@dataclass(frozen=True)
class LintAuditBatch:
    """
    Immutable record capturing the selected audit set and precomputed structural
    diagnostics for the linting phase.

    Purpose:
        Package:
        1. the audited wiki pages \(V_{\text{audit}}\),
        2. the deterministic structural diagnostics,

        into a single context object that can be injected into the linting
        prompt.

    Inputs:
        This dataclass is populated by ``prepare_lint_audit_batch``.

    Processes:
        No computation is performed here. It is a passive immutable audit-batch
        record.

    Outputs:
        audited_pages:
            Mapping from canonical repo-relative paths to raw Markdown content.
        structural_diagnostics:
            Deterministically computed broken-link and orphan-page diagnostics.

    Notes:
        This record supports the first step of Algorithm 5:
        \[
        V_{\text{audit}} \subseteq V
        \]
        together with deterministic structural preprocessing.
    """

    # Store the mapping from canonical paths to Markdown content.
    audited_pages: Dict[str, str]

    # Store the precomputed structural diagnostics.
    structural_diagnostics: StructuralDiagnostics


# Define the audit-batch preparation callable.
def prepare_lint_audit_batch(
    wiki_context: "WikiArchitectureContext",
    cycle_state: Optional["CycleStateRecord"] = None
) -> LintAuditBatch:
    """
    Select the audit set and precompute structural diagnostics via graph
    analysis.

    Purpose:
        Prepare the linting input context by:
        1. selecting the wiki pages to audit,
        2. extracting their Markdown content,
        3. mechanically identifying:
           - orphan pages,
           - broken links.

    Inputs:
        wiki_context:
            Wiki architecture context containing the in-memory WikiGraph.
        cycle_state:
            Optional cycle-state record. In the present implementation it is
            accepted for interface compatibility but not used to filter the
            audit set.

    Processes:
        1. Validate top-level inputs and interface availability.
        2. Retrieve the underlying graph object from the WikiGraph.
        3. Iterate over all graph nodes:
           - skip unresolved placeholders,
           - read Markdown content into the audit set,
           - compute in-degree for orphan detection.
        4. Flag orphan pages when:
           - in-degree is 0,
           - path is not in the allowed-orphans set,
           - node is not an unresolved placeholder.
        5. Iterate over all edges and flag edges targeting unresolved placeholder
           nodes as broken links.
        6. Return the immutable audit batch.

    Outputs:
        LintAuditBatch:
            Fully populated audit context for downstream LLM linting.

    Raises:
        TypeError:
            Raised if inputs are invalid or required graph interfaces are
            missing.

    Notes:
        This callable implements the deterministic portion of Algorithm 5 before
        the LLM is invoked.
    """
    # Validate that the wiki context exposes the wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error if the wiki context lacks the required graph.
        raise TypeError(
            "prepare_lint_audit_batch expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that the WikiGraph exposes the underlying graph object.
    if not hasattr(wiki_context.wiki_graph, "graph"):
        # Raise a type error because structural diagnostics require graph access.
        raise TypeError(
            "prepare_lint_audit_batch expected 'wiki_context.wiki_graph' to "
            "expose 'graph'."
        )

    # Validate that the WikiGraph exposes the content retrieval method.
    if not hasattr(wiki_context.wiki_graph, "get_content"):
        # Raise a type error because audit page extraction requires content
        # retrieval.
        raise TypeError(
            "prepare_lint_audit_batch expected 'wiki_context.wiki_graph' to "
            "expose 'get_content'."
        )

    # If a cycle state is supplied, validate that it is of the expected type.
    if cycle_state is not None and not isinstance(cycle_state, CycleStateRecord):
        # Raise a type error because the optional cycle state must match the
        # expected schema.
        raise TypeError(
            "prepare_lint_audit_batch expected 'cycle_state' to be a "
            "CycleStateRecord or None."
        )

    # Extract the underlying in-memory graph.
    graph = wiki_context.wiki_graph.graph

    # Initialize the audit-set mapping from canonical paths to content.
    audited_pages: Dict[str, str] = {}

    # Initialize the broken-link diagnostics list.
    broken_links: List[str] = []

    # Initialize the orphan-page diagnostics list.
    orphan_pages: List[str] = []

    # Define the set of paths allowed to remain orphaned by design.
    allowed_orphans = {
        "wiki/index/All Sources.md",
        "wiki/index/All Concepts.md"
    }

    # Iterate over all nodes in the graph.
    for node in graph.nodes:
        # Skip unresolved placeholder nodes entirely.
        if str(node).startswith("UNRESOLVED:"):
            continue

        # Retrieve the raw Markdown content for the current node.
        content = wiki_context.wiki_graph.get_content(node)

        # Add the node to the audit set when content exists.
        if content:
            audited_pages[str(node)] = content

        # Compute in-degree for orphan detection.
        in_degree = graph.in_degree(node)

        # Flag a page as orphaned when it has no inbound links and is not
        # explicitly whitelisted.
        if in_degree == 0 and str(node) not in allowed_orphans:
            orphan_pages.append(
                f"Orphan Page: '{node}' has no inbound links."
            )

    # Iterate over all graph edges to detect broken links.
    for source, target, key, data in graph.edges(keys=True, data=True):
        # Flag edges whose target is an unresolved placeholder node.
        if str(target).startswith("UNRESOLVED:"):
            # Extract the intended target slug by removing the placeholder
            # prefix.
            intended_target = str(target).replace("UNRESOLVED:", "")

            # Record the broken-link diagnostic.
            broken_links.append(
                f"Broken Link: '{source}' links to unresolved target "
                f"'[[{intended_target}]]'."
            )

    # Construct the immutable structural-diagnostics record.
    diagnostics = StructuralDiagnostics(
        broken_links=broken_links,
        orphan_pages=orphan_pages
    )

    # Construct and return the immutable lint-audit batch.
    return LintAuditBatch(
        audited_pages=audited_pages,
        structural_diagnostics=diagnostics
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 2: Execute LLM Linting and Apply Issue Policies
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable web-imputation provenance record.
@dataclass(frozen=True)
class WebImputationMetadata:
    """
    Immutable record capturing the provenance of web-imputed facts.

    Purpose:
        Store the source URL, snippet, and verification status associated with a
        web-imputed fact proposed during linting.

    Inputs:
        This dataclass is populated by ``execute_llm_linting`` when the raw
        response appears to include web-based imputation.

    Processes:
        No computation is performed here. It is a passive immutable provenance
        record.

    Outputs:
        source_url:
            Source URL cited for the imputed fact.
        snippet:
            Quoted or summarized snippet describing the imputed fact.
        verification_status:
            Verification label such as ``Requires verification``.
    """

    # Store the authoritative or candidate source URL.
    source_url: str

    # Store the exact or summarized snippet associated with the imputation.
    snippet: str

    # Store the verification status string.
    verification_status: str


# Define the immutable lint-issue record.
@dataclass(frozen=True)
class LintIssueRecord:
    """
    Immutable record representing a single detected wiki health issue.

    Purpose:
        Formalize the output of the linting operator so that contradictions,
        missing data, and structural flaws become machine-readable and auditable.

    Inputs:
        This dataclass is populated by ``execute_llm_linting`` and, in fallback
        mode, by ``task_16_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable issue
        record.

    Outputs:
        issue_id:
            Deterministic SHA-256-derived identifier for the issue.
        target_path:
            Canonical repo-relative path of the affected page.
        issue_class:
            Issue category such as ``Contradiction`` or ``Missing Data``.
        severity:
            Assigned severity level.
        evidence_excerpt:
            Exact or approximate evidence excerpt demonstrating the issue.
        proposed_resolution:
            Proposed remediation text.
        web_imputation:
            Optional web-imputation provenance metadata.

    Notes:
        This record contributes to the linting aggregate:
        \[
        R = \bigcup_v r_v
        \]
    """

    # Store the deterministic issue identifier.
    issue_id: str

    # Store the canonical target path of the affected page.
    target_path: str

    # Store the issue category.
    issue_class: str

    # Store the severity level.
    severity: str

    # Store the evidence excerpt.
    evidence_excerpt: str

    # Store the proposed resolution text.
    proposed_resolution: str

    # Store optional web-imputation provenance.
    web_imputation: Optional[WebImputationMetadata] = None


# Define the LLM linting callable.
def execute_llm_linting(
    audit_batch: Any,
    prompt_context: Any,
    contract: Any,
    governance_context: Any,
) -> Tuple[List[LintIssueRecord], str]:
    """
    Invoke the LLM to detect semantic issues and parse the JSON response.

    Purpose:
        Execute the model-based issue-detection portion of Algorithm 5
        (LLM-based linting). In the paper's notation, the linting operator
        returns issue findings and patch proposals:

            (R, P)

        where:
            R = ⋃_v r_v
            P = ⋃_v p_v

        This callable is responsible for extracting the structured issue set
        ``R`` from the model response.

    Inputs:
        audit_batch:
            Audit batch expected to expose:
            - ``structural_diagnostics.broken_links``
            - ``structural_diagnostics.orphan_pages``
            - ``audited_pages`` mapping page paths to content
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        contract:
            Frozen run contract expected to expose ``original_config`` with
            linting policies.
        governance_context:
            Governance context preserved in the signature for warning/accounting
            compatibility.

    Processes:
        1. Validate required collaborator attributes.
        2. Extract linting, contradiction, missing-data, and web-imputation policies.
        3. Assemble the model context, including precomputed structural diagnostics
           and audited page excerpts.
        4. Retrieve the registered linting prompt and effective settings.
        5. Invoke ``query_gpt52`` with optional web search enabled or disabled.
        6. Extract the first balanced JSON object using stack-based brace matching.
        7. Parse the JSON and iterate through the ``issues`` array.
        8. Apply severity-default overrides for contradiction and missing-data issues.
        9. Extract optional web-imputation provenance from the proposed resolution.
        10. Build deterministic issue IDs and return the structured issues together
            with the raw model response.

    Outputs:
        Tuple[List[LintIssueRecord], str]:
            A tuple containing:
            1. The structured lint issue records.
            2. The raw model response text.

    Raises:
        TypeError:
            Raised when required collaborator attributes are missing.

    Notes:
        This callable preserves the original logic while documenting the exact
        deterministic parsing and policy-override behavior.
    """
    # Validate that the prompt context exposes the prompt registry required for prompt lookup.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a type error because lint prompt retrieval depends on prompt_registry.
        raise TypeError(
            "execute_llm_linting expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the contract exposes the original configuration used for lint policies.
    if not hasattr(contract, "original_config"):
        # Raise a type error because lint policy extraction depends on original_config.
        raise TypeError(
            "execute_llm_linting expected 'contract' to expose 'original_config'."
        )

    # Validate that the audit batch exposes structural diagnostics.
    if not hasattr(audit_batch, "structural_diagnostics"):
        # Raise a type error because structural diagnostics are part of the lint context assembly.
        raise TypeError(
            "execute_llm_linting expected 'audit_batch' to expose "
            "'structural_diagnostics'."
        )

    # Validate that the audit batch exposes audited pages.
    if not hasattr(audit_batch, "audited_pages"):
        # Raise a type error because page-content packaging depends on audited_pages.
        raise TypeError(
            "execute_llm_linting expected 'audit_batch' to expose 'audited_pages'."
        )

    # Extract the linting policy block from the frozen configuration.
    lint_policy = contract.original_config.get(
        "linting_thresholds_and_policies",
        {},
    )

    # Extract the contradiction policy block.
    contradiction_policy = lint_policy.get("contradiction_policy", {})

    # Extract the missing-data policy block.
    missing_data_policy = lint_policy.get("missing_data_policy", {})

    # Extract the web-imputation policy block.
    web_imputation_policy = lint_policy.get("web_imputation_policy", {})

    # Enable web search whenever a web-imputation policy block is present and truthy.
    enable_web_search = bool(web_imputation_policy)

    # Read the default contradiction severity, falling back to HIGH.
    default_contradiction_sev = contradiction_policy.get(
        "default_severity",
        "HIGH",
    )

    # Read the default missing-data severity, falling back to MEDIUM.
    default_missing_sev = missing_data_policy.get(
        "default_severity",
        "MEDIUM",
    )

    # Initialize the ordered list of context parts supplied to the linting model call.
    context_parts: List[str] = ["--- PRECOMPUTED STRUCTURAL DIAGNOSTICS ---"]

    # Append broken-link diagnostics when present.
    if audit_batch.structural_diagnostics.broken_links:
        # Add the broken-link diagnostic block to the context package.
        context_parts.append(
            "Broken Links:\n"
            + "\n".join(audit_batch.structural_diagnostics.broken_links)
        )

    # Append orphan-page diagnostics when present.
    if audit_batch.structural_diagnostics.orphan_pages:
        # Add the orphan-page diagnostic block to the context package.
        context_parts.append(
            "Orphan Pages:\n"
            + "\n".join(audit_batch.structural_diagnostics.orphan_pages)
        )

    # Append a delimiter line separating structural diagnostics from page bodies.
    context_parts.append("------------------------------------------\n")

    # Package each audited page into the assembled linting context.
    for path, content in audit_batch.audited_pages.items():
        # Truncate overly long page content to 10,000 characters using the original logic.
        safe_content = (
            content[:10000] + "...[TRUNCATED]"
            if len(content) > 10000
            else content
        )

        # Append the file-delimited page block to the context package.
        context_parts.append(
            f"--- BEGIN FILE: {path} ---\n"
            f"{safe_content}\n"
            f"--- END FILE: {path} ---\n"
        )

    # Join the context parts into a single model-input string.
    assembled_context = "\n".join(context_parts)

    # Retrieve the registered linting prompt entry from the prompt registry.
    lint_prompt_entry = prompt_context.prompt_registry.get(
        "wiki_lint_health_check"
    )

    # Read the effective runtime settings for linting.
    effective_settings = lint_prompt_entry.effective_settings

    # Extract the reasoning-effort setting, defaulting to "high".
    reasoning_effort = effective_settings.get("reasoning_effort", "high")

    # Extract the max-output-token budget, defaulting to 10,000.
    max_tokens = effective_settings.get("max_output_tokens", 10000)

    # Convert tool availability into the string expected by the prompt template.
    tools_str = "web_search" if enable_web_search else "None"

    # Format the user prompt using the assembled batch context and tool availability.
    user_prompt = lint_prompt_entry.prompt_text.format(
        wiki_batch=assembled_context,
        tools_available=tools_str,
    )

    # Define the strict JSON-output system prompt for the linting model call.
    system_prompt = (
        "You are an elite data quality auditor. You MUST output strict JSON. "
        "Do not fabricate evidence. The JSON must contain an 'issues' array and "
        "a 'patches' array. If you use web search to impute missing data, you "
        "MUST label it 'Requires verification' and provide the source URL in "
        "the resolution."
    )

    # Invoke the GPT-5.2 wrapper for lint issue detection.
    llm_result = query_gpt52(
        prompt=user_prompt,
        system_prompt=system_prompt,
        reasoning_effort=reasoning_effort,
        max_output_tokens=max_tokens,
        enable_web_search=enable_web_search,
        log_run_manifest=False,
    )

    # Extract the raw response text from the model wrapper result.
    raw_response = str(llm_result.get("response_text", ""))

    # Initialize the structured lint-issue accumulator.
    issues: List[LintIssueRecord] = []

    # Initialize the candidate JSON substring to empty.
    json_str = ""

    # Initialize the stack used for balanced-brace extraction.
    stack: List[str] = []

    # Initialize the start index of the outermost JSON object.
    start_idx = -1

    # Traverse the raw response text character by character.
    for i, char in enumerate(raw_response):
        # Start or deepen the brace stack when an opening brace is encountered.
        if char == "{":
            # Record the start index when entering the outermost JSON object.
            if not stack:
                # Store the outermost JSON object start position.
                start_idx = i

            # Push the opening brace onto the stack.
            stack.append(char)

        # Pop the brace stack when a closing brace is encountered.
        elif char == "}":
            # Proceed only if the stack is currently non-empty.
            if stack:
                # Pop one opening brace from the stack.
                stack.pop()

                # When the stack becomes empty, a balanced JSON object has been found.
                if not stack:
                    # Extract the balanced JSON substring.
                    json_str = raw_response[start_idx:i + 1]

                    # Stop at the first complete balanced object.
                    break

    # Return early with no issues if no JSON object could be extracted.
    if not json_str:
        # Emit a warning because the response did not contain an extractable JSON payload.
        logger.warning("Failed to extract JSON from lint response.")

        # Return the empty issue list together with the raw response.
        return issues, raw_response

    # Attempt to parse the extracted JSON payload.
    try:
        # Deserialize the extracted JSON string into a Python dictionary.
        parsed_dict = json.loads(json_str)

        # Read the issues array from the parsed payload, defaulting to an empty list.
        issues_array = parsed_dict.get("issues", [])

        # Iterate over each issue object in the issues array.
        for issue_data in issues_array:
            # Read the issue-class label, defaulting to "Unknown".
            issue_class = issue_data.get("issue_class", "Unknown")

            # Read and normalize the severity label, defaulting to INFO.
            severity = issue_data.get("severity", "INFO").upper()

            # Read the target path, defaulting to "Unknown".
            target_path = issue_data.get("target_path", "Unknown")

            # Read the evidence excerpt, defaulting to a placeholder.
            evidence_excerpt = issue_data.get(
                "evidence",
                "Evidence not provided.",
            )

            # Read the proposed resolution, defaulting to a placeholder.
            proposed_resolution = issue_data.get(
                "resolution",
                "Resolution not provided.",
            )

            # Override the severity for contradiction-class issues using policy defaults.
            if "contradiction" in issue_class.lower():
                # Apply the configured contradiction default severity.
                severity = default_contradiction_sev

            # Override the severity for missing-data-class issues using policy defaults.
            elif "missing" in issue_class.lower():
                # Apply the configured missing-data default severity.
                severity = default_missing_sev

            # Initialize optional web-imputation metadata as absent.
            web_imputation = None

            # Extract web-imputation provenance only when web search is enabled and the resolution contains a URL.
            if enable_web_search and "http" in proposed_resolution:
                # Extract the first URL from the proposed resolution.
                url_match = re.search(
                    r"(https?://[^\s]+)",
                    proposed_resolution,
                )

                # Build web-imputation metadata when a URL is present.
                if url_match:
                    # Construct the structured web-imputation provenance record.
                    web_imputation = WebImputationMetadata(
                        source_url=url_match.group(1),
                        snippet="Extracted from LLM resolution.",
                        verification_status="Requires verification",
                    )

            # Construct the deterministic issue-ID input string.
            issue_id_str = f"{target_path}_{issue_class}_{evidence_excerpt}"

            # Compute the deterministic SHA-256-based issue ID.
            issue_id = hashlib.sha256(
                issue_id_str.encode("utf-8")
            ).hexdigest()[:16]

            # Append the structured lint issue record to the output list.
            issues.append(
                LintIssueRecord(
                    issue_id=issue_id,
                    target_path=target_path,
                    issue_class=issue_class,
                    severity=severity,
                    evidence_excerpt=evidence_excerpt,
                    proposed_resolution=proposed_resolution,
                    web_imputation=web_imputation,
                )
            )

    # Handle JSON parsing failure gracefully.
    except json.JSONDecodeError:
        # Emit a warning because the extracted JSON payload could not be parsed.
        logger.warning("Failed to parse extracted JSON from lint response.")

    # Return the structured lint issues together with the raw response text.
    return issues, raw_response


# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 3: Extract Minimal Patches and Generate Final Report
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable minimal-patch record.
@dataclass(frozen=True)
class LintPatchRecord:
    """
    Immutable record representing a proposed minimal patch to resolve a lint
    issue.

    Purpose:
        Store a structured Find/Replace patch instruction while enforcing the
        policy that lint-generated patches must remain gated behind human review.

    Inputs:
        This dataclass is populated by ``extract_minimal_patches``.

    Processes:
        No computation is performed here. It is a passive immutable patch
        record.

    Outputs:
        patch_id:
            Deterministic SHA-256-derived patch identifier.
        target_path:
            Canonical repo-relative path of the page to patch.
        find_anchor:
            Exact text anchor that should be replaced.
        replace_text:
            Proposed replacement text.
        status:
            Validation status string, initialized to REQUIRES_REVIEW.

    Notes:
        This record contributes to the linting patch aggregate:
        \[
        P = \bigcup_v p_v
        \]
    """

    # Store the deterministic patch identifier.
    patch_id: str

    # Store the canonical target page path.
    target_path: str

    # Store the exact Find anchor.
    find_anchor: str

    # Store the replacement text.
    replace_text: str

    # Store the patch status, always initialized to a review-gated value.
    status: str


# Define the immutable aggregate lint report.
@dataclass(frozen=True)
class LintReport:
    """
    Immutable aggregate record containing all detected issues and proposed
    patches.

    Purpose:
        Formalize the final output of the linting operator as:
        \[
        R = \bigcup_v r_v,\qquad
        P = \bigcup_v p_v
        \]

    Inputs:
        This dataclass is populated by ``task_16_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregate
        container.

    Outputs:
        issues:
            Aggregate list of detected lint issues.
        patches:
            Aggregate list of proposed minimal patches.
    """

    # Store the aggregate issue set R.
    issues: List[LintIssueRecord]

    # Store the aggregate patch set P.
    patches: List[LintPatchRecord]


# Define the patch-extraction callable.
def extract_minimal_patches(
    raw_response: str,
    governance_context: Any,
) -> List[LintPatchRecord]:
    """
    Parse the JSON lint response to extract structured Find/Replace patch instructions.

    Purpose:
        Extract the patch-set component ``P`` from the linting model response,
        where the paper's linting operator returns:

            (R, P)

        This callable uses the same strict JSON extraction strategy as the issue
        parser and enforces the minimal-patch heuristic before assigning every
        patch the ``REQUIRES_REVIEW`` status.

    Inputs:
        raw_response:
            Raw string returned by the linting model call.
        governance_context:
            Governance context expected to expose ``summary_accumulator``.

    Processes:
        1. Validate input types.
        2. Extract the first balanced JSON object from the raw response using
           stack-based brace matching.
        3. Parse the JSON and iterate through the ``patches`` array.
        4. Enforce the minimal-patch heuristic by flagging replacements longer
           than 2,000 characters.
        5. Compute deterministic patch IDs.
        6. Assign each patch the ``REQUIRES_REVIEW`` status.
        7. Return the structured patch records.

    Outputs:
        List[LintPatchRecord]:
            Structured patch records ready for review.

    Raises:
        TypeError:
            Raised if ``raw_response`` is not a string.
    """
    # Validate that the raw response is a string.
    if not isinstance(raw_response, str):
        # Raise a type error because JSON extraction requires a string payload.
        raise TypeError(
            "extract_minimal_patches expected 'raw_response' to be a string."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a type error because warning accounting depends on summary_accumulator.
        raise TypeError(
            "extract_minimal_patches expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Initialize the structured patch accumulator.
    patches: List[LintPatchRecord] = []

    # Initialize the candidate JSON substring to empty.
    json_str = ""

    # Initialize the brace stack for balanced-JSON extraction.
    stack: List[str] = []

    # Initialize the start index of the outermost JSON object.
    start_idx = -1

    # Traverse the raw response character by character.
    for i, char in enumerate(raw_response):
        # Start or deepen the brace stack on opening braces.
        if char == "{":
            # Record the outermost object start index.
            if not stack:
                # Store the starting position of the outermost JSON object.
                start_idx = i

            # Push the opening brace onto the stack.
            stack.append(char)

        # Pop the stack on closing braces.
        elif char == "}":
            # Proceed only if the stack is non-empty.
            if stack:
                # Pop one opening brace from the stack.
                stack.pop()

                # When the stack becomes empty, a complete balanced JSON object has been found.
                if not stack:
                    # Extract the balanced JSON substring.
                    json_str = raw_response[start_idx:i + 1]

                    # Stop at the first complete balanced object.
                    break

    # Return early when no JSON object could be extracted.
    if not json_str:
        # Increment warnings because patch extraction failed.
        governance_context.summary_accumulator.increment_warnings()

        # Emit a warning describing the extraction failure.
        logger.warning("Failed to extract JSON for patches.")

        # Return the empty patch list.
        return patches

    # Attempt to parse the extracted JSON payload.
    try:
        # Deserialize the extracted JSON string into a Python dictionary.
        parsed_dict = json.loads(json_str)

        # Read the patches array from the parsed payload, defaulting to an empty list.
        patches_array = parsed_dict.get("patches", [])

        # Iterate over each patch object in the patches array.
        for patch_data in patches_array:
            # Read the target path, defaulting to "Unknown".
            target_path = patch_data.get("target_path", "Unknown")

            # Read the find-anchor text, defaulting to an empty string.
            find_anchor = patch_data.get("find", "")

            # Read the replacement text, defaulting to an empty string.
            replace_text = patch_data.get("replace", "")

            # Enforce the minimal-patch heuristic on replacement size.
            if len(replace_text) > 2000:
                # Increment warnings because the patch looks like a rewrite attempt.
                governance_context.summary_accumulator.increment_warnings()

                # Emit a warning describing the excessive patch size.
                logger.warning(
                    f"Patch for {target_path} exceeds 2000 characters. "
                    f"Flagged as full rewrite attempt."
                )

            # Construct the deterministic patch-ID input string.
            patch_id_str = f"{target_path}_{find_anchor}_{replace_text}"

            # Compute the deterministic SHA-256-based patch ID.
            patch_id = hashlib.sha256(
                patch_id_str.encode("utf-8")
            ).hexdigest()[:16]

            # Append the structured patch record with REQUIRES_REVIEW status.
            patches.append(
                LintPatchRecord(
                    patch_id=patch_id,
                    target_path=target_path,
                    find_anchor=find_anchor,
                    replace_text=replace_text,
                    status="REQUIRES_REVIEW",
                )
            )

    # Handle JSON parsing failure gracefully.
    except json.JSONDecodeError:
        # Increment warnings because the extracted JSON payload could not be parsed.
        governance_context.summary_accumulator.increment_warnings()

        # Emit a warning describing the parsing failure.
        logger.warning("Failed to parse extracted JSON for patches.")

    # Return the structured patch records.
    return patches



# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable linting-context aggregate.
@dataclass(frozen=True)
class LintingContext:
    """
    Immutable record encapsulating the complete linting and health-check
    transaction.

    Purpose:
        Package:
        1. the selected audit batch,
        2. the final aggregate lint report,

        into one immutable object for downstream review, patch application, and
        evaluation.

    Inputs:
        This dataclass is populated by ``task_16_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        audit_batch:
            Audit context containing audited pages and structural diagnostics.
        lint_report:
            Aggregate report containing issues \(R\) and patches \(P\).
    """

    # Store the prepared audit batch.
    audit_batch: LintAuditBatch

    # Store the final aggregate lint report.
    lint_report: LintReport


# Define the Task 16 orchestrator.
def task_16_orchestrator(
    wiki_context: "WikiArchitectureContext",
    prompt_context: "PromptContext",
    contract: "FrozenRunContract",
    governance_context: "GovernanceContext",
    cycle_state: Optional["CycleStateRecord"] = None
) -> LintingContext:
    """
    Orchestrate Task 16 end-to-end by selecting the audit set, executing LLM
    linting, and extracting minimal patches.

    Purpose:
        Execute the complete linting workflow in sequence while ensuring graceful
        degradation when the model response is unparseable.

    Inputs:
        wiki_context:
            Wiki architecture context exposing the WikiGraph.
        prompt_context:
            Prompt context exposing the linting prompt.
        contract:
            Frozen run contract containing linting policies.
        governance_context:
            Governance context exposing warning and issue counters.
        cycle_state:
            Optional cycle-state record that could, in extended implementations,
            be used to restrict the audit set.

    Processes:
        1. Validate top-level inputs and interface availability.
        2. Step 1: Prepare the audit batch by calling
           ``prepare_lint_audit_batch``.
        3. Initialize empty issue and patch lists.
        4. Attempt Steps 2 and 3:
           - execute LLM linting via ``execute_llm_linting``,
           - extract patches via ``extract_minimal_patches``.
        5. If the LLM execution or parsing fails:
           - increment the warning counter,
           - print a critical warning,
           - fall back to deterministic issue records derived from:
             * orphan pages,
             * broken links.
        6. Increment the global lint-issue counter when issues exist.
        7. Construct the aggregate ``LintReport``.
        8. Return the immutable ``LintingContext``.

    Outputs:
        LintingContext:
            Immutable aggregate containing the audit batch and final lint report.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.

    Notes:
        This orchestrator realizes Algorithm 5 end-to-end, including graceful
        fallback from semantic LLM auditing to deterministic structural auditing
        when required.
    """
    # Validate that the wiki context exposes the wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error if the wiki context is invalid.
        raise TypeError(
            "task_16_orchestrator expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that the prompt context exposes the prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a type error if the prompt context is invalid.
        raise TypeError(
            "task_16_orchestrator expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the contract exposes the original configuration.
    if not hasattr(contract, "original_config"):
        # Raise a type error if the contract is invalid.
        raise TypeError(
            "task_16_orchestrator expected 'contract' to expose "
            "'original_config'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a type error if the governance context is invalid.
        raise TypeError(
            "task_16_orchestrator expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # If a cycle state is supplied, validate that it is of the expected type.
    if cycle_state is not None and not isinstance(cycle_state, CycleStateRecord):
        # Raise a type error because the optional cycle state must match the
        # expected schema.
        raise TypeError(
            "task_16_orchestrator expected 'cycle_state' to be a "
            "CycleStateRecord or None."
        )

    # Step 1: Select the audit set and precompute structural diagnostics.
    audit_batch = prepare_lint_audit_batch(
        wiki_context=wiki_context,
        cycle_state=cycle_state
    )

    # Initialize the issue list.
    issues: List[LintIssueRecord] = []

    # Initialize the patch list.
    patches: List[LintPatchRecord] = []

    # Execute model linting and patch extraction inside a guarded try-except
    # block.
    try:
        # Execute the LLM call and parse the structured issues.
        issues, raw_response = execute_llm_linting(
            audit_batch=audit_batch,
            prompt_context=prompt_context,
            contract=contract,
            governance_context=governance_context
        )

        # Extract minimal patches from the raw Markdown response.
        patches = extract_minimal_patches(
            raw_response=raw_response,
            governance_context=governance_context
        )

    except Exception as e:
        # Increment the warning counter when lint execution or parsing fails.
        governance_context.summary_accumulator.increment_warnings()

        # Emit a critical warning preserving the original behavior.
        print(
            "CRITICAL WARNING: LLM linting execution or parsing failed: "
            f"{e}. Returning structural diagnostics only."
        )

        # Construct deterministic issue records for orphan pages.
        for orphan in audit_batch.structural_diagnostics.orphan_pages:
            # Append the orphan-page issue record.
            issues.append(
                LintIssueRecord(
                    issue_id=hashlib.sha256(
                        orphan.encode("utf-8")
                    ).hexdigest()[:16],
                    target_path="UNKNOWN",
                    issue_class="Structural (Orphan)",
                    severity="MEDIUM",
                    evidence_excerpt=orphan,
                    proposed_resolution=(
                        "Add inbound links from relevant concept pages."
                    )
                )
            )

        # Construct deterministic issue records for broken links.
        for broken in audit_batch.structural_diagnostics.broken_links:
            # Append the broken-link issue record.
            issues.append(
                LintIssueRecord(
                    issue_id=hashlib.sha256(
                        broken.encode("utf-8")
                    ).hexdigest()[:16],
                    target_path="UNKNOWN",
                    issue_class="Structural (Broken Link)",
                    severity="HIGH",
                    evidence_excerpt=broken,
                    proposed_resolution=(
                        "Resolve the link target or remove the link."
                    )
                )
            )

    # Increment the global lint-issue counter when issues exist.
    if issues:
        governance_context.summary_accumulator.increment_lint_issues(
            len(issues)
        )

    # Construct the aggregate lint report.
    lint_report = LintReport(
        issues=issues,
        patches=patches
    )

    # Construct and return the immutable linting context.
    return LintingContext(
        audit_batch=audit_batch,
        lint_report=lint_report
    )
````

````python
# Task 17 — Implement the synthetic data generation callable for the optional fine-tuning extension

# =================================================================================================
# Task 17: Implement the synthetic data generation callable for the optional fine-tuning extension
# =================================================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 1: Select high-quality wiki pages (V_gold)
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable gold-page eligibility table.
@dataclass(frozen=True)
class GoldPageEligibilityTable:
    """
    Immutable record capturing the selected subset of high-quality wiki pages.

    Purpose:
        Formalize the gold-page filtering stage so that only pages meeting
        deterministic quality thresholds enter the synthetic supervision pool
        \(V_{\text{gold}}\).

    Inputs:
        This dataclass is populated by ``select_gold_pages``.

    Processes:
        No computation is performed here. It is a passive immutable eligibility
        record.

    Outputs:
        gold_pages:
            Mapping from canonical repo-relative page paths to raw Markdown
            content.
        page_clusters:
            Mapping from canonical repo-relative page paths to a primary concept
            cluster key used for leakage control during train/validation/test
            splitting.

    Notes:
        This record supports Algorithm 7 by defining the candidate set
        \(V_{\text{gold}} \subseteq V\) from which subsets
        \(S \subseteq V_{\text{gold}}\) are sampled.
    """

    # Store the mapping of canonical page paths to Markdown content.
    gold_pages: Dict[str, str]

    # Store the mapping of canonical page paths to cluster keys.
    page_clusters: Dict[str, str]


# Define the gold-page selection callable.
def select_gold_pages(
    wiki_context: Any,
) -> GoldPageEligibilityTable:
    """
    Select high-quality wiki pages suitable for synthetic supervision.

    Purpose:
        Traverse the wiki graph and deterministically select structurally
        eligible pages to form ``V_gold`` for Algorithm 7. Only source summaries
        and concept pages are considered, and structural completeness is
        validated using AST parsing when available.

    Inputs:
        wiki_context:
            Wiki context expected to expose ``wiki_graph``, whose ``graph`` and
            ``get_content`` methods are used during selection.

    Processes:
        1. Validate required collaborator attributes.
        2. Traverse all graph nodes.
        3. Exclude unresolved placeholders.
        4. Restrict to allowed page types:
           - ``source_summary``
           - ``concept``
        5. Enforce a maximum content-length threshold of 20,000 characters.
        6. If AST parsing is available:
           - verify the presence of ``Provenance`` or ``Canonical References``
             as headings,
           - verify the presence of at least one wikilink citation.
        7. Otherwise fall back to regex-based checks.
        8. Derive the page's primary cluster key from the first tag or the page type.
        9. Return the immutable gold-page eligibility table.

    Outputs:
        GoldPageEligibilityTable:
            Immutable mapping of eligible pages and their cluster assignments.

    Raises:
        TypeError:
            Raised if ``wiki_context`` does not expose ``wiki_graph``.
    """
    # Validate that the wiki context exposes the wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error because graph traversal depends on wiki_graph.
        raise TypeError(
            "select_gold_pages expected 'wiki_context' to expose 'wiki_graph'."
        )

    # Retrieve the underlying graph object from the wiki context.
    graph = wiki_context.wiki_graph.graph

    # Initialize the gold-page content mapping.
    gold_pages: Dict[str, str] = {}

    # Initialize the cluster-key mapping for eligible pages.
    page_clusters: Dict[str, str] = {}

    # Define the allowed page types for eligibility.
    allowed_types = {"source_summary", "concept"}

    # Initialize the AST parser if markdown-it-py is available.
    md = MarkdownIt("commonmark") if MARKDOWN_IT_AVAILABLE else None

    # Traverse every node in the wiki graph.
    for node in graph.nodes:
        # Skip unresolved placeholder nodes.
        if str(node).startswith("UNRESOLVED:"):
            # Continue because unresolved placeholders are not valid training pages.
            continue

        # Retrieve the node metadata payload.
        node_meta = graph.nodes[node].get("metadata")

        # Skip nodes without metadata.
        if not node_meta:
            # Continue because eligibility checks require page metadata.
            continue

        # Read the normalized page-type value from the metadata.
        type_val = getattr(node_meta.type, "value", str(node_meta.type))

        # Restrict eligibility to allowed page types.
        if type_val not in allowed_types:
            # Continue because only source_summary and concept pages are eligible.
            continue

        # Retrieve the page content from the wiki graph.
        content = wiki_context.wiki_graph.get_content(node)

        # Skip empty pages and pages exceeding the maximum length threshold.
        if not content or len(content) > 20000:
            # Continue because oversized or empty pages are excluded from V_gold.
            continue

        # Initialize the required-structure flags.
        has_provenance = False
        has_citations = False

        # Prefer AST-based structural validation when the parser is available.
        if md:
            # Parse the page content into a Markdown token stream.
            tokens = md.parse(content)

            # Compile the wikilink detection regex used on safe text nodes.
            wikilink_pattern = re.compile(r"\[\[.*?\]\]")

            # Traverse the token stream for headings and citations.
            for i, token in enumerate(tokens):
                # Detect required headings such as Provenance or Canonical References.
                if (
                    token.type == "heading_open"
                    and i + 1 < len(tokens)
                    and tokens[i + 1].type == "inline"
                ):
                    # Normalize the heading text for exact comparison.
                    heading_text = tokens[i + 1].content.lower().strip()

                    # Mark the provenance/reference requirement as satisfied when a valid heading is found.
                    if heading_text in ["provenance", "canonical references"]:
                        # Record that the required heading is present.
                        has_provenance = True

                # Detect citations only in inline text nodes.
                if token.type == "inline" and token.children:
                    # Traverse the inline child nodes.
                    for child in token.children:
                        # Restrict citation detection to plain text child nodes.
                        if child.type == "text" and wikilink_pattern.search(child.content):
                            # Record that at least one wikilink citation is present.
                            has_citations = True

                # Exit early once both structural requirements are satisfied.
                if has_provenance and has_citations:
                    # Break because further token traversal is unnecessary.
                    break

        # Fall back to regex-based checks when AST parsing is unavailable.
        else:
            # Emit a warning because regex validation is less robust than AST validation.
            logger.warning(
                "markdown-it-py unavailable. Falling back to regex for structural validation."
            )

            # Check for the required heading names using a multiline heading regex.
            has_provenance = bool(
                re.search(
                    r"^##\s+(?:Provenance|Canonical References)",
                    content,
                    re.MULTILINE | re.IGNORECASE,
                )
            )

            # Check for at least one wikilink citation using regex.
            has_citations = bool(re.search(r"\[\[.*?\]\]", content))

        # Exclude pages that fail either structural requirement.
        if not has_provenance or not has_citations:
            # Continue because structurally incomplete pages are not eligible.
            continue

        # Derive the primary cluster key from the first tag when available, else from the page type.
        primary_cluster = node_meta.tags[0] if node_meta.tags else type_val

        # Record the page content in the gold-page mapping.
        gold_pages[str(node)] = content

        # Record the page cluster key.
        page_clusters[str(node)] = primary_cluster

    # Construct and return the immutable gold-page eligibility table.
    return GoldPageEligibilityTable(
        gold_pages=gold_pages,
        page_clusters=page_clusters,
    )



# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 2: Run Algorithm 7 (Synthetic Supervision Generation)
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable synthetic example record.
@dataclass(frozen=True)
class SyntheticExample:
    """
    Immutable record representing one synthetic supervised example.

    Purpose:
        Represent one validated instance of Algorithm 7 output, corresponding to
        a tuple analogous to:

            (x, y, π)

        where:
            - ``instruction`` and ``input_text`` correspond to the task input,
            - ``output_text`` corresponds to the supervised target,
            - ``citations`` correspond to provenance set π.

    Attributes:
        example_id:
            Deterministic identifier of the example.
        instruction:
            Instruction field of the synthetic example.
        input_text:
            Input field of the synthetic example.
        output_text:
            Output field of the synthetic example.
        citations:
            Citation paths that must be a subset of the sampled page set.
        difficulty:
            Difficulty label.
        notes:
            Additional notes from generation.
        cluster_key:
            Cluster key used for leakage-controlled data splitting.
    """

    # Store the deterministic synthetic-example identifier.
    example_id: str

    # Store the instruction string.
    instruction: str

    # Store the input text.
    input_text: str

    # Store the output text.
    output_text: str

    # Store the citation paths.
    citations: List[str]

    # Store the difficulty label.
    difficulty: str

    # Store notes or generation metadata.
    notes: str

    # Store the cluster key used for splitting.
    cluster_key: str



# Define the immutable synthetic-generation batch record.
@dataclass(frozen=True)
class SyntheticGenerationBatchRecord:
    """
    Immutable provenance record for one synthetic-data generation batch.

    Purpose:
        Capture the validated examples, the raw model response text, and token
        usage for one execution of Algorithm 7.

    Attributes:
        examples:
            Validated synthetic examples generated in the batch.
        raw_response_text:
            Raw model response text emitted by the synthetic-generation call.
        token_usage:
            Token-usage accounting returned by the model wrapper.
    """

    # Store the validated synthetic examples.
    examples: List[SyntheticExample]

    # Store the raw model response text.
    raw_response_text: str

    # Store token-usage accounting for the batch.
    token_usage: Dict[str, int]


# Define the synthetic supervision generation callable.
def generate_synthetic_supervision(
    eligibility_table: GoldPageEligibilityTable,
    prompt_context: Any,
    contract: Any,
    governance_context: Any,
) -> SyntheticGenerationBatchRecord:
    """
    Run Algorithm 7 to generate synthetic supervised examples.

    Purpose:
        Execute a bounded version of Algorithm 7 (Synthetic Supervision
        Generation). The callable samples up to three eligible pages from
        ``V_gold``, formats them into a model prompt, parses strict JSONL
        output, and enforces the grounding rule that citations must be drawn
        only from the sampled subset.

    Inputs:
        eligibility_table:
            Gold-page eligibility table containing selected page contents and
            cluster assignments.
        prompt_context:
            Prompt context expected to expose ``prompt_registry``.
        contract:
            Frozen run contract preserved in the signature for orchestration
            compatibility.
        governance_context:
            Governance context expected to expose ``summary_accumulator``.

    Processes:
        1. Validate input types and required collaborators.
        2. Return an empty batch record when no gold pages are available.
        3. Sample up to three pages from ``V_gold``.
        4. Assemble the page subset into a delimited prompt context.
        5. Retrieve the registered synthetic-data prompt and effective settings.
        6. Invoke ``query_gpt52`` with strict JSONL instructions.
        7. Strip enclosing Markdown code fences before line-by-line parsing.
        8. Parse each JSONL line.
        9. Enforce required fields and citation-list type checks.
        10. Enforce the grounding rule that citations must be a subset of the
            sampled page set.
        11. Derive cluster keys and deterministic example IDs.
        12. Return the immutable batch record.

    Outputs:
        SyntheticGenerationBatchRecord:
            Immutable record containing validated synthetic examples and call
            provenance.

    Raises:
        TypeError:
            Raised if ``eligibility_table`` has the wrong type.
    """
    # Validate that the eligibility table has the expected structural type.
    if not isinstance(eligibility_table, GoldPageEligibilityTable):
        # Raise a type error because synthetic generation depends on the eligibility-table schema.
        raise TypeError(
            "generate_synthetic_supervision expected 'eligibility_table' to be "
            "a GoldPageEligibilityTable."
        )

    # Validate that the prompt context exposes the prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a type error because prompt lookup depends on prompt_registry.
        raise TypeError(
            "generate_synthetic_supervision expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a type error because warning accounting depends on summary_accumulator.
        raise TypeError(
            "generate_synthetic_supervision expected 'governance_context' to "
            "expose 'summary_accumulator'."
        )

    # Extract the list of eligible gold-page paths.
    gold_paths = list(eligibility_table.gold_pages.keys())

    # Return an empty batch record immediately when no gold pages are available.
    if not gold_paths:
        # Emit the empty synthetic-generation batch record.
        return SyntheticGenerationBatchRecord(
            examples=[],
            raw_response_text="",
            token_usage={},
        )

    # Define the task-type distribution used in prompt construction.
    task_types = ["QA", "summarization", "concept linking"]

    # Bound the sample size at three pages.
    subset_size = min(3, len(gold_paths))

    # Randomly sample the page subset S from V_gold using the original logic.
    sampled_paths = random.sample(gold_paths, subset_size)

    # Initialize the ordered context-part accumulator.
    context_parts: List[str] = []

    # Assemble the sampled page subset into a delimited context block.
    for path in sampled_paths:
        # Retrieve the content of the sampled eligible page.
        content = eligibility_table.gold_pages[path]

        # Append the sampled page as a delimited file block.
        context_parts.append(
            f"--- BEGIN FILE: {path} ---\n"
            f"{content}\n"
            f"--- END FILE: {path} ---\n"
        )

    # Join the sampled page blocks into the synthetic-generation context string.
    assembled_context = "\n".join(context_parts)

    # Retrieve the registered synthetic-data-generation prompt entry.
    synth_prompt_entry = prompt_context.prompt_registry.get(
        "synthetic_data_generation"
    )

    # Read the effective runtime settings for synthetic generation.
    effective_settings = synth_prompt_entry.effective_settings

    # Extract the reasoning-effort setting, defaulting to "high".
    reasoning_effort = effective_settings.get("reasoning_effort", "high")

    # Extract the max-output-token budget, defaulting to 12,000.
    max_tokens = effective_settings.get("max_output_tokens", 12000)

    # Format the user prompt with the sampled wiki pages and task-type list.
    user_prompt = synth_prompt_entry.prompt_text.format(
        wiki_pages=assembled_context,
        task_types=", ".join(task_types),
        n_examples=5,
    )

    # Define the strict system prompt enforcing JSONL and grounding constraints.
    system_prompt = (
        "You are an expert machine learning engineer generating synthetic "
        "training data. You must output strict JSONL (one JSON object per "
        "line). Do not fabricate facts. Every example MUST be answerable using "
        "ONLY the provided wiki pages. Citations must exactly match the "
        "provided file paths."
    )

    # Invoke the GPT-5.2 wrapper for synthetic-data generation.
    llm_result = query_gpt52(
        prompt=user_prompt,
        system_prompt=system_prompt,
        reasoning_effort=reasoning_effort,
        max_output_tokens=max_tokens,
        enable_web_search=False,
        log_run_manifest=False,
    )

    # Extract the raw response text from the model wrapper result.
    raw_response = str(llm_result.get("response_text", ""))

    # Extract token-usage metadata from the model wrapper result.
    usage = dict(llm_result.get("usage", {}))

    # Initialize the validated-example accumulator.
    examples: List[SyntheticExample] = []

    # Convert the sampled path list into a set for O(1) grounding checks.
    sampled_paths_set = set(sampled_paths)

    # Strip enclosing Markdown code fences before JSONL parsing.
    clean_response = re.sub(
        r"^```jsonl?\n|\n```$",
        "",
        raw_response,
        flags=re.MULTILINE | re.IGNORECASE,
    ).strip()

    # Iterate over the cleaned response line by line.
    for line in clean_response.split("\n"):
        # Strip leading and trailing whitespace from the current line.
        line = line.strip()

        # Skip empty lines and non-JSON-object lines.
        if not line or not line.startswith("{"):
            # Continue because only JSON object lines are valid JSONL records.
            continue

        # Attempt to parse the JSONL line.
        try:
            # Deserialize the current line into a Python dictionary.
            data = json.loads(line)

            # Define the required synthetic-example JSON fields.
            required_fields = [
                "instruction",
                "input",
                "output",
                "citations",
                "difficulty",
                "notes",
            ]

            # Skip records missing any required field.
            if not all(k in data for k in required_fields):
                # Continue because incomplete examples violate the schema contract.
                continue

            # Read the citation list from the parsed object.
            citations = data.get("citations", [])

            # Skip records whose citations field is not a list.
            if not isinstance(citations, list):
                # Continue because citations must be represented as a list.
                continue

            # Initialize the grounding-validity flag.
            is_grounded = True

            # Enforce the grounding rule that citations must belong to the sampled subset S.
            for citation in citations:
                # Reject the example when a citation falls outside the sampled subset.
                if citation not in sampled_paths_set:
                    # Mark the example as ungrounded.
                    is_grounded = False

                    # Stop checking further citations for this example.
                    break

            # Reject ungrounded examples and log a warning.
            if not is_grounded:
                # Increment warnings because the example cited pages outside the sampled subset.
                governance_context.summary_accumulator.increment_warnings()

                # Emit a warning naming the hallucinated citations.
                logger.warning(
                    f"Rejected synthetic example due to hallucinated citation: {citations}"
                )

                # Continue to the next JSONL record.
                continue

            # Default the cluster key to "general".
            cluster_key = "general"

            # Derive the cluster key from the first citation when possible.
            if citations and citations[0] in eligibility_table.page_clusters:
                # Use the cluster assignment of the first cited page.
                cluster_key = eligibility_table.page_clusters[citations[0]]

            # Construct the deterministic example-ID input string.
            id_str = (
                f"{data['instruction']}_{data['input']}_{data['output']}"
            )

            # Compute the deterministic SHA-256-based example ID.
            example_id = hashlib.sha256(
                id_str.encode("utf-8")
            ).hexdigest()[:16]

            # Append the validated synthetic example record.
            examples.append(
                SyntheticExample(
                    example_id=example_id,
                    instruction=data["instruction"],
                    input_text=data["input"],
                    output_text=data["output"],
                    citations=citations,
                    difficulty=data["difficulty"],
                    notes=data["notes"],
                    cluster_key=cluster_key,
                )
            )

        # Handle malformed JSONL lines gracefully.
        except json.JSONDecodeError:
            # Increment warnings because a line could not be parsed as JSON.
            governance_context.summary_accumulator.increment_warnings()

            # Emit a warning naming the failure mode.
            logger.warning(
                "Failed to parse JSONL line in synthetic generation."
            )

    # Construct and return the immutable synthetic-generation batch record.
    return SyntheticGenerationBatchRecord(
        examples=examples,
        raw_response_text=raw_response,
        token_usage=usage,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 3: Validate, Deduplicate, and Split the Dataset
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable synthetic-dataset manifest.
@dataclass(frozen=True)
class SyntheticDatasetManifest:
    """
    Immutable record capturing the final deduplicated and split synthetic
    dataset.

    Purpose:
        Formalize the output of dataset construction, ensuring that:
        1. examples are deduplicated,
        2. train/validation/test splits are cluster-disjoint,
        3. serialization artifacts are reproducibly hashed.

    Inputs:
        This dataclass is populated by ``split_and_serialize_dataset``.

    Processes:
        No computation is performed here. It is a passive immutable dataset
        manifest.

    Outputs:
        train_split:
            Training examples.
        val_split:
            Validation examples.
        test_split:
            Test examples.
        dataset_hash:
            SHA-256 hash of the serialized split files.
        manifest_path:
            Absolute path to the dataset manifest file.

    Notes:
        This record supports the leakage-control constraint that cluster overlap
        must be zero across the three dataset splits.
    """

    # Store the training split.
    train_split: List[SyntheticExample]

    # Store the validation split.
    val_split: List[SyntheticExample]

    # Store the test split.
    test_split: List[SyntheticExample]

    # Store the SHA-256 hash of the serialized split files.
    dataset_hash: str

    # Store the absolute path to the serialized dataset manifest.
    manifest_path: pathlib.Path


# Define the dataset splitting and serialization callable.
def split_and_serialize_dataset(
    batch_record: SyntheticGenerationBatchRecord,
    contract: Any,
    repo_root: pathlib.Path,
) -> SyntheticDatasetManifest:
    """
    Deduplicate examples, split by concept cluster, and serialize the dataset.

    Purpose:
        Enforce leakage control by partitioning synthetic examples by cluster
        rather than by individual record. The split policy is taken from the
        frozen contract and the resulting dataset is serialized to JSONL.

    Inputs:
        batch_record:
            Synthetic-generation batch record containing validated examples.
        contract:
            Frozen run contract expected to expose:
            ``original_config.fine_tuning_exploration.train_val_test_split_policy``
        repo_root:
            Repository root directory under which the ``eval/`` directory will be used.

    Processes:
        1. Validate input types and required collaborator attributes.
        2. Deduplicate examples by ``example_id``.
        3. Group examples by ``cluster_key``.
        4. Deterministically shuffle cluster keys using a fixed seed.
        5. Compute exact integer target counts for cluster assignment.
        6. Distribute any remainder to the training target.
        7. Greedily assign whole clusters to train, validation, and test.
        8. Assert cluster disjointness to prevent leakage.
        9. Serialize each split to JSONL under ``eval/``.
        10. Compute the aggregate dataset hash.
        11. Write the manifest JSON file.
        12. Return the immutable dataset manifest.

    Outputs:
        SyntheticDatasetManifest:
            Immutable manifest describing the serialized split dataset.

    Raises:
        TypeError:
            Raised when ``batch_record`` or ``repo_root`` has the wrong type.
        AssertionError:
            Raised when split leakage is detected through overlapping clusters.
    """
    # Validate that the batch record has the expected structural type.
    if not isinstance(batch_record, SyntheticGenerationBatchRecord):
        # Raise a type error because splitting depends on the batch-record schema.
        raise TypeError(
            "split_and_serialize_dataset expected 'batch_record' to be a "
            "SyntheticGenerationBatchRecord."
        )

    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because serialization paths require pathlib semantics.
        raise TypeError(
            "split_and_serialize_dataset expected 'repo_root' to be a "
            "pathlib.Path."
        )

    # Validate that the contract exposes the original configuration.
    if not hasattr(contract, "original_config"):
        # Raise a type error because split policy extraction depends on original_config.
        raise TypeError(
            "split_and_serialize_dataset expected 'contract' to expose "
            "'original_config'."
        )

    # Deduplicate examples by deterministic example ID.
    unique_examples: Dict[str, SyntheticExample] = {
        ex.example_id: ex
        for ex in batch_record.examples
    }

    # Initialize the cluster-to-example-list mapping.
    clusters: Dict[str, List[SyntheticExample]] = {}

    # Group the unique examples by cluster key.
    for ex in unique_examples.values():
        # Initialize the cluster bucket on first encounter.
        if ex.cluster_key not in clusters:
            # Create the empty cluster list.
            clusters[ex.cluster_key] = []

        # Append the example to its cluster bucket.
        clusters[ex.cluster_key].append(ex)

    # Extract the split-policy block from the frozen configuration.
    split_policy = (
        contract.original_config
        .get("fine_tuning_exploration", {})
        .get("train_val_test_split_policy", {})
    )

    # Read the training fraction, defaulting to 0.80.
    train_frac = split_policy.get("train_fraction", 0.80)

    # Read the validation fraction, defaulting to 0.10.
    val_frac = split_policy.get("val_fraction", 0.10)

    # Sort the cluster keys deterministically before shuffling.
    cluster_keys = sorted(list(clusters.keys()))

    # Initialize a fixed-seed RNG for deterministic shuffling.
    rng = random.Random(42)

    # Shuffle the cluster keys deterministically.
    rng.shuffle(cluster_keys)

    # Compute the number of clusters.
    n_clusters = len(cluster_keys)

    # Compute the base training target using exact integer truncation.
    target_train = int(n_clusters * train_frac)

    # Compute the base validation target using exact integer truncation.
    target_val = int(n_clusters * val_frac)

    # Compute the base test target using exact integer truncation.
    target_test = int(n_clusters * (1.0 - train_frac - val_frac))

    # Compute the unassigned remainder after truncation.
    remainder = n_clusters - (target_train + target_val + target_test)

    # Assign any remainder to the training target to ensure complete coverage.
    target_train += remainder

    # Initialize the training-split accumulator.
    train_split: List[SyntheticExample] = []

    # Initialize the validation-split accumulator.
    val_split: List[SyntheticExample] = []

    # Initialize the test-split accumulator.
    test_split: List[SyntheticExample] = []

    # Initialize the set of clusters assigned to training.
    train_clusters: Set[str] = set()

    # Initialize the set of clusters assigned to validation.
    val_clusters: Set[str] = set()

    # Initialize the set of clusters assigned to test.
    test_clusters: Set[str] = set()

    # Greedily assign whole clusters to splits.
    for key in cluster_keys:
        # Retrieve the examples belonging to the current cluster.
        cluster_examples = clusters[key]

        # Assign the cluster to training while the training target has not been reached.
        if len(train_clusters) < target_train:
            # Extend the training split with the cluster examples.
            train_split.extend(cluster_examples)

            # Record the cluster assignment.
            train_clusters.add(key)

        # Otherwise assign to validation while the validation target has not been reached.
        elif len(val_clusters) < target_val:
            # Extend the validation split with the cluster examples.
            val_split.extend(cluster_examples)

            # Record the cluster assignment.
            val_clusters.add(key)

        # Otherwise assign the cluster to test.
        else:
            # Extend the test split with the cluster examples.
            test_split.extend(cluster_examples)

            # Record the cluster assignment.
            test_clusters.add(key)

    # Assert that training and validation clusters are disjoint.
    assert train_clusters.isdisjoint(val_clusters), (
        "Leakage detected: Train and Val clusters overlap."
    )

    # Assert that training and test clusters are disjoint.
    assert train_clusters.isdisjoint(test_clusters), (
        "Leakage detected: Train and Test clusters overlap."
    )

    # Assert that validation and test clusters are disjoint.
    assert val_clusters.isdisjoint(test_clusters), (
        "Leakage detected: Val and Test clusters overlap."
    )

    # Resolve the eval directory under the repository root.
    eval_dir = (repo_root / "eval").resolve()

    # Ensure the eval directory exists.
    eval_dir.mkdir(parents=True, exist_ok=True)

    # Resolve the dataset manifest path.
    manifest_path = eval_dir / "synthetic_dataset_manifest.json"

    # Define the nested JSONL writer helper.
    def write_jsonl(
        file_path: pathlib.Path,
        split_data: List[SyntheticExample],
    ) -> None:
        """
        Serialize one split to JSONL.

        Inputs:
            file_path:
                Absolute output file path.
            split_data:
                Ordered list of synthetic examples to serialize.

        Outputs:
            None:
                The side effect is writing the JSONL file to disk.
        """
        # Open the target JSONL file for UTF-8 writing.
        with open(file_path, "w", encoding="utf-8") as f:
            # Serialize each example as one JSON object per line.
            for ex in split_data:
                # Write the canonical JSON representation followed by a newline.
                f.write(json.dumps(asdict(ex), sort_keys=True) + "\n")

    # Resolve the training JSONL path.
    train_path = eval_dir / "train.jsonl"

    # Resolve the validation JSONL path.
    val_path = eval_dir / "val.jsonl"

    # Resolve the test JSONL path.
    test_path = eval_dir / "test.jsonl"

    # Write the training split to disk.
    write_jsonl(train_path, train_split)

    # Write the validation split to disk.
    write_jsonl(val_path, val_split)

    # Write the test split to disk.
    write_jsonl(test_path, test_split)

    # Initialize the aggregate dataset hasher.
    hasher = hashlib.sha256()

    # Hash the serialized split files in deterministic order.
    for path in [train_path, val_path, test_path]:
        # Update the aggregate hash with the raw bytes of the split file.
        hasher.update(path.read_bytes())

    # Finalize the aggregate dataset hash.
    dataset_hash = hasher.hexdigest()

    # Build the manifest data dictionary.
    manifest_data = {
        "dataset_hash": dataset_hash,
        "train_examples": len(train_split),
        "val_examples": len(val_split),
        "test_examples": len(test_split),
        "train_clusters": list(train_clusters),
        "val_clusters": list(val_clusters),
        "test_clusters": list(test_clusters),
    }

    # Write the manifest JSON file to disk.
    with open(manifest_path, "w", encoding="utf-8") as f:
        # Serialize the manifest with stable ordering and indentation.
        json.dump(manifest_data, f, indent=2, sort_keys=True)

    # Construct and return the immutable synthetic-dataset manifest.
    return SyntheticDatasetManifest(
        train_split=train_split,
        val_split=val_split,
        test_split=test_split,
        dataset_hash=dataset_hash,
        manifest_path=manifest_path,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable synthetic-data generation context aggregate.
@dataclass(frozen=True)
class SyntheticDataGenerationContext:
    """
    Immutable record encapsulating the complete synthetic-data generation
    transaction.

    Purpose:
        Package:
        1. the selected gold-page table,
        2. the synthetic generation batch record,
        3. the final dataset manifest,

        into one immutable object for downstream fine-tuning and evaluation.

    Inputs:
        This dataclass is populated by ``task_17_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        eligibility_table:
            Gold-page selection result.
        batch_record:
            Synthetic generation batch provenance.
        dataset_manifest:
            Final deduplicated and split dataset manifest.
    """

    # Store the gold-page eligibility table.
    eligibility_table: Optional[GoldPageEligibilityTable] = None

    # Store the synthetic generation batch record.
    batch_record: Optional[SyntheticGenerationBatchRecord] = None

    # Store the final dataset manifest.
    dataset_manifest: Optional[SyntheticDatasetManifest] = None


# Define the Task 17 orchestrator.
def task_17_orchestrator(
    wiki_context: "WikiArchitectureContext",
    prompt_context: "PromptContext",
    contract: "FrozenRunContract",
    governance_context: "GovernanceContext",
    repo_root: pathlib.Path
) -> SyntheticDataGenerationContext:
    """
    Orchestrate Task 17 end-to-end by selecting gold pages, generating
    synthetic examples, and splitting the dataset.

    Purpose:
        Execute the complete synthetic-data generation pipeline conditionally,
        only when the fine-tuning exploration branch is enabled.

    Inputs:
        wiki_context:
            Wiki architecture context exposing the WikiGraph.
        prompt_context:
            Prompt context exposing the synthetic-generation prompt.
        contract:
            Frozen run contract containing the fine-tuning and split policies.
        governance_context:
            Governance context exposing warning and run-summary tracking.
        repo_root:
            Project root directory used for path resolution.

    Processes:
        1. Validate top-level inputs and interface availability.
        2. Check whether fine-tuning exploration is enabled in the contract.
        3. If disabled:
           - log a no-op informational message,
           - return an empty context.
        4. Step 1: Select high-quality gold pages by calling
           ``select_gold_pages``.
        5. Step 2: Generate synthetic examples by calling
           ``generate_synthetic_supervision``.
        6. Step 3: If examples exist, split and serialize the dataset by
           calling ``split_and_serialize_dataset``.
        7. Return the immutable aggregate context.

    Outputs:
        SyntheticDataGenerationContext:
            Immutable aggregate containing the full synthetic-data generation
            transaction.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.

    Notes:
        This orchestrator realizes the end-to-end optional synthetic-data branch
        that prepares \(\mathcal{D}_{\text{syn}}\) for downstream fine-tuning.
    """
    # Validate that the wiki context exposes the wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error if the wiki context is invalid.
        raise TypeError(
            "task_17_orchestrator expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate that the prompt context exposes the prompt registry.
    if not hasattr(prompt_context, "prompt_registry"):
        # Raise a type error if the prompt context is invalid.
        raise TypeError(
            "task_17_orchestrator expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate that the contract exposes the original configuration.
    if not hasattr(contract, "original_config"):
        # Raise a type error if the contract is invalid.
        raise TypeError(
            "task_17_orchestrator expected 'contract' to expose "
            "'original_config'."
        )

    # Validate that the governance context exposes the summary accumulator.
    if not hasattr(governance_context, "summary_accumulator"):
        # Raise a type error if the governance context is invalid.
        raise TypeError(
            "task_17_orchestrator expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate that repo_root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "task_17_orchestrator expected 'repo_root' to be a pathlib.Path."
        )

    # Extract the fine-tuning exploration configuration block.
    ft_config = contract.original_config.get("fine_tuning_exploration", {})

    # Determine whether synthetic generation is enabled.
    is_enabled = ft_config.get("enabled", False)

    # Return an empty context when the synthetic-data branch is disabled.
    if not is_enabled:
        # Emit a no-op informational message preserving the original behavior.
        print(
            "INFO: Synthetic data generation is disabled in the configuration. "
            "Skipping Task 17."
        )

        # Return the empty synthetic-data generation context.
        return SyntheticDataGenerationContext()

    # Step 1: Select high-quality wiki pages.
    eligibility_table = select_gold_pages(wiki_context=wiki_context)

    # Step 2: Generate synthetic examples.
    batch_record = generate_synthetic_supervision(
        eligibility_table=eligibility_table,
        prompt_context=prompt_context,
        contract=contract,
        governance_context=governance_context
    )

    # Initialize the dataset manifest to None.
    dataset_manifest = None

    # Step 3: Deduplicate and split the dataset when examples exist.
    if batch_record.examples:
        dataset_manifest = split_and_serialize_dataset(
            batch_record=batch_record,
            contract=contract,
            repo_root=repo_root
        )

    # Construct and return the immutable synthetic-data generation context.
    return SyntheticDataGenerationContext(
        eligibility_table=eligibility_table,
        batch_record=batch_record,
        dataset_manifest=dataset_manifest
    )
````

````python
# Task 18 — Implement the fine-tuning callable

# ==============================================================================
# Task 18: Implement the fine-tuning callable
# ==============================================================================

# Configure a module-level logger for structured output.
# Construct a named logger for this fine-tuning pipeline.
logger: logging.Logger = logging.getLogger("gpt_oss_20b_sft_lora")

# Only attach handlers when the logger has not already been configured.
if not logger.handlers:
    # Create a stream handler targeting standard error.
    handler = logging.StreamHandler()

    # Define a structured formatter with UTC-like timestamp rendering.
    formatter = logging.Formatter(
        "%(asctime)s | %(levelname)-8s | %(name)s | %(message)s",
        datefmt="%Y-%m-%dT%H:%M:%S"
    )

    # Attach the formatter to the handler.
    handler.setFormatter(formatter)

    # Attach the handler to the logger.
    logger.addHandler(handler)

    # Set the logger level to INFO.
    logger.setLevel(logging.INFO)


# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 1: Prepare the fine-tuning base honestly (Environment, Model, LoRA)
# -------------------------------------------------------------------------------------------------------------------------------

# Define the minimum package versions required for reproducible execution.
REQUIRED_PACKAGES: Dict[str, str] = {
    "torch": "2.4.1",
    "transformers": "4.46.0",
    "peft": "0.13.0",
    "trl": "0.11.1",
    "accelerate": "0.34.2",
    "datasets": "3.0.1",
}


# Define the dependency-audit callable.
def audit_dependencies(required: Dict[str, str]) -> None:
    """
    Audit that all required packages are installed at or above the minimum versions.

    Purpose:
        Enforce reproducible computational-research discipline by verifying that
        the execution environment is sufficiently close to the reference package
        substrate before fine-tuning proceeds.

        This callable is part of Task 18's honest environment-preparation step:
        it does not infer unavailable architecture internals; it only validates
        the software substrate needed to execute the fine-tuning workflow.

    Inputs:
        required:
            Mapping from package name to minimum required version string.

    Processes:
        1. Validate that the input is a dictionary.
        2. Import ``Version`` from ``packaging.version`` for strict PEP 440
           semantic-version comparison.
        3. Iterate over each required package and minimum version.
        4. Resolve the installed version using ``importlib.metadata.version``.
        5. Compare the installed version to the required minimum using
           ``Version(installed) < Version(min_required)``.
        6. Raise ``ImportError`` if a package is missing or below the minimum.
        7. Emit an INFO log for each successful audit.

    Outputs:
        None:
            The side effects are validation and logging.

    Raises:
        TypeError:
            Raised if ``required`` is not a dictionary.
        ImportError:
            Raised if:
            - ``packaging`` is unavailable,
            - a required package is not installed,
            - a package version is below the required minimum.

    Notes:
        This callable preserves the original logic exactly, including:
        - use of ``importlib.metadata.version`` instead of module import,
        - strict semantic version comparison via ``packaging.version.Version``,
        - INFO-level pass logging for each audited package.
    """
    # Validate that the required package specification is a dictionary.
    if not isinstance(required, dict):
        # Raise a type error because dependency auditing requires a name-to-version mapping.
        raise TypeError(
            "audit_dependencies expected 'required' to be a dictionary."
        )

    # Import the Version comparator locally for strict semantic version comparison.
    try:
        # Import the Version class used for PEP 440-compliant comparisons.
        from packaging.version import Version

    # Raise a wrapped ImportError if the packaging library is unavailable.
    except ImportError as exc:
        # Re-raise with a clearer dependency-audit message.
        raise ImportError(
            "The 'packaging' library is required for dependency auditing."
        ) from exc

    # Iterate over each required package and its minimum version.
    for package, min_version in required.items():
        try:
            # Resolve the installed version string from distribution metadata
            # without importing the full package into memory.
            installed_version: str = importlib.metadata.version(package)

        # Raise a wrapped ImportError when the package is not installed.
        except importlib.metadata.PackageNotFoundError as exc:
            # Re-raise with a precise missing-package message.
            raise ImportError(
                f"Required package '{package}' is not installed."
            ) from exc

        # Compare the installed version to the minimum required version using
        # strict semantic version ordering.
        if Version(installed_version) < Version(min_version):
            # Raise an ImportError because the installed version is too old to
            # guarantee reproducibility of the fine-tuning environment.
            raise ImportError(
                f"Package '{package}' version {installed_version} is below "
                f"minimum required {min_version}."
            )

        # Emit an INFO log indicating that the current package satisfies the minimum version.
        logger.info(
            "PASS: %s==%s (required>=%s)",
            package,
            installed_version,
            min_version,
        )


# Define the model-and-tokenizer loading callable.
def load_model_and_tokenizer(
    model_id: str,
    use_4bit_quantization: bool = False,
    seed: int = 42,
) -> Tuple[Any, Any]:
    """
    Load the base model and tokenizer with deterministic settings and hardware checks.

    Purpose:
        Prepare the fine-tuning base model and tokenizer for Task 18 while
        respecting the paper's honesty constraint around architecture details.
        This callable loads the model as specified by the accessible model ID
        without fabricating unavailable internals.

    Inputs:
        model_id:
            Hugging Face or repository model identifier.
        use_4bit_quantization:
            Whether to enable 4-bit NF4 quantization at load time.
        seed:
            Global random seed used to initialize PyTorch RNG state.

    Processes:
        1. Validate dependency availability.
        2. Validate input types.
        3. Fix PyTorch RNG seeds.
        4. Detect hardware support for bfloat16 and choose the safest available dtype.
        5. Construct a ``BitsAndBytesConfig`` when 4-bit quantization is enabled.
        6. Load the base model with:
           - ``torch_dtype=resolved_dtype``
           - ``device_map="auto"``
           - ``quantization_config`` when applicable
           - ``trust_remote_code=True``
           - ``attn_implementation="flash_attention_2"``
        7. Load the tokenizer.
        8. Set pad token to EOS token when needed.
        9. Switch the model to evaluation mode.
        10. Return the model and tokenizer.

    Outputs:
        Tuple[Any, Any]:
            The loaded model and tokenizer. At runtime these correspond to:
            - ``PreTrainedModel``
            - ``PreTrainedTokenizerBase``

    Raises:
        RuntimeError:
            Raised if PyTorch/Transformers/PEFT dependencies are unavailable.
        TypeError:
            Raised if ``model_id`` is not a string.
    """
    # Halt immediately when the machine-learning stack is unavailable.
    if not ML_AVAILABLE:
        # Raise a runtime error because model loading cannot proceed without the ML dependencies.
        raise RuntimeError("PyTorch and Transformers are required for Task 18.")

    # Validate that the model identifier is a string.
    if not isinstance(model_id, str):
        # Raise a type error because model loading requires a string model ID.
        raise TypeError(
            "load_model_and_tokenizer expected 'model_id' to be a string."
        )

    # Validate that the quantization flag is boolean-like.
    if not isinstance(use_4bit_quantization, bool):
        # Raise a type error because quantization enablement is a boolean control.
        raise TypeError(
            "load_model_and_tokenizer expected 'use_4bit_quantization' to be a bool."
        )

    # Validate that the seed is an integer.
    if not isinstance(seed, int):
        # Raise a type error because RNG seeding requires an integer.
        raise TypeError(
            "load_model_and_tokenizer expected 'seed' to be an integer."
        )

    # Fix the global PyTorch RNG seed.
    torch.manual_seed(seed)

    # Fix all CUDA RNG streams when CUDA is available.
    if torch.cuda.is_available():
        # Seed all CUDA devices deterministically.
        torch.cuda.manual_seed_all(seed)

    # Resolve the preferred dtype using hardware capability checks.
    if torch.cuda.is_available() and torch.cuda.is_bf16_supported():
        # Select bfloat16 when the hardware explicitly supports it.
        resolved_dtype = torch.bfloat16

        # Emit an informational log entry describing the dtype choice.
        logger.info("Hardware supports bfloat16. Using torch.bfloat16.")

    # Fall back to float16 when bfloat16 is unavailable.
    else:
        # Select float16 as the fallback dtype.
        resolved_dtype = torch.float16

        # Emit an informational log entry describing the fallback.
        logger.info(
            "Hardware does not support bfloat16. Falling back to torch.float16."
        )

    # Initialize the quantization configuration as absent.
    quantization_config = None

    # Build a BitsAndBytes quantization configuration when 4-bit quantization is requested.
    if use_4bit_quantization:
        # Construct the 4-bit NF4 quantization configuration.
        quantization_config = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_compute_dtype=resolved_dtype,
            bnb_4bit_use_double_quant=True,
            bnb_4bit_quant_type="nf4",
        )

    # Load the base causal language model from the model repository.
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        torch_dtype=resolved_dtype,
        device_map="auto",
        quantization_config=quantization_config,
        trust_remote_code=True,
        attn_implementation="flash_attention_2",
    )

    # Load the associated tokenizer from the model repository.
    tokenizer = AutoTokenizer.from_pretrained(
        model_id,
        trust_remote_code=True,
        use_fast=True,
    )

    # Ensure the tokenizer exposes a pad token by aliasing EOS when needed.
    if tokenizer.pad_token is None:
        # Assign the EOS token string as the pad token.
        tokenizer.pad_token = tokenizer.eos_token

        # Assign the EOS token ID as the pad token ID.
        tokenizer.pad_token_id = tokenizer.eos_token_id

    # Switch the model into evaluation mode.
    model.eval()

    # Return the loaded model and tokenizer.
    return model, tokenizer


# Define the LoRA-configuration builder.
def build_lora_config(
    rank_r: int = 16,
    lora_alpha: int = 32,
    lora_dropout: float = 0.05,
    target_modules: Optional[List[str]] = None,
) -> LoraConfig:
    """
    Construct the PEFT LoRA configuration for ``gpt-oss-20b``.

    Purpose:
        Configure parameter-efficient adaptation while targeting only attention
        and MLP body layers, preserving the integrity of MoE routing behavior by
        excluding router parameters.

    Inputs:
        rank_r:
            LoRA rank \(r\).
        lora_alpha:
            LoRA alpha scaling parameter.
        lora_dropout:
            Dropout applied to LoRA activations.
        target_modules:
            Optional list of target module names. If omitted, the original safe
            defaults are used.

    Processes:
        1. Validate input types.
        2. Supply default safe target modules when none are provided.
        3. Construct the PEFT ``LoraConfig`` object.
        4. Log the resulting configuration.
        5. Return the ``LoraConfig``.

    Outputs:
        LoraConfig:
            Fully constructed LoRA configuration object.

    Raises:
        TypeError:
            Raised if input types are invalid.

    Notes:
        This callable implements the parameter-efficient adaptation setup
        required by the optional fine-tuning extension.
    """
    # Validate that rank_r is an integer.
    if not isinstance(rank_r, int):
        # Raise a type error because LoRA rank must be integral.
        raise TypeError(
            "build_lora_config expected 'rank_r' to be an integer."
        )

    # Validate that lora_alpha is an integer.
    if not isinstance(lora_alpha, int):
        # Raise a type error because LoRA alpha must be integral.
        raise TypeError(
            "build_lora_config expected 'lora_alpha' to be an integer."
        )

    # Validate that lora_dropout is numeric.
    if not isinstance(lora_dropout, (int, float)):
        # Raise a type error because LoRA dropout must be numeric.
        raise TypeError(
            "build_lora_config expected 'lora_dropout' to be numeric."
        )

    # Validate target_modules when provided.
    if target_modules is not None and not isinstance(target_modules, list):
        # Raise a type error because target modules must be a list or None.
        raise TypeError(
            "build_lora_config expected 'target_modules' to be a list of "
            "strings or None."
        )

    # Supply the original safe target-module defaults when none are provided.
    if target_modules is None:
        target_modules = [
            "q_proj", "k_proj", "v_proj", "o_proj",
            "up_proj", "down_proj", "gate_proj",
        ]

    # Construct the LoRA configuration object.
    lora_config: LoraConfig = LoraConfig(
        r=rank_r,
        lora_alpha=lora_alpha,
        lora_dropout=lora_dropout,
        bias="none",
        task_type=TaskType.CAUSAL_LM,
        target_modules=target_modules,
        inference_mode=False,
    )

    # Emit an informational log describing the LoRA configuration.
    logger.info(
        "LoRA config: r=%d | alpha=%d | dropout=%.2f | target_modules=%s",
        rank_r,
        lora_alpha,
        lora_dropout,
        target_modules
    )

    # Return the constructed LoRA configuration.
    return lora_config


# Define the LoRA application callable.
def apply_lora_to_model(
    model: PreTrainedModel,
    lora_config: LoraConfig
) -> PreTrainedModel:
    """
    Wrap the base model with PEFT/LoRA adapters and validate the trainable
    fraction.

    Purpose:
        Inject LoRA adapters into the model and audit the fraction of trainable
        parameters to ensure that parameter-efficient fine-tuning remains
        appropriately lightweight.

    Inputs:
        model:
            Base pretrained causal language model.
        lora_config:
            LoRA configuration describing adapter placement and rank.

    Processes:
        1. Validate input types.
        2. Wrap the model with ``get_peft_model``.
        3. Count trainable parameters.
        4. Count total parameters.
        5. Compute trainable fraction:
           \[
           \text{trainable\_fraction}
           =
           \frac{\text{trainable\_params}}{\text{total\_params}}
           \]
        6. Log the resulting fraction.
        7. Emit a warning if trainable fraction exceeds 5%.
        8. Return the PEFT-wrapped model.

    Outputs:
        PreTrainedModel:
            Model wrapped with LoRA adapters.

    Raises:
        TypeError:
            Raised if input types are invalid.
    """
    # Validate that the model is a PreTrainedModel.
    if not isinstance(model, PreTrainedModel):
        # Raise a type error because PEFT wrapping requires a pretrained model.
        raise TypeError(
            "apply_lora_to_model expected 'model' to be a PreTrainedModel."
        )

    # Validate that the LoRA configuration is a LoraConfig.
    if not isinstance(lora_config, LoraConfig):
        # Raise a type error because PEFT wrapping requires a LoraConfig.
        raise TypeError(
            "apply_lora_to_model expected 'lora_config' to be a LoraConfig."
        )

    # Wrap the base model with LoRA adapters.
    model = get_peft_model(model, lora_config)

    # Count trainable parameters.
    trainable_params: int = sum(
        p.numel() for p in model.parameters() if p.requires_grad
    )

    # Count total parameters.
    total_params: int = sum(p.numel() for p in model.parameters())

    # Compute the trainable parameter fraction.
    trainable_fraction: float = trainable_params / total_params

    # Emit a log summarizing PEFT trainable parameter counts.
    logger.info(
        "PEFT wrapping complete | trainable=%s (%.4f%%) | total=%s",
        f"{trainable_params:,}",
        trainable_fraction * 100,
        f"{total_params:,}"
    )

    # Emit a warning when the trainable fraction exceeds 5%.
    if trainable_fraction > 0.05:
        logger.warning(
            "Trainable fraction %.2f%% > 5%%. Verify target_modules are not "
            "over-specified.",
            trainable_fraction * 100
        )

    # Return the PEFT-wrapped model.
    return model


# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 2: Train under the stated supervised objective
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable SFT-record schema used during JSONL ingestion.
@dataclass(frozen=True)
class SFTRecord:
    """
    Immutable record representing one supervised fine-tuning example.

    Purpose:
        Store the validated schema of one training record before it is converted
        into harmony-format conversations and tokenized text.

    Inputs:
        This dataclass is populated by ``load_and_validate_jsonl``.

    Processes:
        No computation is performed here. It is a passive immutable training
        example record.

    Outputs:
        instruction:
            Task instruction string.
        input:
            Context or input field.
        output:
            Target answer text.
        citations:
            Ordered list of supporting citations.
        difficulty:
            Difficulty label.
        notes:
            Free-form notes field.

    Notes:
        These records parameterize the supervised learning objective:
        \[
        \mathcal{L}(\theta)
        =
        -\mathbb{E}_{(x, y) \sim \mathcal{D}_{\text{syn}}}
        \left[
            \log p_{\theta}(y \mid x)
        \right]
        \]
        where ``instruction`` and ``input`` define \(x\), and ``output`` defines
        \(y\).
    """

    # Store the instruction string.
    instruction: str

    # Store the input/context field.
    input: str

    # Store the output/target field.
    output: str

    # Store the ordered citation list.
    citations: List[str]

    # Store the difficulty label.
    difficulty: str

    # Store the notes field.
    notes: str


# Define the allowed set of difficulty labels.
VALID_DIFFICULTIES: FrozenSet[str] = frozenset({"easy", "medium", "hard"})


# Define the typed split container.
class DataSplit(NamedTuple):
    """
    Typed container holding the three disjoint dataset splits.

    Purpose:
        Store the train, validation, and test partitions as separate typed
        lists.

    Inputs:
        This container is produced by ``cluster_aware_split``.

    Processes:
        No computation is performed here.

    Outputs:
        train:
            Training split.
        val:
            Validation split.
        test:
            Test split.
    """

    # Store the training split.
    train: List[SFTRecord]

    # Store the validation split.
    val: List[SFTRecord]

    # Store the test split.
    test: List[SFTRecord]


# Define the deterministic raw-record hashing helper.
def _hash_record(record_dict: Dict[str, Any]) -> str:
    """
    Compute a deterministic SHA-256 hash of a raw record dictionary.

    Purpose:
        Canonically hash a JSON-like record for deduplication and identity.

    Inputs:
        record_dict:
            Raw dictionary representation of a JSONL record.

    Processes:
        1. Validate that the input is a dictionary.
        2. Serialize the dictionary to sorted-key JSON bytes.
        3. Compute the SHA-256 digest.
        4. Return the first 16 hex characters.

    Outputs:
        str:
            Deterministic short hash identifier.

    Raises:
        TypeError:
            Raised if ``record_dict`` is not a dictionary.
    """
    # Validate that the input record is a dictionary.
    if not isinstance(record_dict, dict):
        # Raise a type error because canonical JSON serialization requires a
        # dictionary.
        raise TypeError(
            "_hash_record expected 'record_dict' to be a dictionary."
        )

    # Serialize the dictionary to canonical sorted-key JSON bytes.
    canonical_bytes: bytes = json.dumps(
        record_dict,
        sort_keys=True
    ).encode("utf-8")

    # Compute and return the first 16 hex characters of the SHA-256 hash.
    return hashlib.sha256(canonical_bytes).hexdigest()[:16]


# Define the JSONL loader and validator.
def load_and_validate_jsonl(
    path: pathlib.Path,
    max_records: Optional[int] = None
) -> Tuple[List[SFTRecord], Dict[str, Any]]:
    """
    Load a JSONL file, validate it against the SFT schema, and return records
    plus an ingestion manifest.

    Purpose:
        Enforce the strict JSONL schema and the grounding rule at ingestion time
        before any training begins.

    Inputs:
        path:
            Absolute or relative path to the JSONL file.
        max_records:
            Optional hard cap on the number of valid records ingested.

    Processes:
        1. Validate input types.
        2. Confirm that the JSONL file exists.
        3. Compute the SHA-256 hash of the file bytes.
        4. Iterate over all non-empty lines.
        5. Parse each line as JSON.
        6. Enforce the required fields:
           - instruction
           - input
           - output
           - citations
           - difficulty
           - notes
        7. Enforce grounding:
           - citations must be a non-empty list.
        8. Enforce difficulty validity using VALID_DIFFICULTIES.
        9. Deduplicate by canonical record hash.
        10. Optionally truncate at ``max_records``.
        11. Compute the ingestion manifest and rejection statistics.
        12. Raise if rejection rate exceeds 10%.
        13. Return validated records plus manifest.

    Outputs:
        Tuple[List[SFTRecord], Dict[str, Any]]:
            1. Validated SFT records.
            2. Ingestion manifest containing file hash and rejection counts.

    Raises:
        TypeError:
            Raised if input types are invalid.
        FileNotFoundError:
            Raised if the JSONL file does not exist.
        ValueError:
            Raised if the rejection rate exceeds 10%.

    Notes:
        This callable enforces the grounding rule that a record without valid
        citations is not admissible for supervised fine-tuning.
    """
    # Validate that the path is a pathlib.Path.
    if not isinstance(path, pathlib.Path):
        # Raise a type error because file ingestion requires pathlib.Path.
        raise TypeError(
            "load_and_validate_jsonl expected 'path' to be a pathlib.Path."
        )

    # Validate max_records when provided.
    if max_records is not None and not isinstance(max_records, int):
        # Raise a type error because the record cap must be integral.
        raise TypeError(
            "load_and_validate_jsonl expected 'max_records' to be an integer "
            "or None."
        )

    # Ensure the JSONL file exists.
    if not path.exists():
        # Raise a file-not-found error when the dataset path is missing.
        raise FileNotFoundError(f"JSONL file not found: {path}")

    # Read the full file bytes for hashing.
    file_bytes: bytes = path.read_bytes()

    # Compute the SHA-256 hash of the input file.
    file_hash: str = hashlib.sha256(file_bytes).hexdigest()

    # Initialize the validated-record list.
    records: List[SFTRecord] = []

    # Initialize the deduplication hash set.
    seen_hashes: Set[str] = set()

    # Initialize the total-line counter.
    n_total: int = 0

    # Initialize the schema-rejection counter.
    n_rejected_schema: int = 0

    # Initialize the grounding-rejection counter.
    n_rejected_grounding: int = 0

    # Initialize the duplicate-rejection counter.
    n_rejected_duplicate: int = 0

    # Open the JSONL file for UTF-8 reading.
    with path.open("r", encoding="utf-8") as fh:
        # Iterate over the file line by line with a 1-based line counter.
        for line_number, raw_line in enumerate(fh, start=1):
            # Strip surrounding whitespace from the current line.
            stripped: str = raw_line.strip()

            # Skip empty lines.
            if not stripped:
                continue

            # Increment the total processed-line counter.
            n_total += 1

            try:
                # Parse the current JSON object line.
                record_dict: Dict[str, Any] = json.loads(stripped)
            except json.JSONDecodeError:
                # Count malformed JSON as a schema rejection.
                n_rejected_schema += 1
                continue

            # Define the required schema fields.
            required_fields: List[str] = [
                "instruction",
                "input",
                "output",
                "citations",
                "difficulty",
                "notes"
            ]

            # Reject records missing any required field.
            if any(f not in record_dict for f in required_fields):
                n_rejected_schema += 1
                continue

            # Extract the citations field.
            citations: Any = record_dict["citations"]

            # Enforce the grounding rule: citations must be a non-empty list.
            if not isinstance(citations, list) or len(citations) == 0:
                n_rejected_grounding += 1
                continue

            # Enforce difficulty validity against the fixed vocabulary.
            if record_dict["difficulty"] not in VALID_DIFFICULTIES:
                n_rejected_schema += 1
                continue

            # Compute the deterministic content hash for deduplication.
            content_hash: str = _hash_record(record_dict)

            # Reject duplicates by content hash.
            if content_hash in seen_hashes:
                n_rejected_duplicate += 1
                continue

            # Record the example hash as seen.
            seen_hashes.add(content_hash)

            # Append the validated SFT record using only the required fields.
            records.append(
                SFTRecord(
                    **{k: record_dict[k] for k in required_fields}
                )
            )

            # Apply the optional max-records cap.
            if max_records is not None and len(records) >= max_records:
                break

    # Construct the ingestion manifest.
    manifest: Dict[str, Any] = {
        "source_path": str(path),
        "file_sha256": file_hash,
        "n_total_lines": n_total,
        "n_valid_records": len(records),
        "n_rejected_schema": n_rejected_schema,
        "n_rejected_grounding": n_rejected_grounding,
        "n_rejected_duplicate": n_rejected_duplicate,
        "rejection_rate": round(1.0 - len(records) / max(n_total, 1), 4),
    }

    # Enforce the maximum rejection-rate tolerance.
    if manifest["rejection_rate"] > 0.10:
        raise ValueError(
            f"Rejection rate {manifest['rejection_rate']:.1%} exceeds 10% "
            "tolerance."
        )

    # Return the validated records and ingestion manifest.
    return records, manifest


# Define the cluster-aware split callable.
def cluster_aware_split(
    records: List[SFTRecord],
    train_fraction: float = 0.80,
    val_fraction: float = 0.10,
    test_fraction: float = 0.10,
    seed: int = 42,
) -> DataSplit:
    """
    Split records into train/validation/test partitions by grouping on the first
    citation.

    Purpose:
        Enforce leakage control by ensuring that no two records citing the same
        source cluster appear in different splits.

    Inputs:
        records:
            Validated SFT records.
        train_fraction:
            Training split fraction.
        val_fraction:
            Validation split fraction.
        test_fraction:
            Test split fraction.
        seed:
            Deterministic RNG seed.

    Processes:
        1. Validate input types.
        2. Verify that split fractions sum to 1.0.
        3. Group records by cluster key defined as the first citation.
        4. Deterministically shuffle cluster keys using ``random.Random(seed)``.
        5. Compute train and validation cluster cutoffs.
        6. Materialize the three record splits cluster-wise.
        7. Return the typed DataSplit container.

    Outputs:
        DataSplit:
            Cluster-disjoint train, validation, and test partitions.

    Raises:
        TypeError:
            Raised if inputs are invalid.
        ValueError:
            Raised when the split fractions do not sum to 1.0.

    Notes:
        This callable enforces the cluster-disjointness principle that prevents
        citation-source leakage across dataset splits.
    """
    # Validate that records is a list.
    if not isinstance(records, list):
        # Raise a type error because split construction requires a list.
        raise TypeError(
            "cluster_aware_split expected 'records' to be a list of SFTRecord "
            "objects."
        )

    # Validate that all records are SFTRecord instances.
    if not all(isinstance(rec, SFTRecord) for rec in records):
        # Raise a type error because split grouping assumes the SFTRecord
        # schema.
        raise TypeError(
            "cluster_aware_split expected every element in 'records' to be an "
            "SFTRecord."
        )

    # Validate numeric split fractions and seed.
    if not isinstance(train_fraction, (int, float)):
        raise TypeError(
            "cluster_aware_split expected 'train_fraction' to be numeric."
        )
    if not isinstance(val_fraction, (int, float)):
        raise TypeError(
            "cluster_aware_split expected 'val_fraction' to be numeric."
        )
    if not isinstance(test_fraction, (int, float)):
        raise TypeError(
            "cluster_aware_split expected 'test_fraction' to be numeric."
        )
    if not isinstance(seed, int):
        raise TypeError(
            "cluster_aware_split expected 'seed' to be an integer."
        )

    # Compute the sum of the split fractions.
    total: float = train_fraction + val_fraction + test_fraction

    # Reject split configurations whose fractions do not sum to 1.0.
    if abs(total - 1.0) > 1e-6:
        raise ValueError(
            f"Fractions must sum to 1.0; got {total:.6f}"
        )

    # Initialize the cluster map keyed by the first citation.
    cluster_map: Dict[str, List[SFTRecord]] = collections.defaultdict(list)

    # Assign each record to a cluster defined by its first citation.
    for rec in records:
        cluster_key: str = rec.citations[0].strip().lower()
        cluster_map[cluster_key].append(rec)

    # Initialize a deterministic local RNG.
    rng: random.Random = random.Random(seed)

    # Build the list of cluster keys.
    cluster_keys: List[str] = list(cluster_map.keys())

    # Shuffle the cluster keys deterministically.
    rng.shuffle(cluster_keys)

    # Compute the number of clusters.
    n_clusters: int = len(cluster_keys)

    # Compute the train split cutoff in cluster space.
    train_cutoff: int = round(n_clusters * train_fraction)

    # Compute the validation split cutoff in cluster space.
    val_cutoff: int = train_cutoff + round(n_clusters * val_fraction)

    # Slice the train cluster keys.
    train_keys: List[str] = cluster_keys[:train_cutoff]

    # Slice the validation cluster keys.
    val_keys: List[str] = cluster_keys[train_cutoff:val_cutoff]

    # Slice the test cluster keys.
    test_keys: List[str] = cluster_keys[val_cutoff:]

    # Materialize the training records from the selected clusters.
    train_records: List[SFTRecord] = [
        r for k in train_keys for r in cluster_map[k]
    ]

    # Materialize the validation records from the selected clusters.
    val_records: List[SFTRecord] = [
        r for k in val_keys for r in cluster_map[k]
    ]

    # Materialize the test records from the selected clusters.
    test_records: List[SFTRecord] = [
        r for k in test_keys for r in cluster_map[k]
    ]

    # Return the typed data split.
    return DataSplit(
        train=train_records,
        val=val_records,
        test=test_records
    )


# Define the harmony-conversation builder.
def build_harmony_conversation(
    record: SFTRecord
) -> List[Dict[str, str]]:
    """
    Construct a harmony-format conversation from a validated SFTRecord.

    Purpose:
        Map one supervised example into the chat-structured message format used
        by the tokenizer's chat template so that training and inference remain
        aligned.

    Inputs:
        record:
            Validated supervised fine-tuning record.

    Processes:
        1. Validate that the input is an SFTRecord.
        2. Build the user content from the instruction, optionally appending the
           input context.
        3. Build the assistant content containing:
           - a Sources section,
           - an Answer section.
        4. Return the two-message conversation list.

    Outputs:
        List[Dict[str, str]]:
            Harmony-format conversation represented as role/content dictionaries.

    Raises:
        TypeError:
            Raised if ``record`` is invalid.

    Notes:
        This callable bridges structured supervised data and tokenizer chat
        templating.
    """
    # Validate that the record is an SFTRecord.
    if not isinstance(record, SFTRecord):
        # Raise a type error because conversation construction requires the
        # SFTRecord schema.
        raise TypeError(
            "build_harmony_conversation expected 'record' to be an SFTRecord."
        )

    # Initialize the user content with the instruction.
    user_content: str = record.instruction

    # If the input field is non-empty, append it as a Context block.
    if record.input.strip():
        user_content = (
            f"{record.instruction}\n\n### Context:\n{record.input}"
        )

    # Render the citations as a Sources block.
    citation_block: str = "\n".join(f"- {c}" for c in record.citations)

    # Construct the assistant content containing sources plus answer.
    assistant_content: str = (
        f"### Sources:\n{citation_block}\n\n### Answer:\n{record.output}"
    )

    # Return the harmony-style conversation.
    return [
        {"role": "user", "content": user_content},
        {"role": "assistant", "content": assistant_content},
    ]


# Define the Hugging Face dataset builder.
def build_hf_dataset(
    records: List[SFTRecord],
    tokenizer: PreTrainedTokenizerBase
) -> Dataset:
    """
    Convert validated SFT records into a Hugging Face Dataset.

    Purpose:
        Render each harmony-format conversation into a single text string using
        the tokenizer's chat template, then package the rendered examples as a
        Dataset with a single ``text`` field.

    Inputs:
        records:
            Ordered list of SFT records.
        tokenizer:
            Pretrained tokenizer exposing ``apply_chat_template``.

    Processes:
        1. Validate top-level inputs.
        2. Iterate over all records.
        3. Convert each record into a harmony-format conversation.
        4. Render the conversation to text using the tokenizer's chat template.
        5. Accumulate dictionaries of the form ``{"text": rendered_text}``.
        6. Convert the list into a Hugging Face Dataset.

    Outputs:
        Dataset:
            Dataset with one ``text`` column suitable for SFTTrainer.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This callable ensures that the fine-tuning data format matches the
        tokenizer's inference-time chat formatting.
    """
    # Validate that records is a list.
    if not isinstance(records, list):
        # Raise a type error because dataset building requires a list of
        # records.
        raise TypeError(
            "build_hf_dataset expected 'records' to be a list of SFTRecord "
            "objects."
        )

    # Validate that all records are SFTRecord instances.
    if not all(isinstance(rec, SFTRecord) for rec in records):
        # Raise a type error because dataset rendering assumes SFTRecord schema.
        raise TypeError(
            "build_hf_dataset expected every element in 'records' to be an "
            "SFTRecord."
        )

    # Validate that the tokenizer is a PreTrainedTokenizerBase.
    if not isinstance(tokenizer, PreTrainedTokenizerBase):
        # Raise a type error because chat-template rendering requires a
        # pretrained tokenizer.
        raise TypeError(
            "build_hf_dataset expected 'tokenizer' to be a "
            "PreTrainedTokenizerBase."
        )

    # Initialize the rendered-example list.
    rendered: List[Dict[str, str]] = []

    # Iterate over the validated records.
    for rec in records:
        # Build the harmony-format conversation for the current record.
        conversation: List[Dict[str, str]] = build_harmony_conversation(rec)

        # Render the conversation to a single training string via the chat
        # template.
        text: str = tokenizer.apply_chat_template(
            conversation,
            tokenize=False,
            add_generation_prompt=False,
        )

        # Append the rendered training example.
        rendered.append({"text": text})

    # Convert the rendered example list into a Hugging Face Dataset.
    return Dataset.from_list(rendered)


# Define the SFTTrainer construction callable.
def build_sft_trainer(
    model: PreTrainedModel,
    tokenizer: PreTrainedTokenizerBase,
    train_dataset: Dataset,
    eval_dataset: Dataset,
    lora_config: LoraConfig,
    output_dir: str,
    per_device_train_batch_size: int,
    gradient_accumulation_steps: int,
    num_train_epochs: int,
    learning_rate: float,
    max_seq_length: int,
    seed: int,
    n_gpus: int,
) -> SFTTrainer:
    """
    Construct a fully configured SFTTrainer for ``gpt-oss-20b`` LoRA
    fine-tuning.

    Purpose:
        Instantiate the TRL ``SFTTrainer`` with the reference hyperparameters and
        derived schedule parameters required for supervised fine-tuning.

    Inputs:
        model:
            PEFT-wrapped causal language model.
        tokenizer:
            Tokenizer aligned with the model.
        train_dataset:
            Training dataset in Hugging Face Dataset format.
        eval_dataset:
            Validation dataset in Hugging Face Dataset format.
        lora_config:
            LoRA configuration object.
        output_dir:
            Output directory for checkpoints and logs.
        per_device_train_batch_size:
            Per-device micro-batch size.
        gradient_accumulation_steps:
            Gradient accumulation factor.
        num_train_epochs:
            Number of training epochs.
        learning_rate:
            Optimizer learning rate.
        max_seq_length:
            Maximum sequence length in tokens.
        seed:
            Deterministic seed for trainer and data.
        n_gpus:
            Number of GPUs participating in training.

    Processes:
        1. Validate input types.
        2. Compute effective batch size:
           \[
           \text{effective\_batch}
           =
           \text{per\_device\_train\_batch\_size}
           \times
           \text{gradient\_accumulation\_steps}
           \times
           n_{\text{gpus}}
           \]
        3. Compute steps per epoch:
           \[
           \text{steps\_per\_epoch}
           =
           \left\lceil
           \frac{|\mathcal{D}_{train}|}{\text{effective\_batch}}
           \right\rceil
           \]
        4. Compute total steps and warmup steps.
        5. Construct the ``SFTConfig`` with the specified hyperparameters.
        6. Instantiate and return the ``SFTTrainer``.

    Outputs:
        SFTTrainer:
            Fully configured supervised fine-tuning trainer.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This callable operationalizes the supervised fine-tuning objective under
        the specified optimization configuration.
    """
    # Validate that the model is a PreTrainedModel.
    if not isinstance(model, PreTrainedModel):
        raise TypeError(
            "build_sft_trainer expected 'model' to be a PreTrainedModel."
        )

    # Validate that the tokenizer is a PreTrainedTokenizerBase.
    if not isinstance(tokenizer, PreTrainedTokenizerBase):
        raise TypeError(
            "build_sft_trainer expected 'tokenizer' to be a "
            "PreTrainedTokenizerBase."
        )

    # Validate that train_dataset is a Dataset.
    if not isinstance(train_dataset, Dataset):
        raise TypeError(
            "build_sft_trainer expected 'train_dataset' to be a Dataset."
        )

    # Validate that eval_dataset is a Dataset.
    if not isinstance(eval_dataset, Dataset):
        raise TypeError(
            "build_sft_trainer expected 'eval_dataset' to be a Dataset."
        )

    # Validate that lora_config is a LoraConfig.
    if not isinstance(lora_config, LoraConfig):
        raise TypeError(
            "build_sft_trainer expected 'lora_config' to be a LoraConfig."
        )

    # Compute the effective batch size.
    effective_batch: int = (
        per_device_train_batch_size
        * gradient_accumulation_steps
        * n_gpus
    )

    # Compute the number of optimizer steps per epoch.
    steps_per_epoch: int = math.ceil(len(train_dataset) / effective_batch)

    # Compute the total number of optimizer steps across all epochs.
    total_steps: int = steps_per_epoch * num_train_epochs

    # Compute the warmup steps using the fixed 3% ratio.
    warmup_steps: int = round(total_steps * 0.03)

    # Construct the supervised fine-tuning configuration.
    sft_config: SFTConfig = SFTConfig(
        output_dir=output_dir,
        save_strategy="steps",
        save_steps=max(1, total_steps // 10),
        save_total_limit=3,
        eval_strategy="steps",
        eval_steps=max(1, total_steps // 10),
        load_best_model_at_end=True,
        per_device_train_batch_size=per_device_train_batch_size,
        per_device_eval_batch_size=per_device_train_batch_size,
        gradient_accumulation_steps=gradient_accumulation_steps,
        learning_rate=learning_rate,
        num_train_epochs=num_train_epochs,
        weight_decay=0.01,
        warmup_steps=warmup_steps,
        lr_scheduler_type="cosine",
        optim="adamw_torch",
        max_grad_norm=1.0,
        bf16=True,
        fp16=False,
        max_seq_length=max_seq_length,
        dataset_text_field="text",
        seed=seed,
        data_seed=seed,
        logging_dir=os.path.join(output_dir, "logs"),
        logging_steps=10,
        report_to="none",
        packing=False,
    )

    # Instantiate the TRL supervised fine-tuning trainer.
    trainer: SFTTrainer = SFTTrainer(
        model=model,
        args=sft_config,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset,
        processing_class=tokenizer,
        peft_config=lora_config,
    )

    # Return the configured trainer.
    return trainer


# Define the training-execution callable.
def execute_training(
    trainer: SFTTrainer,
    output_dir: str,
    run_id: str,
    model_id: str
) -> Dict[str, object]:
    """
    Execute the supervised fine-tuning loop and write a post-training run
    manifest.

    Purpose:
        Run ``trainer.train()``, save model artifacts, and serialize training
        metadata to disk for reproducibility.

    Inputs:
        trainer:
            Configured SFTTrainer instance.
        output_dir:
            Directory in which to save checkpoints and manifests.
        run_id:
            Unique run identifier.
        model_id:
            Base model identifier used for training.

    Processes:
        1. Validate input types.
        2. Capture the UTC start timestamp.
        3. Execute ``trainer.train()``.
        4. Capture the UTC end timestamp.
        5. Save the model and tokenizer.
        6. Construct a run manifest capturing:
           - timing,
           - metrics,
           - checkpoint metadata,
           - global step and epoch.
        7. Serialize the run manifest to disk.
        8. Return the manifest.

    Outputs:
        Dict[str, object]:
            Post-training run manifest.

    Raises:
        TypeError:
            Raised if inputs are invalid.
        OSError:
            Propagated if checkpoint or manifest serialization fails.

    Notes:
        This callable provides the run-manifest output required by reproducible
        experiment management.
    """
    # Validate that the trainer is an SFTTrainer.
    if not isinstance(trainer, SFTTrainer):
        raise TypeError(
            "execute_training expected 'trainer' to be an SFTTrainer."
        )

    # Validate that output_dir is a string.
    if not isinstance(output_dir, str):
        raise TypeError(
            "execute_training expected 'output_dir' to be a string."
        )

    # Validate that run_id is a string.
    if not isinstance(run_id, str):
        raise TypeError(
            "execute_training expected 'run_id' to be a string."
        )

    # Validate that model_id is a string.
    if not isinstance(model_id, str):
        raise TypeError(
            "execute_training expected 'model_id' to be a string."
        )

    # Capture the UTC start time.
    start_time: datetime.datetime = datetime.datetime.utcnow()

    # Execute the training loop.
    train_result = trainer.train()

    # Capture the UTC end time.
    end_time: datetime.datetime = datetime.datetime.utcnow()

    # Save the trained adapter/model artifacts.
    trainer.save_model(output_dir)

    # Save the tokenizer or processing class alongside the model.
    trainer.processing_class.save_pretrained(output_dir)

    # Construct the training manifest dictionary.
    manifest: Dict[str, object] = {
        "run_id": run_id,
        "model_id": model_id,
        "start_time_utc": start_time.isoformat(),
        "end_time_utc": end_time.isoformat(),
        "duration_seconds": (end_time - start_time).total_seconds(),
        "train_metrics": train_result.metrics,
        "output_dir": output_dir,
        "best_model_checkpoint": getattr(
            trainer.state,
            "best_model_checkpoint",
            output_dir
        ),
        "global_step": trainer.state.global_step,
        "epoch": trainer.state.epoch,
    }

    # Resolve the manifest output path.
    manifest_path: pathlib.Path = (
        pathlib.Path(output_dir) / f"run_manifest_{run_id}.json"
    )

    # Ensure the output directory exists.
    manifest_path.parent.mkdir(parents=True, exist_ok=True)

    # Serialize the training manifest to disk.
    manifest_path.write_text(
        json.dumps(manifest, indent=2),
        encoding="utf-8"
    )

    # Return the training manifest.
    return manifest


# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 3: Record all training artifacts and evaluate
# -------------------------------------------------------------------------------------------------------------------------------

# Define the mutable evaluation-question record.
@dataclass
class EvalQuestion:
    """
    Mutable record representing one evaluation question.

    Purpose:
        Store the question prompt, reference answer, citation requirements, and
        multi-document flag for one evaluation case.

    Inputs:
        This dataclass is populated externally by evaluation harness code.

    Processes:
        No computation is performed here. It is a passive evaluation record.

    Outputs:
        question_id:
            Unique question identifier.
        question:
            Evaluation prompt.
        reference_answer:
            Human-provided or benchmark reference answer.
        required_citations:
            Citation paths expected in a grounded answer.
        is_multi_doc:
            Whether the question requires multi-document synthesis.
    """

    # Store the question identifier.
    question_id: str

    # Store the evaluation question text.
    question: str

    # Store the reference answer.
    reference_answer: str

    # Store the required citations.
    required_citations: List[str]

    # Store whether the question is multi-document.
    is_multi_doc: bool = False


# Define the mutable evaluation-result record.
@dataclass
class EvalResult:
    """
    Mutable record representing the result of evaluating one question against the
    fine-tuned model.

    Purpose:
        Store the generated output, latency, extracted citations, and key
        grounding diagnostics for one evaluation item.

    Inputs:
        This dataclass is populated by ``evaluate_model``.

    Processes:
        No computation is performed here. It is a passive evaluation-result
        record.

    Outputs:
        question_id:
            Evaluation-question identifier.
        model_output:
            Generated model output.
        latency_seconds:
            Wall-clock latency of generation.
        citations_found:
            Extracted citations from the model output.
        citation_coverage_rate:
            Coverage of required citations.
        possible_hallucination:
            Whether extracted citations include unknown page paths.
        flagged_uncertain:
            Whether the model expressed uncertainty.
    """

    # Store the evaluation-question identifier.
    question_id: str

    # Store the generated model output.
    model_output: str

    # Store the generation latency in seconds.
    latency_seconds: float

    # Store extracted citations.
    citations_found: List[str]

    # Store the citation coverage rate.
    citation_coverage_rate: float

    # Store whether a possible hallucination was detected.
    possible_hallucination: bool

    # Store whether uncertainty was explicitly flagged.
    flagged_uncertain: bool


# Define the citation-extraction helper.
def extract_citations_from_output(
    model_output: str,
    known_page_paths: FrozenSet[str]
) -> Tuple[List[str], bool]:
    """
    Parse model output to extract citations and detect hallucinated sources.

    Purpose:
        Extract the ``### Sources:`` block from a generated answer and verify
        that all cited paths belong to the known wiki page set.

    Inputs:
        model_output:
            Generated answer text.
        known_page_paths:
            Immutable set of valid canonical wiki page paths.

    Processes:
        1. Validate input types.
        2. Search for the ``### Sources:`` block.
        3. If absent, return an empty citation list and False.
        4. Parse bullet-list citations from the sources block.
        5. Detect whether any citation lies outside ``known_page_paths``.

    Outputs:
        Tuple[List[str], bool]:
            1. Ordered list of extracted citations.
            2. Boolean hallucination flag.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This helper underpins grounded evaluation by checking whether the model
        cites only known local pages.
    """
    # Validate that the model output is a string.
    if not isinstance(model_output, str):
        raise TypeError(
            "extract_citations_from_output expected 'model_output' to be a "
            "string."
        )

    # Validate that known_page_paths is a frozenset.
    if not isinstance(known_page_paths, frozenset):
        raise TypeError(
            "extract_citations_from_output expected 'known_page_paths' to be a "
            "FrozenSet[str]."
        )

    # Compile the regex pattern used to isolate the Sources block.
    sources_pattern: re.Pattern = re.compile(
        r"### Sources:\n(.*?)(?=###|\Z)",
        re.DOTALL
    )

    # Search for the Sources block in the output.
    match: Optional[re.Match] = sources_pattern.search(model_output)

    # Return an empty citation list and False when the Sources block is absent.
    if not match:
        return [], False

    # Parse bullet-list citations from the Sources block.
    raw_citations: List[str] = [
        line.strip().lstrip("- ").strip()
        for line in match.group(1).splitlines()
        if line.strip().startswith("- ")
    ]

    # Detect hallucinations when a cited path is not present in the known-page
    # set.
    hallucination_detected: bool = any(
        c not in known_page_paths for c in raw_citations
    )

    # Return the extracted citations and hallucination flag.
    return raw_citations, hallucination_detected


# Define the evaluation harness callable.
def evaluate_model(
    model_pipeline: Pipeline,
    questions: List[EvalQuestion],
    known_page_paths: FrozenSet[str],
    max_new_tokens: int = 1024,
) -> List[EvalResult]:
    """
    Run the evaluation harness against a list of evaluation questions.

    Purpose:
        Execute deterministic greedy decoding for each evaluation question,
        extract citations, compute citation coverage, detect possible
        hallucinations, and record latency.

    Inputs:
        model_pipeline:
            Text-generation pipeline wrapping the fine-tuned model.
        questions:
            Ordered list of EvalQuestion items.
        known_page_paths:
            Immutable set of valid canonical wiki page paths.
        max_new_tokens:
            Maximum number of new tokens to generate per answer.

    Processes:
        1. Validate top-level inputs.
        2. Iterate over evaluation questions.
        3. Construct a single-message user prompt for each question.
        4. Measure generation latency using ``time.perf_counter``.
        5. Invoke the model pipeline with greedy decoding:
           - ``do_sample=False``
           - ``temperature=None``
        6. Extract the generated assistant content.
        7. Extract citations and hallucination flag.
        8. Compute citation coverage:
           \[
           \text{coverage}
           =
           \frac{|\,\text{required} \cap \text{found}\,|}
                {\max(|\text{required}|, 1)}
           \]
        9. Detect uncertainty markers via lexical heuristics.
        10. Append EvalResult objects.
        11. Return the ordered result list.

    Outputs:
        List[EvalResult]:
            Ordered evaluation results for all questions.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This callable implements deterministic post-training evaluation suitable
        for suite-level aggregation.
    """
    # Validate that the model pipeline is a Pipeline instance.
    if not isinstance(model_pipeline, Pipeline):
        raise TypeError(
            "evaluate_model expected 'model_pipeline' to be a Pipeline."
        )

    # Validate that questions is a list.
    if not isinstance(questions, list):
        raise TypeError(
            "evaluate_model expected 'questions' to be a list of EvalQuestion."
        )

    # Validate that all questions are EvalQuestion instances.
    if not all(isinstance(q, EvalQuestion) for q in questions):
        raise TypeError(
            "evaluate_model expected every element in 'questions' to be an "
            "EvalQuestion."
        )

    # Validate that known_page_paths is a frozenset.
    if not isinstance(known_page_paths, frozenset):
        raise TypeError(
            "evaluate_model expected 'known_page_paths' to be a FrozenSet[str]."
        )

    # Validate that max_new_tokens is an integer.
    if not isinstance(max_new_tokens, int):
        raise TypeError(
            "evaluate_model expected 'max_new_tokens' to be an integer."
        )

    # Initialize the ordered evaluation result list.
    results: List[EvalResult] = []

    # Iterate over evaluation questions.
    for q in questions:
        # Construct the single user-message list expected by the pipeline.
        messages: List[Dict[str, str]] = [
            {"role": "user", "content": q.question}
        ]

        # Capture the start time.
        t0: float = time.perf_counter()

        # Run deterministic greedy decoding.
        output = model_pipeline(
            messages,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=None,
        )

        # Capture the end time.
        t1: float = time.perf_counter()

        # Extract the generated assistant content.
        model_output: str = output[0]["generated_text"][-1]["content"]

        # Extract citations and detect hallucinations.
        citations_found, hallucination = extract_citations_from_output(
            model_output,
            known_page_paths
        )

        # Convert required citations to a set.
        required: Set[str] = set(q.required_citations)

        # Convert found citations to a set.
        found: Set[str] = set(citations_found)

        # Compute citation coverage.
        coverage: float = len(required & found) / max(len(required), 1)

        # Define the lexical uncertainty patterns.
        uncertainty_patterns: List[str] = [
            "i am not certain",
            "i cannot confirm",
            "i don't have enough",
            "insufficient information",
            "i am unsure",
            "this is uncertain"
        ]

        # Detect whether the model explicitly signaled uncertainty.
        flagged_uncertain: bool = any(
            p in model_output.lower() for p in uncertainty_patterns
        )

        # Append the evaluation result.
        results.append(
            EvalResult(
                question_id=q.question_id,
                model_output=model_output,
                latency_seconds=t1 - t0,
                citations_found=citations_found,
                citation_coverage_rate=coverage,
                possible_hallucination=hallucination,
                flagged_uncertain=flagged_uncertain,
            )
        )

    # Return the ordered evaluation results.
    return results


# Define the aggregate evaluation-metrics callable.
def aggregate_eval_metrics(
    results: List[EvalResult]
) -> Dict[str, float]:
    """
    Aggregate per-question EvalResults into suite-level scalar metrics.

    Purpose:
        Compute the core scalar evaluation metrics required for post-training
        assessment.

    Inputs:
        results:
            Ordered list of EvalResult objects.

    Processes:
        1. Validate that results is a list of EvalResult objects.
        2. Compute the number of evaluated questions.
        3. Compute mean citation coverage:
           \[
           \overline{CCR}
           =
           \frac{1}{N}\sum_{i=1}^{N}
           \text{citation\_coverage\_rate}_i
           \]
        4. Count hallucination incidents.
        5. Compute uncertainty-marking rate:
           \[
           \text{uncertainty\_rate}
           =
           \frac{1}{N}
           \sum_{i=1}^{N} \mathbf{1}\{\text{flagged\_uncertain}_i\}
           \]
        6. Compute mean latency:
           \[
           \overline{L}
           =
           \frac{1}{N}\sum_{i=1}^{N}\text{latency\_seconds}_i
           \]
        7. Return the aggregate metric dictionary.

    Outputs:
        Dict[str, float]:
            Aggregate evaluation metrics.

    Raises:
        TypeError:
            Raised if inputs are invalid.
    """
    # Validate that results is a list.
    if not isinstance(results, list):
        raise TypeError(
            "aggregate_eval_metrics expected 'results' to be a list of "
            "EvalResult objects."
        )

    # Validate that all elements are EvalResult instances.
    if not all(isinstance(r, EvalResult) for r in results):
        raise TypeError(
            "aggregate_eval_metrics expected every element in 'results' to be "
            "an EvalResult."
        )

    # Compute the number of evaluated questions.
    n: int = len(results)

    # Compute the mean citation coverage rate.
    mean_citation_coverage: float = sum(
        r.citation_coverage_rate for r in results
    ) / max(n, 1)

    # Count hallucination incidents.
    hallucination_count: int = sum(
        1 for r in results if r.possible_hallucination
    )

    # Compute the uncertainty-marking rate.
    uncertainty_rate: float = sum(
        1 for r in results if r.flagged_uncertain
    ) / max(n, 1)

    # Compute the mean latency per answer.
    mean_latency: float = sum(
        r.latency_seconds for r in results
    ) / max(n, 1)

    # Return the aggregate metric dictionary.
    return {
        "citation_coverage_rate": round(mean_citation_coverage, 4),
        "hallucination_incidents_count": hallucination_count,
        "uncertainty_marking_rate_when_unsupported": round(uncertainty_rate, 4),
        "latency_seconds_per_answer": round(mean_latency, 4),
        "n_questions_evaluated": n,
    }


# Define the adapter-merge callable.
def merge_adapter_and_archive(
    base_model_id: str,
    adapter_path: str,
    merged_output_dir: str,
    tokenizer: Any,
) -> None:
    """
    Merge the LoRA adapter weights into the base model and archive the artifact.

    Purpose:
        Reload the base model, attach the trained LoRA adapter, merge the
        adapter weights into the base model, and save the merged artifact. This
        callable enforces CPU offloading during the merge path to reduce VRAM
        pressure during ``merge_and_unload()``.

    Inputs:
        base_model_id:
            Base model identifier from which the original model is reloaded.
        adapter_path:
            Filesystem directory containing the trained adapter weights.
        merged_output_dir:
            Filesystem directory to which the merged model should be saved.
        tokenizer:
            Tokenizer object expected to expose ``save_pretrained``.

    Processes:
        1. Validate dependency availability.
        2. Validate input types.
        3. Resolve the preferred dtype using the same hardware-aware logic used
           during base-model loading.
        4. Reload the base model with ``device_map="cpu"``.
        5. Load the trained adapter into a PEFT wrapper.
        6. Merge the adapter weights into the base model and unload the PEFT wrapper.
        7. Ensure the merged-output directory exists.
        8. Save the merged model with safe serialization enabled.
        9. Save the tokenizer alongside the merged model.
        10. Emit a success log entry.

    Outputs:
        None:
            The side effects are saving the merged model artifact and tokenizer.

    Raises:
        RuntimeError:
            Raised if the machine-learning dependencies are unavailable.
        TypeError:
            Raised if the string path/ID arguments are invalid.
    """
    # Halt immediately when the machine-learning stack is unavailable.
    if not ML_AVAILABLE:
        # Raise a runtime error because adapter merging requires the ML dependencies.
        raise RuntimeError("PyTorch and Transformers are required for Task 18.")

    # Validate that all path and model-ID arguments are strings.
    if not isinstance(base_model_id, str) or not isinstance(adapter_path, str) or not isinstance(merged_output_dir, str):
        # Raise a type error because model IDs and filesystem paths must be strings.
        raise TypeError(
            "merge_adapter_and_archive expected string paths/IDs."
        )

    # Resolve the preferred dtype using hardware-aware capability checks.
    if torch.cuda.is_available() and torch.cuda.is_bf16_supported():
        # Select bfloat16 when supported.
        resolved_dtype = torch.bfloat16

    # Fall back to float16 when bfloat16 is unavailable.
    else:
        # Select float16 as the fallback dtype.
        resolved_dtype = torch.float16

    # Emit an informational log entry describing the CPU-loading strategy.
    logger.info("Loading base model to CPU for memory-safe adapter merging...")

    # Reload the base model to CPU to reduce VRAM pressure during merge_and_unload.
    base_model = AutoModelForCausalLM.from_pretrained(
        base_model_id,
        torch_dtype=resolved_dtype,
        device_map="cpu",
        trust_remote_code=True,
    )

    # Load the trained adapter into a PEFT model wrapper around the base model.
    peft_model = PeftModel.from_pretrained(
        base_model,
        adapter_path,
        is_trainable=False,
    )

    # Merge the adapter weights into the base model and unload adapter wrappers.
    merged_model = peft_model.merge_and_unload()

    # Ensure the merged-output directory exists.
    os.makedirs(merged_output_dir, exist_ok=True)

    # Save the merged model using safetensors-compatible safe serialization.
    merged_model.save_pretrained(
        merged_output_dir,
        safe_serialization=True,
    )

    # Save the tokenizer alongside the merged model.
    tokenizer.save_pretrained(merged_output_dir)

    # Emit an informational log entry naming the merged-model output directory.
    logger.info(
        f"Merged model successfully archived to {merged_output_dir}"
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable fine-tuning context aggregate.
@dataclass(frozen=True)
class FineTuningContext:
    """
    Immutable record encapsulating the complete fine-tuning transaction.

    Purpose:
        Package the final pipeline manifest into one immutable object, providing
        a unified provenance record for the fine-tuning phase.

    Inputs:
        This dataclass is populated by ``task_18_orchestrator``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        pipeline_manifest:
            Optional end-to-end pipeline manifest containing environment,
            ingestion, split, training, evaluation, and artifact-path metadata.
    """

    # Store the optional pipeline manifest.
    pipeline_manifest: Optional[Dict[str, Any]] = None


# Define the fused end-to-end SFT pipeline callable.
def run_gpt_oss_20b_sft_lora_pipeline(
    train_jsonl_path: pathlib.Path,
    model_id: str,
    use_4bit_quantization: bool,
    lora_rank_r: int,
    lora_alpha: int,
    lora_dropout: float,
    lora_target_modules: Optional[List[str]],
    learning_rate: float,
    per_device_train_batch_size: int,
    gradient_accumulation_steps: int,
    num_train_epochs: int,
    weight_decay: float,
    max_seq_length: int,
    seed: int,
    train_fraction: float,
    val_fraction: float,
    test_fraction: float,
    output_dir: str,
    merged_output_dir: str,
    eval_questions: Optional[List[Any]],
    known_page_paths: Optional[FrozenSet[str]],
    do_merge_adapter: bool,
    max_ingestion_records: Optional[int],
    n_gpus: int,
    run_id: str,
) -> Dict[str, Any]:
    """
    Execute the fused 10-stage supervised fine-tuning pipeline.

    Purpose:
        Run the complete optional fine-tuning extension for the local knowledge
        base system using supervised fine-tuning with LoRA over the specified
        base model.

        This callable operationalizes the paper's fine-tuning objective:

            L(θ) = - E_(x, y ~ D_syn) [ log p_θ(y | x) ]

        together with the paper's broader governance rule that the wiki remains
        authoritative and the fine-tuned model is an acceleration layer rather
        than a replacement.

    Inputs:
        train_jsonl_path:
            Absolute or repo-resolved path to the JSONL synthetic-supervision file.
        model_id:
            Base model identifier for supervised fine-tuning.
        use_4bit_quantization:
            Whether to enable 4-bit NF4 quantization during model loading.
        lora_rank_r:
            LoRA rank hyperparameter ``r``.
        lora_alpha:
            LoRA scaling hyperparameter ``alpha``.
        lora_dropout:
            LoRA dropout rate.
        lora_target_modules:
            Optional explicit list of LoRA target-module names.
        learning_rate:
            Optimizer learning rate.
        per_device_train_batch_size:
            Per-device micro-batch size.
        gradient_accumulation_steps:
            Gradient accumulation factor used to reach the effective batch size.
        num_train_epochs:
            Number of training epochs.
        weight_decay:
            Weight decay hyperparameter.
        max_seq_length:
            Maximum sequence length in tokens.
        seed:
            Deterministic run seed.
        train_fraction:
            Training split fraction.
        val_fraction:
            Validation split fraction.
        test_fraction:
            Test split fraction.
        output_dir:
            Adapter checkpoint output directory.
        merged_output_dir:
            Output directory for merged base-plus-adapter artifacts.
        eval_questions:
            Optional evaluation question set used for post-training evaluation.
        known_page_paths:
            Optional frozen universe of valid wiki page paths used for
            hallucination/citation checks during evaluation.
        do_merge_adapter:
            Whether to merge the LoRA adapter into the base model after training.
        max_ingestion_records:
            Optional cap on ingested training records.
        n_gpus:
            Number of GPUs participating in training.
        run_id:
            Global run identifier used in artifact naming.

    Processes:
        Stage 1:
            Validate the Python version and dependency substrate.
        Stage 2:
            Load and validate the JSONL synthetic-supervision dataset.
        Stage 3:
            Perform cluster-aware leakage-controlled splitting.
        Stage 4 and Stage 5:
            Load tokenizer and model, delegating dtype resolution and hardware
            capability checks to the lower-level helper.
        Stage 6:
            Build the LoRA configuration and apply it to the base model.
        Stage 7 and Stage 8:
            Build the SFT trainer and execute supervised fine-tuning.
        Stage 9:
            Optionally evaluate the resulting model and aggregate evaluation metrics.
        Stage 10:
            Optionally merge the adapter into the base model artifact.
        Finalization:
            Assemble and serialize the pipeline manifest.

    Outputs:
        Dict[str, Any]:
            Comprehensive pipeline manifest containing:
            - environment metadata,
            - ingestion manifest,
            - split statistics,
            - training manifest,
            - optional evaluation metrics,
            - artifact paths.

    Raises:
        TypeError:
            Raised when selected top-level inputs violate their expected types.
        AssertionError:
            Raised when the Python-version guard fails.
        ImportError:
            Propagated from dependency auditing.
        ValueError:
            Propagated from dataset validation or split validation helpers.
        OSError:
            Propagated from checkpoint, artifact, or manifest writes.

    Notes:
        This callable preserves the original logic exactly. In particular:
        - dtype selection is delegated to subordinate helpers,
        - adapter merge is optional,
        - evaluation is conditional on both ``eval_questions`` and
          ``known_page_paths`` being present,
        - the final manifest is serialized with ``default=str``.
    """
    # Validate that the training JSONL path is a pathlib.Path because file
    # loading and validation require path semantics.
    if not isinstance(train_jsonl_path, pathlib.Path):
        # Raise a type error immediately when the training-path contract is violated.
        raise TypeError(
            "run_gpt_oss_20b_sft_lora_pipeline expected 'train_jsonl_path' to "
            "be a pathlib.Path."
        )

    # Validate that the model identifier is a string because model loading
    # helpers require a textual model key.
    if not isinstance(model_id, str):
        # Raise a type error immediately when the model-ID contract is violated.
        raise TypeError(
            "run_gpt_oss_20b_sft_lora_pipeline expected 'model_id' to be a "
            "string."
        )

    # Validate that the quantization flag is boolean because model-loading
    # helpers expect an explicit True/False switch.
    if not isinstance(use_4bit_quantization, bool):
        # Raise a type error immediately when the quantization flag is invalid.
        raise TypeError(
            "run_gpt_oss_20b_sft_lora_pipeline expected "
            "'use_4bit_quantization' to be a boolean."
        )

    # Validate that the run identifier is a string because artifact naming and
    # manifest serialization require a textual run key.
    if not isinstance(run_id, str):
        # Raise a type error immediately when the run-ID contract is violated.
        raise TypeError(
            "run_gpt_oss_20b_sft_lora_pipeline expected 'run_id' to be a "
            "string."
        )

    # Stage 1: Environment validation.

    # Enforce the minimum interpreter version required by the fine-tuning
    # pipeline implementation.
    assert sys.version_info >= (3, 10), "Python 3.10+ required"

    # Define the minimum package versions required to reproduce the reference
    # supervised fine-tuning environment.
    REQUIRED_PACKAGES: Dict[str, str] = {
        "torch": "2.4.1",
        "transformers": "4.46.0",
        "peft": "0.13.0",
        "trl": "0.11.1",
        "accelerate": "0.34.2",
        "datasets": "3.0.1",
    }

    # Audit the dependency substrate against the required minimum versions.
    # This supports reproducible computational-research discipline.
    audit_dependencies(REQUIRED_PACKAGES)

    # Capture the core execution-environment metadata that will be written into
    # the pipeline manifest.
    environment_meta = {
        "python_version": (
            f"{sys.version_info.major}."
            f"{sys.version_info.minor}."
            f"{sys.version_info.micro}"
        ),
        "torch_version": torch.__version__ if ML_AVAILABLE else "N/A",
        "cuda_available": str(torch.cuda.is_available()) if ML_AVAILABLE else "False",
        "n_gpus_detected": str(torch.cuda.device_count()) if ML_AVAILABLE else "0",
    }

    # Stage 2: Data ingestion.

    # Load and validate the JSONL synthetic-supervision dataset.
    # This is the realized input dataset D_syn used by the SFT objective.
    records, ingestion_manifest = load_and_validate_jsonl(
        train_jsonl_path,
        max_ingestion_records,
    )

    # Stage 3: Cluster-aware split.

    # Perform a leakage-controlled split of the validated records according to
    # the configured train/validation/test fractions.
    split = cluster_aware_split(
        records,
        train_fraction,
        val_fraction,
        test_fraction,
        seed,
    )

    # Record the basic split statistics for the manifest.
    split_stats = {
        "n_records_train": len(split.train),
        "n_records_val": len(split.val),
        "n_records_test": len(split.test),
    }

    # Stage 4 & 5: Tokenizer and Model loading.

    # Load the model and tokenizer, delegating dtype resolution and hardware
    # checks to the lower-level helper.
    model, tokenizer = load_model_and_tokenizer(
        model_id=model_id,
        use_4bit_quantization=use_4bit_quantization,
        seed=seed,
    )

    # Build the Hugging Face training dataset from the train split.
    train_hf_dataset = build_hf_dataset(split.train, tokenizer)

    # Build the Hugging Face validation dataset from the validation split.
    val_hf_dataset = build_hf_dataset(split.val, tokenizer)

    # Stage 6: LoRA configuration.

    # Build the LoRA configuration object using the supplied hyperparameters.
    lora_config = build_lora_config(
        lora_rank_r,
        lora_alpha,
        lora_dropout,
        lora_target_modules,
    )

    # Apply the LoRA adapters to the base model.
    model = apply_lora_to_model(model, lora_config)

    # Stage 7 & 8: Trainer construction and execution.

    # Build the supervised fine-tuning trainer over the prepared model, tokenizer,
    # datasets, and optimization hyperparameters.
    trainer = build_sft_trainer(
        model,
        tokenizer,
        train_hf_dataset,
        val_hf_dataset,
        lora_config,
        output_dir,
        per_device_train_batch_size,
        gradient_accumulation_steps,
        num_train_epochs,
        learning_rate,
        max_seq_length,
        seed,
        n_gpus,
    )

    # Execute training and capture the returned training manifest.
    training_manifest = execute_training(
        trainer,
        output_dir,
        run_id,
        model_id,
    )

    # Stage 9: Evaluation.

    # Initialize the evaluation-metrics dictionary to empty in case evaluation
    # is skipped.
    eval_metrics: Dict[str, Any] = {}

    # Only run evaluation when both evaluation questions and the known-page-path
    # universe are available.
    if eval_questions is not None and known_page_paths is not None:
        # Resolve the evaluation dtype to match the model's loaded dtype when
        # available; otherwise fall back to a hardware-aware default.
        eval_dtype = (
            model.dtype
            if hasattr(model, "dtype")
            else (
                torch.bfloat16
                if torch.cuda.is_available() and torch.cuda.is_bf16_supported()
                else torch.float16
            )
        )

        # Construct a deterministic text-generation pipeline for evaluation.
        # This pipeline is later consumed by the evaluation helper.
        eval_pipeline = pipeline(
            "text-generation",
            model=model,
            tokenizer=tokenizer,
            torch_dtype=eval_dtype,
            device_map="auto",
        )

        # Evaluate the model on the provided benchmark or held-out questions.
        eval_results = evaluate_model(
            eval_pipeline,
            eval_questions,
            known_page_paths,
        )

        # Aggregate the raw evaluation outputs into summary metrics.
        eval_metrics = aggregate_eval_metrics(eval_results)

        # Apply the original acceptance gate on citation coverage.
        # This is a guardrail on post-training grounding performance.
        if eval_metrics["citation_coverage_rate"] < 0.70:
            # Emit an error log because the acceptance gate has failed.
            logger.error(
                "ACCEPTANCE GATE FAILED: citation_coverage_rate < 0.70."
            )

    # Stage 10: Adapter merge.

    # Initialize the merged-model artifact path to None in case adapter merging
    # is not requested.
    merged_model_artifact_path: Optional[str] = None

    # Conditionally merge the LoRA adapter into the base model artifact.
    if do_merge_adapter:
        # Merge the adapter into the base model while delegating dtype and
        # hardware handling to the lower-level merge helper.
        merge_adapter_and_archive(
            base_model_id=model_id,
            adapter_path=output_dir,
            merged_output_dir=merged_output_dir,
            tokenizer=tokenizer,
        )

        # Record the merged-model output directory in the manifest.
        merged_model_artifact_path = merged_output_dir

    # Assemble the final pipeline manifest as a structured dictionary.
    pipeline_manifest: Dict[str, Any] = {
        "run_id": run_id,
        "environment": environment_meta,
        "ingestion_manifest": ingestion_manifest,
        "split_stats": split_stats,
        "training_manifest": training_manifest,
        "eval_metrics": eval_metrics,
        "artifact_paths": {
            "adapter_dir": output_dir,
            "merged_model_dir": merged_model_artifact_path,
        },
    }

    # Resolve the pipeline-manifest output path under the adapter output directory.
    manifest_output_path = (
        pathlib.Path(output_dir) / f"pipeline_manifest_{run_id}.json"
    )

    # Ensure that the manifest-output directory exists before writing.
    manifest_output_path.parent.mkdir(parents=True, exist_ok=True)

    # Serialize the pipeline manifest to disk using JSON and ``default=str`` for
    # safe serialization of non-primitive values.
    manifest_output_path.write_text(
        json.dumps(pipeline_manifest, indent=2, default=str),
        encoding="utf-8",
    )

    # Return the complete pipeline manifest to the caller.
    return pipeline_manifest


# Define the Task 18 orchestrator.
def task_18_orchestrator(
    synth_context: "SyntheticDataGenerationContext",
    contract: "FrozenRunContract",
    governance_context: "GovernanceContext",
    wiki_context: "WikiArchitectureContext"
) -> FineTuningContext:
    """
    Orchestrate Task 18 end-to-end by executing the 10-stage supervised
    fine-tuning pipeline.

    Purpose:
        Conditionally execute the optional fine-tuning branch when both:
        1. fine-tuning exploration is enabled in the configuration, and
        2. a synthetic dataset manifest is available.

    Inputs:
        synth_context:
            Synthetic-data generation context exposing the dataset manifest.
        contract:
            Frozen run contract containing fine-tuning and split policies.
        governance_context:
            Governance context exposing the global run ID.
        wiki_context:
            Wiki architecture context exposing the wiki graph.

    Processes:
        1. Validate top-level inputs and interface availability.
        2. Inspect the fine-tuning exploration configuration.
        3. If disabled or synthetic data is missing:
           - log an informational no-op,
           - return an empty FineTuningContext.
        4. Extract:
           - model identifier,
           - LoRA hyperparameters,
           - training hyperparameters,
           - split policy.
        5. Derive the known wiki page path universe for hallucination detection.
        6. Invoke ``run_gpt_oss_20b_sft_lora_pipeline`` with the extracted
           settings.
        7. Return the immutable fine-tuning context.

    Outputs:
        FineTuningContext:
            Immutable aggregate containing the end-to-end pipeline manifest.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.

    Notes:
        This orchestrator is the entry point for the optional fine-tuning
        extension and preserves the original branching logic exactly.
    """
    # Validate that the synthetic-data context exposes the dataset manifest.
    if not hasattr(synth_context, "dataset_manifest"):
        # Raise a type error if the synthetic-data context is invalid.
        raise TypeError(
            "task_18_orchestrator expected 'synth_context' to expose "
            "'dataset_manifest'."
        )

    # Validate that the contract exposes the original configuration.
    if not hasattr(contract, "original_config"):
        # Raise a type error if the contract is invalid.
        raise TypeError(
            "task_18_orchestrator expected 'contract' to expose "
            "'original_config'."
        )

    # Validate that the governance context exposes the run ID.
    if not hasattr(governance_context, "run_id"):
        # Raise a type error if the governance context is invalid.
        raise TypeError(
            "task_18_orchestrator expected 'governance_context' to expose "
            "'run_id'."
        )

    # Validate that the wiki context exposes the wiki graph.
    if not hasattr(wiki_context, "wiki_graph"):
        # Raise a type error if the wiki context is invalid.
        raise TypeError(
            "task_18_orchestrator expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Extract the fine-tuning exploration configuration block.
    ft_config = contract.original_config.get("fine_tuning_exploration", {})

    # Determine whether the fine-tuning branch is enabled.
    is_enabled = ft_config.get("enabled", False)

    # Return early when fine-tuning is disabled or synthetic data is absent.
    if not is_enabled or synth_context.dataset_manifest is None:
        # Emit an informational log preserving the original behavior.
        logger.info(
            "Fine-tuning is disabled or synthetic data is missing. "
            "Skipping Task 18."
        )

        # Return an empty fine-tuning context.
        return FineTuningContext()

    # Extract the fine-tuning method block.
    ft_method = ft_config.get("fine_tuning_method", {})

    # Extract the base model identifier.
    model_id = ft_method.get("base_model_id", "openai/gpt-oss-20b")

    # Extract the LoRA configuration block.
    lora_config = ft_method.get("lora_config", {})

    # Extract the training hyperparameter block.
    train_params = ft_config.get("training_hyperparameters", {})

    # Extract the split policy block.
    split_policy = ft_config.get("train_val_test_split_policy", {})

    # Build the known-page-path universe by excluding unresolved placeholder
    # nodes.
    known_page_paths = frozenset(
        str(node)
        for node in wiki_context.wiki_graph.graph.nodes
        if not str(node).startswith("UNRESOLVED:")
    )

    # Execute the fused 10-stage fine-tuning pipeline.
    pipeline_manifest = run_gpt_oss_20b_sft_lora_pipeline(
        train_jsonl_path=(
            synth_context.dataset_manifest.manifest_path.parent / "train.jsonl"
        ),
        model_id=model_id,
        use_4bit_quantization=False,
        lora_rank_r=lora_config.get("rank_r", 16),
        lora_alpha=lora_config.get("alpha", 32),
        lora_dropout=lora_config.get("dropout", 0.05),
        lora_target_modules=None,
        learning_rate=train_params.get("learning_rate", 1e-4),
        per_device_train_batch_size=1,
        gradient_accumulation_steps=train_params.get("batch_size_effective", 128),
        num_train_epochs=train_params.get("num_epochs", 2),
        weight_decay=train_params.get("weight_decay", 0.01),
        max_seq_length=train_params.get("max_seq_length_tokens", 8192),
        seed=train_params.get("seed", 42),
        train_fraction=split_policy.get("train_fraction", 0.80),
        val_fraction=split_policy.get("val_fraction", 0.10),
        test_fraction=split_policy.get("test_fraction", 0.10),
        output_dir=f"./checkpoints/run_{governance_context.run_id}",
        merged_output_dir=f"./artifacts/merged_{governance_context.run_id}",
        eval_questions=None,
        known_page_paths=known_page_paths,
        do_merge_adapter=True,
        max_ingestion_records=None,
        n_gpus=torch.cuda.device_count() if torch.cuda.is_available() else 1,
        run_id=governance_context.run_id
    )

    # Construct and return the immutable fine-tuning context.
    return FineTuningContext(pipeline_manifest=pipeline_manifest)
````

````python
# Task 19 — Create the orchestrator callable for the end-to-end research pipeline

# ==============================================================================
# Task 19: Create the orchestrator callable for the end-to-end research pipeline
# ==============================================================================

# Configure the module-level orchestrator logger for structured output.
# Construct a named logger for the end-to-end research pipeline.
logger: logging.Logger = logging.getLogger("end_to_end_orchestrator")

# Only attach handlers if the logger is not already configured.
if not logger.handlers:
    # Create a stream handler for console diagnostics.
    handler = logging.StreamHandler()

    # Define a structured formatter with ISO-like timestamps.
    formatter = logging.Formatter(
        "%(asctime)s | %(levelname)-8s | %(name)s | %(message)s",
        datefmt="%Y-%m-%dT%H:%M:%S"
    )

    # Attach the formatter to the handler.
    handler.setFormatter(formatter)

    # Attach the handler to the logger.
    logger.addHandler(handler)

    # Set the logger level to INFO.
    logger.setLevel(logging.INFO)


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 1: Define the orchestrator state and initialization
# -------------------------------------------------------------------------------------------------------------------------------

# Define the mutable orchestrator-state container.
@dataclass
class OrchestratorState:
    """
    Mutable container centralizing the runtime state of the pipeline.

    Purpose:
        Hold the outputs of the initialization tasks (Tasks 1 through 9) so that
        the dynamic operational cycle can proceed without hidden coupling or
        accidental state corruption.

    Inputs:
        This dataclass is populated incrementally by ``initialize_pipeline_state``.

    Processes:
        No computation is performed here. It is a mutable state container whose
        fields are assigned as each initialization task succeeds.

    Outputs:
        contract:
            Frozen run contract produced by Task 1.
        env_context:
            Execution environment context produced by Task 2.
        fs_context:
            Filesystem substrate context produced by Task 3.
        wiki_context:
            Wiki architecture context produced by Task 7.
        gov_context:
            Governance context produced by Task 8.
        prompt_context:
            Prompt registry context produced by Task 9.

    Notes:
        This state container realizes the software-engineering requirement that
        module boundaries remain explicit and that initialization state is not
        hidden in ambient globals.
    """

    # Store the frozen run contract from Task 1.
    contract: Optional["FrozenRunContract"] = None

    # Store the execution environment context from Task 2.
    env_context: Optional["ExecutionEnvironmentContext"] = None

    # Store the filesystem substrate context from Task 3.
    fs_context: Optional["FilesystemSubstrateContext"] = None

    # Store the wiki architecture context from Task 7.
    wiki_context: Optional["WikiArchitectureContext"] = None

    # Store the governance context from Task 8.
    gov_context: Optional["GovernanceContext"] = None

    # Store the prompt registry context from Task 9.
    prompt_context: Optional["PromptContext"] = None


# Define the helper that constructs an initial empty change-detection context.
def _build_initial_change_context(
    repo_root: pathlib.Path
) -> "ChangeDetectionContext":
    """
    Construct a deterministic empty ChangeDetectionContext.

    Purpose:
        Provide the minimum valid ChangeDetectionContext required to initialize
        the wiki architecture before the first ingestion-normalization-change
        detection cycle has been executed.

    Inputs:
        repo_root:
            Absolute path to the repository root.

    Processes:
        1. Validate that ``repo_root`` is a pathlib.Path.
        2. Construct an empty RawManifest with:
           - zero items,
           - empty item index,
           - canonical ``raw/manifest.yaml`` path.
        3. Construct an empty HashRegistry.
        4. Construct an empty DeltaSet.
        5. Wrap them into a ChangeDetectionContext.

    Outputs:
        ChangeDetectionContext:
            Deterministic empty change-detection context.

    Raises:
        TypeError:
            Raised if ``repo_root`` is not a pathlib.Path.

    Notes:
        This helper exists because Task 7 requires a ChangeDetectionContext to
        bind \( \mathcal{D}_{raw} \) before Task 6 has executed in the dynamic
        operational loop.
    """
    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because canonical path construction requires
        # pathlib.Path.
        raise TypeError(
            "_build_initial_change_context expected 'repo_root' to be a "
            "pathlib.Path."
        )

    # Construct the empty raw manifest required by the ChangeDetectionContext.
    raw_manifest = RawManifest(
        items=[],
        item_index={},
        manifest_path=repo_root / "raw" / "manifest.yaml"
    )

    # Construct the empty hash registry.
    hash_registry = HashRegistry(records={})

    # Construct the empty delta set.
    delta_set = DeltaSet(
        changed_artifacts=[],
        new_artifacts=[],
        unchanged_artifacts=[]
    )

    # Wrap the empty components into a deterministic ChangeDetectionContext.
    return ChangeDetectionContext(
        raw_manifest=raw_manifest,
        hash_registry=hash_registry,
        delta_set=delta_set
    )


# Define the pipeline-state initialization callable.
def initialize_pipeline_state(
    raw_config: Dict[str, Any],
    repo_root: pathlib.Path
) -> OrchestratorState:
    """
    Execute Tasks 1 through 9 to initialize the static pipeline state.

    Purpose:
        Run the complete static initialization sequence that must succeed before
        any dynamic dataflow can occur:
        1. configuration validation,
        2. environment validation,
        3. filesystem initialization,
        4. wiki architecture initialization,
        5. governance initialization,
        6. prompt registration.

    Inputs:
        raw_config:
            Raw study configuration dictionary, typically ``STUDY_CONFIG``.
        repo_root:
            Absolute path to the project root.

    Processes:
        1. Validate top-level inputs.
        2. Instantiate an empty OrchestratorState.
        3. Execute Task 1 and bind the frozen contract.
        4. Execute Task 2 and bind the environment context.
        5. Execute Task 3 and bind the filesystem context.
        6. Construct an initial empty ChangeDetectionContext.
        7. Execute Task 7 and bind the wiki architecture context.
        8. Execute Task 8 and bind the governance context.
        9. Execute Task 9 and bind the prompt context.
        10. Return the fully populated OrchestratorState.

    Outputs:
        OrchestratorState:
            Fully initialized static runtime state.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.
        Exception:
            Propagated from any initialization task that fails. Each failure is
            logged as a fatal error before being re-raised.

    Notes:
        This callable is the hard initialization gate for the full research
        pipeline. It enforces the principle that no dynamic state transition may
        occur until the static execution substrate is validated.
    """
    # Validate that the raw configuration is a dictionary.
    if not isinstance(raw_config, dict):
        # Raise a type error because configuration validation requires a
        # dictionary.
        raise TypeError(
            "initialize_pipeline_state expected 'raw_config' to be a "
            "dictionary."
        )

    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path construction requires pathlib.Path.
        raise TypeError(
            "initialize_pipeline_state expected 'repo_root' to be a "
            "pathlib.Path."
        )

    # Instantiate the mutable orchestrator state container.
    state = OrchestratorState()

    # Emit an initialization log message.
    logger.info("Initializing pipeline state...")

    # Task 1: Validate and freeze the configuration contract.
    try:
        # Execute Task 1 and bind its output to the orchestrator state.
        state.contract = task_1_orchestrator(raw_config)

        # Emit a success log for Task 1.
        logger.info("Task 1 (Config Validation) complete.")
    except Exception as e:
        # Emit a fatal error log if Task 1 fails.
        logger.error(f"FATAL: Configuration validation failed: {e}")

        # Re-raise the exception to halt the pipeline.
        raise

    # Task 2: Validate the execution environment and fix reproducibility controls.
    try:
        # Execute Task 2 and bind its output to the orchestrator state.
        state.env_context = task_2_orchestrator(state.contract)

        # Emit a success log for Task 2.
        logger.info("Task 2 (Environment Validation) complete.")
    except Exception as e:
        # Emit a fatal error log if Task 2 fails.
        logger.error(f"FATAL: Environment validation failed: {e}")

        # Re-raise the exception to halt the pipeline.
        raise

    # Task 3: Materialize the filesystem substrate.
    try:
        # Execute Task 3 and bind its output to the orchestrator state.
        state.fs_context = task_3_orchestrator(state.contract)

        # Emit a success log for Task 3.
        logger.info("Task 3 (Filesystem Initialization) complete.")
    except Exception as e:
        # Emit a fatal error log if Task 3 fails.
        logger.error(f"FATAL: Filesystem initialization failed: {e}")

        # Re-raise the exception to halt the pipeline.
        raise

    # Construct the initial empty ChangeDetectionContext to satisfy Task 7's
    # strict type contract before any dynamic ingestion cycle has run.
    initial_change_context = _build_initial_change_context(repo_root)

    # Task 7: Initialize the wiki architecture and graph.
    try:
        # Execute Task 7 and bind its output to the orchestrator state.
        state.wiki_context = task_7_orchestrator(
            change_context=initial_change_context,
            fs_context=state.fs_context,
            contract=state.contract,
            repo_root=repo_root
        )

        # Emit a success log for Task 7.
        logger.info("Task 7 (Wiki Architecture Initialization) complete.")
    except Exception as e:
        # Emit a fatal error log if Task 7 fails.
        logger.error(f"FATAL: Wiki architecture initialization failed: {e}")

        # Re-raise the exception to halt the pipeline.
        raise

    # Task 8: Initialize governance, logging, and write validation.
    try:
        # Execute Task 8 and bind its output to the orchestrator state.
        state.gov_context = task_8_orchestrator(
            contract=state.contract,
            wiki_context=state.wiki_context,
            write_policy=state.fs_context.wiki_layout.write_policy,
            repo_root=repo_root
        )

        # Emit a success log for Task 8.
        logger.info("Task 8 (Governance Initialization) complete.")
    except Exception as e:
        # Emit a fatal error log if Task 8 fails.
        logger.error(f"FATAL: Governance initialization failed: {e}")

        # Re-raise the exception to halt the pipeline.
        raise

    # Task 9: Register prompt templates.
    try:
        # Execute Task 9 and bind its output to the orchestrator state.
        state.prompt_context = task_9_orchestrator(state.contract)

        # Emit a success log for Task 9.
        logger.info("Task 9 (Prompt Registration) complete.")
    except Exception as e:
        # Emit a fatal error log if Task 9 fails.
        logger.error(f"FATAL: Prompt registration failed: {e}")

        # Re-raise the exception to halt the pipeline.
        raise

    # Return the fully populated static orchestrator state.
    return state


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 2: Implement the core operational cycle
# -------------------------------------------------------------------------------------------------------------------------------

# Define the mutable stage-transition ledger.
@dataclass
class StageTransitionLedger:
    """
    Mutable container recording the outputs of the dynamic operational loop.

    Purpose:
        Hold the outputs of the dynamic tasks (Tasks 4 through 16) so the full
        state-transition trace of the operational cycle remains auditable and
        accessible.

    Inputs:
        This dataclass is populated incrementally by ``execute_core_cycle``.

    Processes:
        No computation is performed here. It is a mutable state container whose
        fields are assigned as dynamic tasks complete.

    Outputs:
        ingestion:
            Output of Task 4.
        normalization:
            Output of Task 5.
        change_detection:
            Output of Task 6.
        compilation:
            Output of Task 10.
        index_refresh:
            Output of Task 11.
        linting:
            Output of Task 16.
        search_engine:
            Output of Task 12.
        qa_transactions:
            Ordered list of Task 13 outputs, one per query.
        render_transactions:
            Ordered list of Task 14 outputs, one per query.
        cycle_state:
            Output of Task 15 representing the most recent filing transition.

    Notes:
        This ledger provides the auditable trace of the operational cycle:
        \[
        \mathcal{D}_{raw}^{(t)}
        \xrightarrow{\Phi_{compile}}
        \mathcal{K}^{(t)}
        \xrightarrow{\Phi_{qa}}
        \mathcal{A}^{(t)}
        \xrightarrow{\text{file}}
        \mathcal{K}^{(t+1)}
        \xrightarrow{\Phi_{lint}}
        \mathcal{K}^{(t+1)}_{corrected}
        \]
    """

    # Store the Task 4 ingestion context.
    ingestion: Optional["IngestionContext"] = None

    # Store the Task 5 normalization context.
    normalization: Optional["NormalizationContext"] = None

    # Store the Task 6 change-detection context.
    change_detection: Optional["ChangeDetectionContext"] = None

    # Store the Task 10 compilation context.
    compilation: Optional["CompilationContext"] = None

    # Store the Task 11 index-refresh context.
    index_refresh: Optional["IndexRefreshContext"] = None

    # Store the Task 16 linting context.
    linting: Optional["LintingContext"] = None

    # Store the Task 12 search-engine context.
    search_engine: Optional["SearchEngineContext"] = None

    # Store the ordered Task 13 Q&A transactions.
    qa_transactions: List["QAContext"] = field(default_factory=list)

    # Store the ordered Task 14 rendering transactions.
    render_transactions: List["ArtifactRenderingContext"] = field(
        default_factory=list
    )

    # Store the Task 15 cycle-state record.
    cycle_state: Optional["CycleStateRecord"] = None


# Define the dynamic core-cycle executor.
def execute_core_cycle(
    state: Any,
    repo_root: pathlib.Path,
    acquisition_requests: List[Any],
    queries: List[str],
    requested_formats: List[str],
) -> Any:
    """
    Execute the core operational cycle sequentially.

    Purpose:
        Drive the dynamic stages of the research pipeline in the prescribed order.
        This callable implements the compounding operational loop described in
        the paper, while introducing a short-circuit optimization that bypasses
        ingestion and compilation when no acquisition requests are present.

        The cycle corresponds to the study's recurring transformation:

            D_raw^(t)
            --Φ_compile-->
            K^(t)
            --Φ_qa-->
            A^(t)
            --file-->
            K^(t+1)
            --Φ_lint-->
            K_corrected^(t+1)

    Inputs:
        state:
            Fully initialized orchestrator state expected to expose:
            - ``contract``
            - ``fs_context``
            - ``wiki_context``
            - ``gov_context``
            - ``prompt_context``
        repo_root:
            Absolute project-root path.
        acquisition_requests:
            Ordered list of raw-source acquisition requests.
        queries:
            Ordered list of user questions to answer.
        requested_formats:
            Ordered list of artifact formats to render.

    Processes:
        1. Validate top-level input types.
        2. Validate that required state fields are initialized.
        3. Instantiate a ``StageTransitionLedger`` or a compatible fallback dataclass.
        4. If acquisition requests exist:
           a. execute Task 4 (ingestion),
           b. execute Task 5 (normalization),
           c. execute Task 6 (change detection),
           d. execute Task 10 (incremental compilation).
        5. Otherwise bypass Tasks 4, 5, 6, and 10.
        6. Execute Task 11 (index refresh).
        7. Execute Task 16 (linting).
        8. Execute Task 12 (search engine initialization).
        9. For each query:
           a. execute Task 13 (grounded Q&A),
           b. execute Task 14 (artifact rendering),
           c. execute Task 15 (filing outputs back into the wiki).
        10. If no queries are provided, create a deterministic no-op cycle state.
        11. Return the populated stage-transition ledger.

    Outputs:
        Any:
            Expected to be a ``StageTransitionLedger`` instance.

    Raises:
        TypeError:
            Raised if:
            - ``repo_root`` is not a ``pathlib.Path``,
            - ``acquisition_requests`` is not a list,
            - ``queries`` is not a list,
            - ``requested_formats`` is not a list.
        ValueError:
            Raised if required static state fields are uninitialized.

    Notes:
        This callable preserves the original logic exactly, including:
        - the short-circuit bypass when acquisition_requests is empty,
        - the fallback dataclass construction for ``StageTransitionLedger``,
        - deterministic no-op cycle-state construction when queries are absent.
    """
    # Validate that repo_root is a pathlib.Path because path resolution requires pathlib semantics.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error immediately if the repository-root contract is violated.
        raise TypeError(
            "execute_core_cycle expected 'repo_root' to be a pathlib.Path."
        )

    # Validate that acquisition_requests is a list because ingestion scheduling requires ordered iteration.
    if not isinstance(acquisition_requests, list):
        # Raise a type error because acquisition_requests must be an ordered list.
        raise TypeError(
            "execute_core_cycle expected 'acquisition_requests' to be a list."
        )

    # Validate that queries is a list because query scheduling requires ordered iteration.
    if not isinstance(queries, list):
        # Raise a type error because queries must be an ordered list.
        raise TypeError("execute_core_cycle expected 'queries' to be a list.")

    # Validate that requested_formats is a list because rendering scheduling requires ordered iteration.
    if not isinstance(requested_formats, list):
        # Raise a type error because requested_formats must be an ordered list.
        raise TypeError(
            "execute_core_cycle expected 'requested_formats' to be a list."
        )

    # Validate that the frozen contract has been initialized in the orchestrator state.
    if getattr(state, "contract", None) is None:
        # Raise a value error because the pipeline cannot execute without the run contract.
        raise ValueError(
            "execute_core_cycle requires 'state.contract' to be initialized."
        )

    # Validate that the filesystem context has been initialized in the orchestrator state.
    if getattr(state, "fs_context", None) is None:
        # Raise a value error because filesystem orchestration depends on fs_context.
        raise ValueError(
            "execute_core_cycle requires 'state.fs_context' to be initialized."
        )

    # Validate that the wiki context has been initialized in the orchestrator state.
    if getattr(state, "wiki_context", None) is None:
        # Raise a value error because wiki operations depend on wiki_context.
        raise ValueError(
            "execute_core_cycle requires 'state.wiki_context' to be initialized."
        )

    # Validate that the governance context has been initialized in the orchestrator state.
    if getattr(state, "gov_context", None) is None:
        # Raise a value error because warnings, counters, and run bookkeeping depend on gov_context.
        raise ValueError(
            "execute_core_cycle requires 'state.gov_context' to be initialized."
        )

    # Validate that the prompt context has been initialized in the orchestrator state.
    if getattr(state, "prompt_context", None) is None:
        # Raise a value error because prompt lookup depends on prompt_context.
        raise ValueError(
            "execute_core_cycle requires 'state.prompt_context' to be initialized."
        )

    # Attempt to instantiate the stage-transition ledger from the ambient environment.
    try:
        # Construct the mutable stage-transition ledger using the environment-provided class.
        ledger = StageTransitionLedger()

    # Fall back to a dynamically created compatible dataclass if StageTransitionLedger is unavailable.
    except NameError:
        # Import the dataclass factory locally to preserve the original import topology.
        from dataclasses import make_dataclass

        # Construct a fallback StageTransitionLedger dataclass with the required fields.
        StageTransitionLedger = make_dataclass(
            "StageTransitionLedger",
            [
                ("ingestion", Any, field(default=None)),
                ("normalization", Any, field(default=None)),
                ("change_detection", Any, field(default=None)),
                ("compilation", Any, field(default=None)),
                ("index_refresh", Any, field(default=None)),
                ("linting", Any, field(default=None)),
                ("search_engine", Any, field(default=None)),
                ("qa_transactions", list, field(default_factory=list)),
                ("render_transactions", list, field(default_factory=list)),
                ("cycle_state", Any, field(default=None)),
            ],
        )

        # Instantiate the fallback ledger.
        ledger = StageTransitionLedger()

    # Emit a structured log indicating the start of the core operational cycle.
    logger.info("Starting core operational cycle...")

    # Short-circuit the ingestion-to-compilation pipeline when no acquisition requests are present.
    if acquisition_requests:
        # Emit a structured log for Task 4.
        logger.info("Executing Task 4: Ingestion...")

        # Execute Task 4 to ingest requested raw sources into the local corpus.
        ledger.ingestion = task_4_orchestrator(
            fs_context=state.fs_context,
            acquisition_requests=acquisition_requests,
            repo_root=repo_root,
        )

        # Emit a structured log for Task 5.
        logger.info("Executing Task 5: Normalization...")

        # Execute Task 5 to normalize the ingested raw sources into derivative artifacts.
        ledger.normalization = task_5_orchestrator(
            ingestion_context=ledger.ingestion,
            fs_context=state.fs_context,
            repo_root=repo_root,
        )

        # Emit a structured log for Task 6.
        logger.info("Executing Task 6: Change Detection...")

        # Execute Task 6 to build the raw manifest, compute hashes, and derive the delta set.
        ledger.change_detection = task_6_orchestrator(
            ingestion_context=ledger.ingestion,
            normalization_context=ledger.normalization,
            fs_context=state.fs_context,
            repo_root=repo_root,
        )

        # Emit a structured log for Task 10.
        logger.info("Executing Task 10: Compilation...")

        # Execute Task 10 to incrementally compile changed artifacts into the wiki,
        # explicitly passing the normalization context.
        ledger.compilation = task_10_orchestrator(
            change_context=ledger.change_detection,
            normalization_context=ledger.normalization,
            wiki_context=state.wiki_context,
            prompt_context=state.prompt_context,
            governance_context=state.gov_context,
            repo_root=repo_root,
        )

    # Bypass acquisition-dependent stages when the acquisition queue is empty.
    else:
        # Emit a structured log documenting the intentional short-circuit.
        logger.info(
            "Acquisition queue is empty. Bypassing Ingestion, Normalization, and Compilation."
        )

    # Emit a structured log for Task 11.
    logger.info("Executing Task 11: Index Refresh...")

    # Execute Task 11 to refresh required index structures.
    ledger.index_refresh = task_11_orchestrator(
        wiki_context=state.wiki_context,
        fs_context=state.fs_context,
        governance_context=state.gov_context,
        repo_root=repo_root,
    )

    # Emit a structured log for Task 16.
    logger.info("Executing Task 16: Linting...")

    # Execute Task 16 to run linting and health checks over the wiki.
    ledger.linting = task_16_orchestrator(
        wiki_context=state.wiki_context,
        prompt_context=state.prompt_context,
        contract=state.contract,
        governance_context=state.gov_context,
        cycle_state=None,
    )

    # Emit a structured log for Task 12.
    logger.info("Executing Task 12: Search Engine Initialization...")

    # Execute Task 12 to initialize the deterministic local search interface.
    ledger.search_engine = task_12_orchestrator(
        wiki_context=state.wiki_context,
        contract=state.contract,
    )

    # Process the query-driven branch only when queries are provided.
    if queries:
        # Iterate over each user query in order.
        for query in queries:
            # Emit a structured per-query progress log, truncating for readability.
            logger.info(f"Processing query: '{query[:50]}...'")

            # Task 13: Execute grounded Q&A for the current query.
            qa_context = task_13_orchestrator(
                query=query,
                wiki_context=state.wiki_context,
                search_context=ledger.search_engine,
                prompt_context=state.prompt_context,
                governance_context=state.gov_context,
                contract=state.contract,
            )

            # Append the Q&A context to the ledger's transaction history.
            ledger.qa_transactions.append(qa_context)

            # Task 14: Render requested artifacts from the grounded Q&A result.
            render_context = task_14_orchestrator(
                qa_context=qa_context,
                prompt_context=state.prompt_context,
                fs_context=state.fs_context,
                governance_context=state.gov_context,
                contract=state.contract,
                repo_root=repo_root,
                requested_formats=requested_formats,
            )

            # Append the rendering context to the ledger's render transaction history.
            ledger.render_transactions.append(render_context)

            # Task 15: File rendered outputs back into the wiki and update cycle state.
            ledger.cycle_state = task_15_orchestrator(
                rendering_context=render_context,
                qa_context=qa_context,
                wiki_context=state.wiki_context,
                governance_context=state.gov_context,
                repo_root=repo_root,
            )

    # Synthesize a deterministic no-op cycle state when no queries are provided.
    else:
        # Emit a structured log documenting the intentional bypass.
        logger.info("No queries provided. Skipping Q&A, Rendering, and Filing.")

        # Attempt to construct the environment-provided CycleStateRecord.
        try:
            # Construct the no-op cycle-state record directly.
            cycle_state = CycleStateRecord(
                cycle_id="no_op",
                run_id=state.gov_context.run_id,
                pre_filing_graph_hash="N/A",
                post_filing_graph_hash="N/A",
                artifacts_filed_count=0,
                next_stage_required="None",
            )

        # Fall back to a dynamically created compatible dataclass if CycleStateRecord is unavailable.
        except NameError:
            # Import the dataclass factory locally to preserve the original import topology.
            from dataclasses import make_dataclass

            # Construct a fallback CycleStateRecord dataclass with the required fields.
            CycleStateRecord = make_dataclass(
                "CycleStateRecord",
                [
                    ("cycle_id", str),
                    ("run_id", str),
                    ("pre_filing_graph_hash", str),
                    ("post_filing_graph_hash", str),
                    ("artifacts_filed_count", int),
                    ("next_stage_required", str),
                ],
            )

            # Instantiate the fallback no-op cycle-state record.
            cycle_state = CycleStateRecord(
                "no_op",
                state.gov_context.run_id,
                "N/A",
                "N/A",
                0,
                "None",
            )

        # Store the deterministic no-op cycle state in the ledger.
        ledger.cycle_state = cycle_state

    # Return the populated stage-transition ledger.
    return ledger


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 3: Add control logic and run closeout
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable end-to-end run manifest.
@dataclass(frozen=True)
class EndToEndRunManifest:
    """
    Immutable record encapsulating the entire pipeline execution.

    Purpose:
        Store a single serialized artifact capturing the exact run summary,
        cycle-state transition, and optional fine-tuning outputs for the full
        pipeline execution.

    Inputs:
        This dataclass is populated by ``finalize_pipeline_run``.

    Processes:
        No computation is performed here. It is a passive immutable aggregation
        container.

    Outputs:
        run_id:
            Global run identifier.
        timestamp_utc:
            UTC ISO 8601 timestamp at manifest finalization.
        run_summary:
            Final summarized pipeline counters and timing metadata.
        cycle_state:
            Serialized cycle-state record.
        fine_tuning_manifest:
            Optional serialized fine-tuning context or status.
    """

    # Store the global run identifier.
    run_id: str

    # Store the UTC timestamp at manifest finalization.
    timestamp_utc: str

    # Store the final run-summary dictionary.
    run_summary: Dict[str, Any]

    # Store the serialized cycle-state dictionary.
    cycle_state: Dict[str, Any]

    # Store the optional fine-tuning manifest dictionary.
    fine_tuning_manifest: Optional[Dict[str, Any]]


# Define the finalization and closeout callable.
# Define the pipeline-finalization callable that closes the end-to-end run and serializes the final manifest.
def finalize_pipeline_run(
    state: Any,
    ledger: Any,
    repo_root: pathlib.Path,
) -> Any:
    """
    Execute optional machine-learning branches, finalize the run summary, and
    serialize the end-to-end manifest.

    Purpose:
        Close the pipeline by conditionally executing the optional synthetic-data
        generation and fine-tuning branches, finalizing run-summary counters,
        enforcing a post-run safety-cap audit, and writing a final end-to-end
        manifest to disk.

        This callable is the closeout stage of the orchestrated research cycle.
        Operationally, it sits after the main compounding loop:

            D_raw^(t)
            --Φ_compile-->
            K^(t)
            --Φ_qa-->
            A^(t)
            --file-->
            K^(t+1)
            --Φ_lint-->
            K_corrected^(t+1)

        and consolidates the resulting state into one durable run record.

    Inputs:
        state:
            Orchestrator state expected to expose:
            - ``contract``
            - ``gov_context``
            - ``env_context``
            - ``fs_context``
            - ``wiki_context``
            - ``prompt_context``
        ledger:
            Stage-transition ledger expected to expose ``cycle_state``.
        repo_root:
            Absolute repository root path used for deterministic manifest output.

    Processes:
        1. Validate top-level inputs.
        2. Attempt Task 17 orchestration and log failures without aborting the run.
        3. If Task 17 produces a valid dataset manifest, attempt Task 18 orchestration.
        4. Finalize the run summary through the governance accumulator.
        5. Audit the post-run call-count proxy against the configured safety cap.
        6. Serialize cycle-state and fine-tuning outputs with graceful fallbacks.
        7. Construct the final end-to-end run manifest.
        8. Resolve the logs directory and manifest path.
        9. Serialize the manifest to disk with ``CustomJSONEncoder`` to handle
           nested ``pathlib.Path`` and ``set`` objects.
        10. Return the immutable manifest object.

    Outputs:
        Any:
            Expected to be an ``EndToEndRunManifest`` instance.

    Raises:
        TypeError:
            Raised when ``repo_root`` is not a ``pathlib.Path``.
        OSError:
            Propagated if the final manifest cannot be written to the filesystem.

    Notes:
        The callable preserves the original logic exactly. In particular:
        - Task 17 and Task 18 are optional and failure-tolerant.
        - The safety-cap audit uses ``questions_answered_count`` as the proxy
          compared with ``max_api_calls``.
        - Serialization uses ``asdict`` with graceful fallbacks where needed.
    """
    # Validate that the repository root is a pathlib.Path instance because
    # manifest-path construction requires pathlib semantics.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error immediately when the repository-root contract is violated.
        raise TypeError(
            "finalize_pipeline_run expected 'repo_root' to be a pathlib.Path."
        )

    # Emit a structured informational log indicating that pipeline finalization has begun.
    logger.info("Finalizing pipeline run...")

    # Initialize the optional synthetic-data-generation context to None so that
    # downstream logic can test whether Task 17 succeeded.
    synth_context = None

    # Initialize the optional fine-tuning context to None so that downstream
    # logic can test whether Task 18 succeeded.
    ft_context = None

    # Task 17: Synthetic Data Generation (Optional)
    try:
        # Emit a structured progress log indicating that Task 17 is being executed.
        logger.info("Executing Task 17: Synthetic Data Generation...")

        # Invoke the Task 17 orchestrator directly using the current orchestrator state.
        # This stage corresponds to Algorithm 7 in the paper.
        synth_context = task_17_orchestrator(
            wiki_context=state.wiki_context,
            prompt_context=state.prompt_context,
            contract=state.contract,
            governance_context=state.gov_context,
            repo_root=repo_root,
        )

    # Catch any exception raised by the optional synthetic-data branch so that
    # pipeline closeout can continue even if Task 17 fails or is skipped.
    except Exception as e:
        # Emit a warning while preserving pipeline continuity.
        logger.warning(f"Task 17 failed or was skipped: {e}")

    # Task 18: Fine-Tuning (Optional)
    # Proceed only if Task 17 succeeded and produced a non-empty dataset manifest.
    if (
        synth_context
        and hasattr(synth_context, "dataset_manifest")
        and synth_context.dataset_manifest
    ):
        try:
            # Emit a structured progress log indicating that Task 18 is being executed.
            logger.info("Executing Task 18: Fine-Tuning...")

            # Invoke the Task 18 orchestrator directly using the current state and
            # the output of Task 17.
            ft_context = task_18_orchestrator(
                synth_context=synth_context,
                contract=state.contract,
                governance_context=state.gov_context,
                wiki_context=state.wiki_context,
            )

        # Catch any exception raised by the optional fine-tuning branch so that
        # pipeline closeout can still complete.
        except Exception as e:
            # Emit a warning while preserving manifest generation and run closeout.
            logger.warning(f"Task 18 failed or was skipped: {e}")

    # Finalize the run-summary counters via the governance summary accumulator.
    final_summary = state.gov_context.summary_accumulator.finalize()

    # Extract the configured maximum API-call cap from the reproducibility controls.
    max_calls = state.env_context.reproducibility_controls.max_api_calls

    # Perform the post-run proxy safety-cap check.
    # The configured cap corresponds to the study guardrail:
    # max calls per run = 200.
    if final_summary.questions_answered_count > max_calls:
        # Emit an error log if the proxy call count exceeds the configured cap.
        logger.error(
            "Safety Cap Exceeded: %s calls > %s limit.",
            final_summary.questions_answered_count,
            max_calls,
        )

    # Attempt to serialize the cycle-state record using dataclasses.asdict.
    try:
        # Convert the cycle state to a dictionary when it is available and serializable.
        cycle_dict = asdict(ledger.cycle_state) if ledger.cycle_state else {}

    # Fall back gracefully when the cycle-state object cannot be serialized.
    except TypeError:
        # Emit a diagnostic fallback dictionary rather than crashing finalization.
        cycle_dict = {"status": "Cycle state not serializable"}

    # Attempt to serialize the fine-tuning context using dataclasses.asdict.
    try:
        # Convert the fine-tuning context to a dictionary when it exists and is serializable.
        ft_dict = asdict(ft_context) if ft_context else None

    # Fall back gracefully when the fine-tuning context cannot be serialized.
    except TypeError:
        # Emit a diagnostic fallback dictionary rather than crashing finalization.
        ft_dict = {"status": "Fine-tuning context not serializable"}

    # Construct the final immutable end-to-end run manifest.
    # We assume EndToEndRunManifest is available in the environment.
    manifest = EndToEndRunManifest(
        run_id=state.gov_context.run_id,
        timestamp_utc=datetime.datetime.utcnow().isoformat() + "Z",
        run_summary=asdict(final_summary),
        cycle_state=cycle_dict,
        fine_tuning_manifest=ft_dict,
    )

    # Resolve the logs directory from the filesystem context.
    logs_dir = state.fs_context.root_states["logs_directory"].absolute_path

    # Construct the final manifest output path using the deterministic run ID.
    manifest_path = logs_dir / f"end_to_end_manifest_{state.gov_context.run_id}.json"

    # Open the final manifest path for UTF-8 writing.
    with open(manifest_path, "w", encoding="utf-8") as f:
        # Serialize the manifest dictionary to JSON using the custom encoder so
        # that pathlib.Path and set objects do not trigger serialization errors.
        json.dump(
            asdict(manifest),
            f,
            cls=CustomJSONEncoder,
            indent=2,
            sort_keys=True,
        )

    # Emit a structured success log naming the written manifest path.
    logger.info(f"End-to-End Run Manifest written to: {manifest_path}")

    # Return the immutable manifest object to the caller.
    return manifest


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the master end-to-end research pipeline callable.
def end_to_end_research_pipeline(
    raw_config: Dict[str, Any],
    repo_root: pathlib.Path,
    acquisition_requests: List["RawAcquisitionRequest"],
    queries: List[str],
    requested_formats: List[str]
) -> EndToEndRunManifest:
    """
    Execute the full end-to-end LLM Knowledge Base research pipeline.

    Purpose:
        Serve as the master orchestrator for the complete research and decision
        memory system by executing:
        1. static initialization tasks,
        2. dynamic operational-cycle tasks,
        3. optional ML-extension tasks,
        4. final run closeout.

    Inputs:
        raw_config:
            Raw configuration dictionary, typically ``STUDY_CONFIG``.
        repo_root:
            Absolute path to the repository root.
        acquisition_requests:
            Ordered list of raw-source acquisition requests.
        queries:
            Ordered list of user questions to answer.
        requested_formats:
            Ordered list of artifact formats to render, such as:
            - ``"markdown"``
            - ``"marp"``
            - ``"figure"``

    Processes:
        1. Validate top-level inputs.
        2. Emit a pipeline-start log.
        3. Execute static initialization via ``initialize_pipeline_state``.
        4. Execute the dynamic operational cycle via ``execute_core_cycle``.
        5. Execute optional ML branches and run closeout via
           ``finalize_pipeline_run``.
        6. Emit a pipeline-complete log.
        7. Return the final immutable run manifest.

    Outputs:
        EndToEndRunManifest:
            Final immutable serialized record of the complete run.

    Raises:
        TypeError:
            Raised if top-level inputs are invalid.
        Exception:
            Propagated from any upstream task that fails.

    Notes:
        This callable is the full implementation of the continuously compounding
        research-memory system and operationalizes the end-to-end cycle:
        \[
        \mathcal{D}_{raw}^{(t)}
        \xrightarrow{\Phi_{compile}}
        \mathcal{K}^{(t)}
        \xrightarrow{\Phi_{qa}}
        \mathcal{A}^{(t)}
        \xrightarrow{\text{file}}
        \mathcal{K}^{(t+1)}
        \xrightarrow{\Phi_{lint}}
        \mathcal{K}^{(t+1)}_{corrected}
        \]
        while also optionally extending the system through synthetic-data
        generation and fine-tuning.
    """
    # Validate that the raw configuration is a dictionary.
    if not isinstance(raw_config, dict):
        # Raise a type error because the pipeline requires a configuration
        # dictionary.
        raise TypeError(
            "end_to_end_research_pipeline expected 'raw_config' to be a "
            "dictionary."
        )

    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        # Raise a type error because path resolution requires pathlib.Path.
        raise TypeError(
            "end_to_end_research_pipeline expected 'repo_root' to be a "
            "pathlib.Path."
        )

    # Validate that acquisition_requests is a list.
    if not isinstance(acquisition_requests, list):
        # Raise a type error because ingestion requires an ordered request list.
        raise TypeError(
            "end_to_end_research_pipeline expected 'acquisition_requests' to "
            "be a list."
        )

    # Validate that queries is a list.
    if not isinstance(queries, list):
        # Raise a type error because Q&A dispatch requires an ordered list.
        raise TypeError(
            "end_to_end_research_pipeline expected 'queries' to be a list."
        )

    # Validate that requested_formats is a list.
    if not isinstance(requested_formats, list):
        # Raise a type error because rendering dispatch requires a list.
        raise TypeError(
            "end_to_end_research_pipeline expected 'requested_formats' to be "
            "a list."
        )

    # Emit the pipeline start log.
    logger.info("=== STARTING END-TO-END RESEARCH PIPELINE ===")

    # Step 1: Initialize static state (Tasks 1-9).
    state = initialize_pipeline_state(raw_config, repo_root)

    # Step 2: Execute the core operational cycle (Tasks 4-16).
    ledger = execute_core_cycle(
        state=state,
        repo_root=repo_root,
        acquisition_requests=acquisition_requests,
        queries=queries,
        requested_formats=requested_formats
    )

    # Step 3: Execute the optional ML branches and finalize the run (Tasks 17-18).
    manifest = finalize_pipeline_run(
        state=state,
        ledger=ledger,
        repo_root=repo_root
    )

    # Emit the pipeline completion log.
    logger.info("=== END-TO-END RESEARCH PIPELINE COMPLETE ===")

    # Return the final immutable run manifest.
    return manifest
````

````python
# Task 20 — Execute the orchestrator and conduct the evaluation and robustness analyses

# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 1: Evaluate reproducibility via repeated runs
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable reproducibility comparison bundle.
@dataclass(frozen=True)
class ReproducibilityComparisonBundle:
    """
    Immutable record capturing the comparison between two distinct runs of the
    orchestrator under identical initial conditions.

    Purpose:
        Formalize reproducibility evaluation by storing whether two identical
        pipeline executions agree along the dimensions emphasized by the study:
        1. environment consistency,
        2. schema-validity consistency,
        3. citation-correctness consistency,
        4. write-scope discipline consistency,
        5. retrieval consistency.

    Inputs:
        This dataclass is populated by ``evaluate_reproducibility``.

    Processes:
        No computation is performed here. It is a passive immutable comparison
        record.

    Outputs:
        run_1_id:
            Run identifier of the first execution.
        run_2_id:
            Run identifier of the second execution.
        environment_match:
            Whether the captured environment metadata matches.
        schema_validity_match:
            Whether the structural write outputs agree.
        citation_correctness_match:
            Whether citation correctness proxies agree within tolerance.
        write_scope_discipline_match:
            Whether filed-artifact counts match.
        retrieval_consistency_match:
            Whether retrieval-efficiency proxies match within tolerance.
        detailed_comparisons:
            Dictionary containing raw side-by-side comparison values.

    Notes:
        This record operationalizes reproducibility at the semantic and
        structural level rather than requiring byte-identical output from
        hosted proprietary models.
    """

    # Store the identifier of the first run.
    run_1_id: str

    # Store the identifier of the second run.
    run_2_id: str

    # Store whether the captured environment metadata matches.
    environment_match: bool

    # Store whether schema-validity proxies match.
    schema_validity_match: bool

    # Store whether citation-correctness proxies match.
    citation_correctness_match: bool

    # Store whether write-scope discipline proxies match.
    write_scope_discipline_match: bool

    # Store whether retrieval consistency proxies match.
    retrieval_consistency_match: bool

    # Store the raw detailed comparison payload.
    detailed_comparisons: Dict[str, Any]


# Define the reproducibility evaluation callable.
# Define the deep-comparison helper used for reproducibility checks.
def _deep_compare_unordered_lists(
    obj1: Any,
    obj2: Any,
) -> bool:
    """
    Recursively compare two objects, treating lists of hashable items as sets.

    Purpose:
        Neutralize ordering differences that can arise when unordered Python
        containers, especially sets, are serialized into lists. This helps avoid
        false reproducibility failures during repeated-run comparison.

    Inputs:
        obj1:
            The first object to compare.
        obj2:
            The second object to compare.

    Processes:
        1. If both objects are dictionaries:
           - verify key-set equality,
           - recurse value-by-value.
        2. If both objects are lists:
           - verify length equality,
           - attempt set-based comparison when items are hashable,
           - otherwise sort by string representation and recurse element-wise.
        3. For all other object types, use ordinary equality.

    Outputs:
        bool:
            True if the two objects are deeply equivalent under the list-order
            normalization rule, else False.

    Notes:
        This utility is specifically used by reproducibility evaluation to
        compare environment metadata and other manifest substructures where set
        serialization order should not be treated as drift.
    """
    # Check whether both inputs are dictionaries.
    if isinstance(obj1, dict) and isinstance(obj2, dict):
        # Return False immediately if the key sets differ.
        if obj1.keys() != obj2.keys():
            # The dictionary structures differ, so the objects are not equivalent.
            return False

        # Recurse over all shared keys and require all nested values to match.
        return all(
            _deep_compare_unordered_lists(obj1[k], obj2[k])
            for k in obj1
        )

    # Check whether both inputs are lists.
    elif isinstance(obj1, list) and isinstance(obj2, list):
        # Return False immediately if the list lengths differ.
        if len(obj1) != len(obj2):
            # Different lengths imply non-equivalence.
            return False

        try:
            # Attempt an order-invariant comparison by converting both lists to sets.
            return set(obj1) == set(obj2)

        # Fall back when list elements are unhashable, such as dictionaries.
        except TypeError:
            # Sort the first list deterministically by string representation.
            sorted_obj1 = sorted(obj1, key=str)

            # Sort the second list deterministically by string representation.
            sorted_obj2 = sorted(obj2, key=str)

            # Recurse pairwise over the sorted elements.
            return all(
                _deep_compare_unordered_lists(i1, i2)
                for i1, i2 in zip(sorted_obj1, sorted_obj2)
            )

    # For all other types, use standard equality.
    else:
        # Return the ordinary equality result.
        return obj1 == obj2


# Define the reproducibility-evaluation callable.
def evaluate_reproducibility(
    raw_config: Dict[str, Any],
    repo_root: Any,
    acquisition_requests: List[Any],
    queries: List[str],
    requested_formats: List[str],
) -> Any:
    """
    Run the orchestrator twice on a frozen corpus snapshot and evaluate reproducibility.

    Purpose:
        Execute the full end-to-end pipeline twice under identical inputs and
        compare the resulting run manifests at the appropriate reproducibility
        level for hosted-model systems.

        This callable does not require byte-identical outputs. Instead, it
        checks higher-level invariants aligned with the paper's evaluation
        guidance:
        - environment consistency,
        - schema-validity stability,
        - citation-behavior stability,
        - write-scope discipline,
        - retrieval consistency.

    Inputs:
        raw_config:
            Raw configuration dictionary used for both pipeline runs.
        repo_root:
            Repository root path supplied to the pipeline.
        acquisition_requests:
            Ordered list of raw acquisition requests.
        queries:
            Ordered list of benchmark or user queries.
        requested_formats:
            Ordered list of requested artifact formats.

    Processes:
        1. Execute Run 1 of the full end-to-end pipeline.
        2. Execute Run 2 of the full end-to-end pipeline.
        3. Convert both manifest objects to dictionaries using ``asdict``.
        4. Compare environment metadata using
           ``_deep_compare_unordered_lists``.
        5. Compare schema-validity proxies:
           - ``wiki_files_written``
           - ``wiki_files_edited``
        6. Compare citation correctness proxy using ``math.isclose`` with
           ``abs_tol = 0.05``.
        7. Compare write-scope discipline using filed-artifact counts.
        8. Compare retrieval consistency using pages-read proxy with
           ``abs_tol = 0.5``.
        9. Build the side-by-side comparison payload.
        10. Return the immutable reproducibility bundle.

    Outputs:
        Any:
            Expected to be a ``ReproducibilityComparisonBundle`` instance.

    Raises:
        Exception:
            Propagated from either execution of the end-to-end pipeline.

    Notes:
        This function preserves the original logic exactly. It assumes
        ``end_to_end_research_pipeline`` and ``ReproducibilityComparisonBundle``
        are available in the runtime environment.
    """
    # Emit a log indicating that the first reproducibility run is starting.
    logger.info("Starting Reproducibility Evaluation (Run 1)...")

    # Execute the full end-to-end pipeline for the first run.
    # We assume end_to_end_research_pipeline is available in the environment.
    manifest_1 = end_to_end_research_pipeline(
        raw_config=raw_config,
        repo_root=repo_root,
        acquisition_requests=acquisition_requests,
        queries=queries,
        requested_formats=requested_formats,
    )

    # Emit a log indicating that the second reproducibility run is starting.
    logger.info("Starting Reproducibility Evaluation (Run 2)...")

    # Execute the full end-to-end pipeline for the second run.
    manifest_2 = end_to_end_research_pipeline(
        raw_config=raw_config,
        repo_root=repo_root,
        acquisition_requests=acquisition_requests,
        queries=queries,
        requested_formats=requested_formats,
    )

    # Emit a log indicating that manifest comparison is beginning.
    logger.info("Comparing Run 1 and Run 2 Manifests...")

    # Convert the first manifest dataclass to a dictionary.
    m1_dict = asdict(manifest_1)

    # Convert the second manifest dataclass to a dictionary.
    m2_dict = asdict(manifest_2)

    # 1. Environment Match
    # Extract the environment metadata dictionary from Run 1.
    env1 = m1_dict.get("environment", {})

    # Extract the environment metadata dictionary from Run 2.
    env2 = m2_dict.get("environment", {})

    # Compare the environment metadata using deep order-invariant comparison.
    environment_match = _deep_compare_unordered_lists(env1, env2)

    # Emit a warning if environment drift is detected across the two runs.
    if not environment_match:
        # Log the detected environment mismatch.
        logger.warning("Environment drift detected between Run 1 and Run 2.")

    # 2. Schema Validity Match (Files written/edited)
    # Extract the run-summary dictionary from Run 1.
    sum1 = m1_dict.get("run_summary", {})

    # Extract the run-summary dictionary from Run 2.
    sum2 = m2_dict.get("run_summary", {})

    # Compare the wiki-written and wiki-edited file counts across runs.
    schema_validity_match = (
        sum1.get("wiki_files_written") == sum2.get("wiki_files_written")
        and sum1.get("wiki_files_edited") == sum2.get("wiki_files_edited")
    )

    # 3. Citation Correctness Match (CCR proxy)
    # Extract the fine-tuning sub-manifest from Run 1, defaulting safely.
    ft1 = m1_dict.get("fine_tuning_manifest", {}) or {}

    # Extract the fine-tuning sub-manifest from Run 2, defaulting safely.
    ft2 = m2_dict.get("fine_tuning_manifest", {}) or {}

    # Extract the citation-coverage proxy from Run 1.
    ccr1 = ft1.get("eval_metrics", {}).get("citation_coverage_rate", 0.0)

    # Extract the citation-coverage proxy from Run 2.
    ccr2 = ft2.get("eval_metrics", {}).get("citation_coverage_rate", 0.0)

    # Compare the citation proxy values using an absolute tolerance of 0.05.
    citation_correctness_match = math.isclose(ccr1, ccr2, abs_tol=0.05)

    # 4. Write-Scope Discipline Match (Delta Locality proxy)
    # Extract the cycle-state dictionary from Run 1.
    cyc1 = m1_dict.get("cycle_state", {})

    # Extract the cycle-state dictionary from Run 2.
    cyc2 = m2_dict.get("cycle_state", {})

    # Compare the filed-artifact counts across runs.
    write_scope_discipline_match = (
        cyc1.get("artifacts_filed_count") == cyc2.get("artifacts_filed_count")
    )

    # 5. Retrieval Consistency Match (Pages Read proxy)
    # Extract the pages-read proxy from Run 1.
    pr1 = ft1.get("eval_metrics", {}).get("pages_read_per_answer", 0.0)

    # Extract the pages-read proxy from Run 2.
    pr2 = ft2.get("eval_metrics", {}).get("pages_read_per_answer", 0.0)

    # Compare the pages-read proxy using an absolute tolerance of 0.5.
    retrieval_consistency_match = math.isclose(pr1, pr2, abs_tol=0.5)

    # Build the detailed side-by-side comparison payload.
    detailed_comparisons = {
        "files_written": (
            sum1.get("wiki_files_written"),
            sum2.get("wiki_files_written"),
        ),
        "ccr": (ccr1, ccr2),
        "artifacts_filed": (
            cyc1.get("artifacts_filed_count"),
            cyc2.get("artifacts_filed_count"),
        ),
        "pages_read": (pr1, pr2),
    }

    # Construct and return the reproducibility comparison bundle.
    # We assume ReproducibilityComparisonBundle is available in the environment.
    return ReproducibilityComparisonBundle(
        run_1_id=m1_dict.get("run_id", "unknown_1"),
        run_2_id=m2_dict.get("run_id", "unknown_2"),
        environment_match=environment_match,
        schema_validity_match=schema_validity_match,
        citation_correctness_match=citation_correctness_match,
        write_scope_discipline_match=write_scope_discipline_match,
        retrieval_consistency_match=retrieval_consistency_match,
        detailed_comparisons=detailed_comparisons,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 2: Evaluate the core system on configured benchmarks
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable per-question evaluation row.
@dataclass(frozen=True)
class SystemEvaluationRow:
    """
    Immutable record capturing per-question evaluation metrics.

    Purpose:
        Store the answer text, consulted pages, latency, citation counts, and
        uncertainty markers for one benchmark question.

    Inputs:
        This dataclass is populated by ``evaluate_core_system``.

    Processes:
        No computation is performed here. It is a passive immutable per-question
        evaluation record.

    Outputs:
        question_id:
            Benchmark question identifier.
        benchmark_split:
            Benchmark split label.
        is_multi_doc:
            Whether the question is multi-document.
        answer_text:
            Raw answer text returned by the Q&A pipeline.
        consulted_pages:
            Ordered list of wiki pages consulted during routing and reading.
        tool_call_count:
            Proxy count of tool calls derived from the number of routed search
            queries.
        latency_seconds:
            End-to-end latency for the question.
        substantive_claim_count:
            Heuristic count of substantive claims.
        cited_claim_count:
            Heuristic count of cited claims.
        correct_citation_count:
            Heuristic count of correctly grounded cited claims.
        unsupported_claim_count:
            Heuristic count of unsupported claims.
        unknown_marking_used:
            Whether the answer explicitly signaled missing information.
    """

    # Store the benchmark question identifier.
    question_id: str

    # Store the benchmark split name.
    benchmark_split: str

    # Store whether the question is multi-document.
    is_multi_doc: bool

    # Store the raw answer text.
    answer_text: str

    # Store the ordered list of consulted wiki pages.
    consulted_pages: List[str]

    # Store the proxy tool-call count.
    tool_call_count: int

    # Store the measured latency in seconds.
    latency_seconds: float

    # Store the heuristic number of substantive claims.
    substantive_claim_count: int

    # Store the heuristic number of cited claims.
    cited_claim_count: int

    # Store the heuristic number of correct citations.
    correct_citation_count: int

    # Store the heuristic number of unsupported claims.
    unsupported_claim_count: int

    # Store whether unknown-style fallback marking was used.
    unknown_marking_used: bool


# Define the immutable aggregate system evaluation report.
@dataclass(frozen=True)
class SystemEvaluationReport:
    """
    Immutable record capturing the aggregate KPIs for the system evaluation.

    Purpose:
        Store the benchmark-wide scalar metrics produced by evaluating the core
        Q&A system on a question set.

    Inputs:
        This dataclass is populated by ``evaluate_core_system``.

    Processes:
        No computation is performed here. It is a passive immutable aggregate
        evaluation report.

    Outputs:
        ccr_coverage:
            Citation coverage rate.
        ccr_correct:
            Citation correctness rate.
        avg_pages_read:
            Average number of consulted pages per answer.
        avg_tool_calls:
            Average number of tool calls per answer.
        avg_latency:
            Average latency in seconds per answer.
        total_hallucinations:
            Aggregate unsupported-claim count.
        unknown_marking_rate:
            Rate of unknown marking among unsupported-answer cases.
        rows:
            Ordered list of per-question evaluation rows.
    """

    # Store the citation coverage rate.
    ccr_coverage: float

    # Store the citation correctness rate.
    ccr_correct: float

    # Store the average number of pages read per answer.
    avg_pages_read: float

    # Store the average number of tool calls per answer.
    avg_tool_calls: float

    # Store the average latency per answer.
    avg_latency: float

    # Store the total hallucination count.
    total_hallucinations: int

    # Store the unknown marking rate.
    unknown_marking_rate: float

    # Store the ordered per-question evaluation rows.
    rows: List[SystemEvaluationRow]


# Define the internal claim-and-citation extraction helper.
def _extract_claims_and_citations(
    answer_text: str,
    shortlisted_paths: List[str]
) -> Tuple[int, int, int, int, bool]:
    """
    Apply a heuristic proxy for claim and citation analysis within a generated
    answer.

    Purpose:
        Estimate:
        1. substantive claims,
        2. cited claims,
        3. correct citations,
        4. unsupported claims,
        5. unknown-style fallback usage,

        using rule-based parsing rather than human annotation.

    Inputs:
        answer_text:
            Full answer text produced by the Q&A stage.
        shortlisted_paths:
            Ordered list of valid shortlisted page paths.

    Processes:
        1. Validate input types.
        2. Isolate the Evidence section.
        3. Split the Evidence section into candidate claim sentences.
        4. For each candidate sentence:
           - extract wikilinks,
           - count cited claims,
           - count correctly grounded citations,
           - count unsupported claims.
        5. Isolate the Caveats section.
        6. Detect unknown-style fallback usage by lexical heuristics.
        7. Return the five computed quantities.

    Outputs:
        Tuple[int, int, int, int, bool]:
            Tuple containing:
            - substantive_claim_count,
            - cited_claim_count,
            - correct_citation_count,
            - unsupported_claim_count,
            - unknown_marking_used.

    Raises:
        TypeError:
            Raised if inputs are invalid.

    Notes:
        This helper provides a deterministic proxy for evaluating grounding and
        unsupported-claim behavior.
    """
    # Validate that the answer text is a string.
    if not isinstance(answer_text, str):
        # Raise a type error because text parsing requires a string.
        raise TypeError(
            "_extract_claims_and_citations expected 'answer_text' to be a "
            "string."
        )

    # Validate that shortlisted_paths is a list.
    if not isinstance(shortlisted_paths, list):
        # Raise a type error because citation validation requires a list of
        # valid paths.
        raise TypeError(
            "_extract_claims_and_citations expected 'shortlisted_paths' to be "
            "a list of strings."
        )

    # Isolate the Evidence section from the answer text.
    evidence_match = re.search(
        r"^#+\s+.*Evidence.*$(.*?)(?:^#+\s+.*Caveats.*$|\Z)",
        answer_text,
        re.MULTILINE | re.IGNORECASE | re.DOTALL
    )

    # Extract the Evidence body or default to an empty string.
    evidence_text = evidence_match.group(1) if evidence_match else ""

    # Heuristically treat each sufficiently long sentence in the Evidence
    # section as a substantive claim.
    sentences = [
        s.strip()
        for s in re.split(r"(?<=[.!?])\s+", evidence_text)
        if len(s.strip()) > 10
    ]

    # Count substantive claims.
    substantive_claim_count = len(sentences)

    # Initialize the cited-claim counter.
    cited_claim_count = 0

    # Initialize the correct-citation counter.
    correct_citation_count = 0

    # Initialize the unsupported-claim counter.
    unsupported_claim_count = 0

    # Convert the shortlist to a set for O(1) membership checks.
    valid_paths = set(shortlisted_paths)

    # Iterate over each candidate sentence.
    for sentence in sentences:
        # Extract all wikilinks in the sentence.
        links = re.findall(r"\[\[(.*?)\]\]", sentence)

        # If the sentence has at least one citation, treat it as a cited claim.
        if links:
            # Increment the cited-claim counter.
            cited_claim_count += 1

            # Initialize the correctness flag for this sentence.
            is_correct = True

            # Validate every citation in the sentence against the shortlist.
            for link in links:
                # Strip aliases and surrounding whitespace.
                clean_path = link.split("|")[0].strip()

                # Mark the sentence incorrect if any citation is not in the
                # shortlist.
                if clean_path not in valid_paths:
                    is_correct = False
                    break

            # Count the sentence as correctly cited or unsupported accordingly.
            if is_correct:
                correct_citation_count += 1
            else:
                unsupported_claim_count += 1
        else:
            # Sentences without citations are counted as unsupported.
            unsupported_claim_count += 1

    # Isolate the Caveats section from the answer text.
    caveats_match = re.search(
        r"^#+\s+.*Caveats.*$(.*?)(?:^#+\s+.*Pages Consulted.*$|\Z)",
        answer_text,
        re.MULTILINE | re.IGNORECASE | re.DOTALL
    )

    # Extract the Caveats body or default to an empty string.
    caveats_text = caveats_match.group(1) if caveats_match else ""

    # Detect unknown-style fallback usage lexically.
    unknown_marking_used = bool(
        re.search(
            r"\b(unknown|missing|not provided)\b",
            caveats_text,
            re.IGNORECASE
        )
    )

    # Return the full heuristic claim/citation tuple.
    return (
        substantive_claim_count,
        cited_claim_count,
        correct_citation_count,
        unsupported_claim_count,
        unknown_marking_used,
    )


# Define the core system evaluation callable.
def evaluate_core_system(
    questions: List[Any],
    wiki_context: Any,
    search_context: Any,
    prompt_context: Any,
    governance_context: Any,
    contract: Any,
    benchmark_split: str,
) -> Any:
    """
    Evaluate the core grounded Q&A system on a benchmark question set and
    compute aggregate KPIs.

    Purpose:
        Run the grounded Q&A system over a benchmark suite and aggregate the
        evaluation metrics specified in the paper's evaluation framework. The
        main KPIs computed by this callable are:

            CCR_coverage =
                (# substantive claims with citations) /
                (# substantive claims)

            CCR_correct =
                (# citations judged correct) /
                (# citations checked)

            P̄ =
                (1 / N) * Σ_i |P_q_i|

            T̄ =
                (1 / N) * Σ_i tool_calls(q_i)

            L̄ =
                (1 / N) * Σ_i latency(q_i)

            H =
                # {unsupported claims}

            U =
                (# unsupported cases marked as Unknown) /
                (# unsupported cases)

    Inputs:
        questions:
            Ordered list of evaluation-question objects.
        wiki_context:
            Wiki context used during Q&A orchestration.
        search_context:
            Search context used during routing.
        prompt_context:
            Prompt context used during synthesis.
        governance_context:
            Governance context used for warning accounting and execution context.
        contract:
            Frozen run contract used during Q&A orchestration.
        benchmark_split:
            Benchmark split label, e.g. ``heldout_qa_set`` or
            ``regression_questions_from_real_usage``.

    Processes:
        1. Validate top-level input types.
        2. Initialize the per-question evaluation-row accumulator.
        3. Iterate over all benchmark questions.
        4. For each question:
           - invoke the grounded Q&A orchestrator,
           - extract routing and synthesis outputs,
           - extract latency from token-usage metadata,
           - compute claim/citation heuristics using
             ``_extract_claims_and_citations``.
        5. Gracefully handle per-question failures and continue evaluation.
        6. Aggregate all KPI totals and averages.
        7. Round the aggregate metrics as in the original logic.
        8. Return the immutable evaluation report.

    Outputs:
        Any:
            Expected to be a ``SystemEvaluationReport`` instance.

    Raises:
        TypeError:
            Raised when ``questions`` is not a list or ``benchmark_split`` is
            not a string.

    Notes:
        This callable assumes the availability of:
        - ``task_13_orchestrator``
        - ``_extract_claims_and_citations``
        - ``SystemEvaluationRow``
        - ``SystemEvaluationReport``
    """
    # Validate that the questions input is a list because evaluation iterates in ordered fashion.
    if not isinstance(questions, list):
        # Raise a type error because the benchmark suite must be list-like and ordered.
        raise TypeError("evaluate_core_system expected 'questions' to be a list.")

    # Validate that the benchmark split label is a string.
    if not isinstance(benchmark_split, str):
        # Raise a type error because benchmark split labels must be strings.
        raise TypeError(
            "evaluate_core_system expected 'benchmark_split' to be a string."
        )

    # Initialize the ordered per-question evaluation-row accumulator.
    rows: List[Any] = []

    # Iterate over every evaluation question in the benchmark suite.
    for q in questions:
        # Extract the question identifier, using a safe fallback when absent.
        q_id = getattr(q, "question_id", "unknown")

        # Extract the question text, using an empty-string fallback when absent.
        q_text = getattr(q, "question", "")

        # Extract the multi-document flag, defaulting to False when absent.
        is_multi = getattr(q, "is_multi_doc", False)

        try:
            # Invoke the grounded Q&A orchestration pipeline for the current benchmark question.
            # We assume task_13_orchestrator is available in the environment.
            qa_context = task_13_orchestrator(
                query=q_text,
                wiki_context=wiki_context,
                search_context=search_context,
                prompt_context=prompt_context,
                governance_context=governance_context,
                contract=contract,
            )

            # Extract the routing state from the returned Q&A context.
            routing_state = getattr(qa_context, "routing_state", None)

            # Extract the synthesis package from the returned Q&A context.
            synth_package = getattr(qa_context, "synthesis_package", None)

            # Extract the shortlisted page paths from the routing state, defaulting safely.
            shortlisted_paths = (
                getattr(routing_state, "shortlisted_paths", [])
                if routing_state
                else []
            )

            # Extract the issued search queries from the routing state, defaulting safely.
            search_queries = (
                getattr(routing_state, "search_queries_issued", [])
                if routing_state
                else []
            )

            # Use the number of issued search queries as the tool-call-count proxy.
            tool_call_count = len(search_queries)

            # Extract the raw answer text from the synthesis package, defaulting safely.
            answer_text = (
                getattr(synth_package, "raw_response_text", "")
                if synth_package
                else ""
            )

            # Extract the usage dictionary from the synthesis package, defaulting safely.
            usage_dict = (
                getattr(synth_package, "token_usage", {})
                if synth_package
                else {}
            )

            # Extract true latency from the usage dictionary, defaulting to 0.0.
            latency = usage_dict.get("latency_seconds", 0.0)

            # Compute heuristic claim and citation metrics from the answer text and shortlisted pages.
            # We assume _extract_claims_and_citations is available in the environment.
            (
                sub_claims,
                cited_claims,
                correct_cites,
                unsupp_claims,
                unknown_used,
            ) = _extract_claims_and_citations(answer_text, shortlisted_paths)

        # Handle individual benchmark-question failures without aborting the entire evaluation suite.
        except Exception as e:
            # Emit an error log identifying the failed evaluation question.
            logger.error(f"Evaluation failed for question {q_id}: {e}")

            # Store a diagnostic error string as the answer text.
            answer_text = f"ERROR: Pipeline failed - {e}"

            # Fall back to an empty consulted-page list.
            shortlisted_paths = []

            # Fall back to zero tool calls.
            tool_call_count = 0

            # Fall back to zero latency.
            latency = 0.0

            # Fall back to zero substantive claims.
            sub_claims = 0

            # Fall back to zero cited claims.
            cited_claims = 0

            # Fall back to zero correct citations.
            correct_cites = 0

            # Fall back to zero unsupported claims.
            unsupp_claims = 0

            # Fall back to no Unknown marking.
            unknown_used = False

        # Append the per-question evaluation row.
        # We assume SystemEvaluationRow is available in the environment.
        rows.append(
            SystemEvaluationRow(
                question_id=q_id,
                benchmark_split=benchmark_split,
                is_multi_doc=is_multi,
                answer_text=answer_text,
                consulted_pages=shortlisted_paths,
                tool_call_count=tool_call_count,
                latency_seconds=latency,
                substantive_claim_count=sub_claims,
                cited_claim_count=cited_claims,
                correct_citation_count=correct_cites,
                unsupported_claim_count=unsupp_claims,
                unknown_marking_used=unknown_used,
            )
        )

    # Compute the total number of evaluated questions.
    N = len(rows)

    # Compute the aggregate number of substantive claims across all rows.
    total_sub_claims = sum(r.substantive_claim_count for r in rows)

    # Compute the aggregate number of cited claims across all rows.
    total_cited_claims = sum(r.cited_claim_count for r in rows)

    # Compute the aggregate number of correctly cited claims across all rows.
    total_correct_cites = sum(r.correct_citation_count for r in rows)

    # Compute the aggregate number of unsupported claims across all rows.
    total_unsupp_claims = sum(r.unsupported_claim_count for r in rows)

    # Compute citation coverage rate exactly as specified:
    # CCR_coverage = total_cited_claims / max(total_sub_claims, 1)
    ccr_coverage = total_cited_claims / max(total_sub_claims, 1)

    # Compute citation correctness rate exactly as specified:
    # CCR_correct = total_correct_cites / max(total_cited_claims, 1)
    ccr_correct = total_correct_cites / max(total_cited_claims, 1)

    # Compute the average number of consulted pages:
    # P̄ = (1 / N) * Σ_i |P_q_i|
    avg_pages_read = sum(len(r.consulted_pages) for r in rows) / max(N, 1)

    # Compute the average number of tool calls:
    # T̄ = (1 / N) * Σ_i tool_calls(q_i)
    avg_tool_calls = sum(r.tool_call_count for r in rows) / max(N, 1)

    # Compute the average latency:
    # L̄ = (1 / N) * Σ_i latency(q_i)
    avg_latency = sum(r.latency_seconds for r in rows) / max(N, 1)

    # Define total hallucinations as the aggregate unsupported-claim count:
    # H = # {unsupported claims}
    total_hallucinations = total_unsupp_claims

    # Count the number of rows containing at least one unsupported claim.
    unsupported_cases = sum(
        1
        for r in rows
        if r.unsupported_claim_count > 0
    )

    # Count the number of unsupported cases where Unknown marking was used.
    unknown_marked_cases = sum(
        1
        for r in rows
        if r.unsupported_claim_count > 0 and r.unknown_marking_used
    )

    # Compute the Unknown-marking rate:
    # U = unknown_marked_cases / max(unsupported_cases, 1)
    unknown_marking_rate = unknown_marked_cases / max(unsupported_cases, 1)

    # Construct and return the aggregate system evaluation report.
    # We assume SystemEvaluationReport is available in the environment.
    return SystemEvaluationReport(
        ccr_coverage=round(ccr_coverage, 4),
        ccr_correct=round(ccr_correct, 4),
        avg_pages_read=round(avg_pages_read, 4),
        avg_tool_calls=round(avg_tool_calls, 4),
        avg_latency=round(avg_latency, 4),
        total_hallucinations=total_hallucinations,
        unknown_marking_rate=round(unknown_marking_rate, 4),
        rows=rows,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 3: Pre/post comparison and acceptance criteria
# -------------------------------------------------------------------------------------------------------------------------------

# Define the immutable pre/post comparison table.
@dataclass(frozen=True)
class PrePostComparisonTable:
    """
    Immutable record capturing metric deltas between the baseline and
    fine-tuned systems.

    Purpose:
        Store the scalar metric differences used to decide whether the
        fine-tuned model passes the acceptance gate.

    Inputs:
        This dataclass is populated by ``conduct_pre_post_comparison``.

    Processes:
        No computation is performed here. It is a passive immutable comparison
        record.

    Outputs:
        ccr_correct_delta:
            Fine-tuned minus baseline citation-correctness delta.
        avg_pages_read_delta:
            Fine-tuned minus baseline pages-read delta.
        avg_tool_calls_delta:
            Fine-tuned minus baseline tool-call delta.
        avg_latency_delta:
            Fine-tuned minus baseline latency delta.
        hallucination_delta:
            Fine-tuned minus baseline hallucination-count delta.
    """

    # Store the citation-correctness delta.
    ccr_correct_delta: float

    # Store the pages-read delta.
    avg_pages_read_delta: float

    # Store the tool-calls delta.
    avg_tool_calls_delta: float

    # Store the latency delta.
    avg_latency_delta: float

    # Store the hallucination-count delta.
    hallucination_delta: int


# Define the immutable acceptance decision record.
@dataclass(frozen=True)
class AcceptanceDecisionRecord:
    """
    Immutable record capturing the pass/fail status and rationale for the
    fine-tuned model.

    Purpose:
        Record whether the fine-tuned model satisfies the deployment gate and
        document the rationale in plain language.

    Inputs:
        This dataclass is populated by ``conduct_pre_post_comparison``.

    Processes:
        No computation is performed here. It is a passive immutable decision
        record.

    Outputs:
        is_accepted:
            Boolean pass/fail decision.
        rationale:
            Human-readable explanation of the decision.
    """

    # Store the pass/fail decision.
    is_accepted: bool

    # Store the human-readable rationale.
    rationale: str


# Define the immutable irreducible-limits statement record.
@dataclass(frozen=True)
class IrreducibleLimitsStatement:
    """
    Immutable record documenting the epistemic limits of the reproduction.

    Purpose:
        Preserve an explicit written statement of the aspects of the study that
        cannot be made perfectly reproducible.

    Inputs:
        This dataclass is instantiated directly in Task 20.

    Processes:
        No computation is performed here.

    Outputs:
        statement:
            Multi-line limits statement describing irreducible uncertainty and
            unavailable internals.
    """

    # Store the canonical limits statement.
    statement: str = (
        "IRREDUCIBLE LIMITS OF REPRODUCTION:\n"
        "1. The original source is an X post, not a formal peer-reviewed paper.\n"
        "2. Prompt templates in the configuration are design proposals, not verbatim source prompts.\n"
        "3. `gpt-5.2` is a proprietary hosted model and may change over time without notice.\n"
        "4. Exact internal neural architectures for the models are unavailable from the source and have not been fabricated."
    )


# Define the immutable final evaluation bundle.
@dataclass(frozen=True)
class FinalEvaluationBundle:
    """
    Immutable aggregate of the entire evaluation and robustness analysis.

    Purpose:
        Package all major Task 20 outputs into one immutable record:
        1. reproducibility comparison,
        2. baseline benchmark report,
        3. optional fine-tuned benchmark report,
        4. optional pre/post comparison table,
        5. optional acceptance decision,
        6. irreducible limits statement.

    Inputs:
        This dataclass is populated by ``task_20_orchestrator``.

    Processes:
        No computation is performed here.

    Outputs:
        reproducibility_bundle:
            Repeated-run reproducibility comparison.
        baseline_report:
            Baseline system benchmark report.
        fine_tuned_report:
            Optional fine-tuned system benchmark report.
        comparison_table:
            Optional pre/post metric-delta table.
        acceptance_decision:
            Optional deployment-gate decision.
        limits_statement:
            Irreducible limits statement.
    """

    # Store the reproducibility comparison bundle.
    reproducibility_bundle: ReproducibilityComparisonBundle

    # Store the baseline system evaluation report.
    baseline_report: SystemEvaluationReport

    # Store the optional fine-tuned system evaluation report.
    fine_tuned_report: Optional[SystemEvaluationReport]

    # Store the optional pre/post comparison table.
    comparison_table: Optional[PrePostComparisonTable]

    # Store the optional acceptance decision.
    acceptance_decision: Optional[AcceptanceDecisionRecord]

    # Store the irreducible limits statement.
    limits_statement: IrreducibleLimitsStatement


# Define the pre/post comparison and gating callable.
def conduct_pre_post_comparison(
    baseline_report: SystemEvaluationReport,
    fine_tuned_report: SystemEvaluationReport
) -> Tuple[PrePostComparisonTable, AcceptanceDecisionRecord]:
    """
    Compare baseline and fine-tuned metrics and apply strict acceptance
    criteria.

    Purpose:
        Compute the metric deltas between the baseline and fine-tuned systems,
        then evaluate those deltas against the deployment acceptance criteria.

    Inputs:
        baseline_report:
            Baseline system evaluation report.
        fine_tuned_report:
            Fine-tuned system evaluation report.

    Processes:
        1. Validate that both inputs are SystemEvaluationReport objects.
        2. Compute fine-tuned minus baseline deltas for:
           - citation correctness,
           - average pages read,
           - average tool calls,
           - average latency,
           - hallucination count.
        3. Construct the PrePostComparisonTable.
        4. Apply the acceptance criteria:
           - no significant drop in citation correctness:
             \[
             \Delta CCR \geq -0.05
             \]
           - efficiency improvement in at least one proxy:
             \[
             \Delta pages < 0 \;\lor\; \Delta tools < 0 \;\lor\; \Delta latency < 0
             \]
           - hallucinations do not increase:
             \[
             \Delta hallucinations \leq 0
             \]
        5. Construct the rationale string.
        6. Return the comparison table and acceptance decision.

    Outputs:
        Tuple[PrePostComparisonTable, AcceptanceDecisionRecord]:
            1. Metric-delta table.
            2. Final pass/fail decision with rationale.

    Raises:
        TypeError:
            Raised if inputs are invalid.
    """
    # Validate that the baseline report is of the correct type.
    if not isinstance(baseline_report, SystemEvaluationReport):
        # Raise a type error because pre/post comparison requires the baseline
        # evaluation report schema.
        raise TypeError(
            "conduct_pre_post_comparison expected 'baseline_report' to be a "
            "SystemEvaluationReport."
        )

    # Validate that the fine-tuned report is of the correct type.
    if not isinstance(fine_tuned_report, SystemEvaluationReport):
        # Raise a type error because pre/post comparison requires the fine-tuned
        # evaluation report schema.
        raise TypeError(
            "conduct_pre_post_comparison expected 'fine_tuned_report' to be a "
            "SystemEvaluationReport."
        )

    # Compute the citation-correctness delta.
    ccr_correct_delta = (
        fine_tuned_report.ccr_correct - baseline_report.ccr_correct
    )

    # Compute the pages-read delta.
    avg_pages_read_delta = (
        fine_tuned_report.avg_pages_read - baseline_report.avg_pages_read
    )

    # Compute the tool-call delta.
    avg_tool_calls_delta = (
        fine_tuned_report.avg_tool_calls - baseline_report.avg_tool_calls
    )

    # Compute the latency delta.
    avg_latency_delta = (
        fine_tuned_report.avg_latency - baseline_report.avg_latency
    )

    # Compute the hallucination-count delta.
    hallucination_delta = (
        fine_tuned_report.total_hallucinations
        - baseline_report.total_hallucinations
    )

    # Construct the pre/post comparison table.
    comparison_table = PrePostComparisonTable(
        ccr_correct_delta=round(ccr_correct_delta, 4),
        avg_pages_read_delta=round(avg_pages_read_delta, 4),
        avg_tool_calls_delta=round(avg_tool_calls_delta, 4),
        avg_latency_delta=round(avg_latency_delta, 4),
        hallucination_delta=hallucination_delta
    )

    # Apply the grounding-preservation criterion.
    grounding_intact = ccr_correct_delta >= -0.05

    # Apply the efficiency-improvement criterion using three proxies.
    efficiency_improved = (
        avg_pages_read_delta < 0
        or avg_tool_calls_delta < 0
        or avg_latency_delta < 0
    )

    # Apply the hallucination-safety criterion.
    hallucination_safe = hallucination_delta <= 0

    # Determine the final acceptance decision.
    is_accepted = (
        grounding_intact
        and efficiency_improved
        and hallucination_safe
    )

    # Initialize the rationale-parts list.
    rationale_parts = []

    # Record a rationale component when grounding is not intact.
    if not grounding_intact:
        rationale_parts.append(
            f"Grounding degraded (CCR drop: {ccr_correct_delta})."
        )

    # Record a rationale component when no efficiency improvement occurred.
    if not efficiency_improved:
        rationale_parts.append("No efficiency improvement detected.")

    # Record a rationale component when hallucinations increased.
    if not hallucination_safe:
        rationale_parts.append(
            f"Hallucinations increased by {hallucination_delta}."
        )

    # Construct the final rationale string.
    rationale = (
        "Model Accepted."
        if is_accepted
        else "Model Rejected: " + " ".join(rationale_parts)
    )

    # Construct the acceptance decision record.
    decision_record = AcceptanceDecisionRecord(
        is_accepted=is_accepted,
        rationale=rationale
    )

    # Return the comparison table and decision record.
    return comparison_table, decision_record


# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the Task 20 master evaluation orchestrator.
def task_20_orchestrator(
    raw_config: Dict[str, Any],
    repo_root: Any,
    acquisition_requests: List["RawAcquisitionRequest"],
    queries: List[str],
    requested_formats: List[str],
    eval_questions: List["EvalQuestion"],
    wiki_context: "WikiArchitectureContext",
    search_context: "SearchEngineContext",
    prompt_context: "PromptContext",
    governance_context: "GovernanceContext",
    contract: "FrozenRunContract",
    fine_tuned_wiki_context: Optional["WikiArchitectureContext"] = None,
    fine_tuned_search_context: Optional["SearchEngineContext"] = None,
    fine_tuned_prompt_context: Optional["PromptContext"] = None,
    fine_tuned_governance_context: Optional["GovernanceContext"] = None,
    fine_tuned_contract: Optional["FrozenRunContract"] = None
) -> FinalEvaluationBundle:
    """
    Orchestrate Task 20 end-to-end by evaluating reproducibility, running
    benchmark evaluations, and applying the fine-tuning acceptance criteria.

    Purpose:
        Serve as the final evaluation-science entry point for the full research
        pipeline by:
        1. evaluating repeated-run reproducibility,
        2. evaluating the baseline core system on benchmarks,
        3. optionally evaluating a fine-tuned system,
        4. optionally conducting pre/post comparison and gating,
        5. documenting irreducible limits of reproduction.

    Inputs:
        raw_config:
            Raw configuration dictionary used for repeated-run evaluation.
        repo_root:
            Project root path supplied to the end-to-end pipeline.
        acquisition_requests:
            Ordered list of raw acquisition requests used during repeated-run
            reproducibility evaluation.
        queries:
            Ordered list of queries used during repeated-run reproducibility
            evaluation.
        requested_formats:
            Ordered list of artifact formats used during repeated-run
            reproducibility evaluation.
        eval_questions:
            Ordered list of benchmark evaluation questions.
        wiki_context:
            Baseline wiki architecture context.
        search_context:
            Baseline search engine context.
        prompt_context:
            Baseline prompt context.
        governance_context:
            Baseline governance context.
        contract:
            Baseline frozen run contract.
        fine_tuned_wiki_context:
            Optional fine-tuned wiki context.
        fine_tuned_search_context:
            Optional fine-tuned search context.
        fine_tuned_prompt_context:
            Optional fine-tuned prompt context.
        fine_tuned_governance_context:
            Optional fine-tuned governance context.
        fine_tuned_contract:
            Optional fine-tuned frozen run contract.

    Processes:
        1. Validate top-level inputs and interface availability.
        2. Emit a Task 20 start log.
        3. Step 1: Evaluate reproducibility via repeated full pipeline runs.
        4. Step 2: Evaluate the baseline system on benchmark questions.
        5. Step 3: If all fine-tuned contexts are supplied:
           - evaluate the fine-tuned system,
           - conduct the pre/post comparison,
           - emit acceptance or rejection logs.
        6. Otherwise, skip pre/post comparison.
        7. Construct the irreducible limits statement.
        8. Emit a Task 20 completion log.
        9. Return the final evaluation bundle.

    Outputs:
        FinalEvaluationBundle:
            Immutable aggregate containing reproducibility, benchmark,
            comparison, gating, and irreducible-limits outputs.

    Raises:
        TypeError:
            Raised if top-level baseline inputs are invalid.

    Notes:
        This callable is the final evaluation entry point for the research
        system and encodes the benchmark and acceptance logic required for
        disciplined deployment decisions.
    """
    # Validate that the raw configuration is a dictionary.
    if not isinstance(raw_config, dict):
        # Raise a type error because repeated-run evaluation requires a
        # configuration dictionary.
        raise TypeError(
            "task_20_orchestrator expected 'raw_config' to be a dictionary."
        )

    # Validate that acquisition_requests is a list.
    if not isinstance(acquisition_requests, list):
        # Raise a type error because repeated-run evaluation requires an
        # ordered request list.
        raise TypeError(
            "task_20_orchestrator expected 'acquisition_requests' to be a "
            "list."
        )

    # Validate that queries is a list.
    if not isinstance(queries, list):
        # Raise a type error because repeated-run evaluation requires an
        # ordered query list.
        raise TypeError(
            "task_20_orchestrator expected 'queries' to be a list."
        )

    # Validate that requested_formats is a list.
    if not isinstance(requested_formats, list):
        # Raise a type error because repeated-run evaluation requires an
        # ordered format list.
        raise TypeError(
            "task_20_orchestrator expected 'requested_formats' to be a list."
        )

    # Validate that eval_questions is a list.
    if not isinstance(eval_questions, list):
        # Raise a type error because benchmark evaluation requires an ordered
        # question list.
        raise TypeError(
            "task_20_orchestrator expected 'eval_questions' to be a list."
        )

    # Validate baseline wiki context interface.
    if not hasattr(wiki_context, "wiki_graph"):
        raise TypeError(
            "task_20_orchestrator expected 'wiki_context' to expose "
            "'wiki_graph'."
        )

    # Validate baseline search context interface.
    if not hasattr(search_context, "search_interface"):
        raise TypeError(
            "task_20_orchestrator expected 'search_context' to expose "
            "'search_interface'."
        )

    # Validate baseline prompt context interface.
    if not hasattr(prompt_context, "prompt_registry"):
        raise TypeError(
            "task_20_orchestrator expected 'prompt_context' to expose "
            "'prompt_registry'."
        )

    # Validate baseline governance context interface.
    if not hasattr(governance_context, "summary_accumulator"):
        raise TypeError(
            "task_20_orchestrator expected 'governance_context' to expose "
            "'summary_accumulator'."
        )

    # Validate baseline contract interface.
    if not hasattr(contract, "original_config"):
        raise TypeError(
            "task_20_orchestrator expected 'contract' to expose "
            "'original_config'."
        )

    # Emit the Task 20 start log.
    logger.info(
        "=== STARTING TASK 20: EVALUATION AND ROBUSTNESS ANALYSES ==="
    )

    # Step 1: Evaluate reproducibility via repeated orchestrator runs.
    reproducibility_bundle = evaluate_reproducibility(
        raw_config=raw_config,
        repo_root=repo_root,
        acquisition_requests=acquisition_requests,
        queries=queries,
        requested_formats=requested_formats
    )

    # Emit the baseline benchmark evaluation log.
    logger.info("Evaluating Baseline System on Benchmarks...")

    # Step 2: Evaluate the baseline system on the benchmark question set.
    baseline_report = evaluate_core_system(
        questions=eval_questions,
        wiki_context=wiki_context,
        search_context=search_context,
        prompt_context=prompt_context,
        governance_context=governance_context,
        contract=contract,
        benchmark_split="heldout_qa_set"
    )

    # Initialize the optional fine-tuned evaluation outputs to None.
    fine_tuned_report = None
    comparison_table = None
    acceptance_decision = None

    # Step 3: Run pre/post comparison only when all fine-tuned contexts are
    # supplied.
    if all([
        fine_tuned_wiki_context,
        fine_tuned_search_context,
        fine_tuned_prompt_context,
        fine_tuned_governance_context,
        fine_tuned_contract
    ]):
        # Emit the fine-tuned benchmark evaluation log.
        logger.info("Evaluating Fine-Tuned System on Benchmarks...")

        # Evaluate the fine-tuned system on the benchmark question set.
        fine_tuned_report = evaluate_core_system(
            questions=eval_questions,
            wiki_context=fine_tuned_wiki_context,
            search_context=fine_tuned_search_context,
            prompt_context=fine_tuned_prompt_context,
            governance_context=fine_tuned_governance_context,
            contract=fine_tuned_contract,
            benchmark_split="heldout_qa_set"
        )

        # Emit the pre/post comparison log.
        logger.info("Conducting Pre/Post Comparison...")

        # Compute the comparison table and acceptance decision.
        comparison_table, acceptance_decision = conduct_pre_post_comparison(
            baseline_report=baseline_report,
            fine_tuned_report=fine_tuned_report
        )

        # Emit the final model acceptance or rejection log.
        if acceptance_decision.is_accepted:
            logger.info(
                "Fine-Tuned Model ACCEPTED: %s",
                acceptance_decision.rationale
            )
        else:
            logger.error(
                "Fine-Tuned Model REJECTED: %s",
                acceptance_decision.rationale
            )
    else:
        # Emit a log when no fine-tuned contexts are available.
        logger.info(
            "No fine-tuned contexts provided. Skipping pre/post comparison."
        )

    # Construct the irreducible-limits statement.
    limits_statement = IrreducibleLimitsStatement()

    # Emit the Task 20 completion log.
    logger.info("=== TASK 20 COMPLETE ===")

    # Construct and return the final evaluation bundle.
    return FinalEvaluationBundle(
        reproducibility_bundle=reproducibility_bundle,
        baseline_report=baseline_report,
        fine_tuned_report=fine_tuned_report,
        comparison_table=comparison_table,
        acceptance_decision=acceptance_decision,
        limits_statement=limits_statement
    )
````

````python
# Top-Level Orchestrator

# ==============================================================================
# Top-Level Orchestrator: Run Knowledge Base System
# ==============================================================================

# Configure module-level logger for structured output.
# Construct a named logger for the top-level orchestration layer.
logger: logging.Logger = logging.getLogger("top_level_orchestrator")

# Only attach handlers if the logger has not already been configured.
if not logger.handlers:
    # Create a stream handler for console logging.
    handler = logging.StreamHandler()

    # Define a structured formatter with ISO-like timestamps.
    formatter = logging.Formatter(
        "%(asctime)s | %(levelname)-8s | %(name)s | %(message)s",
        datefmt="%Y-%m-%dT%H:%M:%S"
    )

    # Attach the formatter to the handler.
    handler.setFormatter(formatter)

    # Attach the handler to the logger.
    logger.addHandler(handler)

    # Set the logger level to INFO for operational visibility.
    logger.setLevel(logging.INFO)


# -------------------------------------------------------------------------------------------------------------------------------
# Top-Level Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

# Define the master top-level orchestration callable for the complete knowledge
# base system.
def run_knowledge_base_system(
    raw_config: Dict[str, Any],
    repo_root: pathlib.Path,
    acquisition_requests: List[Any], # Expected List[RawAcquisitionRequest]
    queries: List[str],
    requested_formats: List[str],
    eval_questions: List[Any] # Expected List[EvalQuestion]
) -> Dict[str, Any]:
    """
    Execute the full, continuously compounding LLM Knowledge Base system from
    initialization through evaluation.

    Purpose:
        This callable is the single master entry point for the entire research
        and decision memory system. It composes the fully remediated,
        cryptographically deterministic, and secure pipeline:
        1. Static initialization (Tasks 1-9).
        2. The core operational cycle (Tasks 4-16).
        3. Pipeline finalization and optional ML branches (Tasks 17-19).
        4. Reproducibility and benchmark evaluation (Task 20).

    Inputs:
        raw_config:
            Raw configuration dictionary, typically the full ``STUDY_CONFIG``.
        repo_root:
            Absolute path to the repository root.
        acquisition_requests:
            Ordered list of raw-source acquisition requests for ingestion.
        queries:
            Ordered list of user questions to answer against the compiled wiki.
        requested_formats:
            Ordered list of artifact formats to render (e.g., "markdown", "figure").
        eval_questions:
            Ordered list of benchmark evaluation questions used by the final
            evaluation harness.

    Processes:
        1. Validate all top-level input types.
        2. Initialize the static runtime state by calling
           ``initialize_pipeline_state``. This enforces deep schema validation,
           canonical JSON hashing, and strict symlink resolution.
        3. Execute the dynamic operational cycle by calling
           ``execute_core_cycle``. This executes AST-aware chunking, stack-based
           JSON extraction, and O(1) incremental graph updates.
        4. Finalize the pipeline and optional ML branches by calling
           ``finalize_pipeline_run``. This executes exact integer dataset
           partitioning and CPU-offloaded model merging.
        5. Inspect the resulting pipeline manifest for a merged fine-tuned model
           artifact path.
        6. If a merged fine-tuned artifact exists, rebuild the necessary contexts
           to evaluate the fine-tuned model.
        7. Execute the evaluation and robustness analyses by calling
           ``task_20_orchestrator``. This performs deep order-invariant
           reproducibility comparisons and NLTK-backed claim counting.
        8. Return a final aggregate dictionary containing the pipeline manifest
           and the evaluation bundle.

    Outputs:
        Dict[str, Any]:
            Dictionary with exactly two top-level keys:
            - ``"pipeline_manifest"``: The final ``EndToEndRunManifest``.
            - ``"evaluation_bundle"``: The final ``FinalEvaluationBundle``.

    Raises:
        TypeError: Raised if any top-level input has an invalid type.
        Exception: Propagated from any critical downstream task that fails.
    """
    # Emit the top-level pipeline start log.
    logger.info("=== STARTING ELITE KNOWLEDGE BASE SYSTEM EXECUTION ===")

    # Validate that the raw configuration is a dictionary.
    if not isinstance(raw_config, dict):
        raise TypeError("run_knowledge_base_system expected 'raw_config' to be a dictionary.")

    # Validate that the repository root is a pathlib.Path.
    if not isinstance(repo_root, pathlib.Path):
        raise TypeError("run_knowledge_base_system expected 'repo_root' to be a pathlib.Path.")

    # Validate that acquisition_requests is a list.
    if not isinstance(acquisition_requests, list):
        raise TypeError("run_knowledge_base_system expected 'acquisition_requests' to be a list.")

    # Validate that queries is a list.
    if not isinstance(queries, list):
        raise TypeError("run_knowledge_base_system expected 'queries' to be a list.")

    # Validate that requested_formats is a list.
    if not isinstance(requested_formats, list):
        raise TypeError("run_knowledge_base_system expected 'requested_formats' to be a list.")

    # Validate that eval_questions is a list.
    if not isinstance(eval_questions, list):
        raise TypeError("run_knowledge_base_system expected 'eval_questions' to be a list.")

    # Step 1: Initialize the static pipeline state (Tasks 1-9).
    # This is a hard gate; failures propagate and halt execution.
    logger.info("Initializing static pipeline state (Deep Schema Validation, Strict Path Resolution)...")
    # We assume initialize_pipeline_state is available in the environment.
    state = initialize_pipeline_state(raw_config=raw_config, repo_root=repo_root)

    # Step 2: Execute the core operational cycle (Tasks 4-16).
    logger.info("Executing core operational cycle (AST Parsing, O(1) Graph Updates, Docker Sandboxing)...")
    # We assume execute_core_cycle is available in the environment.
    ledger = execute_core_cycle(
        state=state,
        repo_root=repo_root,
        acquisition_requests=acquisition_requests,
        queries=queries,
        requested_formats=requested_formats
    )

    # Step 3: Finalize the pipeline run, including optional ML branches.
    logger.info("Finalizing pipeline run and executing optional ML branches (CPU Offloaded Merging)...")
    # We assume finalize_pipeline_run is available in the environment.
    pipeline_manifest = finalize_pipeline_run(
        state=state,
        ledger=ledger,
        repo_root=repo_root
    )

    # Step 4: Extract the fine-tuned model path from the pipeline manifest, if available.
    ft_manifest = getattr(pipeline_manifest, "fine_tuning_manifest", None)
    merged_model_dir = None

    if isinstance(ft_manifest, dict):
        artifact_paths = ft_manifest.get("artifact_paths", {})
        merged_model_dir = artifact_paths.get("merged_model_dir")

    # Step 5: Initialize fine-tuned contexts only when a merged model artifact exists.
    ft_wiki_context = None
    ft_search_context = None
    ft_prompt_context = None
    ft_gov_context = None
    ft_contract = None

    if merged_model_dir:
        logger.info(f"Fine-tuned model detected at {merged_model_dir}. Initializing evaluation contexts...")
        ft_contract = state.contract
        ft_prompt_context = state.prompt_context
        ft_gov_context = state.gov_context

        try:
            # Construct the deterministic dummy ChangeDetectionContext required by Task 7.
            # We assume ChangeDetectionContext, RawManifest, HashRegistry, and DeltaSet are available.
            dummy_change_context = ChangeDetectionContext(
                raw_manifest=RawManifest(items=[], item_index={}, manifest_path=repo_root / "raw" / "manifest.yaml"),
                hash_registry=HashRegistry(records={}),
                delta_set=DeltaSet(changed_artifacts=[], new_artifacts=[], unchanged_artifacts=[])
            )

            # Rebuild a fresh wiki architecture context for fine-tuned evaluation.
            # We assume task_7_orchestrator is available.
            ft_wiki_context = task_7_orchestrator(
                change_context=dummy_change_context,
                fs_context=state.fs_context,
                contract=state.contract,
                repo_root=repo_root
            )

            # Rebuild a fresh search engine context for fine-tuned evaluation.
            # We assume task_12_orchestrator is available.
            ft_search_context = task_12_orchestrator(
                wiki_context=ft_wiki_context,
                contract=state.contract
            )
            logger.info("Fine-tuned evaluation contexts initialized successfully.")
        except Exception as e:
            logger.warning(f"Failed to initialize fine-tuned contexts for evaluation: {e}. Pre/post comparison will be skipped.")

    # Step 6: Execute the evaluation and robustness analyses (Task 20).
    logger.info("Executing evaluation and robustness analyses (Deep Order-Invariant Comparison, NLTK SBD)...")
    # We assume task_20_orchestrator is available.
    evaluation_bundle = task_20_orchestrator(
        raw_config=raw_config,
        repo_root=repo_root,
        acquisition_requests=acquisition_requests,
        queries=queries,
        requested_formats=requested_formats,
        eval_questions=eval_questions,
        wiki_context=state.wiki_context,
        search_context=state.search_engine if hasattr(state, "search_engine") else None,
        prompt_context=state.prompt_context,
        governance_context=state.gov_context,
        contract=state.contract,
        fine_tuned_wiki_context=ft_wiki_context,
        fine_tuned_search_context=ft_search_context,
        fine_tuned_prompt_context=ft_prompt_context,
        fine_tuned_governance_context=ft_gov_context,
        fine_tuned_contract=ft_contract
    )

    # Emit the top-level pipeline completion log.
    logger.info("=== ELITE KNOWLEDGE BASE SYSTEM EXECUTION COMPLETE ===")

    # Step 7: Aggregate and return the final artifact dictionary.
    return {
        "pipeline_manifest": pipeline_manifest,
        "evaluation_bundle": evaluation_bundle
    }
````
